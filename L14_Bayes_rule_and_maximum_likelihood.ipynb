# **Abstract:**

Today we meet the Bayesian approach to probability and to analyzing the relationship between different random variables. First, we illustrate the central concepts -- **joint, marginal and conditional probability** -- with some simple examples (**Scottish High school** and **Colored Dice**).

# **Encoding and decoding**


# Encoding:
**Sensory stimulus** $\longrightarrow$ **Variable neural activity**


Sensory stimuli evoke variable activity.  Average evoked activity depends systematically on stimulus properties.  We desire to know the **probability** of different levels of activity, given a particular sensory stimulus.


# Decoding:
**Variable neural activity** $\longrightarrow$ **Stimulus estimate**


Variable neural activity determines choice response, which represents a (coarse) stimulus alternative.  More generally, neural activity lets us estimate stimulus properties.  The reliability of such estimates depends on encoding.   We desire to know the **likelihood** of different stimulus estimates, given a particular neural activity.

# **1. Bayes' Rule**

We consider two random variables and wonder what the value of one variable can tell us about the other (statistically speaking).  To this end, we will employ 'Bayes' Rule' or 'Bayes' Law'.

For example, the two random variables could be a stimulus value and the activity of a neuron, which we happen to be observing simultaneously.  In this case, we would like to know what the neuron's activity at a given time tells us about the stimulus value at the same moment, or vice versa.

Obviously the question raised here is very general and applies to any set of paired observations.

*(dt.: Verbundwahrscheinlichkeit)*

# Marginal probability

*(dt.: a priori Wahrscheinlichkeit)*


The  marginal probability is the probability that **one random** variable assumes a particular value, independent of the value of the other.  It is a **one-dimensional** distribution.

In the context of considering two random variables, there are two marginal probability distributions.

# Conditional probability

*(dt. bedingte od. a posteriori Wahrscheinlichkeit)*

The **conditional probability** is defined for subset of outcomes, in which the second variable takes a given value (the given condition).  

It is the probability that the first variable takes different values. It is a **two-dimensional distribution**, because it depends also on the value of the condition.

# **Independent random variables**


Red and white dice are **independent**! Thus, the joint probability of outcomes is simply the product of the individual probabilities:

$\begin{eqnarray}
\qquad\qquad\qquad\qquad P(n_\mathit{red}, n_\mathit{white}) = P(n_\mathit{red}) \cdot P(n_\mathit{white})
\end{eqnarray}$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Independent_joint.png" width="800">

(Slide-14)

# Marginal probabilities

The marginal probabilities are obtained by summation:

$\begin{eqnarray}
\qquad\qquad P(n_\mathit{red}) = \sum_{n_\mathit{white}} \, P(n_\mathit{red}, n_\mathit{white}) \qquad\qquad P(n_\mathit{white}) = \sum_{n_\mathit{red}} \, P(n_\mathit{red}, n_\mathit{white})
\end{eqnarray}$

(Slide-15



# Conditional probability of red, given white

The conditional probability of $n_\mathit{red}$ eyes, given that the **white** dice show $n_\mathit{white}$ eyes,  is

$\begin{eqnarray}
P(n_\mathit{red}|n_\mathit{white} ) = \frac{P(n_\mathit{red}, n_\mathit{white})}{P(n_\mathit{white})}
\end{eqnarray}$

Because of **independence**, conditional and marginal probabilities are the same:

$\begin{eqnarray}
P(n_\mathit{red}|n_\mathit{white} ) = \frac{P(n_\mathit{red}) \, P(n_\mathit{white})}{P(n_\mathit{white})} = P(n_\mathit{red})
\end{eqnarray}$

# Summary Bayes' Rule

**Joint**

$\begin{eqnarray}
\qquad\qquad\qquad\qquad\qquad P(r,s)
\end{eqnarray}$

**Marginal**

$\begin{eqnarray}
\qquad P(r) = \sum_s \, P(r,s) \qquad\qquad\qquad\qquad P(s) = \sum_r \, P(r,s)
\end{eqnarray}$

**Conditional**

$\begin{eqnarray}
\qquad P(r|s) = \frac{P(r,s)}{P(s)}\qquad\qquad\qquad\qquad P(s|r) = \frac{P(r,s)}{P(r)}
\end{eqnarray}$

**Bayes' Rule**

$\begin{eqnarray}
\qquad\qquad\qquad\qquad P(r|s) \, P(s)  = P(s|r) \, P(r)
\end{eqnarray}$

# **4. Joint decoding of two neurons with different tuning**

We now apply conditional probability and Bayes' rule to decoding neural responses.  In general, neuronal populations are heterogeneous, in that each neuron preferes a different tuning (*e.g.*, prefers a different direction of motion).

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_JMcrop_1.png" width="600">

This schematic illustration shows multiple neurons (about 50) tuned to different directions of motion (arrows). The black dots represent the individual responses.

Note that neurons are sorted by direction preference, so that nearby neurons have similar preference.

The blue and orange curves represent average responses to two particular directions (blue and orange arrows).

How can we decode this population response? How can we decide which direction is most likely to have evoked this response.

(Slide-26)

# Combine likelihoods, not responses!

Adding the responses of different neurons does not help, because the summed response would not be informative.

Instead, we need to infer from each individual response two stimulus  *likelihoods* (one for orange, another for blue), which we can the combine across the population.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_JMcrop_2.png" width="600">

(Slide-27)

# Tuning function

Consider neurons $i$ tuned for directions of visual motion.  We approximate the *tuning curve* (**average response**) by a Gaussian function

<br>

$\begin{eqnarray}
r_i = f_i(\theta)  = r_\mathit{max}\, \exp\left[ -\frac{(\theta-\theta_i)^2}{2 \kappa^2} \right]
\end{eqnarray}$

<br>

where $\theta_i$ is the preferred direction of neuron $i$ and $\kappa$ determines the tuning width.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Illu01.png" width="800">

(Slide-28)

# Response variability

Assuming independence, we approximate the **response variability** with a Poisson distribution

<br>

$\begin{eqnarray}
p(x|r_i) = \frac{(r_i)^x}{x!} e^{-r_i}
\end{eqnarray}$

<br>

where $x$ is the actual and $r_i$ is the average response:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Illu02.png" width="800">

Note: while the Poisson assumption is realistic, the independence assumption is not.

(Slide-29)

## Conditional response probability

Combining tuning function and response variability, we obtain the  **conditional response probability**: the **probability** of a response, given the stimulus:

<br>

$\begin{eqnarray}
p(x|\theta) = \frac{(r_i)^x}{x!} \, e^{-r_i} = \frac{\left[f_i(\theta)\right]^x}{x!} e^{-f_i(\theta)}
\end{eqnarray}$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Illu03.png" width="800">

(Slide-30)

# Stimulus likelihood

If all possible stimuli $\theta$ are equally probable, the conditional response probability can be re-interpreted as a **stimulus likelihood**: the **likelihood** of stimulus $\theta$ having evoked the observed response $x_1$

<br>

$\begin{eqnarray}
L(\theta|x_1) =  p(x_1|\theta) = \frac{\left[f_1(\theta)\right]^{x_1}}{x_1!} \, e^{-f_1(\theta)}
\end{eqnarray}$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Illu04.png" width="800">

(Slide-31)

# Multiple stimulus likelihoods

From multiple responses (of different neurons), we obtain multiple likelihoods

<br>

$\begin{eqnarray}
\qquad L(\theta|x_1) \qquad\qquad\qquad\qquad\qquad L(\theta|x_2)
\end{eqnarray}$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Illu05.png" width="800">

(Slide-32)

# Combined or joint likelihood

Assuming that the response variability of different neurons is independent, the product of the individual likelihoods provides the **joint likelihood** that stimulus $\theta$ evoked **both** observed responses $x_1$ and $x_2$, as

<br>

$\begin{eqnarray}
L(\theta | \{x_1, x_2\}) = L(\theta|x_1) \, L(\theta|x_2)
\end{eqnarray}$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec14_Illu06.png" width="800">

(Slide-33)

# Summary joint decoding of two neurons


- The combined response of a heterogeneous population is uninformative about the stimulus.

- To proceed, we rely on prior information about each neuron $i$, namely, its conditional response probability
$P(x_i|\theta)$.

- This equals the likelihood $L(\theta|x_i)$ of an observed response $x_i$ being caused by stimulus $\theta$, provided all $\theta$ are equally probable.

- The joint likelihood $L(\theta|x_1, x_2 )$ of two observed responses, $x_1, x_2$, having been caused by stimulus $\theta$, is obtained by multiplication

$\begin{eqnarray}
\qquad\qquad\qquad L(\theta|x_1, x_2 )= L(\theta|x_1) \, L(\theta|x_2 )
\end{eqnarray}$

(Slide-34)


# **5. Generalization to many neurons with different tuning (advanced)**

In general, the likelihood of stimulus $\theta$, given an observed population response $\{x_1, x_2, \ldots x_n\}$, is

<br>
$\begin{eqnarray}
\qquad\qquad L(\theta | \{x_1, x_2, \ldots x_n\}) = \prod_i \, L( \theta | x_i )
\end{eqnarray}$

if response variability is independent.

<br>

To find the most likely stimulus $\theta$, we seek the maximum of this expression or, equivalently, the maximum of
its logarithm (*'log likelihood'*) :

<br>
$\begin{eqnarray}
\log L(\theta | \{x_1, x_2, \ldots x_n\}) = \log \prod_i \, L( \theta | x_i ) = \sum_i \, \log L(\theta | x_i)
\end{eqnarray}$

The logarithm is convenient because it lets us compute sums rather than products.

(Slide-35)

# Gaussian tuning

Consider neurons $i$ tuned for directions of visual motion.  We approximate the *tuning curve* by a Gaussian function

<br>

$\begin{eqnarray}
r_i = f_i(\theta)  = r_\mathit{max}\, \exp\left[ -\frac{(\theta-\theta_i)^2}{2 \kappa^2} \right]
\end{eqnarray}$

<br>

where $\theta_i$ is the preferred direction of neuron $i$ and $\kappa$ determines the tuning width.

(Slide-36)

```python
from inspect import signature
# @title Population with Gaussian tuning and Poisson variability {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def TuningCurve( theta0, theta_k, kappa, rmax ):

    # initialize array of population responses
    tuning_k = np.zeros(len(theta_k));

    # angle difference between preferred theta_k and actual theta0, on circular range
    delta_k = abs(theta_k - theta0);
    delta_k[delta_k > np.pi] = 2*np.pi - delta_k[delta_k > np.pi];

    # compute tuning (mean response) of population
    tuning_k = rmax * np.exp( - delta_k**2 / (2*kappa**2) )

    return tuning_k

def PoissonVariability( tuning_k ):

    x_k = np.random.poisson(tuning_k)

    return x_k


def plot_population_with_gaussian_tuning(Kappa=0.5, Nneuron=20, toggle=0):
    """(Slide-37)"""

    Rmax = 10

    # range of range of stimulus preferences
    theta_i = np.linspace(-np.pi, np.pi, Nneuron, endpoint=False );

    # random stimulus
    theta0 = 2 * np.pi * np.random.rand() - np.pi ;

    # compute tuning curve for each neuron
    Nstimulus = 200;
    theta_j = np.linspace(-np.pi, np.pi, Nstimulus, endpoint=False );
    f_ij = np.zeros((Nneuron,Nstimulus))

    for i in range(Nneuron):
        f_ij[i,:] = TuningCurve( theta_i[i], theta_j, Kappa, Rmax )

    # get average response
    f_i = TuningCurve( theta0, theta_i, Kappa, Rmax )

    # get noisy response
    x_i = PoissonVariability( f_i )

    # average response above 0.5
    idx = np.where( f_i>0.5 )
    theta_k = theta_i[idx]
    f_k = f_i[idx]
    x_k = x_i[idx]

    # Plot results
    fig, axs = plt.subplots( 1,1, figsize=(8,4))

    # Moments
    for i in range(Nneuron):
        axs.plot( theta_j, f_ij[i,:], color=(0.5,0.5,0.5), linewidth=1 )
    for k in range(len(theta_k)):
        axs.plot( [theta_k[k],theta0], [f_k[k],f_k[k]], ':', color=(0.5,0.5,0.5), linewidth=1 )

    axs.plot( theta_i, f_i, color=(0.75,0.5,0.5), linewidth=2, label='expected response')
    axs.plot( theta_i, x_i, 'o', color=(0.75,0.25,0.25), markersize=5, label='noisy response')
    axs.plot( [theta0,theta0], [0,Rmax], ':', color=(0.75,0.25,0.25), linewidth=2, label='true stimulus')
    axs.legend(loc='upper left')
    axs.set_xlabel('direction of motion')
    axs.set_box_aspect(1/2)
    axs.set_ylabel('response')
    axs.set_title('Population response', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_population_with_gaussian_tuning, Kappa = (0.2, 1.5, 0.1), Nneuron = (10, 50, 10), toggle = (0, 1, 1) )
```

# Poisson variability

Assuming Poisson-distributed variability, we have

<br>

$\begin{eqnarray}
\qquad L(\theta | x_i) = \frac{f_i(\theta)^{x_i}}{x_i!} e^{- f_i(\theta)}
\end{eqnarray}$

<br>

$\begin{eqnarray}
\log L(\theta | x_i) = x_i \, \log f_i(\theta) - \log x_i! - f_i(\theta)
\end{eqnarray}$

for the log likelihood of an individual response $x_i$. The summed log likelihood of of many responses $x_i$ is

<br>

$\begin{eqnarray}
\sum_i \, \log L(\theta|x_i)  =  \sum_i \, x_i \, \log f_i(\theta)  - \sum_i  \, \log x_i! - \sum_i \, f_i(\theta)
\end{eqnarray}$

which turns out to simplify as follows

$\begin{eqnarray}
\qquad \sum_i \, \log L(\theta|x_i) \quad  =  \quad \sum_i \, x_i \, \log f_i(\theta) + \mathit{constant}
\end{eqnarray}$

because other terms are independent of $\theta$ (see below).

(Slide-38)

# Why are other terms independent of $\theta$?

The second term

$\begin{eqnarray}
\qquad\qquad \sum_i  \, \log x_i!
\end{eqnarray}$

depends only on the observed response $\{x_1 , \ldots , x_n\}$, not on $\theta$.

<br>

The third term

$\begin{eqnarray}
\qquad\qquad \sum_i \, f_i(\theta)
\end{eqnarray}$

takes nearly the same value for all $\theta$, provided the neural population covers all possible stimuli 'uniformly'.

```python
from inspect import signature
# @title Total expected response {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def TuningCurve( theta0, theta_k, kappa, rmax ):

    # initialize array of population responses
    tuning_k = np.zeros(len(theta_k));

    # angle difference between preferred theta_k and actual theta0, on circular range
    delta_k = abs(theta_k - theta0);
    delta_k[delta_k > np.pi] = 2*np.pi - delta_k[delta_k > np.pi];

    # compute tuning (mean response) of population
    tuning_k = rmax * np.exp( - delta_k**2 / (2*kappa**2) )

    return tuning_k

def PoissonVariability( tuning_k ):

    x_k = np.random.poisson(tuning_k)

    return x_k


def plot_total_expected_response(Kappa=0.5, Nneuron=20, toggle=0):
    """(Slide-39)"""

    Rmax = 10

    # range of range of stimulus preferences
    theta_i = np.linspace(-np.pi, np.pi, Nneuron, endpoint=False );

    # random stimulus
    theta0 = 2 * np.pi * np.random.rand() - np.pi ;

    # compute tuning curve for each neuron
    Nstimulus = 200;
    theta_j = np.linspace(-np.pi, np.pi, Nstimulus, endpoint=False );
    f_ij = np.zeros((Nneuron,Nstimulus))

    for i in range(Nneuron):
        f_ij[i,:] = TuningCurve( theta_i[i], theta_j, Kappa, Rmax )

    # get average response
    f_i = TuningCurve( theta0, theta_i, Kappa, Rmax )

    # get noisy response
    x_i = PoissonVariability( f_i )

    # total expected response and total observed
    f_mean = np.sum(f_i)
    x_mean = np.sum(x_i)

    # Plot results
    fig, axs = plt.subplots( 1,1, figsize=(8,4))


    for i in range(Nneuron):
        axs.plot( theta_j, f_ij[i,:], color=(0.5,0.5,0.5), linewidth=1 )
    axs.plot( theta_i, f_i, color=(0.75,0.5,0.5), linewidth=2, label='expected response')
    axs.plot( theta_i, x_i, 'o', color=(0.75,0.25,0.25), markersize=5, label='noisy response')
    axs.plot( [theta0,theta0], [0,Rmax], ':', color=(0.75,0.25,0.25), linewidth=2, label='true stimulus')
    axs.bar( [3.3], f_mean, color=(0.25,0.25,0.75), width=0.3, label='Total expected' )
    axs.bar( [3.65], x_mean, color=(0.75,0.25,0.25), width=0.3, label='Total observed' )
    axs.legend(loc='upper left')
    axs.set_xlabel('direction of motion')
    axs.set_ylim((0.0,50.0))
    axs.set_box_aspect(1/2)
    axs.set_ylabel('response')
    axs.set_title('Total expected response', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_total_expected_response, Kappa = (0.2, 1.5, 0.1), Nneuron = (10, 50, 10), toggle = (0, 1, 1) )
```

# Maximum likelihood stimulus

The most likely stimulus $\theta = \theta_{est}$ to have caused an observed response $\{x_1, \ldots , x_n\}$ is the value that maximizes

<br>

$\begin{eqnarray}
\log L(\theta | \{x_1, x_2, \ldots x_n\})  =  \sum_i \, \log L(\theta|x_i) \approx \sum_i \, x_i \, \log f_i(\theta)
\end{eqnarray}$

or, in other words, the response-weighted average of the log-tuning $\log f_i(\theta)$, which is the logarithm of the expected response.

<br>

This expression 'correlates' **observed response** with the logarithm of the **expected response** to a hypothetical stimulus $\theta$!

<br>

The larger the correlation between observed and expected (log) response, the more likely the hypothetical stimulus!

(Slide-40)

# Prepare for maximization by getting derivative of tuning

To solve for the most likely (hypothetical) $\theta$ to have evoked the observed response $\{x_1, x_2, \ldots x_n\}$, we will need the derivative of our log likelihood expressions:

$\begin{eqnarray}
0 \stackrel{!}{=} \frac{\partial}{\partial \theta} \log L(\theta | \{x_1, x_2, \ldots x_n\})  = \frac{\partial}{\partial \theta} \sum_i \, \log L(\theta|x_i)
\end{eqnarray}$

From the assumption of Gaussian tuning, we obtain the derivative of tuning:

$\begin{eqnarray}
f_i(\theta)  = r_{max}\, e^{-\frac{(\theta-\theta_i)^2}{2 \kappa^2}}
\end{eqnarray}$

$\begin{eqnarray}
\log f_i(\theta) = -\frac{(\theta-\theta_i)^2}{2\kappa^2} + \log r_{max}
\end{eqnarray}$

$\begin{eqnarray}
\frac{\partial}{\partial \theta} \log f_i(\theta) = - \frac{\theta-\theta_i}{\kappa^2}
\end{eqnarray}$

(Slide-41)



# Maximum likelihood estimate


We can now solve the maximum likelihood condition (on previous page):

$\begin{eqnarray}
0 &\stackrel{!}{=}&\frac{\partial}{\partial \theta} \sum_i \, \log L(\theta|x_i) = \frac{\partial}{\partial \theta} \sum_i \, x_i \, \log f_i(\theta) =
\\
\\
&=& \sum_i \, x_i \,\frac{\partial}{\partial \theta}  \log f_i(\theta)  = \sum_i \, x_i \, \left( -\frac{\theta-\theta_i}{\kappa^2}\right)  =
\\
\\
&=& -\frac{1}{\kappa^2} \, \sum_i \, x_i \, (\theta-\theta_i) = \theta \, \sum_i \, x_i - \sum_i \, x_i \, \theta_i
\\
\\
&\Rightarrow& \qquad\qquad\theta_\mathit{ml} = \frac{\sum_i \, x_i \, \theta_i}{\sum_i \, x_i}
\end{eqnarray}$

$\theta_\mathit{ml}$ is the maximum likelihood estimate for the stimulus that caused the observed population response.  For our assumptions (Poisson variability and Gaussian tuning), it turns out to be simply the response-weighted average over the preferred stimuli $\theta_i$ of all neurons.

(Slide-42)

```python
from inspect import signature
# @title Maximum likelihood estimate {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def TuningCurve( theta0, theta_k, kappa, rmax ):

    # initialize array of population responses
    tuning_k = np.zeros(len(theta_k));

    # angle difference between preferred theta_k and actual theta0, on circular range
    delta_k = abs(theta_k - theta0);
    delta_k[delta_k > np.pi] = 2*np.pi - delta_k[delta_k > np.pi];

    # compute tuning (mean response) of population
    tuning_k = rmax * np.exp( - delta_k**2 / (2*kappa**2) )

    return tuning_k

def PoissonVariability( tuning_k ):

    x_k = np.random.poisson(tuning_k)

    return x_k


def plot_maximum_likelihood_estimate(Kappa=0.5, Nneuron=20, toggle=0):
    """(Slide-43)"""

    Rmax = 10

    # range of range of stimulus preferences
    theta_i = np.linspace(-np.pi, np.pi, Nneuron, endpoint=False );

    # random stimulus
    theta0 = 2 * np.pi * np.random.rand() - np.pi ;

    # compute tuning curve for each neuron
    Nstimulus = 200;
    theta_j = np.linspace(-np.pi, np.pi, Nstimulus, endpoint=False );
    f_ij = np.zeros((Nneuron,Nstimulus))

    for i in range(Nneuron):
        f_ij[i,:] = TuningCurve( theta_i[i], theta_j, Kappa, Rmax )

    # get average response
    f_i = TuningCurve( theta0, theta_i, Kappa, Rmax )

    # get noisy response
    x_i = PoissonVariability( f_i )

    # maximum likelihood estimate
    # this is tricky because the values of theta come from a circular range
    # to deal with the issue, we convert angles into cartesian coordinates,
    # then average the coordinates, and then convert back into angles

    cos_theta_ml = np.sum(x_i * np.cos(theta_i)) / np.sum(x_i)
    sin_theta_ml = np.sum(x_i * np.sin(theta_i)) / np.sum(x_i)
    theta_ml = np.arctan2(sin_theta_ml, cos_theta_ml)

    # Plot results
    fig, axs = plt.subplots( 1,1, figsize=(8,4))


    for i in range(Nneuron):
        axs.plot( theta_j, f_ij[i,:], color=(0.5,0.5,0.5), linewidth=1 )
    axs.plot( theta_i, f_i, color=(0.75,0.5,0.5), linewidth=2, label='ave response')
    axs.plot( theta_i, x_i, 'o', color=(0.75,0.25,0.25), markersize=5, label='noisy response')
    axs.plot( [theta0,theta0], [0,12], ':', color=(0.75,0.25,0.25), linewidth=3, label='true stimulus')
    axs.plot( [theta_ml, theta_ml], [0,12], ':', color=(0.25,0.25,0.75), linewidth=3, label='ML estimate' )
    axs.legend(loc='upper left')
    axs.set_xlabel('direction of motion')
    axs.set_ylim((0.0,15.0))
    axs.set_box_aspect(1/2)
    axs.set_ylabel('response')
    axs.set_title('ML estimate', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_maximum_likelihood_estimate, Kappa = (0.2, 1.5, 0.1), Nneuron = (10, 50, 10), toggle = (0, 1, 1) )
```

# Math aside: averaging values from a circular range

To average values $\theta$ from a circular range, we need to first transform angles to Cartesian coordinates, and then transform the average (center-of-mass) Cartesian coordinate back to an angle:

$\begin{eqnarray}
\cos(\theta_\mathit{ml}) = \frac{\sum_i \, x_i \, \cos(\theta_i)}{\sum_i \, x_i}, \qquad\qquad \sin(\theta_\mathit{ml}) = \frac{\sum_i \, x_i \, \sin(\theta_i)}{\sum_i \, x_i}
\\
\\
\theta_\mathit{ml} = \arctan\left[ \frac{\sin(\theta_\mathit{ml})}{\cos(\theta_\mathit{ml})} \right] \qquad\qquad\qquad
\end{eqnarray}$

(Slide-44)



# **Summary generalization to many neurons**


- To decode a heterogeneous population, we rely on the conditional response probability of each neuron $i$,
$\quad P(x_i|\theta)\approx L(\theta|x_i)$.

- The joint likelihood $L(\theta|\{x_1, \ldots , x_n\} )$ of an observed response $x_1, \ldots, x_n$ being caused by stimulus $\theta$, is

$\begin{eqnarray}
\qquad\qquad L(\theta|x_1, \dots , x_n )= L(\theta|x_1) \, L(\theta|x_2 ) \, \ldots \,  L(\theta|x_n )
\end{eqnarray}$

- The 'maximum likelihood estimate' $\theta_{est}$ is obtained by maximizing either of these expressions

$\begin{eqnarray}
\qquad\qquad \prod_i\,L_i(\theta| x_i) \quad\textrm{or}\quad \sum_i \,  \log L_i(\theta| x_i)
\end{eqnarray}$

- In our example population -- neurons with Gaussian tuning and Poisson variability -- the ML estimate is simply the response-weighted average

$\begin{eqnarray}
\qquad\qquad \theta_{est} = \frac{\sum_i \, x_i \, \theta_i}{\sum_i \, x_i}
\end{eqnarray}$

- *Averaging values over a circular range requires special precautions!*

# **Overall summary**

The Bayesian approach to probability and to the relationship between different random variables relies on three central concepts: **joint, marginal and conditional probability**

These concepts are foundational in all fields of science.

The Bayesian approach informs **maximum likelihood estimation** and lets us determine the most likely stimulus to have caused an observed neuronal response.

Basically, given **detailed prior knowledge** about response tuning and variability, we can estimate a stimulus from the population response of heterogeneous neurons.

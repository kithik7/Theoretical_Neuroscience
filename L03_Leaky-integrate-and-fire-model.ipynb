**Abstract:**

 Today we take another step towards modeling networks of neurons. We start by appreciating that the exponential relaxation of the single-compartment model (**SCM**) is untouched by the presence of **multiple conductances**. Next we solve the SCM for time-varying input currents with the handy technique of **iterative integration**. By adding a simplistic spike mechanism, we obtain the leaky-integrate-and-fire (**LIF**) model. As a simple application of this model, we derive its **response function** (also called input-output function or transfer function). Finally, we add a touch of biological realism with **spike-frequency adaptation**.


**Overview:**

# **1. Generalization of SCM**

If a neuron is "electrotonically compact", it may be treated as a single "electric compartment" with a membrane potential, membrane capacity, and membrane resistance.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig01.png" width="500">

D&A Fig. 5.3

(Slide-1)

# Electrode current flows two ways

An (inward) electrode current $I_e(t)$ can flow in two ways, dividing itself into (outward) resistance and capacitor currents, $I_R(t)$ and $I_C(t)$:

$$\begin{eqnarray}
I_e(t) = I_R(t) + I_C(t) =  \frac{V_m(t)- E_L}{R_m} + C_m \,\, \frac{d V_m(t)}{dt}
\end{eqnarray}$$

With Ohm's Law and the Capacitor Equation, we can link the input current $I_e(t)$ to the membrane potential $V_m(t)$:

$$\begin{eqnarray}
I_R(t) = \frac{V_m(t)- E_L}{R_m} \qquad\qquad\qquad\qquad I_C(t) =  C_m \,\, \frac{d V_m}{dt}
\end{eqnarray}$$

The reversal potential $E_L$ takes into account the ion-specific nature of the membrane resistance.  The inverse of the membrane resistance is sometimes called "leak conductance":
$$\begin{eqnarray}
g_L = \frac{1}{R_m}
\end{eqnarray}$$

(Slide-2)

# Equilibrium potential and time-constant

The resulting linear differential equation describes how the membrane voltage responds to an input current:

$$\begin{eqnarray}
R_m \, C_m \, \frac{d}{dt} V_m(t)= -V_m(t) + E_L + R_m \, I_e(t)
\end{eqnarray}$$

or

$$\begin{eqnarray}
\tau_m \, \frac{d}{dt}V_m(t) = -V_m(t) + V_\infty(t)
\end{eqnarray}$$

where  equilibrium potential $V_\infty(t)$ and membrane time-constant $\tau_m$ are defined as

$$\begin{eqnarray}
\tau_m = R_m \, C_m \qquad\qquad V_\infty(t) = E_L + R_m \, I_e(t)
\end{eqnarray}$$

# Multiple membrane conductances

In addition to a leak conductance and its associated reversal potential
$$\begin{eqnarray}
g_L = \frac{1}{R_m}, \quad E_L
\end{eqnarray}$$

let us also consider further ion-specific conductances for $Na^+$ and $K^+$:
$$\begin{eqnarray}
g_{Na^+}, \quad E_{Na^+}, \qquad\qquad g_{K^+}, \quad E_{K^+}
\end{eqnarray}$$

Will this change the essential behaviour of our model?

Will the membrane potential still respond with an **exponential relaxation**?

(Slide-7)

# Outward membrane current

We write the dynamic equation in terms of specific quantities (per unit membrane area)

$$\begin{eqnarray}
c_m \, \frac{dV_m}{dt} = -i_m + \frac{I_e}{A}
\end{eqnarray}$$

where $i_m$ is the sum of all outward currents through every ion-specific conductance

$$\begin{eqnarray}
i_m =  g_{Na^+} \left(V_m - E_{Na^+} \right) +  g_{K^+} \left(V_m - E_{K^+} \right) +  g_{L} \left(V_m - E_{L} \right)
\end{eqnarray}$$

(Slide-8)

# Dynamic equation

Writing out the dynamic equation

$$\begin{eqnarray}
c_m \, \frac{dV_m}{dt} = -g_{Na^+} \left(V_m - E_{Na^+} \right) -  g_{K^+} \left(V_m - E_{K^+} \right) -  g_{L} \left(V_m - E_{L} \right)  + \frac{I_e}{A}
\end{eqnarray}$$

and rearranging

$$\begin{eqnarray}
c_m \, \frac{dV_m}{dt} = -V_m \left( g_{Na^+} +  g_{K^+}  +  g_{L}\right) +  \left(g_{Na^+} \, E_{Na^+} + g_{K^+} \, E_{K^+} +  g_{L} \, E_{L}  + \frac{I_e}{A} \right)
\end{eqnarray}$$

gives

$$\begin{eqnarray}
\frac{c_m}{g_{Na^+} +  g_{K^+}  +  g_{L}} \, \frac{dV_m}{dt} = -V_m +  \frac{g_{Na^+} \, E_{Na^+} + g_{K^+} \, E_{K^+} +  g_{L} \, E_{L}  + \frac{I_e}{A}}{g_{Na^+} +  g_{K^+}  +  g_{L}}
\end{eqnarray}$$

(Slide-9)

# Exponential relaxation remains!

Accordingly, we can once again write
$$\begin{eqnarray}
\tau_m^\mathit{eff} \, \frac{dV_m}{dt} = - V_m + V_\infty^\mathit{eff}
\end{eqnarray}$$
where $\tau_m$ is an effective membrane time-constant and and $V_\infty$ an effective equilibrium potential.
$$\begin{eqnarray}
\tau_m^\mathit{eff} = \frac{c_m}{g_{Na^+} +  g_{K^+}  +  g_{L}},
\qquad
V_\infty^\mathit{eff} = \frac{g_{Na^+} \, E_{Na^+} + g_{K^+} \, E_{K^+} +  g_{L} \, E_{L}  + \frac{I_e}{A}}{g_{Na^+} +  g_{K^+}  +  g_{L}}
\end{eqnarray}$$


Note that $\tau_m^\mathit{eff}$ depends on *all* conductances, while $V_\infty^\mathit{eff}$ depends on the input current, *all* conductances, and *all* reversal potentials.


The important point is that the membrane potential still responds with an **exponential relaxation** to an input current.  Only the constants of this relaxation have changed.

(Slide-10)


# **2. Time-varying input currents**

We would like to solve the basic equation of the SCM

$$\begin{eqnarray}
\tau_m \, \frac{dV_m}{dt} = - V_m(t) + E_L + r_m \, I_e(t)
\end{eqnarray}$$

for arbitrary, time-varying input currents $I_e(t)$.


To accomplish this, we introduce a handy computational technique, **iterative integration**.

(Slide-11)

# What we want

Given an arbitrary time-varying input current (upper plot), we want to compute the time-varying response of the membrane potential (lower plot).

# Divide and conquer

To solve this task, we divide time into small intervals. If the input current remains effectively constant during on such interval, we can apply our solution for for constant current steps (from the last lecture)

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig02b.png" width="500">

(Slide-13)


# One interval, magnified greatly

Consider one interval $i$ which begins at $t_i$, ends at $t_{i+1}$, and contains all times $t$ with $t_i \leq t   \leq  t_{i+1}$. The interval duration $\Delta t$ is an important parameter, with $t_{i+1} = t_i+\Delta t$ for all $i$.
\ea

If $\Delta t$ is chosen sufficiently small, the input current can be considered constant during each interval $i$:

$$\begin{eqnarray}
I_e(t) \approx I_i \qquad\qquad t_i \leq t   \leq  t_{i+1}
\end{eqnarray}$$

and the equilibrium potential can be considered constant, as well
$$\begin{eqnarray}
V_\infty(t) \approx V_i^\infty = E_L + r_m\, I_i
\end{eqnarray}$$


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig03.png" width="500">

(Slide-14)

# Solution for one interval

Given the membrane potential at the beginning of an interval, $V_i = V_m(t_i)$, and the equilibrium potential during this interval, $V_i^\infty = E_L + r_m I_i$,
we wish to solve for the potential at the end of the interval,  $V_{i+1} = V_m(t_i+\Delta t)$.

The solution for all times $t_i \leq t \leq t_{i+1}$ is

$$\begin{eqnarray}
V_m(t) &=&  V_i^\infty + \left [ V_i - V_i^\infty  \right]  \, \exp\left(-\frac{t-t_i}{\tau_m} \right)
\end{eqnarray}$$

but we care only about the final point $t_{i+1} = t_i + \Delta t$:

$$\begin{eqnarray}
V_{i+1}  &=& V_i^\infty + \left [ V_i - V_i^\infty  \right]  \, \exp\left(-\frac{\Delta t}{\tau_m} \right)
\end{eqnarray}$$


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig03.png" width="300">

(Slide-15)

# Iterative integration

Starting with given initial values, $V_0$ and $V_0^\infty$, we can repeat the procedure for every interval and iteratively compute obtain the time-varying membrane potential at times $t_1$, $t_2$, $\ldots$, $t_i$

$$\begin{eqnarray}
V_0, V_0^\infty \to V_1, V_1^\infty \to V_2, V_2^\infty \to V_3, V_3^\infty \to \ldots \to V_i, V_{i}^\infty
\end{eqnarray}$$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig02b.png" width="500">

(Slide-16)


# Summary of integrative integration

* Convert time-varying input current $I_e(t)$ into a series of values $I_i$ at discrete times $t_i$ spaced at small intervals $\Delta i$.
* Obtain equilibrium potentials $V_i^\infty = E_L + r_m \,I_i$ for all times $t_i$.
* Obtain membrane potentials $V_i$ by iterative integration, starting from a given initial value $V_0$,
$$\begin{eqnarray}
V_{i+1} = V_i^\infty + \left[V_i - V_i^\infty \right] \exp\left(-\frac{\Delta t}{\tau_m} \right)
\end{eqnarray}$$
* If desired, obtain resistance current from Ohm's law,
$$\begin{eqnarray}
i_i^R = \frac{1}{r_m}\left[V_i - E_L \right]
\end{eqnarray}$$
* If desired, obtain capacitor current from capacitor equation,
$$\begin{eqnarray}
i_i^C = c_m \frac{dV(t_i)}{dt} =  \frac{1}{r_m} \left[ - V_i + E_L + r_m \, I_i \right]= - \frac{1}{r_m} \left[ V_i - V_i^\infty \right]
\end{eqnarray}$$
* Double check your computation by confirming that
$$\begin{eqnarray}
I_i = i_i^R + i_i^C\qquad\qquad \forall \, t_i
\end{eqnarray}$$

For small time intervals $\Delta t$, the results approximate smooth curves for both currents (top) and voltages (bottom):

```python
# @title Iterative integration of single-compartment model {"vertical-output":true,"display-mode":"form"}
""" Slide-19 """
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Iterative integration of single-compartment model !

rm = 10;          # specific membrane resistance, in MOhm mm2 (10MOhm mm2)
cm = 5;           # specific membrane capacitance, in nF /mm2 (50 nF/mm2)
i0 = 2.5;         # amplitude of electrode current, in nA / mm2 (25 nA/mm2)
EL = -65;         # reversal potential of leak conductance, in mV (-65 mV)
taum = rm*cm;     # membrane time constant, in seconds (0.05 s or 50 ms)

                  # voltage in units of mV, time in units of ms !!!

dt  = 0.2;                   # time step, in s (0.2 ms)
T   = 200;                   # total time, in s (400 ms)
Ni  = round(T/dt) + 1;       # number of time steps
ti  = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

# cook up a time-varying input current as a superposition of eight Gaussian
# waveforms with random mean and variance

N = 8
iei = np.zeros_like(ti)

for n in range(1, N):
    ien_mean  = T * np.random.rand();
    ien_sigma = T / 32 * (1 + np.random.rand() );
    ien_ampl  = i0 * (0.2 + 0.8 * np.random.rand());

    iei += ien_ampl * np.exp(- (ti - ien_mean) ** 2 / (2 * ien_sigma ** 2)) / np.sqrt(2 * ien_sigma ** 2);


# equilibrium potential, vector with values for each time step

Vi_infty = iei * rm + EL;

# initialize vectors for iteration results (membrane potential, capacitor
# current, membrane current)

Vmi = np.zeros_like(ti);
imi = np.zeros_like(ti);
ici = np.zeros_like(ti);

# iterate over successive time steps, starting at resting potential (reversal
# potential of leak)

Vmi[0] = EL

for i in range(0,Ni):
    # compute membrane and capacitor current at time i
    imi[i] = ( Vmi[i] - EL ) / rm;

    ici[i] = -(1/rm) * (Vmi[i] - Vi_infty[i] );

    # compute membrane potential at time i+1
    if i < len(ti)-1:
        Vmi[i + 1] = Vi_infty[i] + (Vmi[i] - Vi_infty[i]) * np.exp(-dt / taum)


# Create a figure and a grid of subplots (2 rows, 1 columns)
fig, axs = plt.subplots(2, 1, figsize=(6, 4))

# First subplot
axs[0].plot(ti, Vmi, 'r', ti, Vi_infty, 'r--');
axs[0].set_title('Potentials');
axs[0].set_xlabel('time [ms]');
axs[0].set_ylabel('potential [mV]');


# Second subplot
axs[1].plot(ti, imi, 'r', ti, ici, 'b', ti, iei, 'k');
axs[1].set_title('Currents');
axs[1].set_xlabel('time [ms]');
axs[1].set_ylabel('current [nA]');

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()

```

# Summary of SCM generalization

* The single-compartment model (SCM) can be solved iteratively for arbitrary input currents (i.e., electrode or synaptic currents).     
* To do so, we divide time into short intervals, so that the input current may be considered piece-wise constant.

* Intuitively, thes membrane potential sums over (**integrates**) input currents.

* However, the integration is **leaky** rather than perfect, because some current 'leaks out' through the membrane or **leak** conductance $g_L = \frac{1}{r_m}$.

* This explains why this model can be described (and understood) as a **leaky integrator**.

# **3. Simplistic spike mechanism for LIF model**

# Threshold and reset

The spike mechanism is defined as an **exception** to the iterative integration of the membrane potential, which introduces a **discontinuity**.  The mechanism involves a threshold potential $V_\mathit{th}$, which is set somewhat above the resting potential $E_L$
$$\begin{eqnarray}
V_\mathit{th} > E_l
\end{eqnarray}$$
and a reset potential $V_\mathit{reset}$, well below the  the resting potential $E_L$
$$\begin{eqnarray}
V_\mathit{reset} < E_l
\end{eqnarray}$$
The mechanism works by simply resetting the membrane  potential whenever it exceeds the threshold:
$$\begin{eqnarray}
\mathrm{if} \quad V_m(t) > V_\mathit{th} \quad  \mathrm{then} \quad V_m(t+\Delta t) = V_\mathit{reset}
\end{eqnarray}$$
Thus, the membrane potential **never allowed** to rise above threshold.

Optionally, to create a more realistic appearance, an overshoot (instantaneous potential excursion) may be added as decoration.

# Realistic and not-so-realistic aspects

The realistic aspects of our firing mechanism are:

* $V_m$ changes continuously as long as it remains below a threshold value $V_{th}$.
* $V_m$ changes discontinuously whenever $V_{th}$ is reached.
* After the discontinuity, $V_m$ assumes a much lower value.

Of course, many other aspects are non-so-realistic:
* Instantaneous time-course.
* No depolarization during spike.
* No refractory period.
* No effect on membrane conductances.
* No attempt to model the actual biophysical mechanism.

# **4. Response rate to constant input current (advanced)**

The main point of the LIF model is to predict how a neuron spikes (generates action potentials) in response to an input current.

A straightforward application of the LIF model is to compute the rate response (spike frequency) for a constant input current.

The illustration below shows such a prediction for spike rate as a function of input current.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig09.png" width="500">

How exactly was this prediction obtained?

(Slide-34)

# Membrane potential between successive spikes

When the input current $I_e$ is constant, the subthreshold potential always follows the same trajectory, beginning with $V_\mathit{reset}$ and ending at $V_\mathit{th}$. The inter-spike-interval $t_\mathit{isi}$ is how long it takes to get from A to B.

We want to calculate $t_{isi}$, because it will tell us how often spikes occur (i.e., the spike frequency or rate).

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec03_Fig08.png" width="500">

To do this, we start from the solution for a constant input current
$$\begin{eqnarray}
V_\mathit{th} =  V_\infty + \left[ V_\mathit{reset} - V_\infty \right] \, \exp\left( -\frac{t_{isi}}{\tau_m} \right),  \qquad\qquad V_\infty = E_L + r_m \, I_e
\end{eqnarray}$$
and solve for $t_{isi}$, which is the only unknown.

(Slide-35)

# Solving for $t_{isi}$

$$\begin{eqnarray}
V_{th} =  V_\infty + \left[ V_{reset} - V_\infty \right] \, \exp\left( -\frac{t_{isi}}{\tau_m} \right), \qquad\qquad V_\infty = E_L + r_m \, I_e
\end{eqnarray}$$
$$\begin{eqnarray}
\Leftrightarrow
\end{eqnarray}$$
$$\begin{eqnarray}
\frac{V_{th} - V_\infty}{V_{reset} - V_\infty} =  \exp\left( -\frac{t_{isi}}{\tau_m} \right)
\end{eqnarray}$$
$$\begin{eqnarray}
\Leftrightarrow
\end{eqnarray}$$
$$\begin{eqnarray}
t_{isi} =  \tau_m\,  \ln\left( \frac{V_\infty - V_{reset} }{ V_\infty - V_{th} }  \right)  = \tau_m \, \ln\left( \frac{r_m\, I_e + E_L - V_{reset} }{ r_m \, I_e + E_L - V_{th} }  \right)
\end{eqnarray}$$

We find that $t_\mathit{isi}$ is a logarithmic function of a ratio of potentials. At first glance, this resembles the **Nernst equation**, which featured the logarithm of a ratio of concentrations.

However, unlike concentrations, potentials can be **negative**, which implies that the existence of a solution is not guaranteed! (The logarithm is defined only for positive arguments.)

(Slide-36)

# Voltage and current threshold




A finite value for $t_{isi}$ is obtained only if

$$\begin{eqnarray}
 r_m \, I_e + E_L - V_{th} > 0
\end{eqnarray}$$

because input current $I_e$ must raise equilibrium potential $V_\infty = E_L + r_m \, I_e$ above the threshold potential $V_{th}$ in order for spikes to occur.

In other words, only input currents
$$\begin{eqnarray}
I_e > \frac{V_{th} - E_L}{r_m}
\end{eqnarray}$$
are able to produce spikes.

Thus, the voltage threshold of our LIF model predicts the existence of a current threshold!

(Slide-37)

# Current-to-rate transfer function

We are now in a position to compute the rate response of a LIF neuron to a constant input current.

Electrode current $I_e$ represents the {\it input} to a LIF neuron.   In a network model, this may be replaced by synaptic currents.

Spike rate or frequency $r$ represents the {\it output} of a LIF neuron.

$$\begin{eqnarray}
r \, [\mathit{Hz}] = \frac{1}{t_\mathit{isi} \, [\mathit{s}]}
\end{eqnarray}$$


We are interested in the **transfer function**
$$\begin{eqnarray}
r  = F\left( I_e \right) = \frac{1}{t_\mathit{isi}(I_e)} = \tau_m \ln\left( \frac{ r_m \, I_e + E_L - V_{th} }{r_m\, I_e + E_L - V_{reset} }  \right)
\end{eqnarray}$$

(Slide-38)

# Approximate formula



Seeking an approximate formula, we write

$$\begin{eqnarray}
t_{isi} &=  \tau_m \ln\left( \frac{r_m\, I_e + E_L - V_{reset} }{ r_m \, I_e + E_L - V_{th} }  \right)
\\
&= \tau_m \ln\left( \frac{r_m\, I_e + E_L - V_{th} + V_{th} - V_{reset} }{ r_m \, I_e + E_L - V_{th} }  \right)=
\\
&= \tau_m \ln\left( 1+\frac{V_{th} - V_{reset} }{ r_m \, I_e + E_L - V_{th} }\right)
\end{eqnarray}$$

Approximating $\ln(1+z) \approx z$, we obtain

$$\begin{eqnarray}
t_{isi} \approx  \frac{\tau_m \left( V_{th} - V_{reset} \right) }{ r_m \, I_e + E_L - V_{th}}
\end{eqnarray}$$
or
$$\begin{eqnarray}
r  = F(I_e)= \frac{1}{t_{isi}} \approx  -\frac{V_{th} - E_L}{\tau_m \left(V_{th} - V_{reset}\right)} + \frac{1}{c_m \left(V_{th} - V_{reset}\right)} \, I_e
\end{eqnarray}$$

In other words, our transfer function is approximately linear above the current threshold, but zero below this threshold.

(Slide-39)

r \approx - 16.6 s^{-1} + 5 s^{-1}\, \frac{I_e}{nA \, mm^{-2}} \qquad\qquad I_e \geq 3.3 \, nA \, mm^{-2}

# Summary response rate

* For constant input currents, we can compute the inter-spike-interval, $t_{isi}$, exactly.

* This relates the output spike frequency $r = \frac{1}{t_{isi}}$ to the input current.

* Our model predicts an threshold input current, below which no spikes are fired.

*  Above this threshold, it predicts an approximately linear increase of output spike frequency.

Consider a biological neuron at rest. (Slide-45)

# The key idea

In biological neurons, spike-frequency adaptation is thought to invove a $\mathit{Ca}^{2+}$-gated $K^+$-conductance: a conductance specific for $K^+$ ions that is opened by elevated concentrations of intracellular $\mathit{Ca}^{2+}$.

Here is the key idea:

* During each spike, some $\mathit{Ca}^{2+}$ enters the neuron, elevating the $\mathit{Ca}^{2+}$-concentration a little. Repeated spikes further elevate the $\mathit{Ca}^{2+}$-concentration.

* In response to elevated $\mathit{Ca}^{2+}$, certain $K^+$-specific channel proteins open, presumably due to $\mathit{Ca}^{2+}$ binding.

* Due to the negative reversal potential of  $K^+$, this permits a $K^+$ outflow and lowers the membrane potential.

* A lower membrane potential makes it more difficult for the neuron to reach the firing threshold.

* **In short, each spike that occurs makes it more difficult for additional spikes to occur.**

(Slide-46)


# A heuristic model

We introduce a time-varying $K^+$-conductance $g_\mathit{sfa}(t)$, which increases by a constant amount $\Delta g_\mathit{sfa}$ during each spike.

Between spikes, $g_\mathit{sfa}$ relaxes exponentially to zero, with time-constant $\tau_\mathit{sfa}$, mimicking rapid sequestering of free $\mathit{Ca}^{2+}$ inside the cell.

The dynamic equation of $g_\mathit{sfa}(t)$ is as follows

$$\begin{eqnarray}
\tau_\mathit{sfa} \, \frac{dg_\mathit{sfa}}{dt} = -g_\mathit{sfa}\qquad\mathrm{if} \qquad V(t) < V_{th}
\\
\\
g_\mathit{sfa}(t+dt) = g_\mathit{sfa} + \Delta g_\mathit{sfa} \qquad\mathrm{if} \qquad V(t) \geq V_{th}
\end{eqnarray}$$

This equation includes an **exception** when the membrane potential reaches threshold, just like our spike mechanism.

# Extended LIF model with two state variables

The extended LIF model now features two **state variables** (independent actors or dancers).

In addition to the membrane potential $V(t)$, we have the adaptive conductance $g_\mathit{sfa}$. The membrane current is carried by both *constant* leak conductance $g_L = \frac{1}{r_m}$ and the *time-varying* adaptative conductance $g_\mathit{sfa}$:



When $V(t) < V_{th}$:
$$\begin{eqnarray}
c_m \frac{dV}{dt} &=& -g_L\, (V-E_L)  - g_\mathit{sfa} \, (V-E_K) + \frac{I_e}{A}
\\
\tau_\mathit{sfa} \, \frac{dg_\mathit{sfa}}{dt} &=& -g_\mathit{sfa}
\end{eqnarray}$$

When $V(t) \geq V_{th}$:


$$\begin{eqnarray}
V(t+dt) &=& V_{reset}
\\
g_\mathit{sfa}(t+dt) &=& g_\mathit{sfa} + \Delta g_\mathit{sfa}
\end{eqnarray}$$

(Slide-48)


# Why does this work?

When we simulate this extended model with two state variables, we find that it does reproduce spike-frequency adaptation, in the sense that inter-spike-intervals grow progressively longer.

```python
# @title LIF with spike-frequency adaptation {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Leaky-integrate-and fire model with spike-frequency adaptation


rm = 90;         # specific membrane resistance, in MOhm mm2 (90 MOhm mm2)
cm = 0.33;         # specific membrane capacitance, in nF /mm2 (0.33 nF/mm2)
i0 = 0.6;        # amplitude of electrode current, in nA / mm2 (0.5 nA/mm2)
EL = -65;        # reversal potential of leak conductance, in mV (-65 mV)
Vth= -50;        # spike threshold
Vreset = -65;    # reset potential
taum = rm*cm;    # membrane time constant, in seconds (0.03 s or 30 ms)

EK   = -70;      # reversal potential of potassium conductance
Delta_gsfa = 0.6/rm;    # increment of spike-frequency adaptation, as a fraction of leak conductance
tau_sfa = 100;   # time-constant of spike-frequencey adaptation, in ms (100 ms)

                 # voltage in units of mV, time in units of ms !!!

dt  = 0.1;                   # time step, in ms (0.1 ms)
T   = 200;                   # total time, in ms (200 ms)
Ni  = round(T/dt) + 1;       # number of time steps
ti  = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

tn_spike = np.linspace(0, 20, 21);  # maximal time of spikes to be saved
gn_sfa = np.zeros_like(tn_spike);   # conductance after spike

# initialize vectors for iteration results (membrane potential, adaptive conductance, capacitor current, membrane current)

Vmi = np.zeros_like(ti);
Vmi_infty = np.zeros_like(ti);

imi = np.zeros_like(ti);
ici = np.zeros_like(ti);

gi_sfa = np.zeros_like(ti);

gmi    = np.ones_like(ti);

# iterate over successive time steps, starting at resting potential (reversal potential of leak)

Vmi[0] = EL
gi_sfa[0] = 0

n_spike = 0;

for i in range(0,Ni):

    # compute equilibrium potential and effective membrane time constant at time i

    Vmi_infty[i] = (EL + rm * i0 + rm * gi_sfa[i] * EK) / (1 + rm * gi_sfa[i]);

    taum_eff    = taum / (1 + rm * gi_sfa[i]);

    # compute membrane potential at time i+1
    # compute adaptive conductance at time i+1
    if i < len(ti)-1:
        Vmi[i + 1] = Vmi_infty[i] + (Vmi[i] - Vmi_infty[i]) * np.exp(-dt / taum_eff);
        gi_sfa[i + 1] = gi_sfa[i] * np.exp(-dt / tau_sfa);

        # exception for spikes
        if Vth < Vmi[i + 1]:
            Vmi[i + 1]    = Vreset;
            gi_sfa[i + 1] = gi_sfa[i] + Delta_gsfa;

            # save time of spikes and conductance before spike (for plotting)
            n_spike    += 1;
            if n_spike <= len(tn_spike):
                tn_spike[n_spike] = ti[i + 1];
                gn_sfa[n_spike]   = gi_sfa[i];

# Create a figure and a grid of subplots (2 rows, 1 columns)
fig, axs = plt.subplots(2, 1, figsize=(6, 6))

# First subplot
axs[0].plot(ti, Vmi, 'r', ti, Vmi_infty, 'b--');
axs[0].plot(np.array([0, T]), np.array([Vth, Vth]), 'r:');
                              # mark spikes
for n in range(1,n_spike):
    axs[0].plot(np.array([tn_spike[n], tn_spike[n]]), np.array([Vth, -10]), 'k:');

axs[0].set_title('Potentials');
axs[0].set_xlabel('time [ms]');
axs[0].set_ylabel('potential [mV]');


# Second subplot
axs[1].plot(ti, gi_sfa*rm, 'r', ti, gmi, 'b');
                              # mark spikes
for n in range(1,n_spike):
    axs[1].plot(np.array([tn_spike[n], tn_spike[n]]), rm*np.array([0, gn_sfa[n]]), 'k:');
axs[1].set_title('Conductances');
axs[1].set_xlabel('time [ms]');
axs[1].set_ylabel('conductance [1/r_m]');

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

# Why does this work, ctd

The plots above, generated by the python code, display the mechanism behind spike-frequeny adaptation.

Upper plot shows equilibrium potential, $V_\infty$ (blue dashed), membrane potential $V_m$ (red), and threshold potential $V_\mathit{th}$ (red dotted). Spikes are also marked (black dotted).

Lower plot shows adaptive conductances $g_\mathit{sfa}$ (red) and leak conductance $g_L=1/r_m$ (blue).

This is how spike-frequency adaptation is obtained:

* Spikes elevate the SFA conductance (lower plot).
* This additional conductance lowers the equilibrium potential $V_\infty$  (upper plot).
* The membrane potential $V_m$  relaxes upwards towards $V_\infty$ (red curve, upper plot). When $V_\infty$ is lower, this upwards relaxation slows down.
* When $V_\infty$ drops below $V_\mathit{th}$, the neuron becomes 'refractory' and unable to produce a spike.
* Only when $g_\mathit{sfa}$ has decayed far enough (and $V_\infty$ has risen above $V_\mathit{th}$) can another spike be produced.

* As an example, we have added a $K^+$ conductance that is increased by spikes.

* Every additional time-varying conductances introduces another dynamical equation.

* In general, we need one dynamical equation for every independent state variable.

* In typical biological systems, many independent state variables interact with each other.

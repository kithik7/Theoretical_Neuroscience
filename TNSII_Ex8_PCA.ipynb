# Principal Component Analysis (PCA)

Principal component analysis is an extremely useful tool for analysing high-dimensional
data. It reduces dimensionality by identifying correlated sets of variables, so called ‘principal components’.

In this exercise, you will analyse multi-unit activity from a network of spiking neurons.
In such networks, quiescent periods with comparatively little activity are interrupted
spontaneously by transient periods of intense collective activity (‘bursts’). Your task is to
investigate the origin of these bursts. How does excitation spread through the network?
Is there a stereotypical propagation path? Or does each burst start in a unique way?

## Principal Component Analysis

Given $n$ observations of $m$ variables, each with zero mean, collected in a matrix $\textbf{X}$ where columns are observations and rows are variables. Each row has zero mean, which means the sum of each row is zero:

\\

$$
\textbf{X} =\left(  \begin{array}{cccc} x_{11} & x_{12} \ldots & x_{1n} \\ \vdots & \ddots & \ddots & \vdots \\ x_{m1} & x_{m2} & \ldots & x_{mn}  \end{array} \right) \in \mathbb{R}^{m \times n}, \qquad\qquad \sum_j \, x_{ij} = 0, \,\, \forall i
$$

\\

and the covariance matrix

\\

$$
\textbf{C}_{\textbf{X}} = \frac{1}{n-1} \, \textbf{X} \, \textbf{X}^T \in \mathbb{R}^{m \times m}
$$

\\

In order to calculate covariance, we use the formula:

$$
E\textbf{[X X]} - E[\textbf{X}]E[\textbf{X}]
$$

However, we determined that since each variable has a zero mean, the latter two terms
disappear because $E[\textbf{X}]$ is zero. Dividing by the number of variables gives us the mean
which is equivalent to the expected value. The term in front of the product is known as
Bessel’s correction since we divide by $n − 1$ which we should use because we are dealing
with sample variance/covariance, not population. By dividing with a slightly smaller
number, the result becomes slightly larger which we want since sample variance is usually
smaller than the true population variance since samples tend to fall on average closer to
the sample/population mean.

We now seek an orthonormal transformation $\textbf{P}  \in \mathbb{R}^{m \times m}$ such that the transformed observations have diagonal covariance:

\\

$$
\textbf{Y} = \textbf{P} \, \textbf{X}  \in \mathbb{R}^{m \times n}, \qquad\qquad \textbf{C}_{\textbf{Y}} = \frac{1}{n-1} \, \textbf{Y} \, \textbf{Y}^T = \left(  \begin{array}{cccc} y_{11} & 0 &\ldots & 0 \\ 0 & y_{22} & \ldots & \vdots \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \ldots & y_{mm} \end{array} \right)
$$

\\

From the 'skinny' singular value decomposition $\textbf{X}^T = \textbf{U} \textbf{S} \textbf{V}^T$, we obtain the left singular vectors $\textbf{U} \in  \mathbb{R}^{n \times n}$, the singular values  $\textbf{S} \in  \mathbb{R}^{n \times m}$, and the right singular vectors $\textbf{V}  \in \mathbb{R}^{m \times m}$.


The desired orthonormal transformation is

$$
\textbf{P} = \textbf{V}^T  \in \mathbb{R}^{m \times m}
$$

with the **rows** of $\textbf{P}$ being the **principal components**.  The variance captured by each component is revealed by diagonal matrix $\textbf{S}$.  Specifically, element $s_{ii}$ (the $i^{th}$ diagonal element) represents the variance captured by the $i^{th}$ principal component.

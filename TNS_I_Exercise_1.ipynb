# Nernst Equation

The Nernst equation describes the electrochemical equilibrium at an ion-permeable membrane. It holds when the potential difference across the membrane exactly balances the concentration difference of the permeating ion. Speciffcally, the equation relates reversal potential (defined as measured inside relative to outside) and the ion concentrations inside and outside:

\
$$\begin{eqnarray}
E_{rev} = \frac{RT}{zF} \ln \frac{[X]_{out}}{[X]_{in}}
\end{eqnarray}$$

\\

*   Gas constant: $R  = 8.31 $ (J K$^{-1}$ mol$^{-1}$)  
*   Ionic charge: $z$  positive  or negative
*   Body temperature: $T \approx  310$ (K)
*   RT: $\text{~}2500$ (J mol$^{-1}$)
*   Faraday constant: $F  = 96485$ (C mol$^{-1}$)

\\

At physiological temperatures, the constants may be merged and the equation simplified to:

\\
$$\begin{align*}
E_{rev} & = \frac{V_t}{z} \,  \ln \frac{[X]_{out}}{[X]_{in}}\\
V_t & = \frac{RT}{F} \approx 27\, mV
\end{align*}$$

## Driving forces

When the membrane potential does *not* equal the reversal potential, the ion in question is *not* at equilibrium.   Instead, the ion experiences a *driving force*.  As a result, more ions will cross the membrane in one direction than in another.

The driving force is approximately equal to the difference between the membrane potential $V_{mem}$&nbsp; and the reversal potential:

\\

\begin{equation*}
\Delta V = V_{mem} - E_{rev} = V_{mem} - \frac{27 \, mV}{z} \,  \ln \frac{[X]_{out}}{[X]_{in}}
\end{equation*}

## Goldman equation

The resting potential of a cell is a compromise between the reversal potentials of all ions to which the membrane is permeable.  This compromise depends on ion permeabilities.   More permeable ions have more influence than less permeable ones. The compromise is quantified in the **Goldman equation**:

\

\begin{equation*}
E_{rest} = 27 mV \,\, \ln \frac{P_K\,[K^+]_{out} + P_{Na} \, [Na^+]_{out} + P_{Cl}\,[Cl^-]_{in}}{P_K\,[K^+]_{in} + P_{Na} \, [Na^+]_{in} + P_{Cl}\,[Cl^-]_{out}}
\end{equation*}

\

Actually, its enough to know the relative permeabilities, for example, relative to $K^+$:


\begin{align*}
P_K\hspace{0.3cm} & =1\\
P_{Na}\hspace{0.3cm} & =b = 0.03\\
P_{Cl}\hspace{0.3cm} & =c = 0.1
\end{align*}

\

\begin{equation*}
E_{rest} = 27 mV \,\, \ln \frac{[K^+]_{out} + b \, [Na^+]_{out} + c \, [Cl^-]_{in}}{[K^+]_{in} + b \, [Na^+]_{in} + c \, [Cl^-]_{out}} \\
\end{equation*}

# **Neural, mental, and behavioral levels of description**

Animal and human behavior can be studied at different levels. A comprehensive understanding involves all levels.

1. Neural level: Physical and biological processes in the brain and nervous system. How do brains function at the level of neurons, synapses, neurotransmitters, and brain structures. Includes neuroanatomy, neurophysiology, neurochemistry, etc.

3. Behavioral level:

# **Sensation, perception, and representation**

Some concepts and distinctions are helpful in dicussing sensory processing and behavior (vision, audition, somatosensation, olfaction, gustation):

- **Sensation:** Initial detection of physical stimuli before they are organised or given meaning by higher mental functions. Visual sensation is the initial detection of photons in the retina. Auditory sensation the detection of sound waves in the cochlea. More generally, sensation implies that sensory stimuli have been transduced into neural activity, which is a necessary  condition for behavioral responses. "The snake sensed ground vibrations from the approaching footsteps."

- **Perception** (reptiles \& birds \& mammals): Transforms raw sensations into meaningful experiences. The process of becoming aware of, understanding, and recognizing the physical causes of sensory stimulation.  Perception produces the insight that guides behaviour. Key aspects of perception are 'selection', 'organization', and 'interpretation'.

- **Representation**: Internal models, maps, images, or summaries of the external world that the brain creates within itself. Representations are simplified and structured such as to help brains interact more efficiently with the external world.


*Ideally, one would like to balance mental and neural levels of description.*

# Four classes of feature detectors


With increasing RF size in the frog's optic tectum:

- Class 1: sustained contrast detector (RFs 2 deg - 4 deg)

- Class 2: convexity detector (RFs 3 deg - 5 deg)

- Class 3: moving edge detector (RFs 7 deg - 11 deg)

- Class 4: dimming detector (RFs 15 deg).

Class 1 and class 2 ganglion cells project to the optic tectum via unmyelinated fibers having conduction velocities of 20 to 50 centimeters per second. They make up 97 percent of the optic nerve (Maturana - 1959).


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec09_Gruesser_Fig_1.png" width="600">

While Classes 1, 3, and 4 show simple on-, on-off-, and off-responses,
the convexity detector (Class 2) aroused
strong and lasting interest in the vision community. This neuron responded best to a small dark stimulus of a convex shape moving in a stop-and-go fashion and therefore was dubbed **"bug detector".**

Neuroscientists describe the responses of sensory neurons and formulate hypotheses as to their contribution to behaviour.

# Tuning curve

Definition: A "tuning curve" is the dependence of the average response on one particular stimulus parameter.

A tuning curve $r=f(s)$ is the average number of spikes $r$ elicited by presentation of a particular stimulus attribute $s$.

A neuron's response typically depends on many different stimulus parameters.  A tuning curve characterizes the response as a function of just one of these parameters.  It is measured by repeatedly presenting stimuli with different parameter values.

A neuron has many different tuning curves (one for each parameter affecting its response).   Tuning curves describe only one aspect of a neuron's responsiveness.

(Slide-20)



# Gaussian tuning

Tuning for stimulus orientation in the primary visual cortex of monkey:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec09_Tuning_1.png" width="600">

Tuning curve is approximated by Gaussian function with center $s_\mathit{max}$, width $\sigma$, and height $r_\mathit{max}$:

\begin{eqnarray}
r = f(s) = r_{max} \, \exp\left[-\frac{1}{2} \left(\frac{s-s_{max}}{\sigma_s} \right)^2 \right]
\qquad\qquad{\rm Gaussian\,\, tuning }
\end{eqnarray}

(Slide-21)

# Cosine tuning

Tuning for direction of arm movement in the primary motor cortex of monkey:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec09_Tuning_2.png" width="600">

Tuning curve is approximated by cosine function with center $s_\mathit{max}$, maximal height $r_\mathit{max}$, and background activity $r_0$

\begin{eqnarray}
r = f(s) = r_0 + \left(r_{max}-r_0\right)\, \cos\left(s-s_{max} \right)
\qquad\qquad{\rm Cosine\,\, tuning }
\end{eqnarray}

(Slide-22)

# Sigmoidal tuning

Tuning for binocular disparity in the primary visual cortex of monkey (approximated by logistic function):

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec09_Tuning_3.png" width="600">

Tuning curve is approximated by sigmoidal function (here: logistic function) with center $s_{1/2}$, maximal height $r_\mathit{max}$, and maximal slope $r_0$

\begin{eqnarray}
r = f(s) = \frac{r_{max}}{1 + \exp\left(-\frac{s-s_{1/2}}{\Delta_s} \right)}
\qquad\qquad{\rm Sigmoidal\,\, tuning }
\end{eqnarray}

# Section summary


Some heuristic concepts have proven useful in describing the responsiveness and the function of sensory neurons in all species (drosophila, frogs, cats, ...)

- A **"receptive field"** is the area in sensory space (visual, auditory, somatosensory, ...) in which stimulation drives a neuron.


- **"Response variability"** refers to the variability of a neuron's response to repeated presentation of identical stimuli.

- A **"tuning curve"** is average response as a function of one particular stimulus parameters.

- Typically, there are many relevant stimulus parameters and, thus, many different "tuning curves".

- Tuning curves exhibit different shapes and are approximated by various suitable functions (including Gaussian distribution, cosine function, and logistic function).

# Rapid serial stimulation

For now, we consider only one-dimensional stimuli, that vary over time in only one dimension.  Examples are luminance, sound pressure, electric potential. Later, we will extend this approach to higher-dimensional stimuli, such as visual images or movie.

To avoid biasing the results, we randomly generate our stimulus sequence from some ensemble (e.g., uniform distribution, normal distribution).

To maximize data collection, we use a rapid sequence of stimulus values $s_i$ at successive times $t_i$. In other words, we 'bombard' the neuron with a rapid series of stimuli.

The code below illustrates "white noise" sequences with uniform and normal distribution, and "colored noise" with normal distribution.

(Slide-26)

```python
# @title Random stimulus sequences {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Illustrate random stimulus sequences
#

def PoissonSequence(Tend, Rate):
    # Tend [s]
    # Rate [Hz]
    # time interval dt = 1 / Rate [s]
    # Number of time points
    N = 2 * round(Tend * Rate);
    dti = -np.log(np.random.rand(1,N)) / Rate;
    ti_spike = np.cumsum(dti)
    # truncate spike sequence
    ix       = (ti_spike <= Tend);
    ti_spike = ti_spike[ix];
    return ti_spike

def WhiteNoiseUniform(Tend, Rate, Rmax):
    # Tend [s]
    # Rate [Hz]
    # time interval dt = 1 / Rate [s]
    N     = round(Tend * Rate ) + 1;
    si    = Rmax * ( 2 * np.random.rand(N) - 1 );
    return si

def WhiteNoiseNormal(Tend, Rate, Mu, Sigma):
    # Tend [s]
    # Rate [Hz]
    # time interval dt = 1 / Rate [s]
    N     = round(Tend * Rate ) + 1;
    si    = np.random.normal(Mu, Sigma, N);
    return si

def ColoredNoiseNormal(Tend, Rate, Mu, Sigma, Tau):
    # Tend [s]
    # Rate [Hz]
    # time interval
    dt = 1 / Rate;        # [s]
    N     = round(Tend * Rate ) + 1;
    si    = np.zeros(N);
    si[0] = np.random.normal(Mu, Sigma);
    eta_i = np.random.normal(0, 1, N);

    efac1 = 1 - np.exp(-dt/Tau);
    efac2 = np.sqrt( 1 - np.exp(-dt/Tau));

    for i in range(0,N-1):
        si[i+1] = si[i] + efac1 * (Mu - si[i]) + Sigma * efac2 * eta_i[i];

    return si

def SmoothedNoiseNormal(Tend, Rate, Mu, Sigma, Tau):
    # Tend [s]
    # Rate [Hz]
    # time interval
    dt = 1 / Rate;        # [s]
    N     = round(Tend * Rate ) + 1;
    si    = np.zeros(N);
    si[0] = np.random.normal(Mu, Sigma);
    eta_i = np.random.normal(0, 1, N);

    efac1 = 1 - np.exp(-dt/Tau);
    efac2 = np.sqrt( 1 - np.exp(-dt/Tau));

    for i in range(0,N-1):
        si[i+1] = si[i] + efac1 * (Mu - si[i]) + Sigma * efac2 * eta_i[i];

    B = 5;
    sik = np.zeros((N,B));
    for k in range(0,B):
        sik[:,k] = np.roll(si,k);

    si = np.mean(sik, axis=1);

    return si

# Define an interactive plot
def plot_random_stimulus_sequences(Tend = 1.0, Rate = 100):
    """(Slide-27)"""

    # parameters
    Rmax  = 10;
    Mu    = 3;
    Sigma = 2;
    Tau   = 0.05;

    #
    dt  = 1/Rate;                # time step, in s
    Ni  = round(Tend*Rate) + 1;    # number of time steps
    ti  = np.linspace(0, Tend, Ni); # time vector, with successive time steps from 0 to T

    # allocate vectors for variables to be plotted (membrane potential, steady-state potential, activation variables, currents)

    si_WN_unif   = WhiteNoiseUniform(Tend, Rate, Rmax);
    si_WN_norm   = WhiteNoiseNormal(Tend, Rate, Mu, Sigma);
    si_CN_norm   = ColoredNoiseNormal(Tend, Rate, Mu, Sigma, Tau);
    si_SN_norm   = SmoothedNoiseNormal(Tend, Rate, Mu, Sigma, Tau);

    # Create a figure and a grid of subplots (3 rows, 1 columns)
    fig, axs = plt.subplots(4, 1, figsize=(8, 8))

    msz = 2;

    # First subplot
    axs[0].plot((0,Tend), (Rmax,Rmax), 'b--');
    axs[0].plot((0,Tend), (-Rmax,-Rmax), 'b--');
    axs[0].plot(ti, si_WN_unif, 'bo', markersize=msz);
    axs[0].set_title('White noise (uniform)', fontsize=18);
    axs[0].set_xlabel('time [ms]', fontsize=14);
    axs[0].set_ylabel('stimulus', fontsize=14);

    # Second subplot
    axs[1].plot((0,Tend), (Mu,Mu), 'r--');
    axs[1].plot((0,Tend), (Mu+Sigma,Mu+Sigma), 'r:');
    axs[1].plot((0,Tend), (Mu-Sigma,Mu-Sigma), 'r:');
    axs[1].plot(ti, si_WN_norm, 'ro', markersize=msz);
    axs[1].set_title('White noise (normal)', fontsize=18);
    axs[1].set_xlabel('time [ms]', fontsize=14);
    axs[1].set_ylabel('stimulus', fontsize=14);

    # Third subplot
    axs[2].plot((0,Tend), (Mu,Mu), 'm--');
    axs[2].plot((0,Tend), (Mu+Sigma,Mu+Sigma), 'm:');
    axs[2].plot((0,Tend), (Mu-Sigma,Mu-Sigma), 'm:');
    axs[2].plot(ti, si_CN_norm, 'mo', markersize=msz);
    axs[2].set_title('Colored noise (normal)', fontsize=18);
    axs[2].set_xlabel('time [ms]', fontsize=14);
    axs[2].set_ylabel('stimulus', fontsize=14);

    # Fourth subplot
    axs[3].plot((0,Tend), (Mu,Mu), 'c--');
    axs[3].plot((0,Tend), (Mu+Sigma,Mu+Sigma), 'c:');
    axs[3].plot((0,Tend), (Mu-Sigma,Mu-Sigma), 'c:');
    axs[3].plot(ti, si_SN_norm, 'co', markersize=msz);
    axs[3].set_title('Smoothed noise (normal)', fontsize=18);
    axs[3].set_xlabel('time [ms]', fontsize=14);
    axs[3].set_ylabel('stimulus', fontsize=14);


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_random_stimulus_sequences, Tend = (1.0,50.0,1.0), Rate = (10,500,10))

```

# Statistics of stimulus sequences

The statistics of a stimulus sequence $s_i$ is described by mean and variance:
$\begin{eqnarray}
\left\langle s_i \right\rangle = \frac{1}{N} \sum_{i=1}^N s_i \equiv \mu_s,
\qquad
\left\langle \left(s_i - \left\langle s_i \right\rangle\right)^2 \right\rangle = \frac{1}{N} \sum_{i=1}^N \left(s_i-\left\langle s_i \right\rangle\right)^2  \equiv \sigma^2_s,
\end{eqnarray}$

These basic statistics disregard the temporal order of presentation! Their values would remain the same if the order or presentation were scrambled!

A complementatry statistic that captures regularities in presentation order is the "lagged autocorrelation"

$\begin{eqnarray}
R_n = \frac{1}{\sigma_s^2} \left\langle\left(s_i - \left\langle s_i \right\rangle\right)\left(s_{i+n} - \left\langle s_{i+n} \right\rangle\right)\right\rangle
\end{eqnarray}$

the 'correlation coefficient' between the time series and its shifted (lagged) copy.

In other words, this is the average product of deviations (from the mean) at two time points separated by $n$ (lag $n$).

(Slide-28)

```python
from inspect import signature
# @title Lagged autocorrelations {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Illustrate lagged autocorrelation
#

def LaggedAutocorrelation( s_i, Lags_k ):

    R_k    = np.zeros(len(Lags_k));

    Kmax   = np.max(Lags_k);

    Ni     = len(s_i);

    s_i    = s_i - np.mean(s_i);

    sigma2 = np.mean(s_i**2);

    for k in range(0,len(Lags_k)):

        s1i = s_i;
        s2i = np.roll(s_i, Lags_k[k])

        #s1i = s_i[k:Ni-2*Kmax+k];
        #s2i = s_i[Kmax:Ni-Kmax];

        R_k[k] = np.mean( s1i * s2i ) / sigma2;

    return R_k

# Define an interactive plot
def plot_lagged_autocorrelation(Tend = 1.0, Rate = 100):
    """(Slide-29)"""

    # parameters
    Rmax  = 10;
    Mu    = 3;
    Sigma = 2;
    Tau   = 0.5;
    Lagtime = 0.1;

    #
    dt  = 1/Rate;                # time step, in s
    Ni  = round(Tend*Rate) + 1;    # number of time steps
    ti  = np.linspace(0, Tend, Ni); # time vector, with successive time steps from 0 to T

    # allocate vectors for variables to be plotted (membrane potential, steady-state potential, activation variables, currents)

    si_WN_unif   = WhiteNoiseUniform(Tend, Rate, Rmax);
    si_WN_norm   = WhiteNoiseNormal(Tend, Rate, Mu, Sigma);
    si_CN_norm   = ColoredNoiseNormal(Tend, Rate, Mu, Sigma, Tau);
    si_SN_norm   = SmoothedNoiseNormal(Tend, Rate, Mu, Sigma, Tau);

    #si_CN_norm   = np.cos(2 * np.pi * ti);

    si = si_CN_norm;

    si = si - np.mean(si);

    sigma2 = np.mean( (si[7:Ni])**2 );

    # allocate vectors for variables to be plotted (membrane potential, steady-state potential, activation variables, currents

    # allocate vectors for variables to be plotted (membrane potential, steady-state potential, activation variables, currents)
    Kmax   = round( Lagtime*Rate ) + 1;
    Lags_k = np.arange(-Kmax, Kmax + 1)
    T_k    = Lags_k / Rate;

    R_WN_unif_k = LaggedAutocorrelation(si_WN_unif, Lags_k);
    R_WN_norm_k = LaggedAutocorrelation(si_WN_norm, Lags_k);
    R_CN_norm_k = LaggedAutocorrelation(si_CN_norm, Lags_k);
    R_SN_norm_k = LaggedAutocorrelation(si_SN_norm, Lags_k);

    # Create a figure and a grid of subplots (3 rows, 1 columns)
    fig, axs = plt.subplots(4, 1, figsize=(3, 12))

    msz = 2;

    # First subplot
    axs[0].plot(T_k, R_WN_unif_k, 'bo', markersize=msz);
    axs[0].set_title('White noise (uniform)', fontsize=14);
    axs[0].set_xlabel('lag [s]', fontsize=14);
    axs[0].set_ylabel('cc', fontsize=14);

    # Second subplot
    axs[1].plot(T_k, R_WN_norm_k, 'ro', markersize=msz);
    axs[1].set_title('White noise (normal)', fontsize=14);
    axs[1].set_xlabel('lag [s]', fontsize=14);
    axs[1].set_ylabel('cc', fontsize=14);

    # Third subplot
    axs[2].plot(T_k, R_CN_norm_k, 'mo', markersize=msz);
    axs[2].set_title('Colored noise (normal)', fontsize=14);
    axs[2].set_xlabel('lag [s]', fontsize=14);
    axs[2].set_ylabel('cc', fontsize=14);

    # Fourth subplot
    axs[3].plot(T_k, R_SN_norm_k, 'co', markersize=msz);
    axs[3].set_title('Smoothed noise (normal)', fontsize=14);
    axs[3].set_xlabel('lag [s]', fontsize=14);
    axs[3].set_ylabel('cc', fontsize=14);


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_lagged_autocorrelation, Tend = (1.0,50.0,1.0), Rate = (10,500,10))

```

# Spike-triggered average (of the stimulus)

The weakly electric fish Eigenmannia is extremely sensitive to rapid changes in the electric potential changes of the surrounding sweet water. Somatosensory neurons (lateral line lobe) typically respond to a particular wave form of electric potential.

The illustration shows a short stretch of a time-modulated electric potential and the spikes evoked during this stretch (left, top and bottom), as well as the average time-modulation that preceeded a spike (right), where the average was taken over a much longer sequence.  

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec09_STA_DA_Fig1-19.png" width="600">
D&A Fig. 1-9

The **spike-triggered average (STA)** of a stimulus $\left\langle{s(-\tau)}\right\rangle$ is the average value of the stimulus at some time $\tau$ before a spike, where $0\leq \tau \leq \tau_\mathit{max}$:

$\begin{eqnarray}
{\it Spike \,\, times} &\qquad\qquad t_i, \qquad i\in\{1, \ldots , N\}
\\
\\
{\it Stimulus \,\, function} &\qquad\qquad s(t)
\\
\\
{\it reversed \,\, and \,\, shifted} &\qquad\qquad s(t_i-\tau)
\\
\\
{\it averaged} &\qquad\qquad \left\langle s(-\tau)\right\rangle_\mathit{STA} \equiv \frac{1}{N} \sum_{i=1}^{N} \, s(t_i-\tau)
\end{eqnarray}$


Note that $s(t_i-\tau)$ is the stimulus stretching backwards in time from $t_i$, as $\tau$ grows from $0$ to $\tau_\mathit{max}$.

(Slide-30)

```python
from inspect import signature
# @title STA demonstration {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Illustrate spike-triggered stimulus average
#

def SpikingResponse( t_i, s_i, Kernel_j ):

    # sizes
    Ni       = len(s_i);
    Nj       = len(Kernel_j);

    # allocate
    cnvlv_i  = np.zeros(Ni);

    # zero mean
    s_i      = s_i - np.mean(s_i);
    Kernel_j = Kernel_j - np.mean(Kernel_j);

    # linear response (convolution of stimulus and kernel)

    for i in range(Nj,Ni):

        cnvlv_i[i] = np.dot( Kernel_j, s_i[i-Nj:i] );

    # determine spike threshold (largest 1%)
    q = 0.99;
    spike_threshold = np.quantile(cnvlv_i, q);

    # threshold to generate spikes
    spk_i = (cnvlv_i > spike_threshold);

    # extract list of spike times
    spk_idx = np.where(spk_i == 1)[0];

    # typically, the convolution produces large values
    # at several successive time points. For every
    # such period, we want to generate exactly one spike.
    # To achieve this, we thin out list of spike times
    cnt_idx = np.zeros_like(spk_idx);
    new_idx = np.zeros_like(spk_idx)
    cnt = 0;
    for k in range(0,len(spk_idx)-1):
        if spk_idx[k+1] - spk_idx[k] == 1:
            cnt_idx[k+1] = cnt_idx[k] + 1;
        else:
            if cnt_idx[k] > 0:
                kback = round( cnt_idx[k] / 2 );
                new_idx[cnt] = spk_idx[k-kback];
                cnt = cnt + 1;
    new_idx = new_idx[:cnt];

    # extract thinned out list of spike times
    t_spk = t_i[ new_idx];

    return cnvlv_i, t_spk

def MakeSomeKernel( t_j, LambdaMin, LambdaMax, SigmaMin, SigmaMax ):

    # generate a Gabor kernel with randomly chosen
    # period Lambda, duration Sigma, phase Phi, and sign Sign.
    Lambda = LambdaMin + (LambdaMax - LambdaMin) * np.random.rand(1);
    Sigma  = SigmaMin  + (SigmaMax  - SigmaMin ) * np.random.rand(1);
    Phi    = 2 * np.pi * np.random.rand(1);
    Sign   = 2 * np.random.randint(0,2) - 1;

    Kernel_j = Sign * np.exp( -t_j**2 / (2*Sigma**2) ) * np.cos( 2 * np.pi * t_j / Lambda + Phi );

    return Kernel_j

def ExctractStimulusEpisodes( t_i, s_i, t_spk, Twindow ):

    Nwindow = np.sum( t_i <= Twindow );
    Nspike  = len(t_spk);

    # allocate
    sepsd_kj = np.zeros((Nspike, Nwindow));

    # loop over spikes
    for k in range(0,len(t_spk)):

        idx = np.where( ((t_spk[k]-Twindow)<=t_i) & (t_i <= t_spk[k]) )[0];
        sepsd_kj[k,0:len(idx)] = s_i[idx];

    STA_j = np.mean( sepsd_kj, axis=0);

    return STA_j, sepsd_kj


# Define an interactive plot
def plot_spike_triggered_stimulus_average(Tend = 100.0):
    """(Slide-31)"""

    # stimulus parameters
    Rmax  = 10;
    Mu    = 0;
    Sigma = 2;
    Tau   = 0.5;
    Rate  = 20;
    Lagtime = 0.1;

    # timing parameters
    dt  = 1/Rate;                   # time step, in s
    Ni  = round(Tend*Rate) + 1;     # number of time steps
    ti  = np.linspace(0, Tend, Ni); # time vector, with successive time steps from 0 to T

    # generate different types of stimulus sequences

    si_WU   = WhiteNoiseUniform(Tend, Rate, Rmax);
    si_WN   = WhiteNoiseNormal(Tend, Rate, Mu, Sigma);
    si_CN   = ColoredNoiseNormal(Tend, Rate, Mu, Sigma, Tau);

    # generate some kernel function

    LambdaMin = 2.0;
    LambdaMax = 3.0;
    SigmaMin  = 1.5;
    SigmaMax  = 2.5;

    Tkernel   = 10.0;
    Nj        = round(Tkernel*Rate) + 1;
    tj        = np.linspace(0, Tkernel, Nj);

    Kernelj   = MakeSomeKernel( tj, LambdaMin, LambdaMax, SigmaMin, SigmaMax );

    # get spiking responses

    (cnvlvi_WU, tspk_WU) = SpikingResponse( ti, si_WU, Kernelj );
    (cnvlvi_WN, tspk_WN) = SpikingResponse( ti, si_WN, Kernelj );
    (cnvlvi_CN, tspk_CN) = SpikingResponse( ti, si_CN, Kernelj );

    print( len(tspk_CN) )

    # extract stimulus sequences that preceded each spike

    Twindow = Tkernel;

    (STA_j, sepsd_kj) = ExctractStimulusEpisodes( ti, si_CN, tspk_CN, Twindow );

    print( np.shape( sepsd_kj ) )

    # Create a figure and a grid of subplots (1 rows, 2 columns)
    fig, axs = plt.subplots(1, 2, figsize=(6, 3))

    msz = 0.5;

    # First subplot
    axs[0].plot(-tj, Kernelj, 'bo', markersize=msz);
    axs[0].set_title('Ground truth', fontsize=18);
    axs[0].set_xlabel('lag [s]', fontsize=14);
    axs[0].set_ylabel('stimulus', fontsize=14);
    axs[0].set_ylim([-1, 1]);
    axs[0].set_box_aspect(1/1);

    # Second subplot
    for k in range(1,len(tspk_CN)):
        axs[1].plot(-tj, sepsd_kj[k,:], 'b-', linewidth=0.1);
    axs[1].plot(-tj, STA_j, 'ro', markersize=msz);
    axs[1].set_title('STA', fontsize=18);
    axs[1].set_xlabel('lag [s]', fontsize=14);
    axs[1].set_ylim([-2.5, 2.5]);
    axs[1].set_box_aspect(1/1);

    fig, axs = plt.subplots(1, 1, figsize=(6, 3))

    # Third subplot
    axs.plot(ti, si_CN, 'ko', markersize=msz);
    for k in range(1,len(tspk_CN)):
        tk = tspk_CN[k];
        axs.plot((tk, tk), (-2.5,2.5), 'r-', linewidth=0.5);
        axs.plot( tk, 2.5, 'ro', markersize=4*msz);

    axs.set_title('Stimulus and Spikes', fontsize=18);
    axs.set_xlabel('time [s]', fontsize=14);
    axs.set_ylabel('response', fontsize=14);
    axs.set_box_aspect(1/4);


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_spike_triggered_stimulus_average, Tend = (100.0,2000.0,100.0))

```

# Response-weighted average (of the stimulus)

When measuring firing rates rather than individual spikes, one may analogously compute the **response-weighted average stimulus**.

For a continuously time-varying stimulus $s(t)$ and responses $r_i$ recorded periodically at times $t_i$, we can write

$\begin{eqnarray}
\left\langle s(-\tau)\right\rangle_\mathit{RWA} \equiv \frac{1}{N \left\langle r_i \right\rangle} \sum_{i=1}^{N} \, r_i \, s(t_i - \tau), \qquad\qquad \left\langle r_i \right\rangle = \frac{1}{N} \sum_{i=1}^{N} \, r_i
\end{eqnarray}$

where $s(t_i - \tau)$ is the stimulus at an earlier time $t_i-\tau$ and $\left\langle r_i \right\rangle$ is the average response.

(Slide-32)


```python
from inspect import signature
# @title RWA demonstration {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Illustrate spike-triggered stimulus average
#

def ContinuousResponse( t_i, s_i, Kernel_j ):

    # sizes
    Ni       = len(s_i);
    Nj       = len(Kernel_j);

    # allocate
    cnvlv_i  = np.zeros_like(s_i);

    # zero mean
    s_i      = s_i - np.mean(s_i);
    Kernel_j = Kernel_j - np.mean(Kernel_j);

    # linear response (convolution of stimulus and kernel)

    for i in range(Nj,Ni):

        cnvlv_i[i] = np.dot( Kernel_j, s_i[i-Nj:i] );

    # determine inflection of nonlinearity (largest 20%)
    q = (0.2,0.8);
    beta = np.quantile(cnvlv_i, q);

    kappa = 5/(beta[1]-beta[0]);

    # exponent
    expnnt_i = -kappa * (cnvlv_i - beta[1]);
    expnnt_i[ np.where(expnnt_i < -5)[0] ] = -5;
    expnnt_i[ np.where(expnnt_i > +5)[0] ] = +5;

    # static non-linearity
    r_i = 1 / ( 1 + np.exp( expnnt_i ) );

    return r_i


# Define an interactive plot
def plot_response_weighted_stimulus_average(Tend = 100.0):
    """(Slide-33)"""

    # stimulus parameters
    Rmax  = 10;
    Mu    = 0;
    Sigma = 2;
    Tau   = 0.5;
    Rate  = 20;
    Lagtime = 0.1;

    # timing parameters
    dt  = 1/Rate;                   # time step, in s
    Ni  = round(Tend*Rate) + 1;     # number of time steps
    ti  = np.linspace(0, Tend, Ni); # time vector, with successive time steps from 0 to T

    # generate different types of stimulus sequences

    si_WU   = WhiteNoiseUniform(Tend, Rate, Rmax);
    si_WN   = WhiteNoiseNormal(Tend, Rate, Mu, Sigma);
    si_CN   = ColoredNoiseNormal(Tend, Rate, Mu, Sigma, Tau);

    # generate some kernel function

    LambdaMin = 2.0;
    LambdaMax = 3.0;
    SigmaMin  = 1.5;
    SigmaMax  = 2.5;

    Tkernel   = 10.0;
    Nj        = round(Tkernel*Rate) + 1;
    tj        = np.linspace(0, Tkernel, Nj);

    Kernelj   = MakeSomeKernel( tj, LambdaMin, LambdaMax, SigmaMin, SigmaMax );

    # get continuous responses

    ri_WU = ContinuousResponse( ti, si_WU, Kernelj );
    ri_WN = ContinuousResponse( ti, si_WN, Kernelj );
    ri_CN = ContinuousResponse( ti, si_CN, Kernelj );

    # extract stimulus sequences that preceded each spike

    Twindow = Tkernel;

    (M_j, s_kj) = ExctractStimulusEpisodes( ti, si_CN, ti[Nj:Ni], Twindow );

    # response weighted average
    rs_kj = np.zeros_like(s_kj);
    for k in range(Nj,Ni):
        rs_kj[k-Nj,:] = s_kj[k-Nj,:] * ri_CN[k];

    Rmean = np.mean( ri_CN, axis=0 );
    RWA_j = np.nanmean(rs_kj, axis=0) / Rmean;

    # Create a figure and a grid of subplots (1 rows, 2 columns)
    fig, axs = plt.subplots(1, 2, figsize=(6, 3))

    msz = 0.5;

    # First subplot
    axs[0].plot(-tj, Kernelj, 'bo', markersize=msz);
    axs[0].set_title('Ground truth', fontsize=18);
    axs[0].set_xlabel('lag [s]', fontsize=14);
    axs[0].set_ylabel('stimulus', fontsize=14);
    axs[0].set_ylim([-1, 1]);
    axs[0].set_box_aspect(1/1);

    # Second subplot
    axs[1].plot(-tj, RWA_j, 'ro', markersize=msz);
    axs[1].set_title('RWA', fontsize=18);
    axs[1].set_xlabel('lag [s]', fontsize=14);
    axs[1].set_ylim([-1, 1]);
    axs[1].set_box_aspect(1/1);

    fig, axs = plt.subplots(1, 1, figsize=(6, 3))

    # Third subplot
    axs.plot(ti, si_CN, 'k-', linewidth=0.5);
    axs.plot(ti, ri_CN, 'ro', markersize=msz);

    axs.set_title('Stimulus and response', fontsize=18);
    axs.set_xlabel('time [s]', fontsize=14);
    axs.set_ylabel('response', fontsize=14);
    axs.set_box_aspect(1/4);


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_response_weighted_stimulus_average, Tend = (100.0,2000.0,100.0))

```

# Response-weighted average (of stimulus **patterns**)

For continuously time-varying **stimulus patterns** $s(x,y,t)$ and responses $r_i$ recorded periodically at times $t_i$, we can similarly write

$\begin{eqnarray}
\left\langle s(x,y,-\tau)\right\rangle_\mathit{RWA} \equiv \frac{1}{N \left\langle r_i \right\rangle} \sum_{i=1}^{N} \, r_i \, s(x,y,t_i - \tau), \qquad\qquad \left\langle r_i \right\rangle = \frac{1}{N} \sum_{i=1}^{N} \, r_i
\end{eqnarray}$

where $s(x,y,t_i - \tau)$ is the stimulus pattern at an earlier time $t_i-\tau$ and $\left\langle r_i \right\rangle$ is the average response.

```python
from inspect import signature
# @title Random stimulus patterns {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Generate random stimulus patterns
#

def OnSpot(n):
    """
    Returns a stimulus array of size n x n with a bright spot.
    The spot's position (x0, y0) is random.
    """
    # Initialize n x n array with zeros
    S = np.zeros((n, n))

    # Define x and y arrays, centered around 0
    x = -0.5 * (n + 1) + np.arange(1, n + 1)
    y = -0.5 * (n + 1) + np.arange(1, n + 1)

    # Create meshgrid for X and Y
    X, Y = np.meshgrid(x, y)

    # Random spot position within a range scaled by 0.8
    x0 = 0.8 * (np.random.rand() - 0.5) * n
    y0 = 0.8 * (np.random.rand() - 0.5) * n

    # Calculate the stimulus array with a Gaussian-like bright spot
    S = np.exp(-0.5 * ((X - x0) ** 2 + (Y - y0) ** 2))

    return S

def OffSpot(n):
    """
    Returns a stimulus array of size n x n with a dark spot.
    The spot's position (x0, y0) is random.
    """
    # Initialize an n x n array with zeros
    S = np.zeros((n, n))

    # Define x and y arrays, centered around 0
    x = -0.5 * (n + 1) + np.arange(1, n + 1)
    y = -0.5 * (n + 1) + np.arange(1, n + 1)

    # Create meshgrid for X and Y
    X, Y = np.meshgrid(x, y)

    # Random spot position within a range scaled by 0.8
    x0 = 0.8 * (np.random.rand() - 0.5) * n
    y0 = 0.8 * (np.random.rand() - 0.5) * n

    # Calculate the stimulus array with a dark spot
    S = 1 - np.exp(-0.5 * ((X - x0) ** 2 + (Y - y0) ** 2))

    return S

def OnBar(n, theta):
    """
    Returns a stimulus array of size n x n with a bright bar of orientation theta.
    The bar's position is random.
    """
    # Initialize an n x n array with zeros
    S = np.zeros((n, n))

    # Create meshgrid for X and Y
    X, Y = np.meshgrid(-0.5 * (n + 1) + np.arange(1, n + 1),
                       -0.5 * (n + 1) + np.arange(1, n + 1))

    # Compute the stimulus array with a bright bar of random position and given orientation
    S = np.exp(-0.5 * (np.cos(theta) * X + np.sin(theta) * Y + 0.1 * n * np.random.randn()) ** 2)

    return S

def OffBar(n, theta):
    """
    Returns a stimulus array of size n x n with a dark bar of orientation theta.
    The bar's position is random.
    """
    # Initialize an n x n array with zeros
    S = np.zeros((n, n))

    # Create meshgrid for X and Y
    X, Y = np.meshgrid(-0.5 * (n + 1) + np.arange(1, n + 1),
                       -0.5 * (n + 1) + np.arange(1, n + 1))

    # Compute the stimulus array with a dark bar of random position and given orientation
    S = -np.exp(-0.5 * (np.cos(theta) * X + np.sin(theta) * Y + 0.1 * n * np.random.randn()) ** 2)

    return S

def WhiteNoise(n):
    """
    Produces a white noise array of size n x n, where each value
    is normally distributed with mean zero and variance one.
    """
    s = np.random.randn(n, n)  # Generate n x n array of white noise
    return s


# Define an interactive plot
def plot_show_random_stimuli(Nrepeat = 1):
    """(Slide-38)"""

    # size of stimulus patterns
    Nsize = 50;

    for i in range(Nrepeat):

        # Create a 2x3 grid of subplots
        fig, axs = plt.subplots(2, 3, figsize=(9, 6))

        # Plot ON Spot
        S = OnSpot(Nsize);
        axs[0, 0].pcolor(S, cmap='gray', shading='auto');
        axs[0, 0].axis('square');
        axs[0, 0].set_title('ON Spot', fontsize=14);
        # plt.pause(0.2);

        # Plot OFF Spot
        S = OffSpot(Nsize);
        axs[0, 1].pcolor(S, cmap='gray', shading='auto');
        axs[0, 1].axis('square');
        axs[0, 1].set_title('OFF Spot', fontsize=14);
        # plt.pause(0.2);

        # Plot ON Bar with random orientation
        theta = np.random.rand() * np.pi
        S = OnBar(Nsize, theta);
        axs[0, 2].pcolor(S, cmap='gray', shading='auto');
        axs[0, 2].axis('square');
        axs[0, 2].set_title('ON Bar', fontsize=14);
        # plt.pause(0.2);

        # Plot OFF Bar with random orientation
        theta = np.random.rand() * np.pi
        S = OffBar(Nsize, theta);
        axs[1, 0].pcolor(S, cmap='gray', shading='auto');
        axs[1, 0].axis('square');
        axs[1, 0].set_title('OFF Bar', fontsize=14);
        # plt.pause(0.2);

        # Plot White Noise
        S = WhiteNoise(Nsize);
        axs[1, 1].pcolor(S, cmap='gray', shading='auto');
        axs[1, 1].axis('square');
        axs[1, 1].set_title('White Noise', fontsize=14);
        # plt.pause(0.2);

        axs[1, 2].axis('square');
        axs[1, 2].set_xlim([0, 1]);
        axs[1, 2].set_ylim([0, 1]);


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_show_random_stimuli, Nrepeat = (1,20,1))

```

**Abstract:**

Today we apply signal-detection theory to the analysis of **variable neuronal responses**, following the classic work of Shadlen, Newsome, and Movshon on the **discrimination of visual motion** by macaques.

To prepare, we analyse the **two-alternative force-choice task**, which is a particularly useful experimental paradigm.  Next, we explain the animals' **psychometric function** with STD and then apply the same principles to obtain **neurometric functions** from the response distributions of individual neurons. For average neurons in visual cortical area MT, the two measures of sensory information coincide.

This is surprising, because one would naively have expected populations of neurons to be more informative than individual neurons. Apparently, the brain uses a widely distributed, but highly redundant code that supports recovering most sensory information from smaller subpopulations.

We conclude with an SDT analysis of idealized **populations of correlated neurons**.

# **1. Two-alternative forced-choice**

Criterion bias can be minimized with the right experimental design. In a **two-alternative forced-choice** design, both stimulus alternatives are presented on every trial and the observer must decide which stimulus is which.

2AFC paradigms are realized either spatially (two alternatives presented at different locations) or temporally (two alternatives presented at different times).


The advantages are

- Criterion bias is minimized
- ROC analysis is simplified
- Robust results obtained with fewer trials

(Slide-2)


# Example: Spatial frequency discrimination

Discriminating higher- and lower frequency Gabor patches, a **two-alternative** version presents both patches and the observer reports on every trial whether the lower-frequency patch is **left** or **right**.  Showing two patches facilitates the task objectively (more information).  Subjectively, it helps maintain a neutral decision criterion (favouring neither one, nor the other).

# Analytical result

A two-alternative forced-choice experiment produces an interesting analytical result.

Assume that the **'high'** stimulus elicits a neural response $r$ and the **'low'** stimulus a response $r'$.  The proportion of correct responses is the probability $P(r>r')$, or

$\begin{eqnarray}
P_\mathit{correct} &=& \int_{-\infty}^{+\infty} P_\mathit{high}(r)  \,  \int_{-\infty}^{r} P_\mathit{low}(r')  \, dr' \, dr
\end{eqnarray}$

Comparing the 'high' distribution to a criterion $c$ provides the number of hits

$\begin{eqnarray}
H(c) = \int_{c}^{\infty} P_\mathit{high}(r')  \, dr' \qquad \Rightarrow \qquad \frac{dH(c)}{dc} = -P_\mathit{high}(c)
\end{eqnarray}$

Similarly, comparing the 'low' distribution provides the number of false alarms
$\begin{eqnarray}
F(c) = \int_{c}^{\infty} P_\mathit{low}(r')  \, dr' \qquad \Rightarrow \qquad 1 - F(c) = \int_{-\infty}^{c} P_\mathit{low}(r')  \, dr'
\end{eqnarray}$

Accordingly, we can substitute

$\begin{eqnarray}
P_\mathit{correct} &=& \int_{-\infty}^{+\infty} P_\mathit{high}(c)  \,  \int_{-\infty}^{c} P_\mathit{low}(r')  \, dr' \, dc =
\\
&=& - \int_{-\infty}^{+\infty} \,  dH(r) \, \left[1-F(r)  \right]  =
\\
&=& - \int_{1}^{0} [1 - F] \, dH = \int_0^1 [1-F] \, dH
\end{eqnarray}$

We find that the area under the **one-alternative** ROC equals the proportion correct of the **two-alternative** forced choice task.

(Slide-8)




# Area under ROC curve

Thus, the area under the ROC obtained from $P_\mathit{high}$ and $P_\mathit{low}$ predicts the proportion correct in a two-alternative forced-choice task.

$\begin{eqnarray}
P_{correct} = \int_{0}^{1} \,  \left[ 1 - F(H) \right]  \, dH
\end{eqnarray}$

(Slide-12)

# Formulate SDT

We assume that neural responses $r$ are distributed normally, with either $P_\mathit{pos}(r)$ or $P_\mathit{neg}(r)$, depending on the direction of motion (positive or negative).

We further assume that the two distribution means are shifted left and right of zero, by an amount that increases linearly with dot coherence $c$:

$\begin{eqnarray}
\qquad\qquad \left\langle r_\mathit{neg} \right\rangle = -k\,c, \qquad\qquad \left\langle r_\mathit{pos} \right\rangle = +k \, c
\end{eqnarray}$

where $k$ is a proportionality constant (to be fitted).

Finally, we assume that there may be a criterion bias in that the criterion may be shifted left or right of zero by an amount $\theta$ (also to be fitted).


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec13_RDM_psychometric.png" width="600">

(Slide-13)

# Displacements, sensitivity, and proportion correct

With these assumptions, the predicted displacements of the distributions relative to the criterion are

$\begin{eqnarray}
z(H) = kc-\theta, \qquad\qquad z(F) = -kc-\theta
\end{eqnarray}$

and the predicted sensitivity $d'$ is

$\begin{eqnarray}
\qquad\qquad d' =  z(H) - z(F) = 2 \, k\, c
\end{eqnarray}$

The predicted fractions of hits and false alarms are

with
$\begin{eqnarray}
H(c) =  \frac{1}{2} + \frac{1}{2} \, \mathit{erf}\left( \frac{kc-\theta}{\sqrt{2}}\right),\qquad\qquad
F(c) =  \frac{1}{2} + \frac{1}{2} \, \mathit{erf}\left( \frac{-kc-\theta}{\sqrt{2}}\right)
\end{eqnarray}$

and the predicted proportion correct is

$\begin{eqnarray}
P_{correct}(c) = \frac{H(c)}{2} + \frac{1-F(c)}{2}
\end{eqnarray}$



```python
from inspect import signature
# @title Psychometric function {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def SDTposneg( c, k, theta):

    mean_pos = k*c - theta;
    mean_neg = -k*c - theta;

    H = norm.cdf( mean_pos/np.sqrt(2) );
    F = norm.cdf( mean_neg/np.sqrt(2) );

    Pcorr = H/2 + (1-F)/2;

    return H, F, Pcorr


def plot_psychometric_function( k = 0.1, theta=0.1):
    """(Slide-14)"""

    # stepsize for coherence
    dc = 1;

    # range of coherence
    ci = np.arange(1,101,dc);

    logci = np.log10(ci)

    # loop over coherence values
    Hi = np.zeros(len(ci));
    Fi = np.zeros(len(ci));
    Pcorri = np.zeros(len(ci));

    for i, c in enumerate(ci):
        Hi[i], Fi[i], Pcorri[i] = SDTposneg( c, k, theta)


    # illustrate results
    fig, axs = plt.subplots( 1, 2, figsize=(8, 4))

    axs[0].plot(ci, Hi, '-', color=[0,0.5,0], linewidth=1, label='H')
    axs[0].plot(ci, Fi, '-', color=[0.8,0,0], linewidth=1, label='F')
    axs[0].plot(ci, Pcorri, '-', color=[0,0,0], linewidth=1, label='Pcorr')
    axs[0].set_xlabel('coherence [%]')
    axs[0].set_box_aspect(1/1)
    axs[0].set_ylim((-0.05,1.05))
    axs[0].set_ylabel('H, F, P corr')
    axs[0].set_title('Psychometric', fontsize=14);

    axs[1].plot(logci, Hi, '-', color=[0,0.5,0], linewidth=1, label='H')
    axs[1].plot(logci, Fi, '-', color=[0.8,0,0], linewidth=1, label='F')
    axs[1].plot(logci, Pcorri, '-', color=[0,0,0], linewidth=1, label='Pcorr')
    axs[1].set_xlabel('log coherence')
    axs[1].set_box_aspect(1/1)
    axs[1].set_xticks([0,1,2])
    axs[1].set_xticklabels(['1%', '10%', '100%'])
    axs[1].set_ylim((-0.05,1.05))
    axs[1].set_ylabel('H, F, P corr')
    axs[1].set_title('Psychometric', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_psychometric_function, k=(0.1,1.0,0.01), theta=(-0.2,0.2,0.01) )
```

# Averaging over noisy neural responses

Let $r_1$, $r_2$, $\ldots$ $r_n$ be $n$ noisy neural responses with individual expectations *over time* of $\bar r_i = \mathit{E}[r_i]$.

For the average response $\bar r$ *over neurons*, we would like to know the variance $\mathit{Var}(\bar r)$ *over time*

$\begin{eqnarray}
\bar r = \frac{1}{n} \sum_{i=1}^n r_i, \qquad \mathit{Var}(\bar r) = \frac{1}{n^2} \mathit{E} \left[ \left( \sum_{i=1}^n r_i \right)^2 - \left(\sum_{i=1}^n\bar r_i\right)^2 \right]
\end{eqnarray}$

Rearranging sums

$\begin{eqnarray}
\mathit{Var}(\bar r) = \frac{1}{n^2}\left[\sum_{i=1}^n E[ r_i^2 - \bar r_i^2]
- \sum_{i\neq j} E[r_i r_j - \bar r_i \bar r_j] \right]
\end{eqnarray}$

and assuming uniform variance and covariance for simplicity

$\begin{eqnarray}
E[ r_i^2 - \bar r_i^2] = \sigma^2, \qquad \forall i
\\
E[ r_i r_j - \bar r_i \bar r_j] = \rho \sigma^2, \qquad \forall i\neq j
\end{eqnarray}$

with correlation coefficient
$\begin{eqnarray}
\rho = \frac{E[ r_i r_j - \bar r_i \bar r_j]}{\sqrt{E[r_i^2 - \bar r_i^2]} \sqrt{E[r_j^2 - \bar r_j^2]} }
\end{eqnarray}$

we obtain

$\begin{eqnarray}
\mathit{Var}(\bar r)= \frac{1}{n^2}\left[ n \sigma^2 - n(n-1) \rho \sigma^2 \right] = \frac{1 + (n-1)\rho}{n}
 \, \sigma^2
\end{eqnarray}$

with

$\begin{eqnarray}
\lim_{n\to\infty} \mathit{Var}(\bar r)= \rho
 \, \sigma^2
\end{eqnarray}$



```python
from inspect import signature
# @title Response variance of correlated population {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from matplotlib.colors import LinearSegmentedColormap

def plot_variance_with_correlation(Rho=0.5, LogN=1):
    """(Slide-33)"""

    # range of variable numbers
    N = 10**LogN;
    ni = np.arange(1,N+1,1)

    # factor reducing individual variance

    fcorri  = np.zeros(len(ni))
    funcor = np.zeros(len(ni))

    for i in range(0,len(ni)):
        funcor[i] = 1 / ni[i];
        fcorri[i]   = (1+ (ni[i]-1)*Rho) / ni[i];

    # Plot results
    fig, axs = plt.subplots( 1, 1, figsize=(4, 4))

    # showfactor
    axs.plot( ni, funcor, 'b-', linewidth=2, label='independent')
    axs.plot( ni, fcorri, 'r-', linewidth=2, label='correlated')
    axs.plot( ni, Rho*np.ones(len(ni)), 'k:', linewidth=1, label='lower bound' )
    axs.axis('on')
    axs.legend(loc='best')
    axs.set_xlabel('neuron number n')
    axs.set_ylabel('multiplicative reduction')
    axs.set_box_aspect(1/1)
    axs.set_title('Variance reduction', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_variance_with_correlation, Rho = (0, 1, 0.05), LogN = (0,4,0.1) )
```

# Benefit of averaging

The full benefit of averaging the responses of many neurons is obtained only if response variance is independent.
Any correlation in variance limits this benefit sharply. In fact, the correlation coefficient sets the lower bound for multiplicative reduction of variance.

Area MT contains approximately $O(10^4)$ neurons tuned for any given direction of motion. Since the population appears to be comparably informative as (the best) individual neurons, this seems to imply that correlations are essentially complete. However, the pairwise correlation coefficients between neuronal spikes are only approximately $\sim0.1$.

Apparently, the brain uses a widely distributed, but nonetheless redundant code that supports recovering most sensory information from smaller subpopulations.

# SDT with correlated neurons

The following code illustrates signal-detection theory in the presence of correlated response variability.

The response distributions of MT neurons are assumed to change linearly with stimulus coherence

$\begin{eqnarray}
\mu_\mathit{pref} = 20 + c, \qquad\qquad \sigma_\mathit{pref}^2 = 1.5 \cdot \mu_\mathit{pref}
\\
\mu_\mathit{null} = 20 - c / 3, \qquad\qquad\sigma_\mathit{null}^2 = 1.5 \cdot \mu_\mathit{null}
\end{eqnarray}$

for a fano factor of $FF=1.5$.

Given a null distribution and a preferred distribution, we can compute the intersection (optimal criterion) $\mathit{crit}$ by solving a quadratic equation and predict 1AFC proportion correct as

$\begin{eqnarray}
P_{1AFC} = \frac{H(\mathit{crit})}{2} + \frac{1-F(\mathit{crit})}{2}
\end{eqnarray}$

By computing the area under the ROC (plotting H against F for different criteria), we can predict 2AFC proportion correct

$\begin{eqnarray}
P_{2AFC} = \int_0^1 (1-F) \, dH
\end{eqnarray}$

The reduction of variance $\sigma^2$ due to averaging over $n$ correlated neurons follows from

$\begin{eqnarray}
\sigma^2_\mathit{reduced} = \frac{1 + (n-1) \, \rho}{n} \, \sigma^2
\end{eqnarray}$

where $\rho$ is the correlation coefficient.

Normally distributed responses with correlations can be generated from a multivariate **normal** distribution (i.e., zero-mean and unit variance) with a covariance matrix

$\begin{eqnarray}
C = \left(\begin{array}{cccc} 1 & \rho & \rho & \rho \\ \rho & 1 & \rho & \rho \\ \rho & \rho & 1 & \rho \\ \rho & \rho & \rho & 1 \end{array}\right)
\end{eqnarray}$

where $\rho$ is the correlation coefficient. To obtain the desired distribution moments, the responses can be rescaled as follows

̄$\begin{eqnarray}
r_\mathit{rescaled} = r_\mathit{normal} * \sigma + \mu
\end{eqnarray}$


```python
from inspect import signature
# @title SDT with correlated neurons {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def MVNormalResponses( Nneuron, Ntrial, respmean, respvar, rho ):

    # initialize array of zero means
    mean = np.zeros(Nneuron);

    # initialize covariance matrix
    cov_matrix = np.full( (Nneuron, Nneuron), rho );
    np.fill_diagonal(cov_matrix, 1);

    # generate Ntrial samples for each neuron, conforming to mean and variance
    samples = np.random.multivariate_normal(mean, cov_matrix, Ntrial)

    # rescale to obtain desired mean and variance
    samples = samples * np.sqrt(respvar) + respmean

    # Verify correlations
    # print(np.corrcoef(samples, rowvar=False))  # Correlation matrix

    # (Ntrial, Nneuron) = np.shape(samples)
    return samples

def ResponseMomentsAreaMT( coher, direc ):
    # typical response mean and variance of neurons in area MT
    if direc==True:
        rmean = 20 + coher;
    else:
        rmean = 20 - coher / 3;

    rvari = rmean * 1.5;   # Fano factor 1.5

    return rmean, rvari

def OptimalPerformance1D( mu0, sig20, mu1, sig21 ):
    # sig arguments are already squared!
    # first arguments are null, second arguments preferred

    if mu0 < mu1:       # if the distributions differ

        # find optimal criterion by solving quadratic equation
        # for intersection of two normal distributions

        A = 1 / (2*sig21) - 1 / (2*sig20)
        B = mu0 / (sig20) - mu1 / (sig21)
        C = mu1**2 / (2*sig21) - mu0**2 / (2*sig20) + 0.5*np.log(sig21/sig20)

        X = np.roots([A,B,C])

        # choose the intersection that's closest to the middle
        d1 = X[1] - 0.5* (mu1+mu0)
        d0 = X[0] - 0.5* (mu1+mu0)
        if np.abs(d1) < np.abs(d0):
            crit = X[1]
        else:
            crit = X[0]

        # density at the intersection
        pcrit = norm.pdf( crit, mu0, np.sqrt(sig20) )

        # hit and false alarm fractions
        hits   = 1 - norm.cdf( crit, mu1, np.sqrt(sig21) )
        false  = 1 - norm.cdf( crit, mu0, np.sqrt(sig20) )

        # proportion correct
        pcorr = hits/2 + (1-false)/2;

    else:        # if the distributions are identical

        crit  = mu1;
        pcrit = 0.5;
        pcorr = 0.5;

    return crit, pcrit, pcorr

def OptimalPerformance2D( mu0, sig20, mu1, sig21 ):
    # sig arguments are already squared!
    # first arguments are null, second arguments preferred

    if mu0 < mu1:       # if the distributions differ

        # compute proportion correction from area under ROC

        crit = np.arange(0,101,0.2)

        # hit and false alarm fractions
        hits   = 1 - norm.cdf( crit, mu1, np.sqrt(sig21) )
        false  = 1 - norm.cdf( crit, mu0, np.sqrt(sig20) )

        # integrate (1-F) * dH, see Slide-9 for explanation
        pcorr = -np.trapz(1-false, hits)

    else:        # if the distributions are identical

        hits  = np.arange(0,1,0.01)
        false = np.arange(0,1,0.01)
        pcorr = 0.5;

    return hits, false, pcorr

def plot_sdt_with_correlated_neurons(Rho=0.5, Nneuron=20):
    """(Slide-35)"""

    # response moments as a function of coherence (subplot [0,0])
    ci = np.arange(0,26,1)
    (rprefi, vprefi) = ResponseMomentsAreaMT( ci, True )
    (rnulli, vnulli) = ResponseMomentsAreaMT( ci, False )

    # probability of mean response (6% coherence)  (subplot [0,1])
    (rpref, vpref) = ResponseMomentsAreaMT( 6, True)
    (rnull, vnull) = ResponseMomentsAreaMT( 6, False)
    rn = np.arange(0,51,1)
    pprefn = norm.pdf( rn, rpref, np.sqrt(vpref) )
    pnulln = norm.pdf( rn, rnull, np.sqrt(vnull) )

    # criteron for 6% coherence
    (crit, pcrit, pcorr1D) = OptimalPerformance1D( rnull, vnull, rpref, vpref )
    (hits, false, pcorr2D) = OptimalPerformance2D( rnull, vnull, rpref, vpref )

    # neurometric function (one coherence level at a time)
    # 1D for single stimulus presentation
    # 2D for two-alternative forced-choice presentation
    ci = np.arange(0,26,1)
    pcorri1D = np.zeros(len(ci))
    pcorri2D = np.zeros(len(ci))

    for i in range(0,len(ci)):
        (rpref, vpref) = ResponseMomentsAreaMT( ci[i], True )
        (rnull, vnull) = ResponseMomentsAreaMT( ci[i], False )

        # reduce variance due to multiple responses
        vpref = vpref * (1 + (Nneuron - 1) * Rho) / Nneuron;
        vnull = vnull * (1 + (Nneuron - 1) * Rho) / Nneuron;

        if rpref > rnull:
            (cri, pcr, pcorr1) = OptimalPerformance1D( rnull, vnull, rpref, vpref )
            pcorri1D[i] = pcorr1
            (h, f, pcorr2) = OptimalPerformance2D( rnull, vnull, rpref, vpref )
            pcorri2D[i] = pcorr2

        else:
            cri   = rnull;
            pcr   = 0.05;
            pcorri1D[i] = 0.5;
            pcorri2D[i] = 0.5;

    # multivariate normal responses, one coherence leve at a time
    Ntrial = 500;
    pcorrempi = np.zeros(len(ci))
    for i in range(0,len(ci)):
        rprefkj = MVNormalResponses( Nneuron, Ntrial, rprefi[i], vprefi[i], Rho )
        rnullkj = MVNormalResponses( Nneuron, Ntrial, rnulli[i], vnulli[i], Rho )

        # average over neurons
        rprefk = np.mean(rprefkj, axis=1)
        rnullk = np.mean(rnullkj, axis=1)

        # probability that preferred average exceed null average
        pcorrempi[i] = np.mean( rprefk > rnullk )

    # Plot results
    fig, axs = plt.subplots( 1,3, figsize=(12,6))

    # Moments
    axs[0].plot( ci, rprefi, color=(0.25,0.75,0.25), linewidth=1, label='pref mean')
    axs[0].plot( ci, rnulli, color=(0.25,0.25,0.75), linewidth=1, label='null mean')
    axs[0].plot( ci, vprefi, ':', color=(0.25,0.75,0.25), linewidth=1, label='pref vari')
    axs[0].plot( ci, vnulli, ':', color=(0.25,0.25,0.75), linewidth=1, label='null vari')
    axs[0].legend()
    axs[0].set_xlabel('coherence [%]')
    axs[0].set_box_aspect(1/1)
    axs[0].set_ylabel('response')
    axs[0].set_title('Moments of response distribution', fontsize=14);

    # Distributions
    axs[1].plot( rn, pprefn, color=(0.25,0.75,0.25), linewidth=1 )
    axs[1].plot( rn, pnulln, color=(0.25,0.25,0.75), linewidth=1 )
    axs[1].plot( [crit,crit], [0,pcrit], 'k-', linewidth=1 )
    axs[1].plot( crit, pcrit, 'ko', markersize=3 )
    axs[1].set_xlabel('response')
    axs[1].set_ylabel('density')
    axs[1].set_box_aspect(1/1)
    axs[1].set_title('Response distribution, c= 6%', fontsize=14);

    # ROC, append arrays to fill area
    false = np.append(false[::-1], [1])
    hits  = np.append(hits[::-1], [0])

    axs[2].fill( false, hits, color=(1.0,0.5,0.5), linewidth=1 )
    axs[2].set_xlabel('false alarms')
    axs[2].set_ylabel('hits')
    axs[2].set_box_aspect(1/1)
    axs[2].set_title('ROC, c= 6%', fontsize=14);

    # Plot results
    fig, axs = plt.subplots( 1,1, figsize=(12,6))


    # Neurometric
    axs.plot( ci, pcorri1D, ':', color=(0.75,0.25,0.25), linewidth=1, label='1AFC' )
    axs.plot( ci, pcorri2D, color=(0.75,0.25,0.25), linewidth=1, label='2AFC' )
    axs.plot( ci, pcorrempi, 'ro', markersize=3, label='numerical' )
    axs.set_ylim([0.5, 1.05])
    axs.legend(loc='lower right')
    axs.set_xlabel('coherence')
    axs.set_ylabel('P correct')
    axs.set_box_aspect(1/2)
    axs.set_title(f"Neurometric, rho= {Rho}, n= {Nneuron}", fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_sdt_with_correlated_neurons, Rho = (0, 1, 0.1), Nneuron = (1, 1000, 1) )
```

# **Summary too good?**

- We have formulated a simple but realistic model for the selectivity and variability of direction-selective responses in area MT.

- We also take correlations in response variance into account.

- We can predict behavioral performance (\% correct) as a function of stimulus strength (\% coherence), either by assuming a single presentation (1AFC) or a double presentation (2AFC).

- For independent (uncorrelated) responses, behavioral performance increases rapidly with neuron number.

- For correlated responses, behavioral performance saturates with neuron number, limiting the effective population size.

- The similarity of psychometric and neurometric functions -- which suggests a widely distributed, yet highly redundant encoding -- is a typical result and continues to be a topic of current research.

# **Overall summary**

- **Choice performance ($d'$ or $\%$ correct) quantifies sensory information available to organism.**

- **Sets lower limit for information encoded in neural activity.**

- **Signal-detection theory relates choice performance to underlying neural activity (actual or hypothetical).**

- **Comparison of psychometric and neurometric functions identifies candidate neural populations on which choice performance may be based, as shown by Britten and colleagues (1992).**

- **To establish a causal role of neural populations, their responses can be manipulated (e.g., by electrical stimulation), as Salzman and colleagues did (1992).**

- **Typically, neuronal activities are (weakly) correlated and do not vary (fully) independently, consistent with widely distributed, yet highly redundant encoding.**

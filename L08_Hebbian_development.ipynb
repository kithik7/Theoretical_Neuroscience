# **Abstract:**

In visual development, nearby neurons express the same ocular dominance.  Recurrent cooperation-competition between output neurons explains this.  Combining input correlations ${\bf Q}$ with output cooperation-competition $\bf K$ produces alternating 'ocular dominance bands'.

To illustrate the generality of correlation-driven development, we discuss **two classic** papers on the development of simple cells.  One describes development of dominance by 'on-' or 'off-center' input, the other of combined dominance by 'on' or 'off' and by 'left-' or 'right-eye' input.

Finally, we return to **supervised Hebbian learning** (see Lecture 2 and Exercise 4), where input- and output-activity is externally imposed and synaptic weights develop such as to reproduce the imposed input-output correlations.  We consider the (intentionally) artificial example of transforming Cartesian to polar coordinates.

<br>

We will now consider unsupervised learning with multiple (cortical) output neurons ${\bf v}$, interacting recurrently through fixed synaptic matrix ${\bf M}$ (weak competition-cooperation).  Plasticity and learning occurs only in the feedforward matrix ${\bf W}$:

<br>

<br>

We expect plasticity to depend directly on correlations between input and output neurons, but indirectly on both correlations between input neurons and competition-cooperation between output neurons.

(Slide-2)

# **1. Recurrent connections among output neurons**

Consider a linear network in which $N_v$ output neurons receive recurrent input from each other and feedforward input from $N_u$ input neurons.

<br>

$$
\tau_r \, \frac{d{\bf v}}{dt} = -{\bf v} + {\bf W} \cdot {\bf u} + {\bf M} \cdot {\bf v}
$$

<br>

The steady-state condition is

$$
0 \stackrel{!}{=} \tau_r \, \frac{d{\bf v}}{dt} = -{\bf v} + {\bf W} \cdot {\bf u} + {\bf M} \cdot {\bf v}
\qquad \Leftrightarrow \qquad {\bf v}_\mathit{ss} = {\bf W} \cdot {\bf u} + {\bf M} \cdot {\bf v}_\mathit{ss}
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08-DA-Fig-8-6.jpg" width="800">

<br>

(Slide-3)


The steady-state condition

<br>

$$
{\bf v}_\mathit{ss} = {\bf W} \cdot {\bf u} + {\bf M} \cdot {\bf v}_\mathit{ss}
\qquad\Leftrightarrow\qquad {\bf v}_\mathit{ss} - {\bf M} \cdot {\bf v}_\mathit{ss} = {\bf W} \cdot {\bf u}
$$

<br>

has a solution ${\bf v}_\mathit{ss}$ if there exists an inverse matrix ${\bf K}\equiv ({\bf I} - {\bf M})^{-1}$ to solve

<br>

$$\begin{eqnarray}
\left({\bf I} - {\bf M}\right) \cdot {\bf v}_\mathit{ss} &=& {\bf W} \cdot {\bf u}
\\
\\
{\bf v}_\mathit{ss} &=& \left({\bf I} - {\bf M}\right)^{-1} \cdot  {\bf W} \cdot {\bf u}
\\
\\
{\bf v}_\mathit{ss} &=& {\bf K} \cdot {\bf W} \cdot {\bf u}
\end{eqnarray}$$

<br>

where ${\bf I}$ is the identity matrix, with ${\bf v}_\mathit{ss} = {\bf I} \cdot {\bf v}_\mathit{ss}$.  

This condition is met if the real parts of the eigenvalues of ${\bf M}$ are smaller than 1 (i.e., if recurrent interactions are comparatively weak).

<br>

(Slide-4)

With fixed recurrent weights ${\bf M}$ and plastic feedforward weights ${\bf W}$, the Hebbian learning rule (without threshold) becomes

<br>

$$\begin{eqnarray}
\tau_w \, \frac{d{\bf W}}{dt} &=& \left\langle{\bf v}_\mathit{ss} \, {\bf u}\right\rangle =
\\
\\
&=&  {\bf K} \cdot {\bf W} \cdot \left\langle {\bf u} \,{\bf u}\right\rangle =
\\
\\
&=& {\bf K} \cdot {\bf W} \cdot {\bf Q}, \qquad {\bf Q} \equiv \left\langle{\bf u} \, {\bf u}\right\rangle,\qquad{\bf K} \equiv ({\bf I} - {\bf M})^{-1}
\end{eqnarray}$$

<br>

As expected, the development is influenced not only by input correlations ${\bf Q}$, but also by the recurrent connectivity that shapes $\bf K$.

<br>

(Slide-5)

# Role of recurrent connectivity

<br>


**Without** recurrent connectivity, the inputs weights ${\bf W}_i$ to each output neuron $v_i$ would develop independently, reflecting (in part) the random initial conditions of each neuron.

<br>

**With** recurrent connectivity, a structured cortical map can develop, minimizing differences between neighboring neurons.  For example, in the case of ocular dominance, alternating 'bands' of neurons with left- and right-eye dominance can develop.

<br>


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec07_OD_Intro.png" width="800">

<br>

(Slide-6)

# Points to  note

<br>

- Recurrent connectivity in the output layer alters Hebbian learning rules.

<br>

- The correlation rule for feedforward weights now reads


$$
\tau_w \, \frac{d{\bf W}}{dt} = \left\langle{\bf v} \, {\bf u}\right\rangle = {\bf K} \cdot {\bf W} \cdot {\bf Q}
$$

<br>

- Both input correlations ${\bf Q}$ and output interactions $\bf K$ are important.

<br>

- This permits the development of systematic patterns of feedforward connectivity.

<br>

(Slide-7)

# **2. Development of ocular dominance bands**

<br>

We define recurrent cooperation-competition indirectly in terms of $\bf K$ (rather than directly in terms of ${\bf M}$) and assume

<br>

- translational invariance over cortical locations
- mutual excitation between adjacent units (cooperation)
- mutual inhibition between more distant units (competition)

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Kmatrix_1.png" width="800">

<br>

'Difference-of-Gaussian' (DOG) connectivity
$$
K(d) = \frac{1}{\sqrt{2\pi} \, \sigma_E} \, \exp\left(-\frac{d^2}{2\sigma_E^2} \right) - \frac{1}{\sqrt{2\pi} \, \sigma_I} \, \exp\left(-\frac{d^2}{2\sigma_I^2} \right)
$$
where $d$ is cortical distance and $\sigma_I \approx 3 \, \sigma_E$.

<br>

(Slide-8)

# Recurrent cooperation-competition matrix $\bf K$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Kmatrix_2.png" width="800">

<br>

'Heat map' illustration of positive (yellow-white), negative (red-black), and neutral values (red).

<br>

(Slide-9)

# Eigenvectors 1 and 2

<br>

The principal eigenvectors of the matrix ${\bf K}$ are pairs of even and odd harmonic functions of cortical distance $d$

<br>

$$
{\bf e}_{1,2} = \cos\left(\frac{2\pi d}{N_v} - \phi \right), \qquad \phi\in\left[0, \frac{\pi}{2}\right], \qquad \lambda_{1,2} = 3.2
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Kmatrix_3.png" width="800">

<br>

(Slide-10)

# Eigenvectors 3 and 4

<br>

The next eigenvectors of the matrix ${\bf K}$ are also pairs of even and odd harmonic functions of cortical distance, but with twice the spatial frequency

<br>

$$
{\bf e}_\mu = \cos\left(\frac{4\pi d}{N_v} - \phi \right), \qquad \phi\in\left[0, \frac{\pi}{2}\right], \qquad \lambda_{3,4} = 2.8
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Kmatrix_4.png" width="800">

<br>

(Slide-11)

# Principal eigenvectors

<br>

The two principal eigenvectors are degenerate (have identical eigenvalues) and resemble harmonic functions with alternating bands of positive and negative interactions between output units.   

<br>

Any linear combination will also be an
eigenvector

<br>

$$
{\bf K} \cdot {\bf e} = \lambda \, {\bf e}, \qquad\qquad {\bf e} = \alpha \, {\bf e}_1 + (1-\alpha) {\bf e}_2
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Kmatrix_5.png" width="800">

# Formal treatment of development

<br>

We now formalize the development of OD bands in the primary visual cortex of stereoptic mammals.  All output units are stimulated by the same two input units $u_R$ and $u_L$

<br>

$$
{\bf u} = \left( \begin{array}{c} u_R \\ u_L \end{array} \right),\qquad
{\bf v} = {\bf K} \cdot {\bf W} \cdot {\bf u}, \qquad {\bf K} \equiv ({\bf I} - {\bf M})^{-1}
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Supervised_b.png" width="800">

<br>

(Slide-15)

We assume a dynamic equation for weights:

<br>

$$
\tau_w \, \frac{d{\bf W}}{dt} =  {\bf K} \cdot {\bf W} \cdot {{\bf Q}}
- \frac{1}{2} {\bf n}^T \cdot  {\bf K} \cdot {\bf W} \cdot {{\bf Q}} , \qquad {\bf n} = \left( \begin{array}{c} 1 \\  1 \end{array} \right)
$$

<br>

corresponding to a Hebbian rule without threshold and with subtractive normalization.

<br>

The matrix dimensions (rows $\times$ columns) are

<br>

$$
{\bf K}: n\times n\qquad\qquad {\bf W}: n \times 2 \qquad\qquad {{\bf Q}}: 2 \times 2
$$

<br>

(Slide-14)

# Non-zero mean input

<br>

We further assume non-zero mean input with partially anti-correlated statistics

<br>

$$
{\bf Q} = \left(\begin{array}{cc} q_S & q_D \\ q_D & q_S \end{array}\right)=
 \left(\begin{array}{cc} \frac{1}{2} & \frac{1}{8} \\ \frac{1}{8} &  \frac{1}{2} \end{array}\right)
$$

<br>

This is the statistics for which we **failed** to obtain OD bands in Lecture 7!

<br>

(Slide-15)

When ${\bf W}$ is written out in terms of its two columns ${\bf w}_R$ and ${\bf w}_L$

<br>

$$
{\bf W}=\left(\begin{array}{cc} w_{1R} & w_{1L} \\ \vdots & \vdots  \\ w_{nR} & w_{nL} \end{array} \right)
= \left(\begin{array}{cc}  \\  {\bf w}_R & {\bf w}_L \\  \\\end{array} \right)
$$

<br>

the dynamic equations become

<br>

$$
\tau_w \, \frac{d{{\bf w}_R}}{dt} =  {\bf K}\cdot \left( q_S {\bf w}_R + q_D {\bf w}_L\right) - \frac{1}{2} \,   {\bf K}\cdot \left( q_S {\bf w}_R + q_D {\bf w}_L + q_D {\bf w}_R + q_S {\bf w}_L\right)
$$

<br>

$$
\tau_w \, \frac{d{{\bf w}_L}}{dt} =  {\bf K}\cdot \left( q_D {\bf w}_R + q_S {\bf w}_L\right) - \frac{1}{2} \,   {\bf K}\cdot \left( q_S {\bf w}_R + q_D {\bf w}_L + q_D {\bf w}_R + q_S {\bf w}_L\right)
$$

Using the same trick as in Exercise 3, we introduce two vectors ${\bf w}_+$ (sum of weights) and ${\bf w}_-$ (difference of weights):

<br>

$$
{\bf w}_+ = {\bf w}_R+{\bf w}_L\qquad\qquad{\bf w}_- = {\bf w}_R-{\bf w}_L
$$

<br>

In terms of these aligned weight vectors, the dynamic equations are

<br>

$$\begin{eqnarray}
\tau_w \, \frac{d{\bf w}_+}{dt} &=&  0
\\
\\
\tau_w \, \frac{d{\bf w}_-}{dt} &=& {\bf K} \cdot (q_S-q_D) \, {\bf w}_-
\end{eqnarray}$$

<br>

Thus the development of ${\bf w}_-$ is governed by the principal eigenvectors of $\bf K$!


<br>

(Slide-16)

```python
from inspect import signature
# @title Development of ocular dominance {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from ipywidgets import interact

# toggle switches between interactions matrix M and no interactions

# detailflag causes intermediate results to be printed and illustrated

def plot_animate_development(toggle=1, detailflag=0):
    """(Slide-19)"""

    # Parameters
    Nv = 21                 # number of outputs, recurrently connected
    v  = np.zeros((Nv, 1))   # output activities, column vector
    Nu = 21                 # inputs, with plastic feedforward projections
    u  = np.zeros((Nu, 1))   # input activities, column vector
    W  = np.zeros((Nv, Nu))  # feedforward weights
    M  = np.zeros((Nv, Nv))  # recurrent weights
    K  = np.zeros((Nv, Nv))  # inverse recurrent weights

    # Construct matrix K
    x    = np.arange(1, Nv + 1)
    y    = np.arange(1, Nv + 1)
    X, Y = np.meshgrid(x, y)

    # Matrix of pairwise distances with periodic boundary conditions
    D = np.minimum(np.abs(X - Y), Nv - np.abs(X - Y))

    # Plot distance matrix
    if detailflag:
        plt.figure()
        plt.imshow(D, cmap='hot', aspect='equal')
        plt.colorbar()
        plt.xlabel('N_v', fontsize=20)
        plt.ylabel('N_v', fontsize=20)
        plt.title('Distance matrix', fontsize=20)
        plt.show()

    # Difference of Gaussian profile
    sigmaE = 1.25
    sigmaI = 6.0
    scale = 1.0

    z = np.arange(-10, 10.1, 0.1)
    F = np.exp(-z**2 / (2 * sigmaE**2)) / (np.sqrt(2 * np.pi) * sigmaE) - \
        scale * np.exp(-z**2 / (2 * sigmaI**2)) / (np.sqrt(2 * np.pi) * sigmaI)

    # Compute M matrix of connectivity
    M = np.exp(-D**2 / (2 * sigmaE**2)) / (np.sqrt(2 * np.pi) * sigmaE) - \
        scale * np.exp(-D**2 / (2 * sigmaI**2)) / (np.sqrt(2 * np.pi) * sigmaI)

    # Plot DOG profile and M matrix
    fig, axs = plt.subplots( 1,2, figsize=(12,6))

    axs[0].plot(z, F, linewidth=2.0);
    axs[0].set_xlabel('distance d', fontsize=18);
    axs[0].set_ylabel('K(d)', fontsize=18);

    cax = axs[1].imshow(M, cmap='hot', aspect='equal');
    axs[1].set_xlabel('N_v', fontsize=20);
    axs[1].set_ylabel('N_v', fontsize=20);
    axs[1].set_title('Matrix M', fontsize=20)

    fig.colorbar(cax, ax=axs[1])

    plt.tight_layout()
    plt.show()

    # Compute effective matrix K = (I - M)^{-1}
    I = np.eye(Nv)
    K = np.linalg.inv(I-M)

    if detailflag:
        plt.figure()
        plt.imshow(K, cmap='hot', aspect='equal')
        plt.colorbar()
        plt.xlabel('N_v', fontsize=20)
        plt.ylabel('N_v', fontsize=20)
        plt.title('Matrix K', fontsize=20)
        plt.show()

    # Compute eigenvalues of effective matrix K
    Evals, Evecs = np.linalg.eig(K)

    # suppress spurious imaginary parts ...
    Evals = np.real(Evals)
    Evecs = np.real(Evecs)

    # check eigenvectors and eigenvalues
    # for i in range(Nv):
    #     Kv = K @ Evecs[:, i]         # Multiply M by eigenvector
    #     lv = Evals[i] * Evecs[:, i]  # Multiply eigenvalue by eigenvector
    #     print(f"Kv-lv:\n{Kv - lv}\n")

    # Check the steady-state condition for random intial condition and projection weights
    if detailflag:
        u = np.random.rand(Nu, 1)
        W = -0.5 + np.random.rand(Nv, Nu)

        vss = K @ W @ u
        vss2 = W @ u + M @ vss

        print(np.hstack((vss, vss2, vss - vss2)))

    # check eigenvectors scaled by eigenvalues
    Escaled = np.zeros_like(Evecs);
    for i in range(Nv):
        Escaled[:,i] = Evals[i] * Evecs[:, i]  # Multiply eigenvalue by eigenvector

    plt.figure()
    plt.plot(Escaled, linewidth=2.0)
    plt.xlabel('distance d', fontsize=18)
    plt.ylabel('E^K_i', fontsize=18)
    plt.title('Scaled eigenvectors of K', fontsize=18)
    plt.show()


    if toggle == 0:
        K = np.eye(Nv);

    # Simulate development of OD bands
    qS = 0.50
    qD = 0.125
    WR0 = np.random.rand(Nv)  # initial weights
    WL0 = np.random.rand(Nv)

    T = 500
    dt = 0.5
    tau = 5

    ti = np.arange(0,T,dt);

    WRik = np.zeros((len(ti), Nv))
    WLik = np.zeros((len(ti), Nv))

    WRik[0,:] = WR0;     # current weight, set equal to initial
    WLik[0,:] = WL0;

    for i in range(1,len(ti)):

        WRc = WRik[i-1,:];
        WLc = WLik[i-1,:];

        WRn = WRc + 0.5 * (qS * K @ WRc + qD * K @ WLc - qD * K @ WRc - qS * K @ WLc) * dt / tau         # next iteration of weights
        WLn = WLc + 0.5 * (-qS * K @ WRc - qD * K @ WLc + qD * K @ WRc + qS * K @ WLc) * dt / tau

        WRik[i,:] = np.clip(WRn, 0, 1)
        WLik[i,:] = np.clip(WLn, 0, 1)

    # Initialize figure and axis
    fig, ax = plt.subplots()

    xrange = np.arange(1,Nv + 1);


    ax.plot(xrange, WR0, 'r:', xrange, WL0, 'b:', linewidth=2.0)

    # last time point
    ax.plot(xrange, WRik[-1,:], 'r-', xrange, WLik[-1,:], 'b-', linewidth=2.0)

    # set axis limits
    ax.set_xlim([1, Nv])
    ax.set_ylim([0, 1])
    ax.set_xlabel('distance c', fontsize=18)
    ax.set_ylabel('w_R, w_L', fontsize=18)

    plt.show()

# Creating sliders for concentrations and charge
interact( plot_animate_development, toggle = (0, 1, 1 ), detailflag = (0, 1, 1 ) )

```

# Points to  note

<br>

- For realistic inputs ${\bf u}$ (with non-zero mean), the correlation matrix is ${\bf Q}$ is dominated by the 'same' eigenvector.

- This is why, in Lecture 7, we failed to obtain the desired outcome of synaptic weights aligned with the 'different' eigenvector.

- We have now overcome this difficulty by introducing cooperation-competition (difference-of-Gaussians) within the output layer.

- Now the development of feedforward weights is driven by a combination of input correlations ${\bf Q}$ and output interactions $\bf K$.

- This is a plausible model for developing so-called ''functional anatomy'', such as ocular dominance bands, in sensory cortices of mammals.

The development of **ocular dominance bands** requires

The development of **orientation columns** requires

# **4. Supervised Hebbian learning**

In supervised learning, both input activity $\bf u$ and output activity $\bf v$ is imposed externally.  Synaptic projections $\bf W$ develop such as to reproduce correlations $\bf Q = \langle\bf v \, \bf u\rangle$ in imposed activities.  In this 'training' phase, synaptic weights do *not affect* output activity.

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Supervised_a.png" width="800">

<br>

After training, synaptic projections $\bf W$ map input activity $\bf u$ to output activity $\bf v$ in such a way as to *reproduce* the correlations $\bf Q$ experienced earlier.  Only in this 'testing' phase do projections determine output activity, such that $\bf v = F\left( \bf W \cdot \bf u \right)$.

<br>

(Slide-40)

# Learning rule

<br>

An (unstable) Hebbian rule for supervised learning of a single output neuron is

<br>

$$
\tau_w \frac{d{\bf W}}{dt} = \left\langle v \, {\bf u}\right\rangle
$$

where $v$ is single output activity and ${\bf u}$ multiple input activities imposed externally during training.

<br>

As neither $v$ nor ${\bf u}$ depends on ${\bf W}$ (connections are still being trained!), the unstable solution is

<br>

$$
{\bf W}_{(t)} = {\bf W}_0 + \left\langle v \, {\bf u}\right\rangle \, \frac{t}{\tau_w}
$$
In other words, synaptic weights grow linearly with time, with growth rates proportional to (positive or negative) correlations $\left\langle v\, {\bf u}\right\rangle$.

<br>

(Slide-41)

# Stabilization

<br>


This Hebbian rule may be stabilized by a 'relaxation' term, providing for spontaneous 'decay' of synaptic weights

<br>

$$
\tau_w \frac{d{\bf W}}{dt} = -  {\bf W} + \left\langle v \, {\bf u}\right\rangle
$$

<br>

with the steady-state solution

<br>


$$
{\bf W}_{ss} =  \left\langle v \, {\bf u}\right\rangle
$$

<br>

In other words, synaptic weights develop such as to become proportional to input- output correlations $\left\langle v \, {\bf u}\right\rangle$.  After training, when synaptic projections become effective, the effect will tend to *reproduce* the correlations experienced earlier during training.

<br>

Next we use this mechanism to 'train' arbitrary input-output transformations, or 'functional mappings'.

<br>

(Slide-42)

# Cartesian and polar coordinates

<br>

$$
\theta = \arctan\frac{y}{x}, \qquad\qquad r = \sqrt{ x^2+y^2 }
$$

<br>

(Slide-44)

# Cartesian $\rightarrow$ polar with neural populations

We assume a two-dimensional input population with index $i$, tuned in Cartesian coordinates $(x,y)$, projecting to two one-dimensional output populations with indices $j$ and $k$, tuned in polar angle $\theta$ and log polar radius $\ln r$, respectively:

<br>

$$
\theta = \arctan\frac{y}{x}, \qquad\qquad \ln r =\ln \sqrt{ x^2+y^2 }
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec08_Cartesian_Polar_a.png" width="600">

<br>

(Slide-45)

# Training phase: noisy responses $i$, $j$, and $k$

For many (hundreds or thousands) random stimulus positions $(x,y,r,\theta)$, we compute input responses $R_i^c$ and output responses $R_j^\theta$ and $R_k^r$.

<br>

- Input responses show Gaussian tuning -- preferred position $(x_i,y_i)$ and width $\sigma_{xy}$ -- with noise $\eta_{i}$:

$$\begin{eqnarray}
R_i^c & = &\mathfrak{N}\left( \left| x_i - x \right|, \left|y_i - y\right|, \sigma_{xy} \right)+ \eta_i
\end{eqnarray}$$

<br>

- Output responses show Gaussian -- preferred position $\theta_j$ or $\ln r_k$, and width $\sigma_\theta$ or $\sigma_{r}$ -- with noise $\eta_{j}$ or $\eta_{k}$:

$$\begin{eqnarray}
R_j^\theta &=& \mathfrak{N}\left( \left| \theta - \theta_j \right|, \sigma_\theta \right)+ \eta_j
\\
R_k^r &=& \mathfrak{N}\left( \left| \ln r - \ln r_k \right|, \sigma_{r}\right)+ \eta_k
\end{eqnarray}$$

<br>

where $\mathfrak{N}(\mu,\sigma)$ and $\mathfrak{N}(\mu_x, \mu_y,\sigma)$  are normal distributions.

<br>

(Slide-46)

# Training phase: correlations between noisy responses

<br>

- Next, we compute the covariance of noisy responses for every pair of pre- and postsynaptic neurons $i\to j$ and for  $i\to k$: such as to collect the resulting values into covariance matrices $C_{ji}$ and $C_{ki}$:  

<br>

$$\begin{eqnarray}
C_{ji} &=& \left\langle R_j^\theta \, R_i^c\right\rangle - \left\langle R_j^\theta\right\rangle \, \left\langle R_i^c\right\rangle
\\
C_{ki} &=& \left\langle R_k^r \, R_i^c\right\rangle - \left\langle R_k^r\right\rangle \, \left\langle R_i^c\right\rangle
\end{eqnarray}$$

<br>

- Finally, we set feedforward projection matrices $W_{ji}$ and $W_{ki}$ equal to the covariance matrices $C_{ji}$ and $C_{ki}$:

<br>

$$\begin{eqnarray}
W_{ji} &=& C_{ji}
\\
W_{ki} &=& C_{ki}
\end{eqnarray}$$

<br>

- We have thereby 'trained' synaptic weights to *reproduce* the input-output correlations experienced during training.

<br>

(Slide-47)

# Testing phase: noisy responses $i$, $j$, and $k$

<br>

For random stimulus positions $(x,y,r,\theta)$, we compute input responses $R_i^c$ as before.  However, unlike before, we compute responses $R_j^\theta$ and $R_k^r$ as non-linear functions of total synaptic input.

<br>

- Input responses (as before):

<br>

$$\begin{eqnarray}
R_i^c & = &f\left( \left| x_i - x \right|, \left|y_i - y\right| \right)+ \eta_i
\end{eqnarray}$$

<br>

- Output responses (new) as feedforward many-to-many projections, analogous to ${\bf v}=F({\bf W} \cdot{\bf u})$:

<br>

$$\begin{eqnarray}
R_j^\theta &=& \left[ \sum_i \, W_{ji} \cdot R_i^c \right]_+ \\
\\
R_k^r &=& \left[ \sum_i \, W_{ki} \cdot R_i^c \right]_+
\end{eqnarray}$$

<br>

with rectification $[]_+$ as activation function $F()$.

<br>

(Slide-48)

# Testing phase: decoding noisy responses

<br>

To verify whether responses of output population correctly encode stimulus position, we compare the correct polar coordinates coordinates $(\ln r_\mathit{true}, \theta_\mathit{true})$ of a stimulus $(x,y)$

<br>

$$
\theta_\mathit{true} = \arctan\frac{y}{x}
\qquad\qquad
\ln r_\mathit{true} = \ln\sqrt{x^2+y^2}
$$

<br>

with the polar-coordinates decoded from noisy output responses (by response-weighted averaging):

<br>

$$
\theta_\mathit{decoded} = \frac{\sum_j \, R_j^\theta \, \theta_j}{\sum_j \, R_j^\theta}
\qquad\qquad
\ln r_\mathit{decoded} = \frac{\sum_k \, R_k^\theta \, \ln r_k}{\sum_k \, R_k^\theta}
$$

<br>

Recall response-weighted averaging from ML decoding: average over preferred positions weighed by associated responses.

<br>

(Slide-49)

# Points to note

<br>

- We found that an arbitrary 'functional mapping' may be accomplished by supervised Hebbian learning.

<br>

- As an artificial example, we transformed neural responses from Cartesian to polar coordinates.

<br>

- In a training phase, both input and output responses were imposed directly by (random) external stimulation, and covariances established for all input-output pairs.  Feedforward projection strengths were set equal to covariance values.

<br>

- In testing phase, only input response were due to stimulation.  Output responses were determined by feedforward projection strengths.

<br>

- To validate results, noisy output responses were decoded by response-weighted averaging.

**Abstract:**

Today we discuss **psychophysics** and the problem of quantifying the sensory information within subjective experience.
Unfortunately, we cannot spend time on the many clever and beautiful stimuli devised by psychophysicists over the last 150 years.

Instead, we focus on **signal-detection theory**, which predicts the quantitative relation between **noisy neural responses** and the accuracy of behavioral **choice reports**.

We introduce STD in the **forward direction**, starting from neural responses with known statistics to predict the proportion of correct choice reports.

Most applications of STD work in the **backward direction**, starting from the proportion of correct reports to predict the statistics of neural responses, with the help of some simplifying assumptions.

# Detection task

An observer is asked to detect and report the 'presence' or 'absence' of a target in many successive trials. The target could be a dim flash, a faint sound, or (here) a low-contrast target superimposed on a baseline pattern.

A cue (e.g., by sound) indicates when the trial begins and when the report is due. The observer is instructed that targets are 'present' on half the trials (chosen randomly).

The observer is required to report 'present' and 'absent' on every trial.

# **Step One:** Noisy neural responses

For the sake of concreteness, we assume that our stimuli elicit noisy neural responses with a certain mean and variance. Specifically, we assume that neural responses are **discrete** and **Poisson distributed** (i.e., exhibit the typical statistics of spike counts).

# <img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_STD_a.png" width="400">

Responses $k$ exhibit both a mean  $\left\langle{k}\right\rangle = \lambda$ and a variance $\left\langle{(k-\lambda)^2}\right\rangle = \lambda$. This equality characterizes the Poisson distribution.

For completeness:

$\begin{eqnarray}
P_k = \frac{\lambda^k \, e^{-\lambda}}{k!} \qquad\qquad \sum_{k=0}^\infty \, P_k = 1
\end{eqnarray}$

$\begin{eqnarray}
 \left\langle{k}\right\rangle \equiv \sum_{k=0}^\infty \, k \, P_k = \lambda\qquad\qquad \left\langle{(k-\lambda)^2}\right\rangle \equiv \sum_{k=0}^\infty \, (k-\lambda)^2 \, P_k = \lambda
\end{eqnarray}$

(Slide-12)


# 'Absence' or 'Noise' trials

In the absence of a target, we observe an average neural response of $\lambda_{\it Noise}=3$ spikes.



```python
from inspect import signature
# @title Absence or noise trials {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from matplotlib.colors import LinearSegmentedColormap

def plot_poisson_responses_noise(LambdaN=3, Ntrial=20):
    """(Slide-13)"""

    # random responses with mean lambda
    respn = np.random.poisson(LambdaN, Ntrial)
    tn = np.arange(0,2*Ntrial,2)

    Nbin = 20;
    (histn, bin_edges) = np.histogram(respn, bins=Nbin+1, range=(0,Nbin+1) )

    histn = histn / np.sum(histn)

    # Poisson-distribution with mean lambda
    k = np.arange(0, Nbin+1);
    Pk = poisson.pmf( k, LambdaN )


    # Plot results
    fig, axs = plt.subplots( 2, 1, figsize=(8, 6))

    kn  = np.concatenate((np.arange(0,1),k));
    Pkn = np.concatenate((np.arange(0,1),Pk));

    # Samples X
    axs[0].bar(tn, respn, color=(0,0.75,0), linewidth=1)
    axs[0].set_xlabel('time')
    axs[0].set_box_aspect(1/4)
    axs[0].set_xlabel('trials')
    axs[0].set_ylabel('responses')
    axs[0].set_title('Responses over noise trials', fontsize=14);

    # Distribution
    axs[1].bar(k, histn, color=(0,0.75,0), linewidth=1)
    axs[1].step(kn, Pkn, '-', color=(0,0.25,0), linewidth=2)
    axs[1].set_xlabel('strength k')
    axs[1].set_ylabel('probability')
    axs[1].set_box_aspect(1/2)
    axs[1].set_title('Distribution of responses to noise', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_poisson_responses_noise, LambdaN = (1, 10, 1), Ntrial = (20, 200, 10) )
```

# Signal and noise trials intermixed

In a psychophysical experiment, signal and noise trials are randomly intermixed (often with equal probability).

The observer experiences both types of trials as similar, but the experimenter knows which trial is which.



```python
from inspect import signature
# @title Both types of trials intermixed {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from matplotlib.colors import LinearSegmentedColormap

def plot_poisson_responses_intermixed(LambdaN=3, LambdaS=8, Ntrial=40):
    """(Slide-15)"""

    # random responses with mean lambda
    trialn = ( np.random.rand(Ntrial) > 0.5)

    tn     = np.arange(0,2*Ntrial,2)

    tNn    = tn[np.where(trialn==False)[0]]
    tSn    = tn[np.where(trialn==True)[0]]

    respNn = np.random.poisson(LambdaN,len(tNn))
    respSn = np.random.poisson(LambdaS,len(tSn))

    Nbin = 20;
    (histNn, bin_edges) = np.histogram(respNn, bins=Nbin+1, range=(0,Nbin+1) )
    (histSn, bin_edges) = np.histogram(respSn, bins=Nbin+1, range=(0,Nbin+1) )

    histNn = histNn / np.sum(histNn)
    histSn = histSn / np.sum(histSn)

    # Poisson-distribution with mean lambda
    k = np.arange(0, Nbin+1);
    PSk = poisson.pmf( k, LambdaS )
    PNk = poisson.pmf( k, LambdaN )

    # Plot results
    fig, axs = plt.subplots( 3, 1, figsize=(8, 8))

    kn  = np.concatenate((np.arange(0,1),k));
    PSkn = np.concatenate((np.arange(0,1),PSk));
    PNkn = np.concatenate((np.arange(0,1),PNk));

    # Samples X
    axs[0].bar(tNn, respNn, color=(0,0.75,0), linewidth=1)
    axs[0].bar(tSn, respSn, color=(0.75,0,0), linewidth=1)
    axs[0].set_xlabel('time')
    axs[0].set_box_aspect(1/4)
    axs[0].set_xlabel('trials')
    axs[0].set_ylabel('responses')
    axs[0].set_title('Responses over intermixed trials', fontsize=14);

    # Signal distribution
    axs[1].bar(k, histSn, color=(0.75,0,0), linewidth=1)
    axs[1].step(kn, PSkn, '-', color=(0.25,0,0), linewidth=2)
    axs[1].set_xlabel('strength k')
    axs[1].set_ylabel('probability')
    axs[1].set_box_aspect(1/2)
    axs[1].set_title('Distribution of responses to signal', fontsize=14);

    # Noise distribution
    axs[2].bar(k, histNn, color=(0,0.75,0), linewidth=1)
    axs[2].step(kn, PNkn, '-', color=(0,0.25,0), linewidth=2)
    axs[2].set_xlabel('strength k')
    axs[2].set_ylabel('probability')
    axs[2].set_box_aspect(1/2)
    axs[2].set_title('Distribution of responses to noise', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_poisson_responses_intermixed, LambdaN = (1, 10, 1), LambdaS = (1, 10, 1), Ntrial = (40, 400, 20) )
```

- Responses are described by a probability distribution.

- Different types of stimuli (signal or noise) elicit different response distributions.

# **Step Two:** Criterion determines observer report

Assume that subjective experience -- perceived clarity of target -- depends monotonically on neural response and can inform a volitional response.

For example, the observer could (if so instructed) report the perceived clarity on a scale of 1 to 10. Alternatively, the observer could classify (if so instructed) the perceived clarity in terms of 'present' / 'absent' categories.

*An important empirical observation is that -- in well-trained psychophysical observers -- subjective experience closely shadows sensory transduction (neural response). Thus, a psychophysical experiment can open a window to the 'neural hardware'*

By allowing only distinct categories of report, we motivate the observer to adopt consistent **decision criteria**.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_Four_step_STD.png" width="600">

(Slide-17)


# Decision uncertainty

Signal (presence) and noise (absence) distributions typically **overlap**!  Thus, an observer is bound to make mistakes (report incorrectly), either by falsely reporting 'presence' in absence trials, or falsely reporting 'absence' in presence trials. Deciding under conditions of uncertainty means that one cannot be always correct!

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_STD_b.png" width="400">

**We retain the x-axis label 'neural response k', on the assumption that the values map monotonically onto subjective experience informing behavioral reports.**

(Slide-18)

**Criterion for binary decision**

The (provably) optimal strategy of the observer is to adopt a consistent **criterion** and to report 'presence' whenever the neural response exceeds this value and 'absence' when it falls short.

In our specific example -- Poisson varability with $\lambda_\mathit{noise}=3$ and $\lambda_\mathit{signal}=8$ -- the optimal criterion is $c = 6$ (see further below).

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_STD_c.png" width="400">

While some report decisions will be correct (green bars), others will necessarily be incorrect (red  bars).

As mentioned, there are two kinds of possible mistakes: reporting 'presence' in absence trials and 'absence' in presence trials.

(Slide-19)

# Hits and misses

The combination of two possible stimuli (present/absent) and two possible reports (present/absent) makes for a total of four possible trial outcomes.

If we consider only 'present' trials, there are two possible outcomes:

**Hits** when the observer correctly reports 'presence'

**Misses** when the observer mistakenly reports 'absence'

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_STD_e.png" width="400">

Note that the fractions of hits and misses, $H(c)$ and $M(c)$, depend on the choice of criterion $c$.

(Slide-20)

# False alarms and correct rejections

If we consider only  'absent' trials, there are again two possible outcomes:

**False alarms** when the observer mistakenly reports 'presence'

**Correct rejections** when the observer correctly reports 'absence'

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_STD_d.png" width="400">

Once again, the fractions $F(c)$ and $R(c)$ depend on the choice of criterion $c$.

(Slide-21)

# Behavioral results

The fraction of hits and false alarms fully characterize the behavioural results:

$\begin{eqnarray}
H(c) = \sum_{k\geq c}^\infty P(k \mathit{\,if\,present}), \qquad\qquad F(c) = \sum_{k\geq c}^\infty P(k \mathit{\,if\,absent})
\end{eqnarray}$

If there are equally many presence and absence trials, the fraction of 'correct' responses and the fraction of 'present' responses are

$\begin{eqnarray}
P_\mathit{corr}(c) = \frac{H(c) + 1 - F(c)}{2}, \qquad\qquad P_\mathit{pres}(c) = \frac{ H(c) + F(c)}{2}
\end{eqnarray}$

In a more general case, where the number of presence and absence trials are $N_\mathit{pres}$ and $N_\mathit{abs}$, respectively, the fraction of correct responses is

$\begin{eqnarray}
P_\mathit{corr}(c) = \frac{N_\mathit{pres}}{N_\mathit{pres}+ N_\mathit{abs}} \, H(c) + \frac{N_\mathit{abs}}{N_\mathit{pres}+ N_\mathit{abs}} \, [1-F(c)]
\end{eqnarray}$

and the fraction of 'present' responses is

$\begin{eqnarray}
P_\mathit{pres}(c) = \frac{N_\mathit{pres}}{N_\mathit{pres}+ N_\mathit{abs}} \, H(c) + \frac{N_\mathit{abs}}{N_\mathit{pres}+ N_\mathit{abs}} \, F(c)
\end{eqnarray}$

(Slide-22)


# Summary step two: decision criterion

- Behavioural reports necessitate classification of continuum of subjective experience ('yes'/'no', 'present'/'absent', etc.).

- When there are two possible stimuli ('signal' or 'noise') and two possible behavioral reports ('yes' / 'no'), every trial has four possible outcomes: hit, misse, false alarm and correct rejection.

- The fraction of Hits and False alarms encapsulate all information:

$\begin{eqnarray}
\qquad\qquad H = P(\mathrm{'yes'}|\mathrm{'signal'}),\qquad\qquad F = P(\mathrm{'yes'}|\mathrm{'noise'})
\end{eqnarray}$

(Slide-23)




# **Third step:** receiver operating characteristic

A standard psychophysical task is described by two probability distributions and a decision criterion.

- The difference between distributions reflects the **'sensitivity'** of the observer for the stimulus in question. In a well-designed experiment, subjective sensitivity closely shadows sensory transduction.

- The adopted **'criterion'** is up to the observer. Nevertheless, we can influence this choice experimentally.

The **Receiver Operating Characteristic** takes advantage of this possibility in order to disambiguate **sensitivity** and **criterion**.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_Four_step_STD.png" width="600">

(Slide-24)


# Liberal and conservative criteria

Penalizing misses more than false alarms encourages observers to adopt a 'liberal' criterion, for example $c=2$.

Doing the opposite encourages a 'conservative' criterion, for example $c=8$.

In the illustration below, green areas represent correct reports (hits or correct rejections), whereas red areas represent incorrect reports (false alarms or misse).

Varying the criterion changes the relative size of green and red areas.

(Slide-25)


```python
from inspect import signature
# @title Liberal and conservative criteria {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from matplotlib.colors import LinearSegmentedColormap

def ComputeROC( LambdaN, LambdaS, Nbin, Criterion ):

    # range of possible neural responses
    k = np.arange(0, Nbin+1);

    # range of possible criteria
    c = k;
    cidx = np.where(c==Criterion)[0]


    # Poisson-distributions with mean lambda
    k = np.arange(0, Nbin+1);
    PSk = poisson.pmf( k, LambdaS )
    PNk = poisson.pmf( k, LambdaN )

    # Hits, Misses, False alarms, and Correct rejections
    # as a function of criterion

    PSk_hi = np.full((Nbin+1,Nbin+1),np.nan)
    PSk_lo = np.full((Nbin+1,Nbin+1),np.nan)
    PNk_hi = np.full((Nbin+1,Nbin+1),np.nan)
    PNk_lo = np.full((Nbin+1,Nbin+1),np.nan)

    kk_hi  = np.full((Nbin+1,Nbin+1),np.nan)
    kk_lo  = np.full((Nbin+1,Nbin+1),np.nan)

    for i in range(0,Nbin+1):
        crit = c[i];
        for j in range(0,Nbin+1):
            resp = k[j];
            if resp >= crit:
                PSk_hi[i,j] = PSk[j];
                PNk_hi[i,j] = PNk[j];
                kk_hi[i,j] = resp;
            else:
                PSk_lo[i,j] = PSk[j];
                PNk_lo[i,j] = PNk[j];
                kk_lo[i,j] = resp;

    Hc    = np.nansum(PSk_hi, axis=1)
    Mc    = np.nansum(PSk_lo, axis=1)
    FAc   = np.nansum(PNk_hi, axis=1)
    CRc   = np.nansum(PNk_lo, axis=1)

    # extract highlighted values
    PSklo       = PSk_lo[cidx,:]
    PSkhi       = PSk_hi[cidx,:]
    PNklo       = PNk_lo[cidx,:]
    PNkhi       = PNk_hi[cidx,:]

    klo         = kk_lo[cidx,:]
    khi         = kk_hi[cidx,:]

    # remove nans
    klo = klo[~np.isnan(klo)]
    khi = khi[~np.isnan(khi)]
    PSklo = PSklo[~np.isnan(PSklo)]
    PSkhi = PSkhi[~np.isnan(PSkhi)]
    PNklo = PNklo[~np.isnan(PNklo)]
    PNkhi = PNkhi[~np.isnan(PNkhi)]

    # add values for step plot
    #klo   = np.append(klo, [klo[-1]])
    #PSklo = np.append(PSklo, [PSklo[-1]])
    #PNklo = np.append(PNklo, [PNklo[-1]])

    #khi   = np.concatenate(([khi[0]], khi))
    #PSkhi = np.concatenate(([0], PSkhi))
    #PNkhi = np.concatenate(([0], PNkhi))

    # return results
    return klo, khi, Hc, Mc, FAc, CRc, PSklo, PSkhi, PNklo, PNkhi

def plot_liberal_and_conservative(LambdaN=3, LambdaS=8, Criterion=2):
    """(Slide-26)"""

    Nbin = 20;

    # range of possible neural responses
    k = np.arange(0, Nbin+1);

    # range of possible criteria
    c = k;
    cidx = np.where(c==Criterion)[0]

    # compute ROC
    (klo, khi, Hc, Mc, FAc, CRc, PSklo, PSkhi, PNklo, PNkhi) = ComputeROC( LambdaN, LambdaS, Nbin, Criterion );

    # proportion correct and false
    Pcorr = (Hc + 1 - FAc) / 2
    Pinco = (1 - Hc + FAc) / 2

    # Plot results
    fig, axs = plt.subplots( 3, 1, figsize=(6, 8))

    # Signal distribution
    axs[0].bar(klo, PSklo,  color=(0.5,0.0,0.0), linewidth=2, label='M')
    axs[0].bar(khi, PSkhi,  color=(0.0,0.5,0.0), linewidth=2, label='H')
    ylim = axs[0].get_ylim();
    axs[0].plot((c[cidx]-0.5, c[cidx]-0.5), ylim, 'k:', linewidth=2)
    axs[0].legend()
    axs[0].set_xlabel('resp k')
    axs[0].set_box_aspect(1/2)
    axs[0].set_ylabel('P(k)')
    axs[0].set_title('Probability', fontsize=14);

    # Signal distribution
    axs[1].bar(klo, PNklo,  color=(0.5,1.0,0.5), linewidth=2, label='CR')
    axs[1].bar(khi, PNkhi,  color=(1.0,0.5,0.5), linewidth=2, label='FA')
    ylim = axs[1].get_ylim();
    axs[1].plot((c[cidx]-0.5, c[cidx]-0.5), ylim, 'k:', linewidth=2)
    axs[1].legend()
    axs[1].set_xlabel('resp k')
    axs[1].set_box_aspect(1/2)
    axs[1].set_ylabel('P(k)')
    axs[1].set_title('Probability', fontsize=14);


    # Pcorrect and Pincorrect
    axs[2].step(c, Pcorr, color=(0.0,0.5,0.0), linewidth=2, label='correct')
    axs[2].step(c, Pinco, color=(0.5,0.0,0.0), linewidth=2, label='false')
    axs[2].step(c[cidx]-0.5, Pcorr[cidx], 'ko', markersize=5)
    axs[2].step(c[cidx]-0.5, Pinco[cidx], 'ko', markersize=5)
    axs[2].legend()
    axs[2].set_xlabel('liberal <- crit c -> conservative')
    axs[2].set_ylabel('fraction')
    axs[2].set_box_aspect(1/2)
    axs[2].set_title('P correct', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_liberal_and_conservative, LambdaN = (1, 10, 1), LambdaS = (1, 10, 1), Criterion = (1, 20, 1) )
```

# Optimal criterion

To determine the optimal criterion, we need to know the costs and benefits associated with correct and incorrect reports.  Assuming that all incorrect reports ('misses' and 'false alarms') are equally costly, the optimal criterion is the criterion that minimizes the total number of incorrect reports or, equivalently, maximizes the total number of correct reports.

In the illustration above, neither very liberal nor very conservative criterial maximize the green area (minimize the red).

The optimal criterion lies in between!

(Slide-27)



# Receiver-operating characteristic (ROC)

The ROC plots hits' against 'false alarms' (true positives against false positives), for different criterion values.

The area under the ROC (pale red) measures the difference between signal and noise distributions and represents the 'sensitivity' of the observer to the stimuli in question.

Note that this measure does not depend on the 'criterion'.  We will exploit this to disambiguate sensitivity and criterion.









```python
from inspect import signature
# @title Receiver operating characteristic {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from matplotlib.colors import LinearSegmentedColormap

def GetROCArea( c, FAc, Hc):

    # FAc = np.concatenate(([1],FAc))

    # Hc  = np.concatenate(([0],Hc))

    FAc  = FAc[::-1]

    Hc   = Hc[::-1]

    area = np.trapz(Hc, FAc)

    return area

def plot_receiver_operating_characteristic(LambdaN=3, LambdaS=8, Criterion=2):
    """(Slide-28)"""

    Nbin = 20;

    # range of possible neural responses
    k = np.arange(0, Nbin+1);

    # range of possible criteria
    c = k;
    cidx = np.where(c==Criterion)[0]

    # compute ROC
    (klo, khi, Hc, Mc, FAc, CRc, PSklo, PSkhi, PNklo, PNkhi) = ComputeROC( LambdaN, LambdaS, Nbin, Criterion );

    # proportion correct and false
    Pcorr = (Hc + 1 - FAc) / 2
    Pinco = (1 - Hc + FAc) / 2

    # Plot results
    fig, axs = plt.subplots( 2, 2, figsize=(6, 6))

    # Response distributions
    axs[0,0].bar(klo, PNklo,  color=(0.5,1.0,0.5), linewidth=2, label='CR')
    axs[0,0].bar(khi, PNkhi,  color=(1.0,0.5,0.5), linewidth=2, label='FA')
    axs[0,0].bar(klo, PSklo,  color=(0.5,0.0,0.0), linewidth=2, label='M')
    axs[0,0].bar(khi, PSkhi,  color=(0.0,0.5,0.0), linewidth=2, label='H')
    ylim = axs[0,0].get_ylim();
    axs[0,0].plot((c[cidx]-0.5, c[cidx]-0.5), ylim, 'k:', linewidth=2)
    axs[0,0].legend()
    axs[0,0].set_xlabel('resp k')
    axs[0,0].set_box_aspect(1/1)
    axs[0,0].set_ylabel('P(k)')
    axs[0,0].set_title('Probability', fontsize=14);

    # Hits, misses, false alarms, ...
    axs[1,0].step(c, Hc,  '-', color=(0.0,0.5,0.0), linewidth=2, label='H')
    axs[1,0].step(c, Mc,  '-', color=(0.5,0.0,0.0), linewidth=2, label='M')
    axs[1,0].step(c, FAc, '-', color=(1.0,0.5,0.5), linewidth=2, label='FA')
    axs[1,0].step(c, CRc, '-', color=(0.5,1.0,0.5), linewidth=2, label='CR')
    axs[1,0].plot(c[cidx]-0.5, Hc[cidx],  'ko', markersize=5)
    axs[1,0].plot(c[cidx]-0.5, Mc[cidx],  'ko', markersize=5)
    axs[1,0].plot(c[cidx]-0.5, FAc[cidx], 'ko', markersize=5)
    axs[1,0].plot(c[cidx]-0.5, CRc[cidx], 'ko', markersize=5)
    axs[1,0].legend()
    axs[1,0].set_xlabel('crit c')
    axs[1,0].set_ylabel('P(k>c), P(k<c)')
    axs[1,0].set_box_aspect(1/1)
    axs[1,0].set_title('Cumulative', fontsize=14);

    pcor = Pcorr[cidx]
    pinc = Pinco[cidx]

    # Pcorrect and Pincorrect
    axs[0,1].step(c, Pcorr, color=(0.0,0.5,0.0), linewidth=2, label='correct')
    axs[0,1].step(c, Pinco, color=(0.5,0.0,0.0), linewidth=2, label='false')
    axs[0,1].step(c[cidx]-0.5, pcor, 'ko', markersize=5)
    axs[0,1].step(c[cidx]-0.5, pinc, 'ko', markersize=5)
    axs[0,1].legend();
    axs[0,1].text(15, 0.25, f'Pcorr={pcor[0]:.2f}', fontsize=12, ha='center')
    axs[0,1].set_xlabel('crit c')
    axs[0,1].set_ylabel('fraction')
    axs[0,1].set_box_aspect(1/1)
    axs[0,1].set_title('P correct', fontsize=14);

    area = GetROCArea( c, FAc, Hc)

    # ROC
    axs[1,1].fill(np.concatenate(([1],FAc)), np.concatenate(([0],Hc)), color=(1.0,0.7,0.7))
    axs[1,1].plot(FAc, Hc, 'r-', linewidth=2)
    axs[1,1].plot(FAc[cidx], Hc[cidx], 'ko', markersize=5)
    axs[1,1].plot((0,1), (0,1), 'k:', linewidth=2)
    axs[1,1].text(0.65, 0.25, f'Area={area:.2f}', fontsize=12, ha='center')
    axs[1,1].set_xlabel('false alarms')
    axs[1,1].set_ylabel('hits')
    axs[1,1].set_box_aspect(1/1)
    axs[1,1].set_title('ROC', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_receiver_operating_characteristic, LambdaN = (1, 10, 1), LambdaS = (1, 10, 1), Criterion = (1, 20, 1) )
```

# Measures of sensitivity

To measure sensitivity independently of criterion, one typically uses **optimal performance** (i.e., performance for the optimal criterion).

Plotting this a function of the physical stimulus strength $s$ provides the **psychometric function**.

$\begin{eqnarray}
P_{opt}(s) = \frac{1}{2} + \frac{H( c_{opt})-F(c_{opt})}{2}, \qquad c_{opt} = f(s)
\end{eqnarray}$

Alternatively, one can plot **area under the ROC** as a function of physical stimulus strength.

In our synthetic example, physical signal strength $s$ corresponds to the differential neural response $\Delta \lambda$ to 'presence' and 'absence' stimuli:


$\begin{eqnarray}
\qquad\qquad\Delta \lambda = \lambda_\mathit{signal} - \lambda_\mathit{noise}
\end{eqnarray}$

```python
from inspect import signature
# @title Psychometric function {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from matplotlib.colors import LinearSegmentedColormap

def ComputeROCAlt( LambdaN, LambdaS, Nbin ):

    # range of possible neural responses
    k = np.arange(0, Nbin+1);

    # range of possible criteria
    c = k;

    # Poisson-distributions with mean lambda
    k = np.arange(0, Nbin+1);
    PSk = poisson.pmf( k, LambdaS )
    PNk = poisson.pmf( k, LambdaN )

    # Hits, Misses, False alarms, and Correct rejections
    # as a function of criterion

    PSk_hi = np.full((Nbin+1,Nbin+1),np.nan)
    PSk_lo = np.full((Nbin+1,Nbin+1),np.nan)
    PNk_hi = np.full((Nbin+1,Nbin+1),np.nan)
    PNk_lo = np.full((Nbin+1,Nbin+1),np.nan)

    kk_hi  = np.full((Nbin+1,Nbin+1),np.nan)
    kk_lo  = np.full((Nbin+1,Nbin+1),np.nan)

    for i in range(0,Nbin+1):
        crit = c[i];
        for j in range(0,Nbin+1):
            resp = k[j];
            if resp >= crit:
                PSk_hi[i,j] = PSk[j];
                PNk_hi[i,j] = PNk[j];
                kk_hi[i,j] = resp;
            else:
                PSk_lo[i,j] = PSk[j];
                PNk_lo[i,j] = PNk[j];
                kk_lo[i,j] = resp;

    Hc    = np.nansum(PSk_hi, axis=1)
    Mc    = np.nansum(PSk_lo, axis=1)
    FAc   = np.nansum(PNk_hi, axis=1)
    CRc   = np.nansum(PNk_lo, axis=1)

    # proportion correct
    Pcorrc = (Hc + 1 - FAc) / 2
    Pincoc = (1 - Hc + FAc) / 2

    # optimal performance
    Popt   = np.max(Pcorrc)

    # optimal criterion
    copt   = np.where(Pcorrc==Popt)[0]

    # extract optimal values
    PSklo       = PSk_lo[copt,:]
    PSkhi       = PSk_hi[copt,:]
    PNklo       = PNk_lo[copt,:]
    PNkhi       = PNk_hi[copt,:]

    klo         = kk_lo[copt,:]
    khi         = kk_hi[copt,:]

    # remove nans
    klo = klo[~np.isnan(klo)]
    khi = khi[~np.isnan(khi)]
    PSklo = PSklo[~np.isnan(PSklo)]
    PSkhi = PSkhi[~np.isnan(PSkhi)]
    PNklo = PNklo[~np.isnan(PNklo)]
    PNkhi = PNkhi[~np.isnan(PNkhi)]

    # get area under ROC
    ROCA = GetROCArea( c, FAc, Hc);

    # return results
    return klo, khi, PSklo, PSkhi, PNklo, PNkhi, Popt, ROCA

def plot_psychometric_function():
    """(Slide-32)"""

    Nbin = 20;

    # noise level
    LambdaN = 3;

    # signal levels
    LambdaMax = 11;
    LambdaSs = np.arange(LambdaN, LambdaMax, 1);

    LambdaDs = LambdaSs - LambdaN;

    # range of possible neural responses
    k = np.arange(0, Nbin+1);


    # perform ROC analyses for different signal levels

    ROCAs = np.zeros(len(LambdaSs))
    Popts = np.zeros(len(LambdaSs))

    fig, axs = plt.subplots( 3, 1, figsize=(6, 8))

    for s in range( 0, len(LambdaSs) ):

        (klo, khi, PSklo, PSkhi, PNklo, PNkhi, Popt, ROCA) = ComputeROCAlt( LambdaN, LambdaSs[s], Nbin )

        ROCAs[s] = ROCA
        Popts[s] = Popt

        # Plot response distributions
        if s==1:
            axs[0].bar(klo, PNklo,  color=(0.5,1.0,0.5), linewidth=2, label='CR')
            axs[0].bar(khi, PNkhi,  color=(1.0,0.5,0.5), linewidth=2, label='FA')
            axs[0].bar(klo, PSklo,  color=(0.5,0.0,0.0), linewidth=2, label='M')
            axs[0].bar(khi, PSkhi,  color=(0.0,0.5,0.0), linewidth=2, label='H')
            ylim = axs[0].get_ylim();
            axs[0].set_xlabel('resp k')
            axs[0].set_box_aspect(1/3)
            axs[0].set_ylabel('P(k)')
            axs[0].set_title(f'Strength={s:.2f}', fontsize=14);

        if s==4:
            axs[1].bar(klo, PNklo,  color=(0.5,1.0,0.5), linewidth=2, label='CR')
            axs[1].bar(khi, PNkhi,  color=(1.0,0.5,0.5), linewidth=2, label='FA')
            axs[1].bar(klo, PSklo,  color=(0.5,0.0,0.0), linewidth=2, label='M')
            axs[1].bar(khi, PSkhi,  color=(0.0,0.5,0.0), linewidth=2, label='H')
            ylim = axs[1].get_ylim();
            axs[1].set_xlabel('resp k')
            axs[1].set_box_aspect(1/3)
            axs[1].set_ylabel('P(k)')
            axs[1].set_title(f'Strength={s:.2f}', fontsize=14);

        if s==7:
            axs[2].bar(klo, PNklo,  color=(0.5,1.0,0.5), linewidth=2, label='CR')
            axs[2].bar(khi, PNkhi,  color=(1.0,0.5,0.5), linewidth=2, label='FA')
            axs[2].bar(klo, PSklo,  color=(0.5,0.0,0.0), linewidth=2, label='M')
            axs[2].bar(khi, PSkhi,  color=(0.0,0.5,0.0), linewidth=2, label='H')
            ylim = axs[2].get_ylim();
            axs[2].legend()
            axs[2].set_xlabel('resp k')
            axs[2].set_box_aspect(1/3)
            axs[2].set_ylabel('P(k)')
            axs[2].set_title(f'Strength={s:.2f}', fontsize=14);

    plt.tight_layout()
    plt.show()

    # plot psychometric function
    fig, axs = plt.subplots( 1, 1, figsize=(6, 6))

    idx = [1,4,7];

    axs.plot(LambdaDs, ROCAs, 'k-', linewidth=2)
    axs.plot(LambdaDs, Popts, 'k-', linewidth=2)
    axs.plot(LambdaDs, ROCAs, 'ro', markersize=5, label='ROCA')
    axs.plot(LambdaDs[idx], ROCAs[idx], 'ro', markersize=8)
    axs.plot(LambdaDs, Popts, 'bo', markersize=5, label='Popt')
    axs.plot(LambdaDs[idx], Popts[idx], 'bo', markersize=8)

    axs.legend()
    axs.set_xlabel('Strength = Lambda S - Lambda N')
    axs.set_box_aspect(1/2)
    axs.set_ylabel('Popt, ROCA')
    axs.set_title('Psychometric function', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_psychometric_function )
```

# Aside: Quantiles and displacement

'Displacement' $z$: positive or negative distance from *center* of normal distribution.

'Quantile' $R$: area under normal distribution, from $-\infty$ to value $z$.

*Quantile as a function of displacement*

$\begin{eqnarray}
\qquad\qquad R(z) = \frac{1}{2} + \frac{1}{2} \mathrm{erf} \frac{z}{\sqrt{2}}
\end{eqnarray}$

*Displacement as a function of quantile*

$\begin{eqnarray}
\qquad\qquad z(R) = \sqrt{2} \mathrm{erf}^{-1} \left(2 R - 1 \right)
\end{eqnarray}$

The 'displacement' with respect to a normal distribution is often called **z-score**.

Because of symmetry, we have

$\begin{eqnarray}
\qquad\qquad -z(R) = z(1-R)
\end{eqnarray}$

**Proof:**

$\begin{eqnarray}
\mathrm{erf}(x) \equiv \frac{2}{\sqrt{\pi}} \, \int_0^x \, e^{-x'^2} \, dx' \qquad \stackrel{x=z/\sqrt{2}}{\longrightarrow} \qquad \frac{1}{2} \mathrm{erf}\left(\frac{z}{\sqrt{2}} \right) = \frac{1}{\sqrt{2\pi}} \, \int_0^z \, e^{-z'^2/2} \, dz'
\end{eqnarray}$

(Slide-39)



```python
from inspect import signature
# @title Quantiles and displacement {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap


def plot_quantile_and_displacement( R = 0.16):
    """(Slide-40)"""

    # stepsize for displacement
    dz = 0.1;

    # range of displacements
    zi = np.arange(-3-dz/2,3+dz/2,dz);
    # probability density
    Pi = norm.pdf( zi, 0, 1);
    # cumulative probability mass
    Ci = norm.cdf( zi, 0, 1);

    # check Python implementation of error functiobn
    # Ri = norm.cdf( zi/np.sqrt(2) );
    # z2i = np.sqrt(2) * norm.ppf( Ri );

    # cutoff displacement corresponding to R
    zix = np.where(Ci<=R)[0]
    zR = zi[zix[-1]]+0.05

    # append arrays for fill plot
    zifill = zi[zix];
    zifill = np.concatenate((zifill, [zifill[-1]]))

    Pifill = Pi[zix];
    Pifill = np.concatenate((Pifill, [0]))

    # illustrate results
    fig, axs = plt.subplots( 1, 1, figsize=(6, 8))

    s = 0;
    axs.plot((zi[0],zi[-1]),(0.5,0.5),'k:',linewidth=1)
    axs.plot((0,0),(0,1),'k:',linewidth=1)
    axs.fill(zifill+0.05,Pifill,color=(1.0,0.5,0.5))
    axs.plot((zR,zR),(0,1),'r-',linewidth=1)
    axs.step(zi+0.05, Pi,  'k', linewidth=1, label='PDF')
    axs.step(zi+0.05, Ci, '-', color= (0.8,0.8,0.8), linewidth=1, label='CDF')

    ylim = axs.get_ylim();
    axs.set_xlabel('displacement z')
    axs.set_box_aspect(1/1)
    axs.set_ylim((0,0.75))
    axs.set_ylabel('PDF, CDF')
    axs.set_title(f'R={R:.2f}, z={zR:.2f}', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_quantile_and_displacement, R=(0,1,0.01) )
```

# Convert $H$ and $F$ to $z$-scores

Converting the observed $H$- and $F$-quantiles to displacements determines the position of the distributions relative to the criterion.

Note that $H$ and $F$ are quantiles *above* the criterion, while our conversion formulas assume quantiles *below* a cutoff.

For this reason, *positive* displacements shift the distribution *right*, whereas *negative* displacements shift the distribuiton *left*.

# Sensitivity $d'$

When assuming normal distributions, sensory sensitivity corresponds to the distance $d'= z(H) - z(F)$ between 'signal' and `noise' distributions.

This distance tells us how distinguishable the distributions are, independently of criterion.

# Criterion bias $c_\mathit{bias}$

The optimal criterion is in the middle between the two distribution centers.

If the actual criterion is shifted from the optimal criterion, there is a 'criterion bias'.

This 'criterion bias' can also be computed from the quantiles $H$ and $F$:

$\begin{eqnarray}
c_\mathit{bias} = -\frac{z(H) + z(F)}{2}
\end{eqnarray}$

Proof:

$\begin{eqnarray}
z(1-F) = \frac{d'}{2} + c =-z(F), \qquad\qquad z(H) = \frac{d'}{2} - c
\end{eqnarray}$

# Sensitivity $d'$ and criterion bias $c_\mathit{bias}$


**Sensitivity** $d'$ measures signal strength in terms of the mean difference between the 'signal' and 'noise'} distributions.  Its unit is the standard deviation of the underlying distributions.

$\begin{eqnarray}
d' \equiv z(H) - z(F)
\end{eqnarray}$


**Criterion bias** $c_\mathit{bias}$ measures the decision criterion of the observer (conservative, neutral, liberal), relative to the optimal criterion.   Its unit is the standard deviation of the underlying distributions.

$\begin{eqnarray}
 c_\mathit{bias} \equiv -\frac{z(H) + z(F)}{2}
\end{eqnarray}$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec12_STD_q.png" width="400">

(Slide-44)

# Observed and optimal performance

Observed performance follows directly from $H$ and $F$:

$\begin{eqnarray}
P_\mathit{obs} =  \frac{H}{2} + \frac{1-F}{2}
\end{eqnarray}$

Optimal performance is the quantile of $d'/2$:

$\begin{eqnarray}
P_\mathit{opt} = R(d'/2) =  \frac{1}{2} + \frac{1}{2} \mathit{erf} \frac{z(H)- z(F)}{\sqrt{8}}
\end{eqnarray}$

(Slide-45)





# Psychometric function

The psychometric functions that are obtained by assuming normal distributions are typically a good match for experimental observations, provided sensitivity $d'$ is proportional to the **logarithm** of physical stimulus strength $s$:

$\begin{eqnarray}
\qquad\qquad d' = \log(s)
\end{eqnarray}$

This suggests that physical stimuli are typically transduced **logarithmically** into neural responses.


```python
from inspect import signature
# @title Psychometric function (parametric) {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def Quant2Disp( R ):

    # Python for inverse error function
    zR = np.sqrt(2) * norm.ppf( R );

    return zR

def Disp2Quant( zR ):

    # Python for error function
    R = norm.cdf( zR/np.sqrt(2) )

    return R

def ComputeDiscriminability( zN, zS ):

    # center range
    zC = (zN + zS)/2;
    zN = zN - zC;
    zS = zS - zC;

    # discriminability
    dprime = zS - zN;

    # hits and false alarms
    H = 1 - Disp2Quant( zS );
    F = 1 - Disp2Quant( zN );

    # optimal performance
    Popt = Disp2Quant( dprime/2 );

    return dprime, H, F, Popt

def plot_psychometric_parametric( LambdaMax=10 ):
    """(Slide-46)"""

    Nbin = 20;

    # noise level
    LambdaN = 0;

    # signal levels
    LambdaSs = np.arange(LambdaN+0.2, LambdaMax, 0.2);

    # ROC analysis
    Popts = np.zeros(len(LambdaSs))
    Hs = np.zeros(len(LambdaSs))
    Fs = np.zeros(len(LambdaSs))

    # loop over signal strengths
    for s in range( 0, len(LambdaSs) ):

        (dprime, H, F, Popt) = ComputeDiscriminability( LambdaN, LambdaSs[s] );

        Popts[s] = Popt
        Hs[s] = H
        Fs[s] = F


    # illustrate results

    fig, axs = plt.subplots( 1, 2, figsize=(8, 4))

    # Hits and False alarms

    Ss = np.log( LambdaSs )

    axs[0].plot(Ss, Hs, 'g-', linewidth=2, label='Hits')
    axs[0].plot(Ss, Fs, 'r-', linewidth=2, label='False alarms')
    axs[0].legend()
    axs[0].set_xticks([-1,0,1,2])
    axs[0].set_xticklabels(['0.1', '1', '10', '100'])
    axs[0].set_xlabel('Physical strength s')
    axs[0].set_box_aspect(1/1)
    axs[0].set_ylabel('Quantile')

    axs[1].plot(Ss, Popts, 'k-', linewidth=2, label='P opt')
    axs[1].legend()
    axs[1].set_xticks([-1,0,1,2])
    axs[1].set_xticklabels(['0.1', '1', '10', '100'])
    axs[1].set_xlabel('Physical strength s')
    axs[1].set_box_aspect(1/1)
    axs[1].set_ylabel('P correct')

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_psychometric_parametric, LambdaMax=(5,15,1) )
```

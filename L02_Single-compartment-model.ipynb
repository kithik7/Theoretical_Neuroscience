# **1. Equivalent circuits**

The biophysical components of a neuron (i.e., cell membranes, intracellular plasma, channel proteins, etc) behave like electric circuit components and are characterized by electrical **voltage,  current, resistance, conductance,  capacitance**, and so on.


We take advantage of this fact and describe neurons in terms of electric **equivalent circuits**.   

Along the way, we introduce the concepts of resistance, conductance, ion-specific conductance, and capacitance.

(Slide-1)

# Axon lumen (Ohm's Law with resistance)

As a first example, consider an axon segment of length $L$, radius $a$, and crosssectional area $\pi a^2$.  This segment presents an electrical resistance $R_L$ to longitudinal currents:
$$\begin{eqnarray}
R_L = \frac{r_L \, L}{\pi a^2} \qquad\qquad\qquad r_L \,\, \textit{resistivity}
\end{eqnarray}$$
Note that resistance *grows* in proportion to length $L$ but *shrinks*  in inverse proportion to crossectional area $\pi a^2$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec02_Fig01.png" width="500">

D&A Fig. 5.2

For example, given values for length, radius, and resistivity $$L=100\,\mu m \qquad a=2 \,\mu m \qquad r_L \approx 1 \,\, k\Omega mm$$
we can calculate axon resistence as
$$\begin{eqnarray}
R_L = \frac{ 1 k\Omega \,mm \,\cdot \, 0.1 \, mm}{\pi\,\cdot \, 0.002^2 \,mm^2} \approx 8 \, M\Omega
\end{eqnarray}$$

(Slide-2)

# Ohm's Law

Resistance (symbol $R$, unit 'Ohm', written $\Omega$) is defined as the ratio between voltage drop and current or, equivalently, voltage drop is the product of resistance and current:
$$\begin{eqnarray}
 R[{\rm Ohm}] = \frac{V [{\rm Volt}]}{I [{\rm Ampere}] } \qquad \Leftrightarrow \qquad V [{\rm Volt}] = R[{\rm Ohm}] \, I[{\rm Ampere}]  
\end{eqnarray}$$

Causation works both ways!  Voltage drops cause currents to flow.  Flowing currents cause voltage to drop.

For example, the voltage drop $V_L$ required to drive a current $I_L = 1 \, nA$ through this axon segment turns out to be $8\,\rm{mV}$:

$$\begin{eqnarray}
V_L = 8\,M\Omega \cdot 1 \, nA = 8\times 10^{+6} \, \Omega \, \cdot \, 1\times 10^{-9} \, \rm{A} = 8\times 10^{-3}\, {\rm V}  = 8 \, \rm{mV}
\end{eqnarray}$$

The German circuit symbol for a resistence is a box. The international symbol is a coiled wire.

# Channel pore (Ohm's Law with conductance)

Conductance (symbol $G$, unit 'Siemens') is defined as the ratio between current and voltage (inverse of resistance).

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec02_Fig03.png" width="500">

D&A Fig. 5.1

Consider a channel protein with a cylindrical pore of $L_C=6\,{\rm nm}$ length and crosssectional area of $A_C=0.15\,{\rm nm}^2$ and its conductance $G_C$:

$$\begin{eqnarray}
G_C = \frac{1}{R_C} = \frac{A_C}{r_L \, L_C}\approx \frac{0.15 \,{\rm nm}^2}{ 1 {\rm k}\Omega \, {\rm mm} \,\cdot \, 6\,{\rm nm}} \approx \frac{1}{40 {\rm G}\Omega} = 25 \, {\rm pS}\\
\end{eqnarray}$$

With conductance, Ohm's Law takes the form
$$\begin{eqnarray}
V [{\rm Volt}] = \frac{I[{\rm Ampere}]}{G[{\rm Siemens}]} \qquad \Leftrightarrow \qquad I = G \, V\qquad\qquad {\rm S} \equiv \frac{\rm A}{\rm V} = \Omega^{-1}
\end{eqnarray}$$

(Slide-4)


# Ion-specific conductance

In neuronal membranes, any conductance is typically ion-specific, because channel proteins are typically ion-specific.  For any particular, the driving force is therefore the difference between membrane potential and reversal potential. We can take this into account by modifing Ohm's Law as follows:
$$\begin{eqnarray}
I_X \,\, = \,\, G_{X} \left( V_{mem} - E_{X} \right)\\
\end{eqnarray}$$
The ion current is proportional to conductance $G_{X}$ and to driving force $V_{mem} - E_{X}$


Membrane conductance $G_{X}$ may be calculated from specific resistivity $r_X$ (or conductivity $g_X$) and membrane surface area $a$:
$$\begin{eqnarray}
G_X = g_X \, a = \frac{a}{r_X} = \frac{0.01\,{\rm mm}^2}{1\,{\rm M}\Omega mm^2} = 10\, {\rm nS}
\end{eqnarray}$$
where we have used typical values for $r_X$ and $a$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec02_Fig04.png" width="500">

The circuit symbol for an ion-specific resistance is the symbol for a resistance (box) in series with the symbol for a battery (two parallel plates of different thickness and length).

# Membrane capacitance

The membrane capacitance is governed by the capacitor equation (see last lecture). Capacitance (symbol $F$, unit 'Farad') is defined as the ratio between the charge in, and the voltrage across, a capacitor.
$$\begin{eqnarray}
Q[Coulomb] = C_m [Farad] \,\, V_m [Volt]
\end{eqnarray}$$
The capacitance is proportional to the surface area of a capacitor. Its value is the specific capacitance (which depends on the thickness and dielectric properties of the membrane) times the surface area.
$$\begin{eqnarray}
C_m [F]= c_m [F \, mm^{-2}]\,\, A [mm^2], \qquad\qquad
c_m \approx 10 \,\, nF / mm^2
\end{eqnarray}$$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec02_Fig05a.png" width="300">

Typically, electrostatic repulsion resists any attempt at concentrating charges in one place. A capacitor is different and provides a comparatively comfortabe 'parking lot' for charges. It readily accumulate charges because it offers a large surface area with opposite charges close by.

(Slide-7)

# Time-derivative of the capacitor equation

Typically, we are interested in situations where a current $I_m$ flows through the membrane capacitance and changes the charge $Q$ over time. In fact, current is defined as the rate of change of charge:
$$\begin{eqnarray}
\frac{dQ}{dt} = I_m \qquad\qquad 1 {\rm A} = \frac{1 {\rm Coulomb}}{1 s}
\end{eqnarray}$$

In other words, by taking the time derivative of the capacitor equation, we may obtain an relation between current *through* and voltage *across* the membrane capacitance:

$$\begin{eqnarray}
I_m = \frac{dQ}{dt} = C_m \, \frac{d V_m}{dt}
\end{eqnarray}$$

Note that, in a **capacitance**, current is proportional to voltage **change**. This is fundamentally different from a **resistance**, where current is proportional to **voltage**.

# **3. Single-compartment model**

At the soma of a neuron, the membrane potential is relatively uniform across the entire surface.  The soma is **"electrotonically compact"** and can be treated as a single electrical **"compartment"**.  Such a compartment is characterized by a single membrane potential, membrane capacity, and membrane resistance.  Typical values are:
$$\begin{eqnarray}
R_m \approx \frac{1\,M\Omega mm^2}{0.1{\rm \,\,to\,\,}0.01\,mm^2} = 10{\rm \,\,to\,\,}100\,M\Omega
\end{eqnarray}$$

$$\begin{eqnarray}
C_m \approx 10\,nF/mm^2 \,\cdot \, \left( 0.1{\rm \,\,to\,\,}0.01 \, mm^2 \right)= 0.1{\rm \,\,to\,\,}1 \,nF
\end{eqnarray}$$

$$\begin{eqnarray}
\tau_m = R_m \, C_m \approx 10\,ms {\rm \,\,to\,\,} 100\,ms.
\end{eqnarray}$$

(Slide-15)


# Basic equation

We now derive the basic equation of the single-compartment model.  In addition to membrane resistance and capacitance, we consider an electrode, which has been inserted in the soma and which can serve two functions: it can deliver current and it can measure voltage.   As always, we define the membrane current as an **outward** current.   For an electrode current, the convention is different because it is defined as an **inward** current.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec02_Fig06.png" width="600">

D&A Fig. 5.3

Our goal is to determine the effect of input currents on the membrane potential.

(Slide-16)

# Basic equation, ctd.

Due to the convention about current direcdtion, the inward electrode current (which may be zero) must equal the outward membrane currents:

$$\begin{eqnarray}
I_e = I_R + I_C
\end{eqnarray}$$

Using Ohm's Law and the Capacitor Equation, we can express both components of the membrane current in terms of the membrane potential:

$$\begin{eqnarray}
I_e =  \frac{V_m}{R_m} + C_m \,\, \frac{d V_m}{dt}
\end{eqnarray}$$

$$\begin{eqnarray}
I_R = \frac{V_m}{R_m} \qquad\qquad\qquad\qquad I_C =  C_m \,\, \frac{d V_m}{dt}
\end{eqnarray}$$

This is the **basic equation** of the single-compartment model. Even though it is a **differential equation**, it brings our goal closer by relating input current to membrane potential.

Solving this equation will reveal how membrane potential develops over time.

(Slide-17)

# Alternative form

Often, this equation is written not in absolute but in 'specific' quantities (per unit membrane area).  In this
version, the equation reads

$$\begin{eqnarray}
\frac{I_e}{A} = \frac{V_m}{r_m} + c_m \,\, \frac{dV_m}{dt}
\end{eqnarray}$$

Recall that we use capital letters ($R_m$, $C_m$) for absolute resistance and capacitance, but lower case letters ($r_m$, $c_m$) for resistivity and conductivity.

(Slide-18)

# Equivalent circuit

$$\begin{eqnarray}
I_e =  \frac{V_m}{R_m} + C_m \,\, \frac{d V_m}{dt}
\end{eqnarray}$$

Here is the equivalent circuit for our basic equation. Inward electrode current divides into two outward currents (resistance and capacitance current):

# **4. Solving for voltage**

Given an arbitrary input current $I_e(t)$, we'd like to know the following:

* Membrane voltage $V_m(t)$ over time.

* Resistance current $I_R(t)$ over time.

* Capacitance current $I_C(t)$ over time.


Initially, we restrict ourselves to a constant input current $I_e(t) = I_0$.
In the next lecture, we shall generalize to time-varying input currents.

(Slide-20)

# Equilibrium potential

To begin to understand the implications of our basic equation,
we rearrange to isolate the voltage change:

$$\begin{eqnarray}
R_m \, C_m \, \frac{dV(t)}{dt} = -V(t) + R_m \, I_0
\end{eqnarray}$$

This tell us the change of potential $dV$ over time for a constant input current $I_0$. Specifically, the potential will move towards a particular value, namely, $R_m \, I_0$.  As long as $V$ is less than this value, $dV$ is positive and $V$ will increase.  As long as $V$ is larger than this value, $dV$ is negative and $V$ will decrease.

In other words, we see that $R_m \, I_0$ plays the role of a **steady-state** or **equilibrium potential**, $V_\infty = R_m \, I_0$
(dt. Grenzpotential).

$$\begin{eqnarray}
0 \quad \stackrel{!}{=} \quad R_m \, C_m \, \frac{dV(t)}{dt} = -V(t) + R_m \, I_e \qquad\Rightarrow\qquad V_\infty = R_m \, I_0
\end{eqnarray}$$

(Slide-21)

# Membrane time constant

Furthermore, we notice that the multiplicative constant $R_m \, C_m$ has units of time
$$\begin{eqnarray}
R_m \left[\frac{\rm V}{\rm A} \right] \, C_m \left[\frac{\rm A}{\rm V /s} \right]
\end{eqnarray}$$
so that we can interpret it as a **membrane time-constant**
$$\begin{eqnarray}
\tau_m = R_m \, C_m
\end{eqnarray}$$

Thus, the value of $\tau_m$ determines how rapidly the membrane potential approaches its steady-state or equilibrium value.

Can you understand why the time constant would increase with both membrane resistance and membrane capacitance?

(Slide-22)

# New intrepretation

With these two new definitions,

$$\begin{eqnarray}
V_\infty = R_m I_0, \qquad\qquad \tau_m = R_m\, C_m
\end{eqnarray}$$

our equation now reads

$$\begin{eqnarray}
\tau_m \, \frac{dV(t)}{dt} = -V(t) + V_\infty
\end{eqnarray}$$

or

$$\begin{eqnarray}
\frac{dV(t)}{dt} = - \frac{1}{\tau_m} \, \left[ V(t) - V_\infty \right]
\end{eqnarray}$$

(Slide-23)

# Exponential relaxation

The basic equation of our single-compartment model is a common linear differential equation (LDE), which describes an **exponential relaxation**.  We speak of an exponential relaxation when a state variable $x$ always returns (`relaxes') towards some equilibrium value and when the speed of this movement is proportional to the distance from the equilibrium value.

$$\begin{eqnarray}
\frac{d V(t)}{dt} = -\frac{1}{\tau} \left[ V(t) - V_\infty\right]
\end{eqnarray}$$

where $V_\infty$ is the equilibrium value, and $\tau$ is the relaxation time.

(Slide-24)

# Solution

Defining $V (0)=V_0$, the solution is

$$\begin{eqnarray}
V(t) = V_\infty + (V_0 - V_\infty) \, \exp\left(-\frac{t}{\tau} \right)
\end{eqnarray}$$

To check, differentiate the solution to obtain the left-hand side of the LDE:

$$\begin{eqnarray}
LHS = \frac{d V(t)}{dt} = -\frac{1}{\tau} \, (V_0 - V_\infty) \, \exp\left(-\frac{t}{\tau} \right)
\end{eqnarray}$$

and substitute the solution $V(t)$ to obtain the right-hand side:

$$\begin{eqnarray}
RHS = -\frac{1}{\tau} \, \left[V(t) - V_\infty \right] = -\frac{1}{\tau} \, (V_0 - V_\infty) \, \exp\left(-\frac{t}{\tau} \right)
\end{eqnarray}$$

Find that both sides are identical. QED.

(Slide-25)

```python
# @title Exponential relaxation of membrane potential {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# See how the membrane potential develops over time for different parameter choices !

# Define an interactive plot
def plot_exponential_relaxation(V_0=10, V_infty=0, tau_m = 20):
    """ Slide-26 """
    #V_0     = 10                                  # [mV] initial potential
    #V_infty = 0                                   # [mV] equilibrium potential
    #tau_m   = 10                                  # [ms] time constant

    ti   = np.linspace(0, +100, 200)               # [ms] sample range of times

    Vi_infty = np.linspace(V_infty,V_infty,200)    # [mV] equilibrium potentail, intial value V_0
    Vi_infty[0] = V_0

    Vi = Vi_infty + (V_0 - Vi_infty) * np.exp( -ti / tau_m );            # time-course of exponential relaxation

    plt.plot(ti, Vi, label='potential')
    plt.plot([0, 100], [0,0], label='zero')
    plt.plot([0, 100], [V_0,V_0], label='initial value')
    plt.plot(ti, Vi_infty, label='equilibrium value')
    plt.xlabel('time [ms]')
    plt.ylabel('potential [mV]')
    plt.title('Exponential relaxation')
    plt.legend()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_exponential_relaxation, V_0=(-50, +50, 1), V_infty=(-50, +50, 1), tau_m=(1, 100, 1))
```

# Summary

* Basic equation: inward current equals outward current (at all times).

* Electrode current divides into resistance and capacitance currents (at all times).

* Equilibrium is reached when the membrane potential reaches a value where resistance current (outward current) balances the electrode current (inward current).

* This development towards equilibrium is an **exponential relaxation**.

# **5. Solving for currents**

Our basic equation

$$\begin{eqnarray}
\tau_m \frac{d V(t)}{dt} = - V(t) + V_\infty, \qquad\qquad \tau_m = R_m C_m, \qquad V_\infty = R_m I_0
\end{eqnarray}$$

has the solution

$$\begin{eqnarray}
V(t) = V_\infty + (V_0 - V_\infty) \, \exp\left(-\frac{t}{\tau_m} \right)
\end{eqnarray}$$

With this solution for the development of voltage, we can also obtain the development of resistance and capacitance currents.

(Slide-27)

# Solving for currents, ctd.

Specifically, we can use  Ohm's law and the capacitor equation

$$\begin{eqnarray}
I_R(t) = \frac{V(t)}{R_m} \qquad\qquad  I_C(t) = C_m \, \frac{dV(t)}{dt}
\end{eqnarray}$$
to compute the resistance current


$$\begin{eqnarray}
I_R(t) =  \frac{V_\infty}{R_m} \, \left[1 - \exp\left(-\frac{t}{\tau_m} \right)\right] = I_0 \, \left[1 - \exp\left(-\frac{t}{\tau_m} \right)\right]
\end{eqnarray}$$

and the capacitance current

$$\begin{eqnarray}
I_C(t) = C_m \, \frac{-1}{\tau_m} \, \left[ - V_\infty \, \exp\left(-\frac{t}{\tau_m} \right)\right] = I_0 \, \exp\left(-\frac{t}{\tau_m} \right)
\end{eqnarray}$$

(Slide-28)

```python
# @title Single-compartment model {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# See how both membrane potential and membrane currents develop over time for different parameter choices !

# Define an interactive plot
def plot_current_relaxation(V_0=0, R_m=10, C_m = 10, I_e=40):
    """ Slide-29 """
    #V_0     = 10                                  # [mV] initial potential
    #R_m     = 10                                  # [MOhm] membrane resistance
    #C_m     = 10                                  # [nF]   membrane capacitance
    #I_e     = 10                                  # [nA]   electrode current

    ti   = np.linspace(0, +500, 200) ;             # [ms] sample range of times
    V_infty = R_m * I_e ;                          # [mV = MOhm * nA] equilibrium potential
    tau_m = R_m * C_m ;                            # [ms = MOhm * nF] time constant

    Ii_e     = np.linspace(I_e,I_e,200) ;  # [nA] electrode current, intial value zero
    Ii_e[0]  = 0;
    Vi_infty = Ii_e * R_m;

    Vi  = Vi_infty + (V_0 - Vi_infty) * np.exp( -ti / tau_m );            # time-course of exponential relaxation
    ICi = Ii_e * np.exp( -ti / tau_m );
    IRi = Ii_e * (1 - np.exp( -ti / tau_m ) );

    # Create a figure and a grid of subplots (1 rows, 3 columns)
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))

    # First subplot
    axs[0].plot(ti, Vi, 'k', ti, Vi_infty, 'b');
    axs[0].set_title('membrane potential');
    axs[0].set_xlabel('time [ms]');
    axs[0].set_ylabel('potential [mV]');


    # Second subplot
    axs[1].plot(ti, IRi, 'r', ti, Ii_e, 'b');
    axs[1].set_title('resistance current');
    axs[1].set_xlabel('time [ms]');
    axs[1].set_ylabel('current [nA]');

    # Third subplot
    axs[2].plot(ti, ICi, 'g', ti, Ii_e, 'b');
    axs[2].set_title('capacitance current');
    axs[2].set_xlabel('time [ms]');
    axs[2].set_ylabel('current [nA]');


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_current_relaxation, V_0=(-50, +50, 1), R_m=(1, 100, 1), C_m=(1, 100, 1), I_e=(1, 100, 1))
```

# **6. Ion-specific currents (advanced)**

So far, we have ignored the fact that membrane currents are carried by particular ions.   We now make our single-compartment model more realistic by taking into account ion-specific conductances and reversal potentials.  In our basic equation, we make explicit the membrane current *per unit area*, $i_m$:

$$\begin{eqnarray}
c_m \,\, \frac{dV}{dt} = -i_m + \frac{I_e}{A}
\end{eqnarray}$$

In the past, we have calculated $i_m$ from the membrane potential $V$ and the total membrane resistance $r_m$.

$$\begin{eqnarray}
i_m = \frac{V}{r_m}
\end{eqnarray}$$

Now we want to substitute a more detailed expression, which takes ion-specific conductances into account.

(Slide-32)

# Single ion conductance

Total membrane current carried by ion $X$, with reversal potential $E_X$ and conductivity $g_X$  ( $=\frac{1}{r_X}$ ).    

$$\begin{eqnarray}
c_m \,\, \frac{dV_m}{dt} = -i_m + \frac{I_e}{A}
\qquad\qquad
i_m = g_{X} (V_m - E_X)
\end{eqnarray}$$

# Multiple ion conductances

When total membrane current is carried by multiple ions, say $Na^+$ and $K^+$, we need to take into account the reversal potential and specific conductivity of each ion:

$$\begin{eqnarray}
c_m \,\, \frac{dV_m}{dt} = -i_m + \frac{I_e}{A}
\qquad\qquad
i_m = g_{Na} \left(V_m - E_{Na} \right) +  g_{K} \left(V_m - E_{K} \right)
\end{eqnarray}$$

# Synaptic conductances

In biological neurons, the total membrane current includes synaptic currents, which are also carried by particular ions.   In contrast to the ionic conductances we have encountered until now, synaptic conductances change over time (because they are opened by pre-synaptic activity):

$$\begin{eqnarray}
i_m = g_{syn}(t) \left(V_m - E_{syn} \right) + g_{Na} \left(V_m - E_{Na} \right) +  g_{K} \left(V_m - E_{K} \right)
\end{eqnarray}$$

# Generalization

The membrane current is the sum of all currents through the different types of ion channels, indexed by $i$.  This includes leakage channels, voltage-gated channels, and synaptic channels (which are ligand-gated).

For ion-selective channels, the current vanishes when the membrane potential $V$ equals the reversal potential $E_i$.  For other values of $V$, the current increases or decreases linearly with $V-E_i$.  Accordingly, $V-E_i$ can be considered the **driving force** for currents through conductance $i$.

We can now compute the membrane current per unit area $i_m$ from the conductances $g_i$ per unit area and the reversal potentials $E_i$:
$$\begin{eqnarray}
c_m \,\, \frac{dV_m}{dt} = -i_m + \frac{I_e}{A}
\qquad\qquad
i_m = \sum_i \, g_i \left( V - E_i \right)
\end{eqnarray}$$

(Slide-36)

# Example

For example, a particular neuron might have channels for $K^+$, $Na^+$, $Ca^{2+}$, and $Cl^-$.  In this case, we might write:

$$\begin{eqnarray}
i_m \,\, = \,\, g_{K^+} \left( V - E_{K^+} \right) \,\,+\,\, g_{Na^+} \left( V - E_{Na^+} \right) \,\,+\,\, g_{Ca^{2+}} \left( V - E_{Ca^{2+}} \right) \,\,+\,
\\
\\
+\,\, g_{Cl^+} \left( V - E_{Cl^+} \right) \,\,+\,\, {g_L} \left( V - E_L \right)
\end{eqnarray}$$

where ${g_L}$ and $E_L$ are the effective conductance and reversal potential of all remaining channels (including ion pumps), that are not explicitly included.   

Channels with $E_i \approx V_{rest}$ are sometimes called **shunt conductances** (e.g., $Cl^-$).

(Slide-37)

# Summary of "Ion-specific conductances"

* Ion-specific conductances take the ionic reversal potential into account

* Total conductance is sum the of individual conductances

* Total current is the sum of individual currents

* Synaptic conductances are treated in the same way

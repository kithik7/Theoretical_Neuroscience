# **Abstract:**

PCA is a useful way to summarize high-dimensional data (repeated observations of multiple variables).
This lecture provides the underlying linear algebra needed for practical applications.  It also emphasizes consistent notation.  The central ideas of PCA are **orthonormal coordinate** systems, the distinction between **variance and covariance**, and the possibility of choosing an orthonormal basis to **eliminate covariance**.

<br>

Technically, PCA may be performed either by **eigenvector analysis** of the covariance matrix or by **singular value decomposition** of the original observation matrix.  Both variants are described.  

 <br>

 We illustrate the method with a non-biological example (image-denoising) and with a biological example (multi-unit activity, see **Exercise 8**).

**Motivation**


<br>

Principal components analysis (PCA) is one of the most valuable results of applied linear algebra.  

<br>

It is widely used -- from neuroscience to computer graphics -- because it is an easy way to simplify confusing data sets.

With minimal effort, PCA dramatically reduces the dimensionality of a large data sent and potentially reveals a simpler structure hiding within.

<br>

During brain development, an (approximate) PCA of sensory inputs is performed by activity-dependent Hebbian plasticity.

# Variables and observations

We distinguish between *variables* (types of measurements) and *observations* (distinct times or conditions of measurements). Each *variable* is *observed* multiple times.

<br>

Think of variables as 'effects' (possibly mixed, redundant, or noisy) and observations as variations over 'causes' (due to differences in time, condition, or other factors).

<br>

For example, what kind of food people eat is a *variable*, in that it is an 'effect' of many causal factors such as climate, culture, economics, and so on.

<br>

When considering different countries, historical periods, or climate zones,  these causal factors will presumably differ from case to case.    The different cases being compared may be considered *observations*.

# Summary

<br>

In many fields (including neuroscience), we observe a multitude of variables under different conditions or at different times.
It often makes sense to seek a simpler way to describe of the total variance (sum of variances over condition or time).
PCA considers observations (conditions or times) as points in a high-dimensional space, the dimensionality of which is determined by the number of variables.
PCA computes an alternative orthonormal basis for this space and it ranks the basis vectors ('principal components') by the variance captured.
In effect, it identifies the points of view (directions of projection) from which observations appear most variable.

# **2. Orthonormal bases**

Consider  $m$ variables, **each with zero mean** and observed $n$ times, with $i^{th}$ observations grouped into  **column** vectors (for now!)

$$
\hat{\,\bf x_i} = \left( \begin{array}{c} x_{1i} \\ \vdots \\ x_{mi} \end{array}\right)\qquad \sum_{i=1}^{n} \, x_{ji} = 0, \qquad  i=1\ldots n, \qquad j=1\ldots m
$$

and all $n$ observations collected into a matrix with $N$ columns and $M$ rows

$$
{\bf X} =  \left( \hat{\,\bf x_1 }\quad  \hat{\,\bf x_2} \quad \ldots \quad \hat{\,\bf x}_n \right)  =\left(  \begin{array}{ccc} x_{11} & \ldots & x_{1n} \\ \vdots & \ddots & \vdots \\ x_{m1} & \ldots & x_{mn} \end{array} \right)  \in \mathbb{R}^{m \times n}
$$

Note the indexing with $x_{\mathit{row}, \mathit{column}}$ or  $x_{\mathit{variable}, \mathit{observation}}$:

<br>

**rows $\simeq$ variables** and **columns $\simeq$ observations**

# Orthonormal basis

Implictly, this assumes a basis of $m$ orthonormal **row vectors** ${\bf b}_j = (b_{11}  \ldots b_{1m})$, each with $m$ elements, ${\bf B} \in \mathbb{R}^{m\times m}$:

$$
{\bf B}  = \left( \begin{array}{c} {\bf b}_1  \\   {\bf b}_2  \\ \vdots \\  {\bf b}_m   \end{array} \right) = \left( \begin{array}{cccc} b_{11}  & b_{12} & \ldots & b_{1m}  \\  b_{21} & b_{22}  & \ldots & b_{2m}  \\ \vdots & \vdots & \ddots & \vdots \\ b_{m1} & b_{m2} & \ldots & b_{mm} \end{array} \right) = \left( \begin{array}{cccc} 1  & 0 & \ldots & 0  \\  0 & 1  & \ldots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \ldots & 1 \end{array} \right) = {\bf I}
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec13_FigNaiveBasis.png" width="200">

<br>

'Orthonormal' means of unit length and pairwise orthogonal

$$
{\bf b}_j \cdot {\bf b}_k = \delta_{jk} \qquad\qquad {\bf B} \cdot {\bf B}^T = {\bf B}^T \cdot {\bf B} = {\bf I}
$$

Project **columns** $\hat{\,\bf x}_i  \in \mathbb{R}^{m \times 1}$ of *individual* observations onto **rows** ${\bf b}_j$ of basis

<br>

$$
 \hat{\,\bf x}_i = {\bf B} \cdot  \hat{\,\bf x}_i =  \left( \begin{array}{c} {\bf b}_1 \cdot  \hat{\,\bf x}_i \\   {\bf b}_2 \cdot  \hat{\,\bf x}_i  \\ \vdots \\  {\bf b}_m \cdot  \hat{\,\bf x}_i  \end{array} \right) = \left( \begin{array}{c} b_{11} x_{1i} + \ldots + b_{1m} x_{mi}  \\  b_{21} x_{1i} + \ldots + b_{2m} x_{mi} \\ \vdots \\  b_{m1} x_{1i} + \ldots + b_{mm} x_{mi} \end{array} \right) = \left( \begin{array}{c} x_{1i}  \\ x_{2i}  \\ \vdots \\   x_{mi}  \end{array} \right)
$$

<br>

Project entire matrix ${\bf X}  \in \mathbb{R}^{m \times n}$ of *all* observations onto basis

<br>

$$
{\bf X} = {\bf B}\cdot  {\bf X} = \left( {\bf B} \cdot  \hat{\,\bf x}_1 \ldots {\bf B} \cdot \hat{\,\bf x}_n \right)  = \left( \begin{array}{ccc} {\bf b}_1 \cdot  \hat{\,\bf x}_1 & \ldots & {\bf b}_1 \cdot  \hat{\,\bf x}_n  \\   {\bf b}_2 \cdot \hat{\,\bf x}_1  & \ldots  & {\bf b}_2 \cdot  \hat{\,\bf x}_n \\ \vdots & \ddots & \vdots \\  {\bf b}_m \cdot  \hat{\,\bf x}_1 & \ldots & {\bf b}_m \cdot  \hat{\,\bf x}_n \\   \end{array} \right)   \in \mathbb{R}^{m \times n}
$$

# Many orthonormal bases are possible


<br>

The orthonormal basis ${\bf B} \in \mathbb{R}^{m\times m}$ reflects how we happened to collect the observations.

<br>

Many other orthonormal bases ${\bf P}   \in \mathbb{R}^{m\times m}$ may be obtained as linear combinations of ${\bf B}$.

<br>

Perhaps in terms of other orthonormal bases, we can re-express our observations more simply or more meaningfully?


# Alternative basis

<br>

Alternative basis ${\bf P}  \in \mathbb{R}^{m\times m}$ of orthonormal **row** vectors ${\bf p}_j = \left( p_{j1} \ldots p_{jm} \right)$:

<br>

$$
{\bf P}  =  \left( \begin{array}{c} {\bf p}_1 \\ \vdots \\ {\bf p}_m \end{array} \right)  =\left(  \begin{array}{ccc} p_{11} & \ldots & p_{1m} \\ \vdots & \ddots & \vdots \\ p_{m1} & \ldots & p_{mm} \end{array} \right) \in \mathbb{R}^{m \times m}
$$

<br>

$$
 {\bf p}_j \cdot {\bf p}_k = \delta_{jk} \qquad\qquad {\bf P} \cdot  {\bf P}^T = {\bf P}^T \cdot {\bf P}  = {\bf I}
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec13_FigAlternativeBasis.png" width="200">

# Projection to alternative basis

<br>

Transform individual observations ${\bf x}_ i \to {\bf y}_ i$!  Obtain $y$'s as dot products (projections) of ${\bf x}_ i$ on ${\bf p}_j$!   


$$
 \hat{\,\bf y}_i = {\bf P}  \cdot   \hat{\,\bf x}_i = \left( \begin{array}{c} {\bf p}_1 \cdot  \hat{\,\bf x}_i  \\   {\bf p}_2 \cdot  \hat{\,\bf x}_i  \\ \vdots \\  {\bf p}_m \cdot  \hat{\,\bf x}_i  \end{array} \right) = \left( \begin{array}{c} p_{11} x_{1i} + \ldots + p_{1m} x_{mi}  \\  p_{21} x_{1i} + \ldots + p_{2m} x_{mi} \\ \vdots \\  p_{m1} x_{1i} + \ldots + p_{mm} x_{mi} \end{array} \right)
= \left( \begin{array}{c} y_{1i}  \\   y_{2i}  \\ \vdots \\  y_{mi}  \end{array} \right)
$$

<br>

Transform all observations ${\bf X} \to {\bf Y}$

<br>

$$
{\bf Y} = {\bf P} \cdot  {\bf X} = \left( {\bf P} \cdot  \hat{\,\bf x}_1, \ldots ,{\bf P} \cdot  \hat{\,\bf x}_m \right)  = \left( \begin{array}{ccc} {\bf p}_1 \cdot \hat{\,\bf x}_1 & \ldots & {\bf p}_1 \cdot  \hat{\,\bf x}_m  \\   {\bf p}_2 \cdot  \hat{\,\bf x}_1  & \ldots  & {\bf p}_2 \cdot  \hat{\,\bf x}_m \\ \vdots & \ddots & \vdots \\  {\bf p}_m \cdot  \hat{\,\bf x}_1 & \ldots & {\bf p}_m \cdot  \hat{\,\bf x}_m \\   \end{array} \right)  \in \mathbb{R}^{m \times n}
$$

<br>

Keep in mind that $ \hat{\,\bf x}_i$ and $ \hat{\,\bf y}_i$ are **columns**, whereas ${\bf b}_i$ and ${\bf p}_i$ are **rows** !


# Summary

<br>

Observations ${\bf X}$ need not be expressed in native basis ${\bf B}$.

Alternative orthonormal bases ${\bf P}$ are linear combinations of ${\bf B}$.

<br>

Transformation to new basis is performed conveniently by matrix multiplication

$$
{\bf Y} = {\bf P}  \cdot  {\bf X}
$$

because dot products compute projections!

<br>

Row vectors ${\bf p}_j$ will become *principal components* of ${\bf X}$.

<br>

What would be a good choice for ${\bf P}$?

<br>

In which respect could ${\bf Y}$ be an improvement over ${\bf X}$?

# Variance definition

The variance of **zero-mean** variable $j$ over observations $i = 1, \ldots n$ is defined as

<br>

$$\begin{eqnarray}
\sigma_j^2 &=& \frac{1}{n-1} \left[ x_{j1}^2 \, + \, \ldots \, + \,  x_{jn}^2\right] =  
\\
&=& \frac{1}{n-1} \left( x_{j1} \,\, \ldots  \,\, x_{jn}\right) \cdot \left(\begin{array}{c} x_{j1}\\  \vdots  \\ x_{jn}\end{array} \right) =
\\
&=& \frac{1}{n-1} \, {\bf x}_ j \cdot {\bf x}_ j
\end{eqnarray}$$

where ${\bf x}_ j$ represents the  **rows** of ${\bf X}$!

<br>

The total variance over all rows $j$ reflects the total squared norm

$$
 \sum_j \, \sigma_j^2 = \frac{1}{n-1}\sum_j \, {\bf x}_ j \cdot {\bf x}_ j =  \frac{1}{n-1}\sum_{j=1}^m \, \sum_{i=1}^n \, x_{ji}^2
$$

# Covariance matrix

<br>

Covariance of dimensions or rows $j$ and  $k$

$$
\sigma^2_{jk} = \frac{1}{n-1} \left[ x_{j1}x_{k1} + \ldots +  x_{jn}x_{kn}\right] =  \frac{1}{n-1} \left( x_{j1} \,\, \ldots  \,\, x_{jn}\right) \cdot \left(\begin{array}{c} x_{k1}\\  \vdots  \\ x_{kn}\end{array} \right)
$$

<br>

$$\begin{eqnarray}
{\bf C}_{\bf X} &=& \frac{1}{n-1} \left(\begin{array}{cccc}
\sigma_{11}^2 & \sigma_{12} & \ldots & \sigma_{1m} \\
\sigma_{21} & \sigma_{22}^2 & \ldots & \sigma_{2m}  \\
\vdots & \vdots & \ddots & \vdots \\
\sigma_{m1} & \sigma_{m2} & \ldots & \sigma_{mm}^2 \end{array} \right)   =
\\
\\
\\
&=&  \frac{1}{n-1} \left(\begin{array}{cccc}
{\bf x}_ 1 \cdot {\bf x}_ 1 & {\bf x}_ 1 \cdot {\bf x}_ 2 & \ldots & {\bf x}_ 1 \cdot {\bf x}_ m \\
{\bf x}_ 2 \cdot {\bf x}_ 1 & {\bf x}_ 2 \cdot {\bf x}_ 2 & \ldots & {\bf x}_ 2 \cdot {\bf x}_ m \\
\vdots & \vdots & \ddots & \vdots \\
{\bf x}_ m \cdot {\bf x}_ 1 & {\bf x}_ m \cdot {\bf x}_ 2 & \ldots & {\bf x}_ m \cdot {\bf x}_ m \end{array} \right)  =
\end{eqnarray}$$

# ctd

<br>

$$\begin{eqnarray}
&=&  \frac{1}{n-1} \left(\begin{array}{cccc}
{\bf x}_ 1 \cdot {\bf x}_ 1 & {\bf x}_ 1 \cdot {\bf x}_ 2 & \ldots & {\bf x}_ 1 \cdot {\bf x}_ m \\
{\bf x}_ 2 \cdot {\bf x}_ 1 & {\bf x}_ 2 \cdot {\bf x}_ 2 & \ldots & {\bf x}_ 2 \cdot {\bf x}_ m \\
\vdots & \vdots & \ddots & \vdots \\
{\bf x}_ m \cdot {\bf x}_ 1 & {\bf x}_ m \cdot {\bf x}_ 2 & \ldots & {\bf x}_ m \cdot {\bf x}_ m \end{array} \right)  =  
\\
\\
\\
&=& \frac{1}{n-1} \left(\begin{array}{c} {\bf x}_ 1  \\ {\bf x}_ 2  \\ \vdots  \\ {\bf x}_ m  \end{array} \right)
\left(\begin{array}{cccc} {\bf x}_ 1^T  {\bf x}_ 2^T  \ldots   {\bf x}_ m^T \end{array}\right) =
 \frac{1}{n-1} \, {\bf X} \cdot  {\bf X}^T \in \mathbb{R}^{m\times m}
\end{eqnarray}$$

<br>

Recall that ${\bf x}_ j\in \mathbb{R}^{1\times n}$, ${\bf X}\in \mathbb{R}^{m\times n}$, ${\bf X}^T\in \mathbb{R}^{n\times m}$, and ${\bf C}_{\bf X}   \in \mathbb{R}^{m\times m}$.

# Interim summary

<br>

Properties of ${\bf C}_{\bf X} = \frac{1}{n-1} \, {\bf X} \cdot {\bf X}^T$

<br>

- ${\bf C}_{\bf X}$ is a square symmetric $m \times m$ matrix.

<br>

- Its diagonal terms are the *variances* of individual variables (rows of observations ${\bf X}$).

<br>

- Its off-diagonal terms are the *covariances* between different variables (two rows of observations ${\bf X}$).

<br>

- ${\bf C}_{\bf X}$ captures the correlations between all pairs of observations.

<br>

- The diagonal terms represent non-redundant variance (unique to each variable).  The diagonal (trace) equals *total variance*!

<br>

- The off-diagonal terms represent *redundant variance* (shared between two variables).

<br>

- A symmetric matrix has real eigenvalues, is diagonizable and has orthogonal eigenvectors. If it is also real, the eigenvalues are non-negative ('positive semi-definite').

$$\begin{eqnarray}
{\bf C}_{\bf X}
&=& \frac{1}{n-1} \left(\begin{array}{cccc}\sigma_{1}^2 & 0 & \ldots & 0 \\
0 & \sigma_{2}^2 & \ldots & 0  \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0& \ldots & \sigma_{m}^2 \end{array} \right)  +
\\

&+& \frac{1}{n-1} \left(\begin{array}{cccc} 0 & \sigma_{12} & \ldots & \sigma_{1m} \\
\sigma_{21} & 0 & \ldots & \sigma_{2m}  \\

\sigma_{m1} & \sigma_{m2} & \ldots & 0 \end{array} \right)
\end{eqnarray}$$

<br>

$\qquad\qquad\qquad\qquad \mathit{total~variance} \qquad\qquad\qquad\qquad \mathit{redundant~variance}$

# Signal-to-noise ratio

<br>

Sometimes, redundant and total variance are used to define a signal-to-noise ratio
$$
\mathit{SNR} = \frac{\sigma^2_{\mathit{total}}}{\sigma^2_{\mathit{redundant}}}
$$

<br>

It also specifies the fractional distribution of variance between signal and noise

$$
f_{\mathit{signal}} = \frac{\mathit{SNR}}{1+\mathit{SNR}} = \frac{\sigma^2_{\mathit{total}}}{\sigma^2_{\mathit{total}} + \sigma^2_{\mathit{redundant}}}
$$

<br>

$$
f_{\mathit{noise}} = \frac{1}{1+\mathit{SNR}} = \frac{\sigma^2_{\mathit{redundant}}}{\sigma^2_{\mathit{total}} + \sigma^2_{\mathit{redundant}}}
$$

# Choosing transformation ${\bf P}$

<br>

We can now formulate the criteria for choosing a transformation that helps reduce redundancy

$$
{\bf Y} = {\bf P}  \cdot  {\bf X}, \qquad\qquad {\bf C}_{\bf X} \to  {\bf C}_{\bf Y}
$$

<br>

- Seek to concentrate variance within as few dimensions of $\bf Y$ as possible.

<br>

- In identifying the most variable dimensions, we focus on the strongest signals.

<br>

- In identifying redundant dimensions, we discover ways to reduce the dimensionality of our observations.

<br>

Formally, we seek to maximize a few diagonal terms and minimize the other diagonal and all the off-diagnonal terms of ${\bf C}_{\bf Y}$!

<br>

A fully diagonal $ {\bf C}_{\bf Y}$ is always attainable!  How variance distributes over the diagonal depends on our data.



# **4. Diagonalizing variance**

<br>

There are many ways of choosing ${\bf P}$ such as to diagonalize $ {\bf C}_{\bf Y}$

$$
{\bf Y} = {\bf P}  \cdot  {\bf X}, \qquad\qquad {\bf C}_{\bf X} \to  {\bf C}_{\bf Y}
$$

<br>

For example:

<br>

- Select the normalized direction in $\mathbb{R}^{m\times m}$ along which the variance in ${\bf X}$ is maximal.  Save this vector as ${\bf p}_1$.

<br>

- Find a second, orthonormal direction along which the variance  ${\bf X}$ is maximal.   Save this vector as ${\bf p}_2$.

<br>

- Repeat until an orthonormal basis with $m$ vectors has been selected.


<br>

Proceeding in this way does not take full benefit of linear algebra!


# Orthonormal diagonalization: another side to eigenvectors!

<br>

Every square symmetric ${\bf S}$ matrix is *orthonormally diagonizable*

<br>

$$
{\bf S} = {\bf E} \cdot  {\bf D}\cdot  {\bf E}^T \qquad \Leftrightarrow \qquad {\bf S}\cdot  {\bf E} = {\bf E} \cdot  {\bf D}
$$

<br>

with square matrix ${\bf E}$ collecting **column eigenvectors** of ${\bf S}$:

$$
{\bf E} = \left({\bf E}_1 \, {\bf E}_2 \, \ldots \, {\bf E}_m \right),\qquad\qquad {\bf E}\cdot  {\bf E}^T =  {\bf E}^T \cdot {\bf E} = {\bf I}
$$

<br>

and diagonal matrix ${\bf D}$ collecting **eigenvalues** of ${\bf S}$:

<br>

$$
{\bf D} = \left(\begin{array}{ccc} \lambda_1 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \lambda_m \end{array} \right), \qquad\qquad diag({\bf D}) = \left(\lambda_1 \, \lambda_2 \, \ldots \, \lambda_m \right)
$$

<br>

This applies to covariance matrices, as these are square and symmetric!


# "A matrix is what a matrix does!" (Gilbert Strang, MIT)

<br>

$$
{\bf S} = \left\{\begin{array}{ccccc} {\bf E} & \cdot & {\bf D} & \cdot & {\bf E}^T \\
\in \mathbb{R}^{m\times m} & & \in \mathbb{R}^{m \times m}& & \in \mathbb{R}^{m \times m} \\
\mathit{rotate~back} & & \mathit{stretch} & & \mathit{rotate}
\end{array}\right.$$

<br>

After rotation into eigenvector coordinates, $j^{th}$ component is stretched by $\lambda_j = D_{jj}$, and then rotated back into original coordinates.

# Application to covariance

<br>

Given orthonormal diagonalization ${\bf C}_{\bf X} = {\bf E} \cdot {\bf D} \cdot {\bf E}^T$, we can ensure that covariance of ${\bf Y} = {\bf P} \cdot {\bf X}$ becomes diagonal:

<br>

$$\begin{eqnarray}
{\bf C}_{\bf Y} &=& \frac{1}{n-1} {\bf Y} \cdot {\bf Y}^T = \frac{1}{n-1} ({\bf P}  \cdot {\bf X}) \cdot ( {\bf P} \cdot {\bf X})^T =
\\
\\
&=& \frac{1}{n-1} ({\bf P} \cdot {\bf X}) \cdot ( {\bf X}^T \cdot {\bf P}^T) = \frac{1}{n-1} {\bf P}\cdot ( {\bf X} \cdot {\bf X}^T) \cdot {\bf P}^T =
\\
\\
&=& {\bf P}  \cdot {\bf C}_{\bf X} \cdot {\bf P}^T = {\bf P}  \cdot {\bf E} \cdot {\bf D} \cdot {\bf E}^T \cdot {\bf P}^T = {\bf I} \cdot {\bf D} \cdot {\bf I} =  {\bf D}
\end{eqnarray}$$

<br>

if only the **rows** of ${\bf P}$ are chosen as the **columns** of ${\bf E}$:

<br>

$$
{\bf P}  =  {\bf E}^T \quad\Leftrightarrow\quad {\bf P} \cdot  {\bf E} = {\bf I} \quad\Leftrightarrow\quad {\bf E}^T \cdot  {\bf P}^T  = {\bf I}
$$

<br>

The desired principal components are the eigenvectors of $\bf C_x$!

# Sort principal components, aka eigenvectors

<br>

By convention, we sort rows of ${\bf P}$ (columns of ${\bf E}$) from the largest to the smallest eigenvalue (as long as they are non-zero):  

<br>

The *first principal component* ${\bf p}_{1}$ captures the largest variance $\sigma_1^2 = \lambda_1 = D_{1,1}$.

<br>

The *second principal component* ${\bf p}_{2}$, captures the next largest variance $\sigma_2^2= \lambda_2 = D_{2,2}$.

<br>

${\bf
\vdots
}$

<br>

The $m^\mathit{\,th}$ *principal component* ${\bf p}_{m}$, captures the smallest variance $\sigma_m^2= \lambda_m = D_{m,m}$.

# Different roads to Rome


<br>

**Orthonormal diagonalization**: requires covariance

<br>


Form covariance of observations ${\bf C}_{\bf X} = \frac{1}{n-1} {\bf X}\cdot {\bf X}^T  \in \mathbb{R}^{m \times m}$

<br>

Orthonormal diagonalization ${\bf C}_{\bf X} = {\bf E} \cdot {\bf D} \cdot {\bf E}^T$ into eigenvectors ${\bf E}  \in \mathbb{R}^{m \times m}$ and eigenvalues $\mathit{diag}({\bf D})$.

<br>

Obtain principal components as ${\bf P}  = {\bf E}^T$ and variance contributions as $\mathit{diag}({\bf D})$.
  

<br>

**Singular-value decomposition (SVD)**: proceeds directly from observations

<br>

Singular value decomposition of ${\bf X}^T  = {\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T \in \mathbb{R}^{n \times m}$.

<br>

Obtain principal components as ${\bf P}  = {\bf V}^T$ and variance contributions as $\mathit{diag}({\bf \Sigma}^T \cdot {\bf \Sigma}$).


# Singular Value Decomposition

<br>

What the orthonormal diagonalization is for square and symmetric matrices, the *singular value decomposition* is for arbitrary rectangular matrices.  Every rectangular matrix ${\bf A}$ has a singular value decomposition

<br>

$$
{\bf A} = {\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T  \in \mathbb{R}^{n \times m}
$$

<br>

with *left singular vectors*

<br>

$$
{\bf U} \in \mathbb{R}^{n \times n} \qquad\qquad \mathit{orthonormal},  \mathit{square}
$$

<br>

and *singular values*


<br>

$$
{\bf \Sigma} \in \mathbb{R}^{n \times m} \qquad\qquad \mathit{diagonal}, \mathit{rectangular}
$$

<br>

and *right singular vectors*

<br>

$$
{\bf V} \in \mathbb{R}^{m \times m} \qquad\qquad  \mathit{orthonormal}, \mathit{square}
$$

Orthonormal (and square)

<br>

$$
{\bf U}  = \left(\begin{array}{ccc} u_{11} & \ldots & u_{1n} \\ \vdots & \ddots & \vdots \\ u_{n1} & \ldots & u_{nn} \end{array} \right) \in \mathbb{R}^{n \times n} ,
\quad
{\bf V} = \left(\begin{array}{ccc} v_{11} & \ldots & v_{1m} \\ \vdots & \ddots & \vdots \\ v_{m1} & \ldots & v_{mm} \end{array} \right)  \in \mathbb{R}^{m \times m}
$$

<br>

Diagonal (and rectangular)

<br>

$$
{\bf \Sigma} = \left(\begin{array}{ccc} \sqrt{\lambda_1} & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \sqrt{\lambda_m } \\ \vdots & \ldots & \vdots \\ 0 & \ldots & 0 \end{array} \right)  \in \mathbb{R}^{n \times m}
$$

# "A matrix is what a matrix does!" (Gilbert Strang, MIT)

<br>

$$
{\bf A} = \left\{\begin{array}{ccccc} {\bf U} & \cdot & {\bf \Sigma} & \cdot & {\bf V}^T \\
\in \mathbb{R}^{n\times n} & & \in \mathbb{R}^{n \times m}& & \in \mathbb{R}^{m \times m} \\
\mathit{rotate~back} & & \mathit{stretch} & & \mathit{rotate}
\end{array}\right.$$


<br>

After first rotation, $j^{th}$ component is stretched by $\sqrt{\lambda_j} = \Sigma_{jj}$

# Application to PCA

<br>

Choosing

$$
{\bf A} = \frac{1}{\sqrt{n-1}} \, {\bf X}^T  \in \mathbb{R}^{n \times m}, \qquad\mathrm{with} \qquad {\bf A} = {\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T
$$

<br>

The desired orthonormal basis ${\bf P}$ is

<br>

$$
{\bf P}  = {\bf V}^T  \in \mathbb{R}^{m \times m}
$$

<br>

and the diagonal matrix of eigenvalues ${\bf D}$ is

<br>

$$
{\bf D} = \left(\begin{array}{ccc} \lambda_1 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \lambda_m \end{array} \right)=  {\bf \Sigma} \cdot {\bf \Sigma}^T = \left(\begin{array}{ccc} \sigma_1^2 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \sigma_m^2 \end{array} \right)
$$

# Proof

<br>

The choice of ${\bf A} = \frac{1}{\sqrt{n-1}} \, {\bf X}^T = {\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T$ ensures that

<br>

$$\begin{eqnarray}
{\bf C}_{\bf X} &=& \frac{1}{n-1} \, {\bf X} \cdot {\bf X}^T  = {\bf A}^T \cdot {\bf A} =
\\
&=&  ({\bf U}\cdot {\bf \Sigma}\cdot {\bf V}^T)^T \, ({\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T) =
\\
&=& ({\bf V} \cdot {\bf \Sigma}^T \cdot {\bf U}^T) \, ({\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T) =
\\
&=& {\bf V} \cdot {\bf \Sigma}^T \cdot {\bf \Sigma} \cdot {\bf V}^T  
\end{eqnarray}$$

<br>

$$
\Leftrightarrow \qquad\qquad {\bf C}_{\bf X} \cdot {\bf V} = {\bf V} \cdot {\bf \Sigma}^T \cdot {\bf \Sigma} =  {\bf \Sigma}^T \cdot {\bf \Sigma} \cdot {\bf V} = {\bf D} \cdot {\bf V}
$$

<br>

so that ${\bf V}$ are the eigenvectors and $ {\bf \Sigma}^T \cdot {\bf \Sigma}$ the eigenvalues of ${\bf C}_{\bf X}$.

# Summary diagonalizing variance

<br>

Seek linear transformation ${\bf Y} = {\bf P} \cdot {\bf X}$ such that covariance ${\bf C}_{\bf Y}$ is diagonal matrix ${\bf D}$!   

<br>

SVD of rectangular matrix ${\bf A} = \frac{1}{n-1} {\bf X}^T$ into ${\bf A} = {\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T$ gives us

<br>

$$
{\bf P}  = {\bf V}^T, \qquad\qquad {\bf D} ={\bf \Sigma}^T \cdot {\bf \Sigma}
$$

<br>

Orthonormal diagonalization of square and symmetric matrix ${\bf C}_{\bf X}$ into  ${\bf C}_{\bf X} = {\bf E}\cdot {\bf D}\cdot {\bf E}^T$ gives us

<br>

$$
{\bf P}  = {\bf E}^T, \qquad {\bf D} = {\bf D}
$$

# What have we achieved?

<br>

By re-representing observations in PC coordinates, we have eliminated redundancy and revealed main sources of non-redundant variance:

<br>

$$\begin{eqnarray}
{\bf C}_{\bf X}
&=& \frac{1}{n-1} \left(\begin{array}{cccc}
\sigma_{11}^2 & \sigma_{12} & \ldots & \sigma_{1m} \\
\sigma_{21} & \sigma_{22}^2 & \ldots & \sigma_{2m}  \\
\vdots & \vdots & \ddots & \vdots \\
\sigma_{m1} & \sigma_{m2} & \ldots & \sigma_{mm}^2 \end{array} \right)
\\
\\
\sigma^2_\mathit{total} &=& \sigma_{11}^2+\sigma_{22}^2+\ldots+\sigma_{mm}^2
\\
\\
\\
{\bf C}_{\bf Y} &=& \frac{1}{n-1} \left(\begin{array}{cccc}
\rho_{1}^2 & 0 & \ldots & 0 \\
0 & \rho_{2}^2 & \ldots & 0  \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \ldots & \rho_{m}^2 \end{array} \right)
\\
\\
\\
 \sigma^2_\mathit{total} &=& \rho_1^2+\rho_2^2+\ldots+\rho_m^2
\end{eqnarray}$$

# **5. De-nosing with PCA**

<br>

Full projection matrix ${\bf V}^T \in \mathbb{R}^{m\times m}$

<br>

$$\begin{eqnarray}
{\bf Y} &=&  {\bf V}^T \cdot  {\bf X}
\\
\\
 \left(\begin{array}{cccc} y_{11} & y_{12} & \ldots & y_{1n} \\ \vdots & \ddots &  \ddots & \vdots \\ y_{m1} & y_{m2} & \ldots & y_{mn} \end{array} \right) &=&
 \left(\begin{array}{ccc} v_{11} & \ldots & v_{m1} \\ \vdots & \ddots & \vdots \\ v_{1m} & \ldots & v_{mm} \end{array} \right) \left(\begin{array}{cccc} x_{11} & x_{1,2} & \ldots & x_{1n} \\ \vdots & \ddots & \ddots & \vdots \\ x_{m1} & x_{m2} & \ldots & x_{mn} \end{array} \right)
\end{eqnarray}$$

<br>

Truncated projection matrix  $\tilde{\,\bf V}^T \in \mathbb{R}^{r\times m}$, with $r<m$

<br>

$$\begin{eqnarray}
\tilde{\,\bf Y} &=&  \tilde{\,\bf V}^T \cdot  {\bf X}
\\
\\
 \left(\begin{array}{cccc} y_{11} & y_{12} & \ldots & y_{1n} \\  y_{r1} & y_{r2} & \ldots & y_{rn} \end{array} \right) &=&
 \left(\begin{array}{ccc} v_{11} & \ldots & v_{m1} \\ v_{1r} & \ldots & v_{mr} \end{array} \right)  \left(\begin{array}{cccc} x_{11} & x_{1,2} & \ldots & x_{1n} \\ \vdots & \ddots  & \vdots \\  x_{m1} & x_{m2} & \ldots & x_{mn} \end{array} \right)
\end{eqnarray}$$


The truncated projection $\tilde{\,\bf Y}$ combines the first $r$ *principal components* of the original observations ${\bf X}$.  Being smaller, it constitutes a  *compressed* version of ${\bf X}$.  

<br>

It may be *uncompressed* into a matrix $\tilde {\,\bf X}  \in \mathbb{R}^{n\times m}$ by performing the backprojection

<br>

$$\begin{eqnarray}
\tilde{\,\bf X} &=&  \left(\tilde{\,\bf V}^T\right)^{-1} \cdot  \tilde{\,\bf Y}
\\
\\
\left(\begin{array}{cccc} x_{11} & x_{1,2} & \ldots & x_{1n} \\ \vdots & \ddots & \ddots & \vdots \\ x_{m1} & x_{m2} & \ldots & x_{mn} \end{array} \right) &=&
 \left(\begin{array}{cc} v_{11} &  v_{r1} \\ \vdots &  \vdots \\  v_{m1} &  v_{mr} \end{array} \right)  \left(\begin{array}{cccc} y_{11} & y_{12} & \ldots & y_{1n} \\  y_{r1} & y_{r2} & \ldots & y_{rn} \end{array} \right)
\end{eqnarray}$$

<br>

If smaller principal components reflect noise, compressing and uncompressing with larger principal components also serves to *de-noise* the original data.

# Summary de-noising

<br>

- Transformed observations ${\bf Y} = {\bf P} \cdot {\bf X} $ span the entire principal component space.

- It can make sense to consider only the first few dimensions, where most of the variance is concentrated.

- To this end, transformed observations ${\bf Y}$ may be truncated to $\tilde{\,\bf Y}$ by zeroing the unwanted dimensions.

- The results may be visualized by back-transforming the truncated observations into the original space

$$
\tilde {\,\bf X} = {\bf P}^T \cdot \tilde {\,\bf Y}
$$

- The result is a 'de-noised' version of the original observations.

```python
# @title Multi-unit activity {"vertical-output":true,"display-mode":"form"}


# Import necessary libraries
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.io import loadmat
import matplotlib

# Set up matplotlib to mimic MATLAB's plotting style
matplotlib.rcParams['lines.linewidth'] = 2

# mount MyDrive
drive.mount('/content/drive')
filename = '/content/drive/MyDrive/Theoretical Neuroscience II/Data files/MUA_b_t_g.mat'

# Load data (ensure 'MUA_b_t_g.mat' is uploaded to Colab)
# Upload the file in Colab: Files -> Upload
data = loadmat(filename)
MUA_b_t_g = data['MUA_b_t_g']  # Shape: (Nb, Nt, Ng)
Nb = data['Nb'].item()         # Number of bursts
Nt = data['Nt'].item()         # Number of time points
ti = data['ti'].flatten()      # Time points (negative latencies)
Ng = data['Ng'].item()         # Number of neuron groups

# Compute and plot mean timecourse, averaging over bursts
mu_MUA_g_t = np.full((Ng, Nt), np.nan)
for ig in range(Ng):
    mu_MUA_g_t[ig, :] = np.mean(MUA_b_t_g[:, :, ig], axis=0)  # Mean over bursts

X = mu_MUA_g_t  # Ng variables (rows), Nt observations (columns)
m, n = X.shape  # m variables (rows), n observations (columns)
mn = np.mean(X, axis=1, keepdims=True)  # Compute row mean
X0 = X - mn  # Subtract row mean

# Plot mean and zero-mean timecourses
def getcolor(idx):
    clr = 'rgbcmyk'
    cidx = idx % len(clr)
    return clr[cidx]

plt.figure(figsize=(8, 10))
plt.subplot(2, 1, 1)
for ig in range(Ng):
    plt.plot(ti, mu_MUA_g_t[ig, :], getcolor(ig), linewidth=2)
plt.xlim(min(ti), max(ti))
plt.ylim(-50, np.max(mu_MUA_g_t))
plt.xlabel('time [ms]')
plt.ylabel('activity [Hz/neuron]')

plt.subplot(2, 1, 2)
for ig in range(Ng):
    plt.plot(ti, X0[ig, :], getcolor(ig), linewidth=2)
plt.xlim(min(ti), max(ti))
plt.ylim(-50, 400)
plt.xlabel('time [ms]')
plt.ylabel('zero-mean activity')

plt.tight_layout()
plt.show()

# Compute and plot covariance matrix
CX = (X0 @ X0.T) / (n - 1)
plt.figure()
plt.pcolormesh(CX, cmap='hot', vmin=np.min(CX), vmax=np.max(CX))
plt.axis([1, m, 1, m])
plt.gca().set_aspect('equal')
plt.xlabel('neuron group')
plt.ylabel('neuron group')
plt.colorbar()
plt.show()

# Identify principal components in mean timecourses
U, S, V = np.linalg.svd(X0.T, full_matrices=False)  # Skinny SVD
V = V.T  # Match MATLAB's SVD output
print(f"U shape: {U.shape}")
print(f"S shape: {S.shape}")
print(f"V shape: {V.shape}")

# Plot variance of principal components
plt.figure()
plt.bar(range(1, len(S) + 1), np.log(np.diag(np.dot(np.diag(S), np.diag(S).T))))
plt.xlim(1, 20)
plt.ylim(-1, 20)
plt.xlabel('PCs')
plt.ylabel('log variance')
plt.show()

# Project onto principal components
P = V.T  # Principal components as row vectors
Y0 = P @ X0
print(f"Y0 shape: {Y0.shape}")

# Plot projections in PC space
def plot_3_axes(ax):
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    zlim = ax.get_zlim()
    ax.plot3D(xlim, [0, 0], [0, 0], 'k')
    ax.plot3D([0, 0], ylim, [0, 0], 'k')
    ax.plot3D([0, 0], [0, 0], zlim, 'k')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_zlim(zlim)

def plot_2_axes(ax):
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    ax.plot(xlim, [0, 0], 'k')
    ax.plot([0, 0], ylim, 'k')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

plt.figure(figsize=(10, 10))
# PC 1-3
ax1 = plt.subplot(2, 2, 1, projection='3d')
ax1.plot3D(Y0[0, :], Y0[1, :], Y0[2, :], 'r', linewidth=2)
ax1.plot3D([Y0[0, 0]], [Y0[1, 0]], [Y0[2, 0]], 'ro', linewidth=2)
plot_3_axes(ax1)
ax1.set_xlabel('PC 1')
ax1.set_ylabel('PC 2')
ax1.set_zlabel('PC 3')
ax1.view_init(30, -37.5)
ax1.set_box_aspect([1, 1, 1])

# PC 4-6
ax2 = plt.subplot(2, 2, 2, projection='3d')
ax2.plot3D(Y0[3, :], Y0[4, :], Y0[5, :], 'b', linewidth=2)
ax2.plot3D([Y0[3, 0]], [Y0[4, 0]], [Y0[5, 0]], 'bo', linewidth=2)
plot_3_axes(ax2)
ax2.set_xlabel('PC 4')
ax2.set_ylabel('PC 5')
ax2.set_zlabel('PC 6')
ax2.view_init(30, -37.5)
ax2.set_box_aspect([1, 1, 1])

# PC 7-9
ax3 = plt.subplot(2, 2, 3, projection='3d')
ax3.plot3D(Y0[6, :], Y0[7, :], Y0[8, :], 'k', linewidth=2)
ax3.plot3D([Y0[6, 0]], [Y0[7, 0]], [Y0[8, 0]], 'ko', linewidth=2)
plot_3_axes(ax3)
ax3.set_xlabel('PC 7')
ax3.set_ylabel('PC 8')
ax3.set_zlabel('PC 9')
ax3.view_init(30, -37.5)
ax3.set_box_aspect([1, 1, 1])

# PC 10-11
ax4 = plt.subplot(2, 2, 4)
ax4.plot(Y0[9, :], Y0[10, :], 'g', linewidth=2)
ax4.plot(Y0[9, 0], Y0[10, 0], 'go', linewidth=2)
plot_2_axes(ax4)
ax4.set_xlabel('PC 10')
ax4.set_ylabel('PC 11')
ax4.set_aspect('equal')

plt.tight_layout()
plt.show()

# Compute and visualize covariance of transformed activity
CY = (Y0 @ Y0.T) / (n - 1)
plt.figure()
plt.pcolormesh(CY, cmap='hot', vmin=np.min(CY), vmax=np.max(CY))
plt.axis([1, m, 1, m])
plt.gca().set_aspect('equal')
plt.xlabel('group')
plt.ylabel('group')
plt.colorbar()
plt.show()

# De-noise mean timecourse by backprojecting first three principal components
Y0dn = Y0.copy()
Y0dn[3:Ng, :] = 0  # Zero later PCs
X0dn = P.T @ Y0dn  # Project back to groups
Xdn = X0dn + mn  # Add back row mean

plt.figure()
for ig in range(Ng):
    plt.plot(ti, Xdn[ig, :], getcolor(ig), linewidth=2)
plt.xlim(min(ti), max(ti))
plt.ylim(-50, np.max(mu_MUA_g_t))
plt.show()

# Plot for groups 15-17
plt.figure()
for ig in range(14, 17):  # Python is 0-indexed, so 14 corresponds to MATLAB's 15
    plt.plot(ti, Xdn[ig, :], getcolor(ig), linewidth=2)
plt.xlim(min(ti), max(ti))
plt.ylim(-50, np.max(mu_MUA_g_t))
plt.show()

# Visualize timecourse of individual bursts in terms of principal components
MUA_g_t = np.full((Ng, Nt * Nb), np.nan)
for ig in range(Ng):
    MUA_g_t[ig, :] = MUA_b_t_g[:, :, ig].T.flatten()  # Concatenate bursts

X = MUA_g_t
m, n = X.shape
mn = np.mean(X, axis=1, keepdims=True)
X0 = X - mn
Y0 = P @ X0
print(f"Y0 shape (bursts): {Y0.shape}")

plt.figure()
ax = plt.axes(projection='3d')
for ib in range(Nb):
    tidx = slice((ib) * Nt, (ib + 1) * Nt)
    ax.plot3D(Y0[0, tidx], Y0[1, tidx], Y0[2, tidx], getcolor(ib), linewidth=2)
    ax.plot3D([Y0[0, (ib) * Nt]], [Y0[1, (ib) * Nt]], [Y0[2, (ib) * Nt]], getcolor(ib) + 'o', linewidth=2, markersize=10)
plot_3_axes(ax)
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
ax.view_init(30, -37.5)
ax.set_box_aspect([1, 1, 1])
plt.show()

```

**Summary of PCA assumptions**

<br>

*Linearity*

- We frame the problem as a change of basis  $\mathbb{R}^{m\times m}$  space.

<br>

*Variance is a sufficient statistic*

- We assume the variance fully describes the distribution of our (zero-mean) variables.

<br>

*Large variances are important*

- We assume variables with high variance are 'signal', whereas those with low variance are 'noise'.

<br>

*Principal components are orthogonal*

- This lets us perform PCA with linear algebra techniques.

# **Appendix on SVD**

# What does SVD do?

<br>

A singular-value decomposition ${\bf A} = {\bf U} \cdot {\bf \Sigma}\cdot {\bf V}^T$ decomposes the transformation (change of basis) performed by ${\bf A}$ into three distinct and intuitively intelligible steps.  Consider a linear transformation ${\bf x} \to {\bf y}$

<br>

$$
{\bf y} = {\bf A}\cdot {\bf x} = {\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T \cdot {\bf x}, \qquad  {\bf y} \in \mathbb{R}^{n}, \qquad {\bf x} \in \mathbb{R}^{m}, \qquad {\bf A} \in\mathbb{R}^{n \times m}
$$

<br>

Expressing ${\bf y}$ in the orthonormal basis ${\bf U}$ and ${\bf x}$ in the orthonormal basis ${\bf V}$, we have

<br>

$$
\hat{\,\bf y} = {\bf U}\cdot {\bf y} \quad\Leftrightarrow\quad   {\bf y} = {\bf U}^T \cdot \hat{\,\bf y},\qquad\qquad \hat{\,\bf x} = {\bf V}\cdot {\bf x} \quad\Leftrightarrow\quad    {\bf x} = {\bf V}^T \cdot \hat{\,\bf x}
$$

<br>

By left-multiplying the SVD with ${\bf U}^T$, we can reformulate the transformation in terms of $\hat{\,\bf x} \to \hat{\,\bf y}$

<br>

$$
{\bf U}^T \cdot \left| \qquad {\bf y} =  {\bf U}\cdot {\bf \Sigma}\cdot {\bf V}^T \cdot {\bf x}\right.\qquad \Leftrightarrow \qquad \hat{\,\bf y} = {\bf U}^T \cdot {\bf y} = {\bf \Sigma}\cdot \hat{\,\bf x}
$$


Thus, in terms of the orthonormal bases ${\bf U}$ and ${\bf V}$, the transformation *rescales* components $1$ to $m$ with the singular values in ${\bf \Sigma}$ and zeroes all others:

<br>

$$
\left(\begin{array}{c} \hat y_1 \\ \vdots \\ \hat y_m \\ \vdots \\ 0 \end{array}\right) = \left(\begin{array}{ccc} \sqrt{\lambda_1} & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \sqrt{\lambda_m}  \\ \vdots & \ldots & \vdots \\ 0 & \ldots & 0 \end{array} \right)  \left(\begin{array}{c} \hat x_1 \\ \hat x_2 \\ \vdots \\ \hat x_m \end{array}\right)
$$

<br>

Note that ${\bf \Sigma}^T {\bf \Sigma}$ is square and diagonal:

<br>

$$
{\bf \Sigma}^T \cdot {\bf \Sigma} = {\bf \Sigma} \cdot {\bf \Sigma}^T = \left(\begin{array}{ccc} \lambda_1 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \lambda_m  \end{array} \right)  \in \mathbb{R}^{m \times m}
$$

# Derivation of SVD

<br>

Consider a covariance matrix and its orthonormal eigenvectors and real positive eigenvalues

<br>

$$
{\bf X}^T \cdot {\bf X} \cdot {\bf v}_j = \lambda_j \, {\bf v}_j, \qquad\qquad {\bf X} \in \mathbb{R}^{n\times m}, \quad {\bf v}_j  \in \mathbb{R}^{m\times 1}
$$

<br>

Use  ${\bf X}$ to transform $m \times 1$ eigenvectors ${\bf v}_j$ into $n\times 1$ vectors ${\bf u}_j$

<br>

$$
{\bf u}_j = \frac{1}{\sqrt{\lambda_j}} {\bf X} \cdot {\bf v}_j, \qquad\qquad {\bf u}_j  \in \mathbb{R}^{n\times 1}
$$

<br>

Astonishingly, the transformed vectors also form an orthonormal basis!  Their norm is unity!  They are pairwise orthogonal!

<br>

$$
|| {\bf u}_j || = 1, \qquad\qquad
{\bf u}_j \, {\bf u}_k = \delta_{ik}
$$

# Proof

<br>

$$\begin{eqnarray}
 || {\bf u}_j ||^2 &=& {\bf u}_j^T \cdot {\bf u}_j =
 \\
 &=& \frac{1}{\lambda_j}({\bf X} \cdot {\bf v}_j)^T \cdot ({\bf X} \cdot {\bf v}_j) =  \frac{1}{\lambda_j}({\bf v}_j^T \cdot {\bf X}^T )\cdot ({\bf X} \cdot {\bf v}_j) =
 \\
 &=& \frac{1}{\lambda_j} \, {\bf v}_j^T \, \lambda_j \, {\bf v}_j =  {\bf v}_j^T \cdot {\bf v}_j = || {\bf v}_j^2|| = 1
\end{eqnarray}$$

<br>

$$\begin{eqnarray}
{\bf u}_j^T \cdot {\bf u}_k &=&
 \\
 &=& \frac{1}{\sqrt{\lambda_j \lambda_k}}({\bf X} \cdot {\bf v}_j)^T \cdot ({\bf X} \cdot {\bf v}_k) =  \frac{1}{\sqrt{\lambda_j\lambda_k}}({\bf v}_j^T \cdot {\bf X}^T )\cdot ({\bf X} \cdot {\bf v}_k) =
 \\
 &=& \frac{1}{\sqrt{\lambda_j \lambda_k}} \, {\bf v}_j^T \, \lambda_k \, {\bf v}_k =  \sqrt{\frac{\lambda_k}{\lambda_j}} \, {\bf v}_j \cdot {\bf v}_k = \delta_{jk}
\end{eqnarray}$$

# Collect intro matrix

<br>

Collect ${\bf u}$'s into $n \times m$ matrix
$$
{\bf U}'  =  \left( {\bf u}_1 \ldots {\bf u}_m \right) \in \mathit{R}^{n\times m}
$$

<br>

'Pad' with further orthonormal vectors to complete square matrix

<br>

$$
{\bf U} = \left(\begin{array}{cccccc} {\bf u}_1 & \ldots & {\bf u}_m & {\bf u}_{m+1} & \ldots & {\bf u}'_n \end{array} \right) \in \mathit{R}^{n\times n}
$$

<br>

Now have orthonormal basis

<br>

$$
 {\bf U} \cdot {\bf U}^T = {\bf U}^T \cdot {\bf U} = {\bf I}  \qquad \in \mathbb{R}^{n\times n}
 $$

# ctd

<br>

Collect the individual vector transformation

<br>

$$
 {\bf u}_j \, \sqrt{\lambda_j} =  \, {\bf X} \cdot {\bf v}_j
$$

<br>

into matrix transformation

<br>

$$
 {\bf U} \cdot {\bf \Sigma} = {\bf X} \cdot {\bf V} \qquad \in \mathit{R}^{n\times m}
$$

<br>

$$
{\bf U} \in \mathit{R}^{n\times n}, \qquad {\bf \Sigma} \in \mathit{R}^{n\times m}, \qquad {\bf X} \in \mathit{R}^{n\times m}, \qquad {\bf V} \in \mathit{R}^{m\times m},
$$

<br>

$$
{\bf \Sigma}= \left( \begin{array}{cccc} \sqrt{\lambda_1} & 0 &  \ldots & 0 \\ 0 & \sqrt{\lambda_2} & \ddots & \vdots \\ \vdots & \ddots & \ddots & \vdots \\ 0 & 0 & \ldots & \sqrt{\lambda_m}  \\ \vdots & \vdots & \vdots & \vdots \\ 0 & 0 & 0 & 0 \end{array} \right)  \in \mathit{R}^{n\times m}
$$

# ctd

<br>

Finally, right-multiply with ${\bf V}^T$ to obtain

<br>

$$\begin{eqnarray}
 {\bf U} \cdot {\bf \Sigma} &=& \left. {\bf X} \, {\bf V} \qquad \right| \cdot {\bf V}^T
\\
\\
{\bf U} \, {\bf \Sigma} \, {\bf V}^T &=& {\bf X}
\end{eqnarray}$$

<br>

To recapitulate, for a rectangular matrix ${\bf X}  \in \mathit{R}^{n\times m}$, we considered the eigenvectors and eigenvalues of the covariance matrix

<br>

$$
{\bf X}^T \cdot {\bf X} \cdot {\bf v}_j = \lambda_j \, {\bf v}_j
$$

<br>

and obtained the singular value decomposition of our matrix ${\bf X}$

<br>

$$\begin{eqnarray}
{\bf U} \cdot {\bf \Sigma} \cdot {\bf V}^T &=& {\bf X},\qquad\qquad \Sigma_{jj} = \sqrt{\lambda_j}, \qquad {\bf V} = \left( \begin{array}{ccc} {\bf v}_1 & \ldots & {\bf v}_m\end{array} \right)
\end{eqnarray}$$

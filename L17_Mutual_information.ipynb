# **Abstract:**

Today we apply the concepts of Shannon information to quantify the statistical dependence of two random variables.  We'll take it slowly to gain comprehensive understanding. First, we find that the marginal entropies of two variables can **overlap** in some sense. To clarify the implications, we introduce the notion of **conditional entropy**, based on the familiar conditional probability. Then we realize that this constitutes **noise entropy**, in that it tells us nothing about the variable in the condition. With this, it becomes clear that the **overlap** is **mutual information**, or information shared between one variable and another.

Finally, we touch on some more advanced topics by introducing **differential entropy**, which is the extension of Shannon entropy to continuous random variables, and by deriving **differential mutual information** with the help of multivariate Gaussian distributions.

# Random variables with discrete outcomes

We have considered ensembles $X$ of discrete outcomes $x$, each with a probability $P(x)$.

<br>

The *Shannon information* of each outcome $x$ can be measured in *bits*.  It is the minimal number of *binary question*s needed to unambiguously determine the outcome. Of course, this works only if we know the ensemble of all possible outcomes!

<br>

The average information content over all possible outcomes is the *Shannon entropy*:

$\begin{eqnarray}
H(X) = \sum_x \, P(x) \, \log_2\, \frac{1}{P(x)} =-\sum_x \, P(x) \, \log_2\,P(x)
\end{eqnarray}$

This is a probability-weighted average over all possible outcomes, with each individual outcome $x$ contributing

$\begin{eqnarray}
P(x) \, \log_2\, \frac{1}{P(x)}= - P(x) \, \log_2\, {P(x)}
\end{eqnarray}$

(Slide-2)



# Joint and marginal entropies

Given the joint probabilities, we defined the  **joint entropy** of $X$ and $Y$ as

$\begin{eqnarray}
H(X,Y) = \sum_x \, \sum_y \, P(x,y) \, \log_2 \, \frac{1}{P(x,y)}
\end{eqnarray}$

<br>

It compares to the **individual** or  **marginal entropies** of $X$, and of $Y$,

$\begin{eqnarray}
H(X)= \sum_{x} \, P(x) \, \log_2 \, \frac{1}{P(x)}, \qquad H(Y) = \sum_{y} \, P(y) \, \log_2 \, \frac{1}{P(y)}
\end{eqnarray}$

<br>
The joint entropy *cannot be larger* than the sum of the individual or marginal entropies

$\begin{eqnarray}
H(X) + H(Y) - H(X,Y) = D_{KL}\left[P(x,y)||P(x)P(y)\right] \quad \geq 0
\end{eqnarray}$


<br>

*In terms of Shannon information, the whole is **smaller** than the sum of its parts!*

$\begin{eqnarray}
\qquad\qquad\qquad H(X,Y) \leq H(X) + H(Y)
\end{eqnarray}$

(Slide-3)

# **1. Overlap between marginal entropies**

The joint entropy *cannot be larger* than the sum of individual or marginal entropies.

The joint entropy *subsumes* (contains within) the individual entropies.

It follows that *some overlap* between individual entropies is possible.

How should we interpret such an overlap?

# Example of fully independent variables

An ensemble with two fully independent variables has the following joint distribution:

$\begin{eqnarray}\begin{array}{|cc|cc|c|}
\hline
P(x,y) &  & x & &  \\
& & + & - & P(y)\\
\hline
y & + & 1/4 & 1/4 & 1/2 \\
& - & 1/4 & 1/4 & 1/2 \\
\hline
& P(x) & 1/2 & 1/2 & 1\\
\hline
\end{array}
\end{eqnarray}$

Accordingly, we have individual or marginal entropies of

$\begin{eqnarray}
H(X) = H(Y) = \frac{1}{2} \, \log_2 \, 2 + \frac{1}{2} \, \log_2 \, 2 = 1
\end{eqnarray}$

a joint entropy of
$\begin{eqnarray}
H(X,Y) = \frac{1}{4} \, \log_2 \, 4 + \frac{1}{4} \, \log_2 \, 4 + \frac{1}{4} \, \log_2 \, 4 + \frac{1}{4} \, \log_2 \, 4= 2
\end{eqnarray}$

and an overlap of zero:

$\begin{eqnarray}
I_m = H(X)+H(Y) - H(X,Y) = 1+1-2=0
\end{eqnarray}$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_TrivialExamples.png" width="800">

<br>

When two variables are fully independent, so that neither carries information about the other, the overlap is zero!

(Slide-5)

# Example of fully dependent variables

Another ensemble with two fully or *deterministically* dependent variables has the following joint distribution:

$\begin{eqnarray}\begin{array}{|cc|cc|c|}
\hline
P(x,y) &  & x & &  \\
& & + & - & P(y)\\
\hline
y & + & 0 & 1/2 & 1/2 \\
& - & 1/2 & 0 & 1/2 \\
\hline
& P(x) & 1/2 & 1/2 & 1\\
\hline
\end{array}
\end{eqnarray}$

Accordingly, we have individual or marginal entropies of

$\begin{eqnarray}
H(X) = H(Y) = \frac{1}{2} \, \log_2 \, 2 + \frac{1}{2} \, \log_2 \, 2 = 1
\end{eqnarray}$

a joint entropy of
$\begin{eqnarray}
H(X,Y) = \frac{1}{2} \, \log_2 \, 2 + \frac{1}{2} \, \log_2 \, 2 = 1
\end{eqnarray}$

and an overlap equalling either individual entropy:

$\begin{eqnarray}
I_m = H(X)+H(Y) - H(X,Y) = 1+1-1=1
\end{eqnarray}$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_TrivialExamples.png" width="800">

<br>

When two variables are fully or deterministically dependent, so that either variable carries all information about the other, the overlap is full!

(Slide-6)

# Example of partially dependent variables

Another ensemble with two partially or *probabilistically* dependent variables has the following joint distribution:

$\begin{eqnarray}\begin{array}{|cc|cc|c|}
\hline
P(x,y) &  & x & &  \\
& & + & - & P(y)\\
\hline
y & + & 1/8 & 3/8 & 1/2 \\
& - & 3/8 & 1/8 & 1/2 \\
\hline
& P(x) & 1/2 & 1/2 & 1\\
\hline
\end{array}
\end{eqnarray}$

Accordingly, we have individual or marginal entropies of

$\begin{eqnarray}
H(X) = H(Y) = \frac{1}{2} \, \log_2 \, 2 + \frac{1}{2} \, \log_2 \, 2 = 1
\end{eqnarray}$

a joint entropy of
$\begin{eqnarray}
H(X,Y) = \frac{2}{8} \, \log_2 \, 8 + \frac{6}{8} \, \log_2 \, \frac{8}{3} = 1.81
\end{eqnarray}$

and an overlap of some part of the individual entropies:

$\begin{eqnarray}
I_m = H(X)+H(Y) - H(X,Y) = 1+1-1.81=0.19
\end{eqnarray}$

When two variables are partially or probabilistically dependent, so that either variable carries some information about the other, the overlap is partial!

We reach the following tenative conclusion:

<br>

* The overlap between individual entropies relates to information that is shared between the two variables.

# **2. Conditional or noise entropy**


The conditional probability $P(y|x)$ is defined for a subset of outcomes $(x,y)$: it is the probability to find outcome $y$ with in the subset of outcomes $x$.

<br>

The *conditional entropy* is the entropy of the conditional probability.  It reveals how much variability remains in $y$, when $x$ is fixed.  By definition, this part of the entropy of $y$ is  uninformative about $x$ (which is given).  Therefore, this part of the entropy of $y$ is *not shared* with $x$.

<br>

Because the conditional entropy of $y$ is uninformative about $x$, it is also called the *noise entropy* (with respect to $x$).

(Slide-8)

# The conditional/noise entropy of $Y$, given $X=x_0$

$\ldots$ is the entropy of the probability distribution $P(y|x=x_0)$

<br>

$\begin{eqnarray}
H(Y|X=x_0) = \sum_y \, P(y|x=x_0) \, \log_2 \frac{1}{P(y|x=x_0)}
\end{eqnarray}$

(Slide-10)

# The conditional/noise entropy of $Y$, given $X$

$\ldots$ is the average, over all possible $x_0$, of the conditional entropy of $Y$, given $X=x_0$:

$\begin{eqnarray}
H(Y|X) = \sum_{x_0} \, P(x_0) \, \left[ \sum_y \, P(y|x=x_0) \, \log_2 \frac{1}{P(y|x=x_0)} \right] =
\end{eqnarray}$

<br>

$\begin{eqnarray}
= \sum_{xy} \, P(y|x) \, P(x) \, \log_2 \, \frac{1}{P(y|x)} = \sum_{xy} \, P(x,y) \, \log_2 \, \frac{1}{P(y|x)}
\end{eqnarray}$

H(X|y=y_0) = \sum_x \, P(x|y=y_0) \, \log_2 \frac{1}{P(x|y=y_0)}

# The conditional/noise entropy of $X$, given $Y$

$\ldots$ is the average, over $y$, of the conditional entropy of $X$, given $y$:

<br>

$\begin{eqnarray}
H(X|Y) = \sum_{y_0} \, P(y_0) \, \left[ \sum_x \, P(x|y=y_0) \, \log_2 \frac{1}{P(x|y=y_0)} \right]
\end{eqnarray}$

<br>

$\begin{eqnarray}
= \sum_{xy} \, P(x|y) \, P(y) \, \log_2 \, \frac{1}{P(x|y)} = \sum_{xy} \, P(x,y) \, \log_2 \, \frac{1}{P(x|y)}
\end{eqnarray}$

# Conditional or noise entropy

<br>

- Conditional entropy of $A$ (given $B$) is the remaining entropy of $A$ when $B$ is fixed.

- Thus, conditional entropy of $A$ (given $B$) is the part of the entropy of $A$ that is *uninformative* about $B$.

- For this reason, it may also be considered the 'noise entropy' of $A$  (with respect to $B$).

- From additivity, it follows that any entropy of $A$ beyond this 'noise entropy' is indeed *informative* about $B$.

- Thus, the 'informative' entropy of $A$ is the total entropy *minus* the 'noise' entropy.

(Slide-14)


# Chain rule for Shannon information


From the product rule for probabilities

$\begin{eqnarray}
P(x,y) = P(x|y) \, P(y)
\end{eqnarray}$

$\begin{eqnarray}
\log_2 \, \frac{1}{P(x,y)} = \log_2 \, \frac{1}{P(x|y)} + \log_2 \, \frac{1}{P(y)}
\end{eqnarray}$
we obtain

$\begin{eqnarray}
h(x,y) = h(x|y) + h(y)
\end{eqnarray}$

Thus, the information of $x$ and $y$ is the information of $x$ given $y$, plus the information of $y$.

(Slide-15)

# Chain rule for entropy

<br>

The joint, conditional, and marginal entropies are related as follows:

<br>

$\begin{eqnarray}
H(X,Y) = H(X|Y) + H(Y)= H(Y|X) + H(X)
\end{eqnarray}$

<br>

In words, the joint entropy of $X$ and $Y$ is the *conditional entropy* of $X$ given $Y$, plus the marginal entropy of $Y$ (and *vice versa*).

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_ConditionalEntropy1.png" width="800">

(Slide-16)

# Proof of chain rule (optional)


$\begin{eqnarray}
H(X|Y) &+& H(Y) =
\\
&=& \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(x|y)} + \sum_{y} \, P(y) \log_2 \frac{1}{P(y)} =
\\
&=& \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(x|y)} + \sum_{xy} \, P(x,y) \log_2 \frac{1}{P(y)} =
\\
&=& \sum_{xy} \, P(x,y) \, \left[ \log_2 \frac{1}{P(x|y)} + \log_2 \frac{1}{P(y)} \right]=
\\
&=& \sum_{xy} \, P(x,y) \,\log_2 \frac{1}{P(x|y) \, P(y)} =
\\
&=& \sum_{xy} \, P(x,y) \,\log_2 \frac{1}{P(x,y)} = H(X,Y)
\end{eqnarray}$

# **3. Mutual information**

The mutual information between to random variables $X$ and $Y$

$\begin{eqnarray}
I_m = H(X) - H(X|Y) \geq 0
\end{eqnarray}$

is here defined as
- the information that $x$ conveys about $y$.
- the total information conveyed by $x$ minus the noise entropy of $x$.
- the average reduction in uncertainty about $y$ from learning the value of $x$.

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_ConditionalEntropy2.png" width="800">

(Slide-19)

Mutual information is symmetrical and also satisfies

$\begin{eqnarray}
I_m  = H(Y) - H(Y|X) \geq 0
\end{eqnarray}$

which is

- the average amount of information that $y$ conveys about $x$.

- the total information conveyed by $y$ minus the noise entropy of $y$.

- the average reduction in uncertainty about $x$ from learning the value of $y$.

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_ConditionalEntropy2.png" width="800">

(Slide-20)

Yet another way to express the mutual information is

$\begin{eqnarray}
I_m = H(X)+H(Y) - H(X,Y)
\end{eqnarray}$

which is

- the difference between the total *individual entropies* and the *joint* entropy.

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_ConditionalEntropy2.png" width="800">

(Slide-21)

# Alternative expressions for $I_m$

The mutual information can be expressed in several ways!

<br>

**Marginal $x$ minus conditional on $y$:**

$\begin{eqnarray}
I_m = H(X) - H(X|Y) = \sum_{x} \, P(x) \log_2 \frac{1}{P(x)}- \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(x|y)}
\end{eqnarray}$

<br>

**Marginal $y$ minus conditional on $x$:**

$\begin{eqnarray}
I_m = H(Y) - H(Y|X) = \sum_{y} \, P(y) \log_2 \frac{1}{P(y)}- \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(y|x)}
\end{eqnarray}$

<br>

**Total marginal minus joint:**

$\begin{eqnarray}
I_m = H(X)+H(Y) - H(X,Y) = \sum_{xy} \, P(x,y) \, \log_2 \frac{P(x,y)}{P(x)\,P(y)}
\end{eqnarray}$

(Slide-22)

# Proof of equivalence

<br>

$\begin{eqnarray}
 H(X) &-& H(X|Y) =
 \\
&=& \sum_{x} \, P(x) \log_2 \frac{1}{P(x)} - \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(x|y)} =
\\
&=& \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(x)} - \sum_{xy} \, P(x,y) \, \log_2 \frac{1}{P(x|y)} =
\\
&=& \sum_{xy} \, P(x,y) \, \left[ \log_2 \frac{1}{P(x)} - \log_2 \frac{1}{P(x|y)} \right]=
\\
&=& \sum_{xy} \, P(x,y) \,\log_2 \frac{P(x|y)\,P(y)}{P(x) \, P(y)} =
\\
&=& \sum_{xy} \, P(x,y) \,\log_2 \frac{P(x,y)}{P(x) \, P(y) } = H(X) + H(Y) - H(X,Y)
\end{eqnarray}$

(Slide-23)

# **Summary mutual information**

<br>

- Given two random variables, it is useful to compare joint entropy and individual entropies.  

- If combined individual entropies exceed joint entropy, the excess is 'mutual information'  $I_m$.

- $I_m$ is shared between variables, revealed by one about the other, uncertainty reduced about the other.

- $I_m$ quantifies how much information the value of one variable reveals about the value of the other variable.

- Mutual information is necessarily symmetric!

- If sunspots are informative about skirt lengths, then skirt lengths are also informative about sunspots.

# **4. Bivalued stimulus and response**

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_transmission_c.png" width="800">

<br>

The value of $s$ is probabilistic and either '+' or '-':

$\begin{eqnarray}
s \in\{`+`, `-`\}, \qquad \begin{array}{c}s_+ \equiv p\left[s=`+`\right]\\ s_- \equiv p\left[s=`-`\right]\end{array}  \qquad s_+ + s_- = 1
\end{eqnarray}$

<br>

The value of $r$ is similarly probabilistic and either '+' or '-':

$\begin{eqnarray}
r \in\{`+`, `-`\}, \qquad \begin{array}{c}r_+ \equiv p\left[r=`+`\right]\\ r_- \equiv p\left[r=`-`\right]\end{array}  \qquad r_+ + r_- = 1
\end{eqnarray}$

<br>

The same scenario described in an alternative notation:

$\begin{eqnarray}
P(s) = \left\{
\begin{array}{cc}
s_+ & s=`+`\\
1-s_+ & s=`-`\\
\end{array}
\right.
\qquad
P(r) = \left\{
\begin{array}{cc}
r_+ & r=`+`\\
1-r_+ & r=`-`\\
\end{array}
\right.
\end{eqnarray}$

(Slide-25)

# Formalizing the connection

In a joint ensemble $P(R,S)$, the value of $r$ is either probabilistically or deterministically related to the value of $s$:

<br>

$\begin{eqnarray}
P(r|s) = \left\{
\begin{array}{cc}
x & s=r\\
1-x & s\neq r\\
\end{array}
\right.
\end{eqnarray}$

<br>

The coupling parameter $x \in [0,1]$ determines the probability that $r$ and $s$ take equal values.

<br>

For either $x=0$ or $x=1$, the joint ensemble $r,s$ is  *deterministically dependent*; whereas for $0<x<1$ it is  *probabilistically dependent*.

(Slide-26)

# Joint and marginal distributions

The joint distribution is the product of $P(r|s)$ and $P(s)$:

$\begin{eqnarray}
\begin{array}{cc|cc|}
P(r,s)  & s &  \\
& & + & - \\
\hline
r & + & x\,s_+ & (1-x)\,(1-s_+) \\
& - & (1-x)\,s_+ & x\,(1-s_+) \\
\hline\\
\end{array}
\end{eqnarray}$

<br>

and the marginal distributions are $P(r) = \sum_s \, P(r,s)$

$\begin{eqnarray}
P[r=`+`]\equiv r_+ &=& x \, s_+ + (1 - x) \, (1-s_+)
\\
P[r=`-`]\equiv r_- &=& (1-x) \, s_+ + x \, (1-s_+)
\end{eqnarray}$

<br>

and  $P(s) = \sum_r \, P(r,s)$:

$\begin{eqnarray}
P[s=`+`]\equiv s_+, \qquad\qquad P[s=`-`]\equiv 1-s_+
\end{eqnarray}$

(Slide-27)

# Entropies and mutual information

We are now in the position to compute individual and joint entropies, as well as mutual information:

<br>

$\begin{eqnarray}
H(S) &=&  s_+ \, \log_2 \, \frac{1}{s_+} + (1-s_+) \, \log_2 \, \frac{1}{1-s_+}
\\
H(R) &=&  r_+ \, \log_2 \, \frac{1}{r_+} + (1-r_+) \, \log_2 \, \frac{1}{1-r_+}
\end{eqnarray}$

$\begin{eqnarray}
H(S,R) &=& x\,s_+ \, \log_2 \, \frac{1}{x s_+} + (1-x)\,(1-s_+) \, \log_2 \, \frac{1}{(1-x)\,(1-s_+)} +
\\
&+& (1-x)\,s_+ \, \log_2 \, \frac{1}{(1-x)\,s_+} + x\,(1-s_+) \, \log_2 \, \frac{1}{x\,(1-s_+)}
\end{eqnarray}$

<br>

$\begin{eqnarray}
I_m = H(S)+H(R) - H(S,R)
\end{eqnarray}$

# Conditional probabilities and entropies

<br>

The conditional probabilities $P(r|s)$ and $P(s|r)$ are

$\begin{eqnarray}
\begin{array}{cc|cc|}
P(r|s) &  & s &  \\
& & + & - \\
\hline
r & + & x & 1-x \\
& - & 1-x & x \\
\hline
\end{array}
\end{eqnarray}$

<br>

$\begin{eqnarray}
\begin{array}{cc|cc|}
P(s|r)  & & s &  \\
& & + & - \\
\hline
r & + & \frac{x\,s_+}{x\,s_+ + (1-x)\,(1-s_+)} & \frac{(1-x)\,(1-s_+)}{x\,s_+ + (1-x)\,(1-s_+)} \\
& - & \frac{(1-x)\,s_+}{(1-x)\,s_+ + x\,(1-s_+)} & \frac{x\,(1-s_+)}{(1-x)\,s_+ + x\,(1-s_+)} \\
\hline
\end{array}
\end{eqnarray}$


The conditional entropies $H(R|S)$ and $H(S|R)$ are

$\begin{eqnarray}

# **5. Continuous random variables (advanced)**

So far, we have considered only *discrete* random variables.   Shannon's approach can be extended to *continuous* variables by defining a *differential entropy* for a continuous random variable $x$ with density $p(x)$:

<br>

$\begin{eqnarray}
H_\mathit{diff} = -\int_x  p(x) \log_2 p(x) dx
\end{eqnarray}$

Unfortunately, the value of this differential entropy is not immediately meaningful.

<br>

It becomes interpretable either when we consider differences between *two* differential entropies, or when we specify the 'quantization'.

(Slide-37)

# Relation to entropy

Consider a continuous random variable $r$, measured with a precision $\Delta r$. Discretizing $r$ in bins of $\Delta r$ yields a discrete random variable $r'$.  If $\Delta r$ is small, we can approximate

<br>

$\begin{eqnarray}
\qquad\qquad P(r') = \int_{r'}^{r'+\Delta r} p(r) dr \approx p(r') \Delta r, \qquad\qquad 1 = \sum_{r'} P(r') = \sum_{r'} \left[p(r') \Delta r\right]
\end{eqnarray}$

<br>

The Shannon entropy of $r'$ and the differential entropy of $r$ are related:

<br>

$\begin{eqnarray}
H(r') =& -\sum_{r'} P(r') \log_2 P(r') = -\sum_{r'} \left[ p(r') \Delta r\right] \log_2 \left[ p(r') \Delta r\right]  =
\\
\\
=& -\sum_{r'}  \left[ p(r') \Delta r\right] \left[ \log_2 p(r') + \log_2 \Delta r\right]  =
\\
\\
=& -\sum_{r'}  \left[ p(r') \Delta r \right] \left[ \log_2 p(r) \right] - \left[\log_2 \Delta r\right] \sum_{r'} \left[ p(r') \, \Delta r\right] \approx
\\
\\
\approx &  -\int p(r) \log_2 p(r) \, dr - \log_2 \Delta r = H_\mathit{diff}(r) - \log_2 \Delta r
\end{eqnarray}$

<br>

In other words, the discretized entropy equals the differential entropy plus a term reflecting discretization, $-\log_2\Delta r$.

(Slide-38)

# Simple example

For simplicity, consider a continuous random variable $r \in [0,1]$ and an $n$-bit quantization bins of size $\Delta r = 2^{-n}$:

<br>

$\begin{eqnarray}
\qquad H(r') = H_\mathit{diff}(r) - \log_2 \Delta r = H_\mathit{diff}(r)  + n
\end{eqnarray}$

<br>

Discretized entropy and differential entropy differ by $n$, the number of bits of the quantization.


The discretized entropy is higher, by $n$, than the differential entropy.

The differential entropy is lower, by $n$, and often negative!

(Slide-39)



# Differential entropy for two variables

For two continuous random variables, measured with identical precision $\Delta r = \Delta s$, we substitute $p(r)\to \frac{P(r)}{\Delta r}$ and $p(r,s)\to \frac{P(r,s)}{\Delta r\Delta s}$ to find

<br>

$\begin{eqnarray}
 H_\mathit{diff}(r) &=& -\int p(r) \log_2 p(r) dr \qquad\qquad \approx \quad H(r') +\log_2 \Delta r
\\
\\
 H_\mathit{diff}(r,s) &=& -\int p(r,s) \log_2 p(r,s) dr\,ds \qquad \approx \quad H(r',s') +2\log_2 \Delta r
\\
\\
 I_\mathit{diff}(r,s)   &=& -\int p(r,s) \log_2 \frac{p(r)p(s)}{p(r,s)} dr\,ds \quad  \approx \quad I_m(r',s')
\\
\\
 H_\mathit{diff}(r|s) &=& -\int p(r,s) \log_2 \frac{p(r,s)}{p(s)} dr\,ds \qquad \approx \quad  H(r'|s') +\log_2 \Delta r
\end{eqnarray}$

<br>

When differential entropy is specificed, the discretization term is typically omitted.

# Differential entropy for flat distribution

As entropy is maximal when all outcomes (firing levels $r$) occur equally often, we consider a continuous ensemble $R$ of outcomes $r$ with a 'flat' distribution $p(r_)$:

<br>

$\begin{eqnarray}
\qquad\qquad p(r) = \frac{1}{r_\mathit{max}}, \qquad r \in [0, r_\mathit{max}]
\end{eqnarray}$

<br>

$\begin{eqnarray}
H(R) = - \int_0^{r_{max}} p(r) \log_2 p(r) \, dr = \frac{\log_2 r_\mathit{max}}{r_\mathit{max}} \int_0^{r_{max}} \, dr = \log_2 r_\mathit{max}
\end{eqnarray}$

<br>

Note that entropy grows arbitrarily with $r_\mathit{max}$

(Slide-40)

# Differential entropy for exponential distribution

<br>

Next we consider an exponential distribution

$\begin{eqnarray}
p(r) = \frac{1}{r_\mathit{ave}} e^{-r / r_\mathit{ave}}
\end{eqnarray}$

<br>

$\begin{eqnarray}
H(R) &=& - \int_0^{\infty} p(r) \log_2 p(r) \, dr =  
\\
&=& \frac{1}{r_\mathit{ave}} \int_0^{\infty} e^{-r/r_\mathit{ave}} \left[\frac{r}{r_\mathit{ave}} \log_2 e +\log_2 r_\mathit{ave} \right]\, dr =  
\\
&=& \log_2 e \left[\frac{1}{r_\mathit{ave}} \int_0^{\infty} e^{-r/r_\mathit{ave}} \, dr \right] + \log_2 r_\mathit{ave} \left[\frac{1}{r_\mathit{ave}^2} \int_0^{\infty} r \, e^{-r/r_\mathit{ave}} \, dr  \right]=
\\
&=& \log_2 e \left[-e^{-r/r_\mathit{ave}} \right]_0^\infty + \log_2 r_\mathit{ave} \left[\left(-\frac{r}{r_\mathit{ave}} -1\right) e^{-r/r_\mathit{ave}}\right]_0^\infty =
\\
&=& \log_2 e + \log_2 r_\mathit{ave}
\end{eqnarray}$

<br>

Entropy grows arbitrarily with $r_\mathit{ave}$


# Differential entropy for Gaussian distribution

Finally, we consider a continuous variable $r$ with Gaussian distribution

<br>

$\begin{eqnarray}
p(r) = \frac{1}{\sqrt{2\pi} \, \sigma_r} e^{-(r-r_\mathit{ave})^2 / 2\sigma^2_r}
\end{eqnarray}$

<br>

$\begin{eqnarray}
H_\mathit{diff}(R) &=& -\int_{-\infty}^{+\infty}  p(r) \log_2 p(r) \, dr =
\\
&=& - \frac{1}{\sqrt{2\pi} \, \sigma_r} \,
\int_{-\infty}^{+\infty} \, e^{-(r-r_\mathit{ave})^2 / 2\sigma^2_r}  \, \left[-\frac{(r-r_\mathit{ave})^2}{2\sigma_r^2} \log_2 e-\log_2\left(\sqrt{2\pi}  \, \sigma_r\right)\right] \, dr =
\\
&=& \frac{1}{2}\log_2 e + log_2\left(\sqrt{2\pi}  \, \sigma_r\right)=
\\
&=&
\frac{1}{2} \log_2 \left(2\pi e\right) + \log_2  \sigma_r
\\
&=& \frac{1}{2} \log_2 \left[ \left(2\pi e \right) \sigma_r^2\right]
\end{eqnarray}$
  
<br>

Entropy is independent of $r_\mathit{ave}$ and grows arbitrarily with $\sigma_r$

# Pairs of Gaussian signals

For two Gaussian distributed signals ('bivariate Gaussian'), the marginal differential entropies were derived above:


$\begin{eqnarray}
 h_1= \frac{1}{2} \log_2 \left[(2\pi e) \sigma^2_{11}\right],\qquad\qquad\qquad\qquad h_2=  \frac{1}{2} \log_2 \left[(2\pi e) \sigma^2_{22}\right]
\end{eqnarray}$

<br>

The joint differential entropy can be shown (see below) to be


$\begin{eqnarray}
h_{12}&=& \frac{1}{2} \log_2 \left[(2\pi e)^2 \left(\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}\right)\right]
\end{eqnarray}$

<br>

With this, it is easy to obtain the  *differential* mutual information

$\begin{eqnarray}
 i_m = h_1 + h_2 - h_{12} = -\frac{1}{2} \log_2 \frac{\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}}{\sigma^2_{11} \sigma^2_{22}}
\end{eqnarray}$

<br>

Recalling the definition of covariance $\sigma_{21}^2$ and of Pearson's correlation coefficient $\rho_{12}$,

$\begin{eqnarray}
\sigma_{21}^2 = \sigma_{12}^2  &\equiv& \left\langle (x_1-\mu_1)(x_2-\mu_2)\right\rangle
\\
\\
\rho_{12} &\equiv& \frac{\left\langle (x_1-\mu_1)(x_2-\mu_2)\right\rangle}{\sqrt{\left\langle (x_1-\mu_1)^2\right\rangle\left\langle (x_2-\mu_2)^2\right\rangle}}= \frac{\sigma_{12}}{\sigma_{11} \sigma_{22}}= \frac{\sigma_{21}}{\sigma_{11} \sigma_{22}}
\end{eqnarray}$

we can simplify this further to

$\begin{eqnarray}
 i_m  = -\frac{1}{2} \log_2 \frac{\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}}{\sigma^2_{11} \sigma^2_{22}} = -\frac{1}{2} \log_2 \left[1 - \rho^2_{12} \right]
\end{eqnarray}$


<br>

Thus, for bivariate Gaussian signals, the differential mutual information follows directly from the correlation coefficient.

# Pointers to multivariate Gaussian derivation (optional)

For the derivation of the joint entropy, we need multivariate Gaussian distributions, including the covariance matrix, its determinant, and its inverse matrix. Here are some pointers:

The marginal densities are

$\begin{eqnarray}
 p(x_1)= \frac{1}{\sqrt{2\pi \sigma_{11}^2} } \exp\left(-\frac{(x_1-\mu_1)^2}{2\sigma_{11}^2}  \right),\qquad p(x_2)= \frac{1}{\sqrt{2\pi \sigma_{22}^2} } \exp\left(-\frac{(x_2-\mu_2)^2}{2\sigma_{22}^2}  \right)
\end{eqnarray}$

and the joint density is best described in vector notation

$\begin{eqnarray}
p(x_1,x_2) = p({\bf z}) = \frac{1}{2\pi \left|{\bf S}\right|^{1/2} }\exp\left(-\frac{1}{2} \left(\bf z - \bf m \right)^T {\bf S}^{-1} \left(\bf z - \bf m \right) \right),\qquad\qquad
\\
\\
{\bf z} \equiv {{x_1}\choose{x_2}},
\qquad {\bf m} \equiv {{\mu_1}\choose{\mu_2}}, \qquad
{\bf S} \equiv \left(\begin{array}{cc} \sigma_{11}^2 & \sigma_{12}^2 \\ \sigma_{21}^2 & \sigma_{22}^2 \end{array} \right), \qquad |{\bf S}| = \sigma_{11}^2 \sigma_{22}^2 - \sigma_{12}^2 \sigma_{21}^2
\end{eqnarray}$

with means $\mu_{1,2}$, marginal variances $\sigma_{11}^2$ and $\sigma_{22}^2$, and covariances $\sigma_{21}^2=\sigma_{12}^2$:

$\begin{eqnarray}
\sigma_{11}^2 &\equiv& \left\langle (x_1-\mu_1)(x_1-\mu_1)\right\rangle
\\
\sigma_{22}^2 &\equiv& \left\langle (x_2-\mu_2)(x_2-\mu_2)\right\rangle
\\
\sigma_{21}^2 &=& \sigma_{12}^2  \equiv \left\langle (x_1-\mu_1)(x_2-\mu_2)\right\rangle
\end{eqnarray}$

<br>

The determinant and the inverse of covariance matrix ${\bf S}$ are easy to work out:

$\begin{eqnarray}
|{\bf S}| = \sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21},
\qquad\qquad
{\bf S}^{-1} = \frac{1}{|{\bf S}|}\left(\begin{array}{cc} \sigma_{22}^2 & -\sigma_{12}^2 \\ -\sigma_{21}^2 & \sigma_{11}^2 \end{array} \right)
\end{eqnarray}$

<br>

Next, we need to recall our linear algebra and work out the multivariate square term:

$\begin{eqnarray}
\left(\bf z - \bf m \right)^T {\bf S}^{-1} \left(\bf z - \bf m \right)
&=& \frac{1}{|{\bf S}|}\left(\begin{array}{cc}(x_1-\mu_1)&(x_2-\mu_2) \end{array} \right) \left(\begin{array}{cc} \sigma_{22}^2 & -\sigma_{12}^2 \\ -\sigma_{21}^2 & \sigma_{11}^2 \end{array} \right) \left(\begin{array}{c}x_1-\mu_1 \\ x_2-\mu_2 \end{array} \right) =
\\
\\
&=& \frac{(x_1-\mu_1)^2 \sigma_{22}^2 - (\sigma_{12}^2 +\sigma_{21}^2) (x_1-\mu_1)(x_2 - \mu_2)+  (x_2-\mu_2)^2 \sigma_{11}^2}{\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}}
\end{eqnarray}$

The expectation (probabilistic average) or this expression is

$\begin{eqnarray}
\left\langle\left(\bf z - \bf m \right)^T {\bf S}^{-1} \left(\bf z - \bf m \right)\right\rangle
&=& \frac{\langle(x_1-\mu_1)^2\rangle \sigma_{22}^2 - (\sigma_{12}^2 +\sigma_{21}^2) \langle(x_1-\mu_1)(x_2 - \mu_2)\rangle+  \langle(x_2-\mu_2)^2\rangle \sigma_{11}^2}{\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}}=
\\
\\
&=& \frac{\sigma_{11}^2 \sigma_{22}^2 - (\sigma_{12}^2 +\sigma_{21}^2) \sigma_{12}^2 +  \sigma_{22}^2 \sigma_{11}^2}{\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}} = 2
\end{eqnarray}$

<br>

After all these preliminaries, we can now derive the joint differential entropy:


$\begin{eqnarray}
h_{12}&=& -\int p({\bf z}) \log_2 p({\bf z}) \, d{\bf z} =
\\
&=& \int p({\bf z})\left[\frac{1}{2}\log_2\left[(2\pi)^2 \left|{\bf S}\right|\right] + \frac{1}{2} \left(\bf z - \bf m \right)^T {\bf S}^{-1} \left(\bf z - \bf m \right) \log_2 e\right]\, d{\bf z} =
\\
&=& \frac{1}{2}\log_2\left[(2\pi)^2 \left|{\bf S}\right|\right] + \frac{1}{2}\log_2 e \left\langle \left(\bf z - \bf m \right)^T {\bf S}^{-1} \left(\bf z - \bf m \right) \right\rangle =
\\
&=& \frac{1}{2} \log_2 \left[(2\pi)^2 \left(\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}\right)\right] + \log_2 e =
\\
&=& \frac{1}{2} \log_2 \left[(2\pi e)^2 \left(\sigma^2_{11} \sigma^2_{22} - \sigma^2_{12}\sigma^2_{21}\right)\right]
\end{eqnarray}$

with covariances $\sigma^2$ and correlation coefficient $\rho$.

# **6. Measures of statistical dependence**

# Measures of statistical dependence


To measure statistical dependence, we compare mutual information from discretized entropy, mutual information from differential entropy, and correlation coefficients.

<br>

**Discretized entropy**

We discretize the signal into 'bins' $k$ and determine probability $P_k$, bin width $d_k$, and density $p_k = P_k / d_k$.  To compute entropy, we use only bins with $0< P_k < 1$. The discretized entropy $H_x$ and (approximate) differntial entropy $h_x$ are

$\begin{eqnarray}
H_x = -\sum_k \left[ P_k \cdot \log_2 P_k \right], \qquad\qquad h_x = -\int p(x)\cdot \log_2 p(x) \, dx \approx -\sum_k d_k \cdot p_k \cdot \log_2 p_k
\end{eqnarray}$

**Differential entropy of Gaussian signal**

We assume a Gaussian signal and compute entropy directly from the variance or covariance

$\begin{eqnarray}
h_x = -\int p(x)\cdot \log_2 p(x) \, dx = \frac{1}{2} \log_2 \left[(2\pi e) \sigma_x^2 \right]
\end{eqnarray}$

**Correlation coefficient**

$\begin{eqnarray}
\rho_{12} \equiv\frac{\left\langle (x_1-\mu_1)(x_2-\mu_2)\right\rangle}{\sqrt{\left\langle (x_1-\mu_1)^2\right\rangle\left\langle (x_2-\mu_2)^2\right\rangle}}= \frac{\sigma_{12}}{\sigma_{12} \sigma_{22}}
\end{eqnarray}$

For our three signal pairs, all statistical measures of mutual dependency --  mutual information and correlation coefficients -- give comparable results.

In the exercise, you will encounter signal pairs where this is no longer the case!

```python
from inspect import signature
# @title Measures of mutual dependence {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

# generate two signals, X1 and X2, with N samples of each

def TwoSignals( N ):

    # principal component directions
    P = [
        [1/np.sqrt(5), 2/np.sqrt(5)],
        [-2/np.sqrt(5),1/np.sqrt(5)]
    ]

    # inverse matrix
    invP = np.linalg.inv(P);

    # measured directions
    B = [
        [1, 0],
        [0, 1]
        ]

    # square root of covariance matrix in principal components
    SIGMA_P = [
        [2, 0],
        [0, 1/2]
    ]

    COV_P = np.square( SIGMA_P )

    # covariance matrix in measurent directions

    COV_B = np.dot( np.transpose(P), np.dot( COV_P, P ) )

    # generate samples
    mean = [0,0];                    # mean of signal


    # either generate in principal component space ...
    XinP = np.random.multivariate_normal( mean, COV_P, size=N )

    # and subsequently rotate into measurement directions
    XinB = np.dot( XinP, P )

    # or, alternatively, generate directly in measurement space
    XinBalt = np.random.multivariate_normal( mean, COV_B, size=N )

    X1 = XinBalt[:,0];
    X2 = XinBalt[:,1];

    return X1, X2

def MyEntropy( PX, DX ):
    # inputs: probability vector PX and bin size vector DX
    # outputs: discrete entropy HX and differential entropy hX

    # density vector
    pX = PX / DX;

    # restrict to valid probability values (0 or 1 do not contribute to entropy!)
    kk = np.where( (0 < PX ) & ( PX < 1 ) );

    HX = -np.sum( PX[kk] * np.log2( PX[kk] ) );
    hX = -np.sum( DX[kk] * pX[kk] * np.log2( pX[kk] ) );

    return HX, hX;

def DiscreteMutualInformation( X1, X2 ):

    # Compute histogram with Scott's rule for bin width and normalization
    bin_width_1 = 3.49 * np.std(X1) / (len(X1) ** (1/3))  # Scott's rule
    Edges1 = np.arange(np.min(X1), np.max(X1) + bin_width_1, bin_width_1)  # Bin edges
    p1, Edges1 = np.histogram(X1, bins=Edges1, density=True)

    bin_width_2 = 3.49 * np.std(X2) / (len(X2) ** (1/3))  # Scott's rule
    Edges2 = np.arange(np.min(X2), np.max(X2) + bin_width_2, bin_width_2)  # Bin edges
    p2, Edges2 = np.histogram(X2, bins=Edges2, density=True)

    # bin widths
    D1 = np.diff(Edges1);
    D2 = np.diff(Edges2);

    # Normalize to 'probability'
    P1 = p1 * D1  # multiply density by bin widths
    P2 = p2 * D2

    P1 = P1 / np.sum(P1);
    P2 = P2 / np.sum(P2);

    # Marginal entropies
    [H1, h1]   = MyEntropy( P1, D1 );
    [H2, h2]   = MyEntropy( P2, D2 );

    # Compute 2D histogram
    p12, Edges1, Edges2 = np.histogram2d(X1, X2, bins=[Edges1, Edges2], density=True)

    # Normalize to 'probability' (if needed)
    D12 = np.outer(D1, D2)  # Calculate bin areas
    P12 = p12 * D12  # Scale to probabilities

    P12 = P12 / np.sum(P12)

    # Joint entropies

    [H12, h12] = MyEntropy( P12, D12 );

    IM12       = H1 + H2 - H12;

    im12       = h1 + h2 - h12;

    return H1, H2, H12, IM12, h1, h2, h12, im12;

def GaussianDifferentialMutualInformation( X1, X2 ):

    # Covariance matrix
    CovX = np.cov(X1, X2)

    # Correlation coefficients
    RhoX = np.corrcoef(X1, X2);
    Rho12 = RhoX[1,0];

    # Marginal entropies
    hX1    = 0.5 * np.log2( 2*np.pi*np.exp(1) * CovX[0,0] );
    hX2    = 0.5 * np.log2( 2*np.pi*np.exp(1) * CovX[1,1] );

    hX12   = 0.5 * np.log2( (2*np.pi*np.exp(1))**2 * np.linalg.det(CovX ) );
    imX12  = -0.5 * np.log2( np.linalg.det(CovX) / ( CovX[0,0] * CovX[1,1] ) );

    return hX1, hX2, hX12, imX12, Rho12;

def plot_mutual_dependence( log10N=4 ):
    """(Slide-48)"""

    # sample size
    N = round(10**log10N);

    # generate signal pair X1, X2
    (X1, X2) = TwoSignals( N );

    # generated transformed versions
    Y1 = (   X1 + 2*X2)/(np.sqrt(5)*1.6);   # linear transformation (PCA)
    Y2 = (-2*X1 + X2  )/(np.sqrt(5)*0.4);

    Z1 = 0.9*np.tanh(0.5*X1+1);           # non-linear transformation
    Z2 = 0.9*np.tanh(0.5*X2+1);

    # mutual information from discretized samples
    HX1, HX2, HX12, IMX12, hX1, hX2, hX12, imX12 = DiscreteMutualInformation( X1, X2 )
    HY1, HY2, HY12, IMY12, hY1, hY2, hY12, imY12 = DiscreteMutualInformation( Y1, Y2 )
    HZ1, HZ2, HZ12, IMZ12, hZ1, hZ2, hZ12, imZ12 = DiscreteMutualInformation( Z1, Z2 )

    # differential mutual information, assumes Gaussian distribution
    hX1, hX2, hX12, imX12, RhoX12 = GaussianDifferentialMutualInformation( X1, X2 );
    hY1, hY2, hY12, imY12, RhoY12 = GaussianDifferentialMutualInformation( Y1, Y2 );
    hZ1, hZ2, hZ12, imZ12, RhoZ12 = GaussianDifferentialMutualInformation( Z1, Z2 );


    # Plot results
    fig, axs = plt.subplots( 1,3, figsize=(12,4))

    # first subplot

    axs[0].plot( X1, X2, 'r.', markersize=1 )
    str1 = 'IMX=';
    str2 = f"{IMX12:.2f}";
    axs[0].text(-9,+9.0,f"{str1}{str2}",fontsize=12)
    str1 = 'imX=';
    str2 = f"{imX12:.2f}";
    axs[0].text(-9,+7.5,f"{str1}{str2}",fontsize=12)
    str1 = 'rhoX=';
    str2 = f"{RhoX12:.2f}";
    axs[0].text(-9,+6.0,f"{str1}{str2}",fontsize=12)
    axs[0].set_xlim([-10,10])
    axs[0].set_ylim([-10,10])
    axs[0].set_xlabel('x_1')
    axs[0].set_box_aspect(1/1)
    axs[0].set_ylabel('x_2')
    axs[0].set_title('Pair X', fontsize=14);

    # second subplot

    axs[1].plot( Y1, Y2, 'b.', markersize=1 )
    str1 = 'IMY=';
    str2 = f"{IMY12:.2f}";
    axs[1].text(-9,+9.0,f"{str1}{str2}",fontsize=12)
    str1 = 'imY=';
    str2 = f"{imY12:.2f}";
    axs[1].text(-9,+7.5,f"{str1}{str2}",fontsize=12)
    str1 = 'rhoY=';
    str2 = f"{RhoY12:.2f}";
    axs[1].text(-9,+6.0,f"{str1}{str2}",fontsize=12)
    axs[1].set_xlim([-10,10])
    axs[1].set_ylim([-10,10])
    axs[1].set_xlabel('y_1')
    axs[1].set_box_aspect(1/1)
    axs[1].set_ylabel('y_2')
    axs[1].set_title('Pair Y', fontsize=14);

    # third subplot

    axs[2].plot( Z1, Z2, 'k.', markersize=1 )
    str1 = 'IMZ=';
    str2 = f"{IMZ12:.2f}";
    axs[2].text(-0.95,+0.90,f"{str1}{str2}",fontsize=12)
    str1 = 'imZ=';
    str2 = f"{imZ12:.2f}";
    axs[2].text(-0.95,+0.75,f"{str1}{str2}",fontsize=12)
    str1 = 'rhoZ=';
    str2 = f"{RhoZ12:.2f}";
    axs[2].text(-0.95,+0.60,f"{str1}{str2}",fontsize=12)
    axs[2].set_xlim([-1,1])
    axs[2].set_ylim([-1,1])
    axs[2].set_xlabel('z_1')
    axs[2].set_box_aspect(1/1)
    axs[2].set_ylabel('z_2')
    axs[2].set_title('Pair Z', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_mutual_dependence, log10N = (1.0, 7.0, 1 ) )
```

```python
from inspect import signature
# @title Three pairs of neurons {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

# generate responses from four neurons, return three pairs of responses

def FourNeurons(N):

    R1  = np.random.uniform(0,11,N);    # uniformly distributed between 1 and 11
    R1 = np.ceil( R1 );

    Off1 = (4.0*(R1-6.0)/5.0);    # offset correlated with R1

    Mu2 = (5.0 + Off1) / 10;
    N2  = np.ones(N,dtype='int') * 10;
    R2  = np.random.binomial(N2, Mu2);       # binomially distributed around mean R1

    RandomSign = 2*np.round(np.random.uniform(0,1,N))-1;
    Mu3 = (5.0 +  Off1 * RandomSign) / 10;
    R3  =  np.random.binomial( N2, Mu3);       # bimodally and binomially distributed around mean +-R1

    RandomSign = 2*np.round(np.random.uniform(0,1,N))-1;

    Mu4 = (5.0 + 4.0*RandomSign) / 10;  # bimodally and binomially distributed
    R4  =  np.random.binomial( N2, Mu4);

    return R1, R2, R3, R4

def plot_three_pairs_of_neurons( log10N=4 ):
    """(Slide-50)"""

    # sample size
    N = round(10**log10N);

    # generate three pairs of neurons R
    (R1, R2, R3, R4) = FourNeurons( N );

    # mutual information from discretized samples
    HR1, HR2, HR12, IMR12, hR1, hR2, hR12, imR12 = DiscreteMutualInformation( R1, R2 )
    HR1, HR3, HR13, IMR13, hR1, hR3, hR13, imR13 = DiscreteMutualInformation( R1, R3 )
    HR1, HR4, HR14, IMR14, hR1, hR4, hR14, imR14 = DiscreteMutualInformation( R1, R4 )

    # Correlation coefficients
    RhoX = np.corrcoef(R1, R2);
    Rho12 = RhoX[1,0];

    RhoX = np.corrcoef(R1, R3);
    Rho13 = RhoX[1,0];

    RhoX = np.corrcoef(R1, R4);
    Rho14 = RhoX[1,0];

    # obtain histograms
    (Count12, Edges1, Edges2) = np.histogram2d(R1, R2, bins=10, density=True)
    (Count13, Edges1, Edges3) = np.histogram2d(R1, R3, bins=10, density=True)
    (Count14, Edges1, Edges4) = np.histogram2d(R1, R4, bins=10, density=True)

    # Plot results
    # use alternative calls to enable 3D projection

    fig = plt.figure(figsize=(12,4))

    # prepare barplot
    x  = np.linspace(1,10, 10, dtype='int')
    x, y = np.meshgrid(x, x)  # Create a grid
    x = x.flatten()  # Flatten to 1D for bar3d
    y = y.flatten()  # Flatten to 1D for bar3d
    z = np.zeros_like(x)  # z-coordinates (all bars start at 0)
    Count12 = Count12.flatten();
    Count13 = Count13.flatten();
    Count14 = Count14.flatten();

    # first subplot

    axs = fig.add_subplot(131, projection='3d')

    axs.bar3d(x, y, z, 0.6, 0.6, Count12, shade=True, color=(0.5,0.5,1.0))
    str1 = 'IM=';
    str2 = f"{IMR12:.2f}";
    axs.text(1,1,0.12,f"{str1}{str2}",fontsize=12)
    str1 = 'im=';
    str2 = f"{imR12:.2f}";
    axs.text(1,1,0.10,f"{str1}{str2}",fontsize=12)
    str1 = 'rho=';
    str2 = f"{Rho12:.2f}";
    axs.text(1,1,0.08,f"{str1}{str2}",fontsize=12)
    axs.set_xlim([0,11])
    axs.set_ylim([0,11])
    axs.set_xlabel('R_1')
    axs.set_ylabel('R_2')
    axs.set_title('Pair R_1, R_2', fontsize=14);

    # second subplot

    axs = fig.add_subplot(132, projection='3d')

    axs.bar3d(x, y, z, 0.6, 0.6, Count13, shade=True, color=(1.0,0.5,0.5))
    str1 = 'IM=';
    str2 = f"{IMR13:.2f}";
    axs.text(1,1,0.06,f"{str1}{str2}",fontsize=12)
    str1 = 'im=';
    str2 = f"{imR13:.2f}";
    axs.text(1,1,0.05,f"{str1}{str2}",fontsize=12)
    str1 = 'rho=';
    str2 = f"{Rho13:.2f}";
    axs.text(1,1,0.04,f"{str1}{str2}",fontsize=12)
    axs.set_xlim([0,11])
    axs.set_ylim([0,11])
    axs.set_xlabel('R_1')
    axs.set_ylabel('R_3')
    axs.set_title('Pair R_1, R_3', fontsize=14);


    # third subplot

    axs = fig.add_subplot(133, projection='3d')

    axs.bar3d(x, y, z, 0.6, 0.6, Count14, shade=True, color=(0.4, 0.8, 0.4))
    str1 = 'IM=';
    str2 = f"{IMR14:.2f}";
    axs.text(1,1,0.06,f"{str1}{str2}",fontsize=12)
    str1 = 'im=';
    str2 = f"{imR14:.2f}";
    axs.text(1,1,0.05,f"{str1}{str2}",fontsize=12)
    str1 = 'rho=';
    str2 = f"{Rho14:.2f}";
    axs.text(1,1,0.04,f"{str1}{str2}",fontsize=12)
    axs.set_xlim([0,11])
    axs.set_ylim([0,11])
    axs.set_xlabel('R_1')
    axs.set_ylabel('R_4')
    axs.set_title('Pair R_1, R_4', fontsize=14);



    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_three_pairs_of_neurons, log10N = (1.0, 7.0, 1 ) )
```

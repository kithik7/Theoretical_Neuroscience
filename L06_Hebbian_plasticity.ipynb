# **Abstract:**

We consider changes of connectivity ("plasticity") that are driven by activity. This creates a coupled dynamics of activity (faster) and connectivity (slower). **Unsupervised:** only input activity is constrained. **Supervised**: both input and output activity are constrained. **Reinforcement:** input activity is constrained and output activity is modulated by a reinforcement signal ('teacher').  

**Hebbian learning** depends both on pre- and postsynaptic activity ('fire together, wire together').  Success strengthens, failure weakens connection.  The **formal analysis** shows that $\Delta {\bf w}$ grows with average product $\left\langle{v \,{\bf u}}\right\rangle$ of pre- and postsynaptic activity, or activity compared to threshold $\left\langle{(v-\theta_v) \, {\bf u}}\right\rangle$, or $\left\langle{v \, ({\bf u}-\theta_u)}\right\rangle$.  

In consequence, the dynamics is governed by the statistical structure of input patterns (principal eigenvectors of covariance matrix):  connection weights grow fastest along dominant eigenvector.

Unfortunately, the **connection dynamics is unstable** and weights grow beyond all bounds.

# **1. Activity-driven plasticity**

Consider a two-layer network with inputs ${\bf u}$ driving recurrently connected outputs ${\bf v}$.  The activity dynamics is governed by

$$
\tau_a \frac{d  {\bf v}}{dt} = - {\bf v} +  F \left({\bf w} \cdot {\bf u} + {\bf M} \cdot {\bf v} \right)
$$

where ${\bf w}$ and ${\bf M}$ are the feedforward and recurrent synaptic weight matrices, respectively.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec01_many2many_rec.png" height="200">

(Slide-1)

# 'Plastic' connectivity

In general, we speak of "plasticity" when synaptic weights change over time, and of "activity-driven plasticity" when this change depends on network activity.

<br>

'Plastic' feedforward connectivity ${\bf w}(t)$:

$$
\tau_w \frac{d  w_{ij} }{dt} = - w_{ij} +  f( w_{ij}, {\bf u}, {\bf v})
$$

The dynamics of activity ${\bf v}(t)$ depends on feedforward connectivity ${\bf w}(t)$ while, in return, plasticity of feedforward connectivity ${\bf w}(t)$ depends on activity ${\bf v}(t)$.

<br>

This co-evolution of two mutually dependent variables is called 'coupled dynamics'.

(Slide-2)

# "Cells that fire together wire together"

# Hebb's rule


-  The Hebbian rule is local in that it compares pre- and post-synaptic firing.
-  The success of A in triggering B strengthens the connection.
-  The failure of A in triggering B weakens the connection.
-  The Hebbian rule is intrinsically dichotomous, leading to either maximally strong or completely broken connections.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec06_Hebb_principle.png" width="800">

# Non-Hebbian plasticity

Not all plasticity depends on both pre- and post-synaptic activity.  For example, in short-term-depression or -potentiation (STD, STP), the pre-synaptic activity alone modifies the synaptic weight.   In such cases one speaks of "non-Hebbian plasticity"

# Points to note

<br>

-  There are two kinds of activity-dependent plasticity.

<br>

-  "Hebbian" plasticity depends both on directly pre- and directly post-synaptic activity.

<br>

-  "Non-Hebbian" plasticity includes all other possibilities, for example, plasticity that depends only on directly pre-synaptic activity, or only on directly post-synaptic activity.

# **3. Formalization of Hebbian plasticity**


We analyze Hebbian plasticity formally in a feedforward network

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec06_Many2one.png" width="300">

with the dynamic equation

$$
\tau_a \, \frac{dv}{dt} = -{\bf v} + F({\bf w} \cdot {\bf u})
$$

where $\tau_a$ is the time constant of activity.  In order to shift the focus to plasticity, we make some drastic simplifying assumptions.

<br>

-  The activation function is the identity function (linear approximation):

$$
F(x) \to x \qquad\Rightarrow \qquad \frac{dv}{dt} = -{\bf v} + {\bf w} \cdot {\bf u}
$$

<br>

-  Output activity is at steady-state::

<br>

$$
\tau_a \, \frac{dv}{dt} = 0 \qquad \Rightarrow \qquad v = {\bf w} \cdot {\bf u}
$$

<br>

-  Plasticity is driven by ensemble averages of pre-synaptic acticivities $\left\langle{\bf u}\right\rangle$ and of post-synaptic activity $\left\langle v\right\rangle$.  In other words, we consider acticity ensembles and their statistics over a longer period of time, rather than activities at particular points of time.

<br>

- Finally, we assume that the change of connection weights $\bf w$ is an ensemble average over some function $f()$  of pre- and post-synaptic activities

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle f\left( v \,{\bf u} \right) \right\rangle \qquad \Leftrightarrow \qquad \tau_w \, \frac{d w_k}{dt} = \left\langle f\left({v}, \, u_k \right) \right\rangle \qquad \forall k
$$

# Correlation rule


<br>

A simple implementation of Hebb's rule is to make weights ${\bf w}$ grow in proportion to the average product and to use $f(v, {\bf u}) = v \, {\bf u}$:

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle{v \, {\bf u}}\right\rangle, \qquad v = {\bf w} \cdot {\bf u}
$$

<br>

$$
\tau_w \, \frac{dw_k}{dt} = \left\langle{v \, u_k}\right\rangle, \qquad v = \sum_k \, w_k \cdot u_k
$$

In other words, the weight $w_k$ of a synapse $u_k \to v$ grows comparatively rapidly, if both  $u_k$ and $v$ are large, but comparatively slowly when either $u_k$ or $v$ or both are small.

(Slide-14)

# Correlation formulation

Note that the value of of $v$ also depends on $u_k$, so that the value of the product $v\, u_k$ depends **doubly** on $u_k$!

For this reason, the dynamic equation can be reformulated in terms of the correlation matrix ${\bf Q}$ of inputs ${\bf u}$:

<br>

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle{v \, {\bf u}}\right\rangle = \left\langle{{\bf w} \cdot {\bf u} \, {\bf u}}\right\rangle = \left\langle{{\bf u} \, {\bf u} \cdot {\bf w}}\right\rangle = \left\langle{{\bf Q} } \right\rangle \cdot {\bf w}
$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec06_Ensemble_stats.png" width="800">

<br>

where ${\bf Q} \equiv \left\langle{{\bf u} \, {\bf u}}\right\rangle$ is correlation matrix and $\left\langle{}\right\rangle$ is taken only over input activities.  The dimensions of vectors, matrices, and their products are illustrated graphically.

(Slide-15)

# 2D case

Let's prove this for a two dimensional example:

<br>

$$
\left\langle{v}\right\rangle = \left\langle{ {\bf w} \cdot {\bf u}}\right\rangle = {\bf w} \cdot \left\langle{\bf u}\right\rangle = \left(w_1 \quad w_2\right) \cdot
\left(\begin{array}{c}
  \langle{u_1}\rangle \\ \langle{u_2}\rangle \end{array}\right)
= w_1 \left\langle{u_1}\right\rangle + w_2 \left\langle{u_2}\right\rangle
$$

$$\begin{eqnarray}
\langle v \, {\bf u}\rangle &=& \langle{\bf w} \cdot {\bf u} \, {\bf u}\rangle = {\bf w} \cdot \langle {\bf u \, u} \rangle =
(w_1 \quad w_2) \cdot
\left\langle
\left(\begin{array}{c} {u_1} \\ {u_2} \end{array}\right)   \, \left( u_1 \quad u_2\right)
\right\rangle =
\\
&=& (w_1 \quad w_2) \cdot \left( \begin{array}{cc} \left\langle{u_1^2}\right\rangle  & \left\langle{u_1 u_2}\right\rangle \\ \left\langle{u_2 u_1}\right\rangle & \left\langle{u_2^2}\right\rangle \end{array} \right)=
\left(\begin{array}{c}
w_1 \left\langle{u_1^2}\right\rangle + w_2 \left\langle{u_1 u_2}\right\rangle
\\
w_1 \left\langle{u_2 u_1}\right\rangle + w_2 \left\langle{u_2^2}\right\rangle
\end{array} \right)
\end{eqnarray}$$

$$\begin{eqnarray}
\langle{ v {\bf u}}\rangle &=& \langle{{\bf u} \, {\bf u}}\rangle  \cdot {\bf w} = \langle{{\bf Q}}\rangle \cdot {\bf w}
&=&  
\left( \begin{array}{cc}
\left\langle{u_1^2}\right\rangle  & \left\langle{u_1 u_2}\right\rangle
\\
\left\langle{u_2 u_1}\right\rangle & \left\langle{u_2^2}\right\rangle
\end{array} \right)
\cdot
\left( \begin{array}{c}{w_1}\\{w_2}\end{array}\right) =
\left(\begin{array}{c}
{w_1 \left\langle{u_1^2}\right\rangle + w_2 \left\langle{u_1 u_2}\right\rangle}
\\
{w_1 \left\langle{u_2 u_1}\right\rangle + w_2 \left\langle{u_2^2}\right\rangle}
\end{array}\right)
\end{eqnarray}$$

The 2D case illustrates how to go about proving the general case.

(Slide-16)

# No steady-state!

<br>

How will weights develop over time?   Will they reach a steady-state?  Consider the norm of weights

$$
|{\bf w}|^2 ={\bf w} \cdot {\bf w} =\sum_k w_k^2
$$

<br>

It's time derivative is

$$\begin{eqnarray}
\tau_w \, \frac{d |{\bf w}|^2}{dt} &=& \tau_w \, \frac{d ({\bf w} \cdot {\bf w})}{dt} =2\, {\bf w} \cdot  \tau_w \, \frac{d  {\bf w}}{dt} =
\\
\\
&=& 2 \, {\bf w} \cdot \langle{v {\bf u}}\rangle = 2 \langle{{\bf w} \cdot v {\bf u}}\rangle = 2 \left\langle{ v^2}\right\rangle \qquad > 0
\end{eqnarray}$$

<br>

and therefore positive (as long as the activity $v \neq 0$).  Thus, there is no steady-state and weights grow without bound!

(Slide-17)

# Covariance rule

To ensure that some weights decrease while others increase, we can introduce a threshold, say, on the presynaptic side.  A convenient choice is to use the average presynaptic activity

<br>

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle v \, \left({\bf u} - \theta_{\bf u}\right) \right\rangle, \qquad \theta_{\bf u} = \langle{\bf u}\rangle, \qquad v = {\bf w} \cdot {\bf u}
$$

<br>

This can be reformulated

<br>

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle{ {\bf w} \cdot {\bf u} \, \left({\bf u} - \theta_{\bf u}\right) }\right\rangle = {\bf w} \, \langle{{\bf u} \, {\bf u}}\rangle - {\bf w} \, \langle{\bf u}\rangle \langle{\bf u}\rangle = {\bf C} \cdot {\bf w}
$$

<br>

where $\quad{\bf C} \equiv \langle{{\bf u} \, {\bf u}}\rangle - \langle{\bf u}\rangle\langle{\bf u}\rangle\quad$ is the covariance matrix.

(Slide-18)

# Alternative motivation


<br>

Alternatively, we can introduce a threshold on the post-synaptic side.  Surprisingly, the resulting average rule is the same

<br>

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle{(v - \theta_v) \, {\bf u}  }\right\rangle, \qquad  \theta_v = \langle{ v}\rangle, \qquad v = {\bf w} \cdot {\bf u}
$$

<br>

Reformulate

<br>

$$
\tau_w \, \frac{d{\bf w}}{dt} = \left\langle{ \left({\bf w} \cdot {\bf u} - \theta_v \right)\, {\bf u}  }\right\rangle = {\bf w} \langle{{\bf u} \, {\bf u}}\rangle - {\bf w} \langle{\bf u}\rangle \langle{\bf u}\rangle = {\bf C} \cdot {\bf w}
$$

<br>

where $\quad{\bf C} \equiv \langle{{\bf u} \, {\bf u}}\rangle - \langle{\bf u}\rangle\langle{\bf u}\rangle\quad$ is once again the covariance matrix.

(Slide-20)

# No steady-state!

<br>

Although the covariance rule ensures that weights increase and decrease, it still does not provide for stability.  In fact, now the weights grow without bound in both directions.

Consider the norm (sum of squares) of the weights

<br>

$$
|{\bf w}|^2 \equiv {\bf w} \cdot {\bf w}
$$

<br>

It's time derivative is

<br>

$$\begin{eqnarray}
\tau_w \, \frac{d |{\bf w}|^2}{dt} &=& \tau_w \, \frac{d ({\bf w} \cdot {\bf w})}{dt} =2 \, {\bf w} \cdot \tau_w \,  \frac{d  {\bf w}}{dt} =
\\
\\
&=& 2 \, {\bf w} \cdot \left\langle{(v - \theta_v) \, {\bf u}}\right\rangle = 2 \left\langle{v \, (v - \theta_v)}\right\rangle = 2 \left( \left\langle{ v^2}\right\rangle - \left\langle{v}^2\right\rangle \right)  \qquad > 0
\end{eqnarray}$$

<br>

and therefore positive as long as the activity $v$ is not identically zero.  We will return to the problem of instability in the next lecture.

(Slide-21)


# Points to note

<br>

-  In Hebbian plasticity, weight change depends on both pre- and post-synaptic activity.

-  As post-synaptic activity also depends on pre-synaptic activity, weight change in effect depends doubly on pre-synaptic activity.

-  Assuming that weight change is slow and driven by the ensemble average, we obtain the **correlation rule** $\ldots$
$$
\tau_w \, \frac{d{\bf w}}{dt} = \langle{{\bf Q}}\rangle \cdot {\bf w} \qquad\qquad {\bf Q} \equiv \langle{{\bf u} \, {\bf u}}\rangle
$$

-  $\ldots$ or, if we assume a threshold, the **covariance rule**
$$
\tau_w \, \frac{d{\bf w}}{dt} = \langle{{\bf C}}\rangle \cdot {\bf w} \qquad\qquad {\bf C} \equiv \langle{{\bf u} \, {\bf u}}\rangle - \langle{\bf u}\rangle\langle{\bf u}\rangle
$$

<br>

-  Neither rule provides for stable weights, at least not without further ado.

(Slide-22)


# **4. Correlations, covariances, and eigenvectors**

<br>

Let's remind ourselves about "correlations" and "covariances". Here we use  correlation to denote the average product of **absolute values**

<br>

$$
{\bf Q} = \langle{{\bf u} \,\, {\bf u}}\rangle = \langle{u_i\,u_j}\rangle
$$

<br>

and covariance the average product of **deviations from the mean**

<br>

$$
{\bf C} = \left\langle \left( {\bf u} - \langle{\bf u}\rangle\right)^2\right\rangle = \langle{{\bf u} \,{\bf u}}\rangle - \langle{\bf u}\rangle \langle{\bf u}\rangle = \langle{u_i\,u_j}\rangle - \langle{u_i}\rangle\langle{u_j}\rangle
$$

<br>

Correlation and covariance **coefficients** are additionally normalized by individual variance:

<br>

$$
{\bf Q}_N = \frac{\langle{u_i\,u_j}\rangle}{\sqrt{\left\langle{u_i^2}\right\rangle\left\langle{u_j^2}\right\rangle}}, \qquad\qquad
{\bf C}_N = \frac{\langle{u_i\,u_j}\rangle- \langle{u_i}\rangle\langle{u_j}\rangle}{\sqrt{\left\langle{u_i^2}\right\rangle\left\langle{u_j^2}\right\rangle}}
$$

(Slide-23)

# Statistical structure of activity patterns

<br>

-  Statistical structure of activity patterns is described by **correlations** (absolute values) and **covariances** (values relative to the mean).

-  How informative one activity is about another ('are they a team?').

-  Eigenvectors of **correlation / covariance** matrix are typical activity patterns.

-  Dominant eigenvectors are dominant patterns.

-  Statistical structure is often illustrated by scatter plots (see below).

-  Analyze statistical structure in eigenvector space.

# Decomposition of covariance matrix

The correlation matrix decomposes into a sum of eigenvector correlations

<br>

$$
C = E \cdot D \cdot E^T = \left[{\bf e}_1, {\bf e}_2, {\bf e}_3\right] \cdot
\left(\begin{array}{ccc}
\lambda_1 & 0 & 0 \\ 0 & \lambda_2 & 0 \\ 0 & 0 & \lambda_3
\end{array}\right)
\cdot
\left[{\bf e}_1, {\bf e}_2, {\bf e}_3\right]^T \quad =
$$

<br>

$$
= \left(\begin{array}{ccc}
e_{11} & e_{21} & e_{31} \\ e_{12} & e_{22} & e_{32} \\ e_{13} & e_{23} & e_{33}
\end{array}\right)
\cdot
\left(\begin{array}{ccc}
\lambda_1 & 0 & 0 \\ 0 & \lambda_2 & 0 \\ 0 & 0 & \lambda_3
\end{array}\right)
\cdot
\left(\begin{array}{ccc}
e_{11} & e_{12} & e_{13} \\ e_{21} & e_{22} & e_{23} \\ e_{31} & e_{32} & e_{33}
\end{array}\right) =
$$

<br>

$$
= \lambda_1 \left(\begin{array}{ccc}
e_{11}^2 & e_{12} e_{11} & e_{13} e_{11} \\
e_{11} e_{12} & e_{12}^2 & e_{13} e_{12} \\
e_{11} e_{13} & e_{12} e_{13} & e_{13}^2
\end{array}\right)
+
\lambda_2 \left(\begin{array}{ccc}
e_{21}^2 & e_{22} e_{21} & e_{23} e_{21} \\
e_{21} e_{22} & e_{22}^2 & e_{23} e_{22} \\
e_{21} e_{23} & e_{22} e_{23} & e_{23}^2
\end{array}\right)
+
\lambda_3 \left(\begin{array}{ccc}
e_{31}^2 & e_{32} e_{31} & e_{33} e_{31} \\
e_{31} e_{32} & e_{32}^2 & e_{33} e_{32} \\
e_{31} e_{33} & e_{32} e_{33} & e_{33}^2
\end{array}\right)
=
$$

<br>

$$
= \lambda_1 {\bf e}_1 {\bf e}_1 + \lambda_2 {\bf e}_2 {\bf e}_2 + \lambda_3 {\bf e}_3 {\bf e}_3
$$

```python
from inspect import signature
# @title Correlation and eigenvectors {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap

# generate two signals, X1 and X2, with N samples of each


def EVanalysis( M ):
    # matrix is assumed to be symmetric
    # so that eigenvectors and eigenvalues are real

    (Eigvals, Eigvecs) = np.linalg.eig(M)

    # ignore spurious complex parts
    Eigvals = np.real( Eigvals )
    Eigvecs = np.real( Eigvecs )

    # choose sign of eigenvectors for visualization
    Esign = np.sign( np.mean(Eigvecs,axis=0) );
    for k in range( len(Esign) ):
         Eigvecs[:,k] = Esign[k] * Eigvecs[:,k]

    # sort eigenvectors in descending order of eigenvalues
    sidx = np.argsort( Eigvals )
    sidx = sidx[::-1]
    Eigvals = Eigvals[sidx]
    Eigvecs = Eigvecs[:,sidx]


    # extract first two eigenvectors
    E1 = Eigvecs[:,0]
    E2 = Eigvecs[:,1]
    E3 = Eigvecs[:,2]

    return Eigvecs, Eigvals, E1, E2, E3


def ThreeSignals( N, thetaZ, thetaY, lambda1, lambda2 ):

    # Measurement coordinates
    COORD_B = [
        [1, 0, 0],
        [0, 1, 0],
        [0, 0, 1]
    ]

    # rotate about Z
    RotZ = [
        [ np.cos(thetaZ), -np.sin(thetaZ), 0],
        [ np.sin(thetaZ),  np.cos(thetaZ), 0],
        [             0,               0,  1 ]
    ]

    # rotate about Y
    RotY = [
        [ np.cos(thetaY), 0, np.sin(thetaY) ],
        [              0, 1,             0  ],
        [-np.sin(thetaY), 0, np.cos(thetaY) ]
    ]

    # principal component axes by two successive rotations

    COORD_P = np.dot( RotY, np.dot( RotZ, COORD_B ) );

    # construct covariance matrix

    LAMBDA = [
        [lambda1**2,           0, 0],
        [0,           lambda2**2, 0],
        [0,                    0, 1]
    ]

    CORR = np.dot( COORD_P, np.dot( LAMBDA, np.transpose(COORD_P) ) )

    # mean array
    MEAN_P = np.zeros(3)

    # generate directly in measurement space
    XinP = np.random.multivariate_normal( MEAN_P, CORR, size=N, check_valid='warn' )

    X1 = XinP[:,0];
    X2 = XinP[:,1];
    X3 = XinP[:,2];

    return X1, X2, X3, XinP, CORR


def plot_correlation_and_covariance( thetaZ=0, thetaY=0, lambda1=1, lambda2=1 ):
    """(Slide-29)"""

    N = 1000

    # generate samples from covariance matrix


    (X1, X2, X3, XinP, CORR) = ThreeSignals( N, thetaZ, thetaY, lambda1, lambda2 )

    # eigenvectors and eigenvalues
    (Eigvecs, Eigvals, E1, E2, E3) = EVanalysis( CORR )

    E1 = E1 * lambda1
    E2 = E2 * lambda2

    CORR1 = np.outer(E1, E1)
    CORR2 = np.outer(E2, E2)
    CORR3 = np.outer(E3, E3)

    # print( COV )

    print( CORR )

    print( np.corrcoef( np.transpose( XinP )) )


    # Create the plot
    fig  = plt.figure(figsize=(8, 8))

    axs = fig.add_subplot(111, projection='3d')  # Specify 3D projection

    axs.plot( X1, X2, -10+np.zeros_like(X3), '.', color=[0.5,1.0,0.5], linewidth=2 )
    axs.plot( X1, -10+np.zeros_like(X2), X3, '.', color=[0.5,1.0,0.5], linewidth=2 )
    axs.plot( -10+np.zeros_like(X1), X2, X3, '.', color=[0.5,1.0,0.5], linewidth=2 )

    axs.plot( X1, X2, X3, 'r.', linewidth=2 )

    axs.plot( (0,E1[0]), (0,E1[1]), (0,E1[2]), 'b-', linewidth=2 )
    axs.plot( (0,E2[0]), (0,E2[1]), (0,E2[2]), 'g-', linewidth=2 )

    axs.view_init(elev=30, azim=37)  # 30° elevation, 45° azimuth

    axs.set_xlim(-10,10)
    axs.set_ylim(-10,10)
    axs.set_zlim(-10,10)
    axs.set_xlabel('x')
    axs.set_ylabel('y')
    axs.set_zlabel('z')
    axs.set_title('Correlation and eigenvectors', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_correlation_and_covariance, thetaZ=(-3,+3,0.5), thetaY=(-3,+3,0.5), lambda1=(1,10,1), lambda2=(1,10,1) )

```

# **5. Linear systems analysis**


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec06_Many2one.png" width="600">

<br>

Our formulation of the Hebbian plasticity of feedforward connectivity $\bf w$ leads to a linear dynamic system

<br>

$$
\tau \, \frac{d{\bf w}}{dt} = {\bf C} \cdot {\bf w}, \qquad\qquad {\bf C} = \langle{{\bf u} \, {\bf u}}\rangle - \langle{\bf u}\rangle \langle{\bf u}\rangle
$$

<br>

which we can solve formally.  The solution is a linear combination of eigenvectors ${\bf e}_\mu$ of the covariance matrix. As a covariance matrix is symmetric, the eigenvectors are orthogonal and all eigenvalues are real.

<br>

$$
\lambda_\mu \, {\bf e}_\mu = {\bf C} \cdot  {\bf e}_\mu,\qquad\qquad {\bf e}_\mu \cdot {\bf e}_\nu = \delta_{\mu\nu}
$$

<br>

The linear combination of eigenvectors

$$
{\bf w}(t) = \sum_{\mu=1}^N \, C_\mu \, e^{\lambda_\mu \, t / \tau} \, {\bf e}_\mu, \qquad {\bf w}(t) \cdot {\bf e}_\mu = C_\mu \,  e^{\lambda_\mu \, t / \tau}, \qquad C_\mu = {\bf w}(0) \cdot {\bf e}_\mu
$$

has exponentially varying coefficients. Positive eigenvalues will make the weights grow beyond all bounds.

(Slide-37)


# Speed of growth


<br>

Speed of growth in each 'eigendirection' depends on associated eigenvalue and on the projection of the initial condition.

<br>

If the largest (=principal) eigenvalue $\lambda_1$ is unique, weight growth along the associated (=principal) eigenvector ${\bf e}_1$ dominates development:

<br>

$$
{\bf w}(t\rightarrow\infty) \propto {\bf e}_1
$$

<br>

as long as the initial condition does not preclude such development, in other words, as long as

<br>

$${\bf w}(0) \cdot {\bf e}_1 \neq 0$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec06_Eigenvector_growth.png" width="600">

(Slide-38)

# Points to note

<br>

-  Linear analysis of the covariance rule provides useful qualitative} information.

-  The weight vector grows as a weighted sum of the eigenvectors of the covariance matrix:

$$
{\bf w}(t) = \sum_{\mu=1}^{N_{\bf u}} \, C_\mu \, e^{\lambda_\mu t / \tau_w} \, {\bf e}_\mu
$$

-  While growth in all direction is exponential, it is dominated by growth along the principal eigenvector.

-  Thus, the weight vector aligns itself with the principal eigenvector.

-  The **principal eigenvector** identifies the **tightest direction** of the scatter plot of correlations.

# Weight normalization (a hack)

Finally, let's consider a Hebbian covariance rule with weight normalization.  At every time-step, all weights are reduced proportionately to maintain an overall square weight of unity.

This is a hack, but it allows us to visualize the development of synaptic weights.

<br>

$$
\tau_w \, \frac{d{\bf w}}{dt} = v \, \left({\bf u} - \langle{\bf u}\rangle\right), \qquad\qquad  v = {\bf w} \cdot {\bf u}, \qquad\qquad \sqrt{ {\bf w} \cdot {\bf w} } = 1
$$

<br>


For uncorrelated inputs ${\bf u}$, the weights ${\bf w}$ fail to converge.

<br>

Whereas for correlated inputs ${\bf u}$, the weights ${\bf w}$ converge to the principal eigenvector.


(Slide-40)

```python
from inspect import signature
# @title Hebbian plasticity with normalization {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap

# generate two signals, X1 and X2, with N samples of each

def PlasticityIterationStep( w, u, ubar, dt, tau ):

    v = np.dot( w, u )

    # Hebbian plasticity rule
    w_new = w + v * (u-ubar) * dt / tau;

    # weight normalization
    w_new = w_new / np.linalg.norm( w_new )

    return w_new



def plot_hebbian_with_normalization( Tend=1, thetaZ=0, thetaY=0, lambda1=1 ):
    """(Slide-41)"""

    N = 1000;

    dt = Tend / N;

    tau = 0.1;

    # generate samples from covariance matrix

    lambda2 = 1
    lambda3 = 1

    (X1, X2, X3, U_i, CORR) = ThreeSignals( N, thetaZ, thetaY, lambda1, lambda2 )

    # eigenvectors and eigenvalues
    (Eigvecs, Eigvals, E1, E2, E3) = EVanalysis( CORR )

    # iterate plasticity

    print( np.shape( U_i ))

    w_ini = np.random.rand(3);
    w_ini = w_ini / np.linalg.norm( w_ini )

    W_i = np.zeros_like( U_i )

    W_i[0,:] = w_ini

    Ubar = U_i[0,:]

    for n in range( 1, N ):

        W_i[n,:] = PlasticityIterationStep( W_i[n-1,:], U_i[n-1,:], Ubar, dt, tau )

        Ubar = np.mean( U_i[0:n,:], axis=0 )

    # Create the plot
    fig  = plt.figure(figsize=(12, 8))

    axs1 = fig.add_subplot(121, projection='3d')  # Specify 3D projection

    axs1.plot( U_i[:,0], U_i[:,1], -10+np.zeros_like(U_i[:,2]), '.', color=[0.5,1.0,0.5], linewidth=2 )
    axs1.plot( U_i[:,0], -10+np.zeros_like(U_i[:,1]), U_i[:,2], '.', color=[0.5,1.0,0.5], linewidth=2 )
    axs1.plot( -10+np.zeros_like(U_i[:,0]), U_i[:,1], U_i[:,2], '.', color=[0.5,1.0,0.5], linewidth=2 )

    axs1.plot( U_i[:,0], U_i[:,1], U_i[:,2], 'r.', linewidth=2 )

    axs1.plot( (0,E1[0]), (0,E1[1]), (0,E1[2]), 'm-', linewidth=2 )
    axs1.plot( (0,E2[0]), (0,E2[1]), (0,E2[2]), 'b-', linewidth=2 )
    axs1.plot( (0,E3[0]), (0,E3[1]), (0,E3[2]), 'g-', linewidth=2 )

    axs1.view_init(elev=30, azim=37)  # 30° elevation, 45° azimuth

    axs1.set_xlim(-10,10)
    axs1.set_ylim(-10,10)
    axs1.set_zlim(-10,10)
    axs1.set_xlabel('x')
    axs1.set_ylabel('y')
    axs1.set_zlabel('z')
    axs1.set_title('Input samples', fontsize=14);

    axs2 = fig.add_subplot(122, projection='3d')  # Specify 3D projection

    axs2.plot( W_i[:,0], W_i[:,1], W_i[:,2], 'k-', linewidth=1 )
    axs2.plot( W_i[:,0], W_i[:,1], W_i[:,2], 'k.', linewidth=2 )

    axs2.plot( W_i[:,0]-1, W_i[:,1], W_i[:,2], '.', color=[0.8,0.8,0.8], linewidth=1 )
    axs2.plot( W_i[:,0], W_i[:,1]-1, W_i[:,2], '.', color=[0.8,0.8,0.8], linewidth=1 )
    axs2.plot( W_i[:,0], W_i[:,1], W_i[:,2]-1, '.', color=[0.8,0.8,0.8], linewidth=1 )

    axs2.plot( (0,E1[0]), (0,E1[1]), (0,E1[2]), 'm-', linewidth=2 )
    axs2.plot( (0,E2[0]), (0,E2[1]), (0,E2[2]), 'b-', linewidth=2 )
    axs2.plot( (0,E3[0]), (0,E3[1]), (0,E3[2]), 'g-', linewidth=2 )

    axs2.plot( (0,E1[0]), (-1,E1[1]-1), (0,E1[2]), 'm--', linewidth=2 )
    axs2.plot( (-1,E1[0]-1), (0,E1[1]), (0,E1[2]), 'm--', linewidth=2 )
    axs2.plot( (-1,E2[0]-1), (0,E2[1]), (0,E2[2]), 'b--', linewidth=2 )
    axs2.plot( (0,E2[0]), (0,E2[1]), (-1,E2[2]-1), 'b--', linewidth=2 )
    axs2.plot( (0,E3[0]), (-1,E3[1]-1), (0,E3[2]), 'g--', linewidth=2 )
    axs2.plot( (0,E3[0]), (0,E3[1]), (-1,E3[2]-1), 'g--', linewidth=2 )

    axs2.view_init(elev=30, azim=37)  # 30° elevation, 45° azimuth

    axs2.set_xlim(-1,1)
    axs2.set_ylim(-1,1)
    axs2.set_zlim(-1,1)
    axs2.set_xlabel('x')
    axs2.set_ylabel('y')
    axs2.set_zlabel('z')
    axs2.set_title('Weights', fontsize=14);

    plt.tight_layout()
    plt.show()



# Creating sliders for concentrations and charge
interact(plot_hebbian_with_normalization, Tend=(1,5,1), thetaZ=(-3,+3,0.5), thetaY=(-3,+3,0.5), lambda1=(1,10,1) )

```

# Intuitive summary

<br>
-  For simplicity, assume that all inputs are (on average) equally large and equally variable.

<br>

-  The weights of co-varying inputs ("team-members") are strengthened, because they succeed more reliably in triggering post-synaptic activity.

<br>

-  The weights of other inputs ("loners") are weakened, because they succeed less reliably.

<br>

-  The Hebbian covariance rule "internalizes" the statistical input structure: statistical "teams" become groups with large synaptic weights.

# **General summary**


<br>

-  Formal learning models distinguish "unsupervised", "supervised", and "reinforcement" learning.

<br>

-  When the Hebbian rule ("fire together, wire together") is formalized, it leads to the "covariance rule"

<br>

$$
\tau_w \, \dot {{\bf w}} = {\bf C} \cdot {\bf w}
$$

<br>

-  Ignoring instability permits linear analysis and shows that weights ${\bf w}$ converge to principal eigenvector(s) ${\bf e}_{1}$ of the input covariance $\bf C = \langle{\bf u} {\bf u}\rangle - \langle {\bf u}\rangle \langle {\bf u}\rangle$.

<br>

-  Hebbian synapses learn input covariances, internalizing the statistical structure of the outside world.

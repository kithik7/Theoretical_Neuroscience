# **1. Feedforward networks for classification**

<br>

- *Feedforward networks* map activity patterns (inputs) onto other activity patterns (output).

<br>

- A *classification* maps many activity patterns (inputs) onto a few *classes* (categories) of patterns.

<br>

- What *classifications* can be performed by *feedforward networks*?

(Slide-1)


# Feedforward many-to-one

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec01_many2one_ff.png" width="600">

In a feedforward network, $N_u$ input units with rates $u_i$ converge via synapses with weights $w_i$ on a single output units with rate $v$.   For convenience, we collect input units and synaptic weight into vectors $\bf u$ and $\bf w$.  The output rate is then approximated by

$$\begin{eqnarray}
\tau \frac{d v}{dt} = -v + F \left( \bf w \cdot \bf u \right) \qquad {\rm or} \qquad \tau_r \frac{dv}{dt} = - v + F \left( \sum_{i=1}^{N_u} w_{i} \, u_i \right)
\end{eqnarray}$$

At steady-state, the output firing rate is

$$\begin{eqnarray}
v_{ss}  = F\left(  \bf w \cdot \bf u \right)
\end{eqnarray}$$

(Slide-2)

# Vector space interpretation

Consider an $N_i$-dimensional vector space, spanned by input activities $\bf u = \{u_i\}$.  The weights $\bf w = \{w_i\}$ define a vector of unit length in this space ($|w|=1$).

The dot product $\bf u \cdot \bf w = |u| \,\cos\alpha$ is the (positive or negative) length of the  *projection* of the input vector $\bf u$ onto the weight vector $\bf w$.

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Vector_space.png" width="600">

<br>

The steady-state response $v_{ss}  = F\left( \bf w \cdot \bf u \right)$ represents how well input vector $\bf u$ aligns with weight vector $\bf w$.  If the alignment is sufficiently good -- so that $v_{ss}$ is above some threshold $\theta$, $v_{ss}>\theta$ --  the steady-state response approaches unity.   If it is not, the steady-state response remains near zero.

<br>

Thus, a feedforward projection computes how well the input vector matches the weight vector!

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Vector_space_b.png" width="600">

(Slide-3)

# Feedforward many-to-many

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec01_many2many_ff.png" width="600">

In a feedforward network, $N_u$ input units with rates $u_i$ converge via synapses with weights $w_{io}$ onto $N_v$ output units with rates $v_o$.  For convenience, input and output units are collected in vectors $\bf u$ and $\bf v$, and the synaptic weights into a matrix $\bf W$.  The output rate is then approximated by

$$\begin{eqnarray}
\tau \frac{d {\bf v}}{dt} = -{\bf v} + {\bf F} \left( {\bf W} \cdot {\bf u} \right) \qquad {\rm or} \qquad \tau_r \frac{dv_o}{dt} = - v_o + F \left( \sum_{i=1}^{N_u} w_{io} \, u_i \right)
\end{eqnarray}$$

At steady-state, the output firing rate is

$$\begin{eqnarray}
{\bf v}_\mathit{ss}  = {\bf F}\left(  {\bf W \cdot \bf u} \right)
\end{eqnarray}$$

<br>

For each output unit $o$, the steady-state response is determined by a vector dot product (independently of other units)

$$\begin{eqnarray}
v^{ss}_o  = F\left(  {\bf w}_{o} \cdot {\bf u} \right) = \left[w_{1o} \, w_{2o} \, \ldots \, w_{{N_i}o} \right] \cdot \left[\begin{array}{c} u_1 \\ u_2  \\  \vdots \\ u_{N_i} \end{array}\right]
\end{eqnarray}$$

The steady-state response depends on the *projection* of input vector
$\bf u$ onto the weight vector $\bf w_o$, in other words, the $o^\mathit{th}$ row of the weight matrix $\bf W$. This happens multiple times, as this matrix has $N_o$ rows!

<br>


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Vector_space.png" width="300">

(Slide-4)

# **Summary**

- A many-to-one *feedforward network* maps input activity vector ${\bf u}$ onto scalar output activity ${v}$ (steady-state).

- A many-to-many *feedforward network* maps input activity vector ${\bf u}$ onto output activity vector ${\bf v}$ (steady-state).

- The mapping is determined by a synaptic vector ${\bf w}$ or a synaptic matrix ${\bf W}$:

$$\begin{eqnarray}
v_\mathit{ss}  = F\left(  \bf w \cdot \bf u \right)   \qquad\qquad {\bf v}_\mathit{ss}  = F\left(  \bf W \cdot \bf u \right)
\end{eqnarray}$$

- Thus, feedforward networks perform *vector dot product* operations.

- This insight is crucial for understanding the *classifications* that can be performed by feedforward networks.

(Slide-4)

# **2. Perceptrons**

- Idealized feedforward networks can function as *binary classifiers*, mapping input vectors onto binary values of $\pm 1$ (neglecting $0$) .

- Binary classifiers were called *perceptrons* by their inventor, Frank Rosenblatt (1958).  

- Minsky and Papert showed that single-layer perceptron can distinguish only linearly separable patterns (1969, "Perceptrons: an introduction to computational geometry," MIT Press.)

- Multi-layer perceptrons are more capable.

- Perceptrons are ancestral to convolutional networks and the underlying concepts are more relevant than ever.

- Perceptrons are no longer considered a model for cognitive function.

(Slide-5)


# **Hyperbolic tangent**

Several steps of simplification are involved in the transformation of a feedforward to a perceptron. The first step is to replace the activation function with the hyberbolic tangent, $\tanh$, which is a sigmoidal and symmetric function:

$$\begin{eqnarray}
F(x) = \tanh(x) = \frac{e^{x-\theta}-e^{-x+\theta}}{e^{x-\theta}+e^{-x+\theta}}, \qquad\qquad F(x-\theta) = \tanh(x-\theta) = \frac{e^{x-\theta}-e^{-x+\theta}}{e^{x-\theta}+e^{-x+\theta}}
\end{eqnarray}$$

<br>

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Hyperbolic_tangent.png" width="400">

<br>

The threshold and the limits of the $\tanh$ are
$$\begin{eqnarray}
\lim_{x\rightarrow 0} \tanh(x)  =  \frac{1 - 1}{1 + 1} = 0
\qquad
\lim_{x\rightarrow +\infty} \tanh(x)  =  \frac{e^x - 0}{e^x+ 0} = 1
\qquad
\lim_{x\rightarrow -\infty} \tanh(x)  =  \frac{0 - e^x}{0 + e^x} = -1
\end{eqnarray}$$

(Slide-6)




# Steady-state mappings

<br>

With  $\tanh$ for an activation function, the steady-state mapping of a many-to-one feedforward network becomes

$$\begin{eqnarray}
v = \tanh\left[ {\bf w} \cdot {\bf u} \right]
\end{eqnarray}$$

Positive (negative) dot products ${\bf w} \cdot {\bf u}$ map onto positive (negative) outputs $v$.  

<br>

Additional mappings become possible when a threshold $\theta$ is introduced

$$\begin{eqnarray}
 v = \tanh\left[ {\bf w} \cdot {\bf u} + \theta\right]
\end{eqnarray}$$


<br>

In both cases, all possible
inputs $u$ are mapped onto two output *categories* or *classes* (positive and negative).

(Slide-7)




# Sign function

Even easier to analyze are activation functions that approach the "hard delimiter"  $\mathrm{sign}(x)$

$$\begin{eqnarray}
\lim_{\gamma\rightarrow \infty} \, \tanh(\gamma x) = \mathrm{sign}(x) = \left\{ \begin{array}{cc} +1 & x>0 \\ 0 & x=0 \\ -1 & x<0 \end{array}\right.
\end{eqnarray}$$

<br>

As $\gamma$ grows large, the mapping equation becomes

$$\begin{eqnarray}
v =  \tanh \left[  {\bf w} \cdot  {\bf u} \right] =  \tanh \left[ \gamma \, \left( \gamma^{-1} {\bf w} \cdot  {\bf u} \right)\right]
\approx {\mathrm{sign}}\left[\gamma^{-1} {\bf w} \cdot {\bf u} \right]
\end{eqnarray}$$

<br>

Absorbing the factor $\gamma^{-1}$ into the weights by defining ${\bf m} \equiv \gamma^{-1} \, {\bf w}$ gives us

$$\begin{eqnarray}
v =  \tanh \left[  {\bf w} \cdot  {\bf u} \right] \approx {\mathrm{sign}}\left[\gamma^{-1} {\bf w} \cdot {\bf u} \right] = {\mathrm{sign}}\left[ {\bf m} \cdot {\bf u} \right]
\end{eqnarray}$$

(Slide-8)


# Hard feedforward networks

The steady-state mappings of many-to-one and many-to-many feedforward networks with a ``hard'' activation function are:

<br>

**Many-to-one:**
$${\bf u}^\mu \longrightarrow v^\mu = \mathrm{sign}\left[  {\bf m} \cdot  {\bf u}^\mu \right]$$

<br>

**Many-to-many:**

$${\bf u}^\mu \longrightarrow {\bf v}^\mu =\mathbf{sign}\left[ {\bf M} \cdot  {\bf u}^\mu \right]$$

<br>

where $\mu$ indexes different input patterns ${\bf u}^\mu = \{u_1^\mu, \ldots , u_N^\mu \}$.

<br>

Output vectors ${\bf v}^\mu$ assume only two values $\in \{-1, 1 \}$.   The probability of a value $0$ is infinitesimal and can be ignored.

(Slide-9)

We find that perceptrons can realize only linearly separable mapping functions.

For two binaryinput dimensions (see below), perceptrons can realize 14 out of 16 possible dichotomous mappings.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Dichotomy_e.png" width="800">

# All possible dichotomous mappings

Two binary inputs permit four input vectors, which allow 16 possible dichotomous mappings.  14 of these mappings are linearly separable and can be realized with a `hard feedforward' network. The remaining 2 are unseparable.

As the number of inputs increases, the fraction of 'mappable' dichotomies decreases rapidly.


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Dichotomy_f.png" width="800">


In general, the number of linearly separable dichotomous mappings of
$N$ points in  $d$-dimensional space is given by Cover's formula:

$$ M(N,d) = 2 \sum_{k=0}^d \frac{(N-1)!}  {k! \, (N-k-1)!}$$

where

$$n!\equiv 1 \cdot 2 \cdot \ldots \cdot n$$

is the factorial product.

In higher dimensions, the overwhelming majority of mappings are NONSEPARABLE and CANNOT be realized!

(Slide-17)

# Points to note

<br>

- In their most abstract form, feedforward networks become binary classifiers.

<br>

- They map any input vector ${\bf u}^\mu$ to either one of two output classes: $\{+1, -1\}$.

<br>

- The mapping has a straightforward vector interpretation: weights ${\bf m}$ define a normal plane (hyperplane), that bisects the input space.

<br>

- Inputs are mapped to either one or the other side of this plane.

<br>

- Feedforward networks can perform only *linearly separable* classifications, severely restricting their usefulness.

# The reference frame problem

- Translating from perception to action encounters a `reference frame problem'.

- Visual information originates in EYE-centered coordinates.  Due to eye movements, this is *ambiguous* and does not determine relative motion of body and environment.

- To determine relative motion, additional information is needed: gaze direction and vestibular inputs.

-  Visual information is transformed into HEAD-centered coordinates by taking gaze direction into account.  

- This format is more suitable for guiding reaching movements.  It is also suitable for integrating vestibular, auditory, and tactile information.

# **4. Coordinate transformation with feedforward projection**

Neurophysiological studies of the tuning properties in posterior parietal cortex PPC -- by Richard Andersen and others -- consistently showed response tuning in EYE coordinates.  In other words, responses depend on retinal stimulus position $s$ or on stimulus position relative to eye fixation.


Puzzlingly, the projection targets of PPC populations, the premotor cortex PMC populations, show response tuning in HEAD coordinates. In other words, responses depend on $s+g$, where $s$ is retinal position and $g$ is gaze direction $g$ (eye fixation relative to the head).

<br>

The position of an object in space, relative to the head, depends on the direction of gaze $g$, relative to the head, and on the position on the retina $s$, relative to the direction of gaze.

The direction of gaze is often called 'eye fixation'.

In PPC, tuning curves of visually responsive neurons depend on EYE coordinates $s$. However, the amplitude of tuning is modulated by gaze direction. Specifically, tuning is pronounced for $g\gg\theta_g$, but diappears for $g\ll\theta_g$, where $\theta_g$ is some threshold value.

# Feedforward network projecting from EYE-tuned to HEAD-tuned populations.

Zipser & Andersen model these experimental findings with a straightforward feedforward network, projecting from visual neurons with activities $\bf u$ to motor neurons with activities $\bf v$ using projection weights $\bf w$.

The motor population represents different possible values of $s+g$ (different neurons respond best to different values of $s+g$). Activity $v$ is a function of HEAD-coordinates $s+g$:

$$v = H(s+g)$$

The visual population represents different possible combinations of $s$ and $g$ (different neurons respond for different combinations). Activity $u$ is the product of tuning function $f(s)$ and gain modulation $h(g)$:

$$
u = f(s) \cdot h(g)$$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Handeye_h.png" width="800">

[(Salinas & Abbott, 2001)](https://www.sciencedirect.com/science/article/abs/pii/S0079612301300122)

(Slide-25)




# Role of gain modulation

How can visual neurons with receptive fields in EYE coordinates drive motor neurons with receptive fields in HEAD coordinates?

<br>

The answer was supplied by Richard Andersen and colleagues, who explained the mystery with a computational model of 'gain modulation' [(Zipser, Andersen, 1988)](https://www.nature.com/articles/331679a0)

<br>

They showed that when visual responses in EYE-coordinates are 'gain-modulated' by GAZE direction, this permits object position to be 'read out' in HEAD-coordinates.

<br>

Subsequently, the principle of 'gain modulation' was found in other projections as well.  For example, auditory responses in HEAD coordinates are gain modulated by neck direction, facilitating subsequent read-out in BODY-coordinates.

(Slide-26)

# Key idea of gain modulation

To understand the key idea, we illustrate the two populations schematically, arranged by selectivity to retinal stimulus position $s$, to gaze direction $g$, and to stimulus position in head coordinats, $s+g$.

Note that visual neurons positioned along a particular off-diagonal (dotted line) are selective for combinations $(s,g)$ that maintain the same value of $s+g=\theta$. The two dotted lines mark $\theta_1=7$ and $\theta_2=9$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Gainmodulation_b.png" width="800">

This illustration highlights (in blue) three visual neurons that respond to different stimulus directions, but only when gaze direction is larger than simulus direction minus threshold, $g\geq s - \theta_1$. All three visual neurons project to the same motor neuron.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec02_Gainmodulation_c.png" width="800">

In summary, visual neurons respond to particular stimulus positions (in EYE coordinates), but only for some gaze positions.  Motor neurons receive input from all visual neurons along a particular off-diagonal. By combining all of these inputs, their responses become independent of gaze direction.

# **5. Implementation of visuomotor projection**


We construct an input layer of both *'simple' and 'complex'* visual neurons $i$, with average response $r_i$ and $u_i$, respectively.

**Simple responses** depend  *multiplicatively* on stimulus position $s$ (in eye coordinates) and on gaze position $g$ (in head coordinates):

$$
r_i^\pm = f(s - \bar s_i) \, h^\pm( g - \bar g_i)
$$

where $\bar s_i$ and $\bar g_i$ are  *preferred stimulus* and *critical gaze* position of neuron $i$, respectively.  Gaussian dependence (with width $\sigma$) on stimulus position $(s-\bar s_i)$

$$
f(s - \bar s_i) =  r_{max} \,  \exp\left[-\frac{(s-\bar s_i)^2}{2 \sigma^2} \right]
$$

is multiplied with positive or negative sigmoidal dependence (of sharpness $\kappa$) on gaze direction $(g-\bar g_i)$

$$
h^\pm(g - \bar g_i) =  \frac{\exp[\pm\kappa (g - \bar g_i)]}{1+\exp[\pm\kappa (g - \bar g_i)]}
$$

**Complex responses** $u_i$ are constructed from pairs of simple responses with identical preference $\bar s_i$ and $\bar g_i$, but opposite sigmoidal dependence:

$$\begin{eqnarray}
u_i &=& 4 \left[r_i^+\right]\cdot \left[r_i^-\right] =
\\
&=& 4 \left[f(s - \bar s_i) \, h^+( g - \bar g_i)\right] \cdot \left[f(s - \bar s_i) \, h^-( g - \bar g_i)\right]
\end{eqnarray}$$

(Slide-28)

# Feedforward connectivity, general principle

Can we find a connectivity $w_{io} (\bar s_i, \bar g_i)$ such that output activity $v_o$ becomes a function of $s+g$,  $v_o=H(s+g)$ ?  


Assuming full coverage, for any neuron $i$ preferring $(\bar s_i, \bar g_i)$ and any value of $g$, there is another neuron $j$ preferring $(\bar s_j = \bar s_i-g, \bar g_j = \bar g_i+g)$ with identical response:

$$\begin{eqnarray}
f(s-\bar s_i) \, h(g - \bar g_i) = f(s-\bar s_j) \, h(g - \bar g_j) = f(s+g-\bar s_i) \, h(- \bar g_i)
\end{eqnarray}$$

If feedforward weights satisfy

$$\begin{eqnarray}
 w_{io}(\bar s_i, \bar g_i) \stackrel{!}{=} w_{io}( \bar s_i +  \bar g_i )
\end{eqnarray}$$

the output response can be written as

$$\begin{eqnarray}
v_o &=& F\left[ \sum_i  \, w_{io} (\bar s_i-g + \bar g_i+g) \, f(s+g-\bar s_i) \, h(- \bar g_i) \right] =
\\
&=& F\left[ \sum_i  \, w_{io} (\bar s_i + \bar g_i) \, f(s+g-\bar s_i) \, h(- \bar g_i) \right]=
\\
&=&  H(s+g)
\end{eqnarray}$$

and is a function of $s+g$.

# Feedforward connectivity, specifics


Consider a visual neuron with preferences $(\bar s_i, \bar g_i)$ and a motor neuron with preference $\bar h_o$. The projection weight $w_{io}$ depends on the distances $\bar h_o - \bar s_i -\bar g_i$, as follows

$$\begin{eqnarray}
w_{io} = \exp\left[-\frac{\left(\bar h_o - \bar s_i -\bar g_i\right)^2}{2 \lambda^2}\right]
\end{eqnarray}$$

where $\lambda$ is the tuning width of the projection.

# Motor population


Steady-state activity of the output layer of motor neurons $o$, which receives feedforward projections from complex visual neurons $i$, is defined as

$$\begin{eqnarray}
v_o &=& F\left[ \sum_i  \, w_{io} \, u_i  \right], \qquad\qquad F[x] = r_{max}\frac{ x^2}{\alpha^2+ x^2}
\end{eqnarray}$$

where $w_{io}$ are projection weights and $F[x]$ is the activation function with half-maximal abscissa $\alpha$.

```python
# @title Visual and motor activity {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import gamma
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.mplot3d import Axes3D

def plot_visuomotor_activity( s_plus_g=0, lamb=0.01, alpha=0.8, toggle=0):
    """ Slide-33 """

    N_s=40;
    N_g=50;
    N_s_plus_g=80;

    # response of visual populations
    # range of preferred stimulus positions
    smax = 1.5
    smin = -smax

    # range of critical gaze directions
    gmin = smin
    gmax = smax

    # visual tuning
    rmax   = 1;
    sigma  = 0.4;
    kappa  = 4;

    # combinations of preferred stimulus positions and critical gaze directions
    si       = np.linspace(smin, smax, N_s)
    gj       = np.linspace(gmin, gmax, N_g)
    Sij, Gij = np.meshgrid(si, gj);

    SplusGij = Sij + Gij;

    # flatten for computation
    si = np.reshape(Sij,-1);
    gi = np.reshape(Gij,-1);
    splusgi_visual = np.reshape(SplusGij,-1);

    kappa = 4;

    # random stimulus position
    s_true = np.random.uniform(s_plus_g-smax, s_plus_g-smin,1)
    g_true = s_plus_g - s_true;

    Rposi   = RSgausssigmoid(s_true, si, sigma, g_true, gi,  kappa, rmax)
    Rnegi   = RSgausssigmoid(s_true, si, sigma, g_true, gi, -kappa, rmax)
    Ui      = 4 * Rposi * Rnegi

    # reassemble arrays for visualization
    Rposij   = np.reshape(Rposi, (N_g,N_s))
    Rnegij   = np.reshape(Rnegi, (N_g,N_s))
    Uij      = np.reshape(Ui, (N_g,N_s))

    # response of motor populations
    # preferences of motor neurons
    splusgj_motor = np.linspace(smin, smax, N_s_plus_g)

    # feedforward projection
    (Pvisji, Pmotji) = np.meshgrid( splusgi_visual, splusgj_motor );
    Diffji = Pvisji - Pmotji;

    # make connections
    Wji = np.exp( -Diffji**2/(2*lamb**2) )


    # matrix multiplication
    Lj = np.matmul(Wji, Ui)

    # activation function
    Vj = rmax * (Lj**2)/(alpha**2 + Lj**2)

    # motor positions
    hk   = splusgj_motor;
    il   = np.linspace(0,1,10)

    (Hkl, Ikl) = np.meshgrid(hk, il);

    Vkl = np.tile(Vj, (1,10))
    Vkl = np.transpose(Vkl)

    # Create a figure and subplots
    (fig, axs) = plt.subplots(1, 2, figsize=(8,4))  # 1x2 grid of subplots

    # First subplot
    contourA = axs[0].contourf(Sij, Gij, Uij, levels=50, vmin=0, vmax=1,cmap='magma')
    axs[0].plot( [smin,smax], [smax,smin], 'w:')
    axs[0].plot( [smax,0], [0,smax], 'w:')
    axs[0].plot( [0,smin], [smin,0], 'w:')

    axs[0].set_xlabel('retinal position s')
    axs[0].set_ylabel('gaze direction g')
    axs[0].set_aspect(1/1)
    axs[0].set_title('Visual responses')

    fig.colorbar(contourA, ax=axs[0], shrink=0.8)

    # Second subplot
    contourB = axs[1].contourf(Hkl, Ikl, Vkl, levels=50, cmap='magma')

    axs[1].set_xlabel('head position s+g')
    axs[1].set_aspect(1/4)
    axs[1].set_title('Motor population')

    #fig.colorbar(contourB, ax=axs[1], shrink=0.5)

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_visuomotor_activity, s_plus_g=(-1.5,1.5,0.5), lamb=(0.01,0.2,0.01), alpha=(0.2,0.8,0.1), toggle=(0,1,1) )


```

- Motor neurons (in premotor cortex) are tuned in HEAD coordinates.

w_{io}=w( \bar s_i, \bar g_i) \stackrel{!}{=} w( \bar s_i + \bar g_i)

ensure that motor responses depend on $s+g$ (HEAD coordinates).

# **5. Final thoughts**

There have been several reports of **'shifting' receptive fields** in sensory modalities.  For example, visual and auditory receptive fields in parietal cortex may shift with eye position, planned eye position, arm or head position.


In the context of gain modulation, 'shifting' receptive fields present no mystery:

When a target population is driven by a gain-modulated source population, we expect to find neurons with receptive fields that 'shift' with the gain-modulating input (e.g., with gaze direction).

Thus, 'shifting' receptive fields do not imply that  projection weights change (projections are being established or broken)!

(Slide-34)


Many alternative **population coding schemes** are conceivable.  Which are chosen by the brain?  For what reasons?  The choice of a particular scheme may reflect $\ldots$


- $\ldots$ accuracy: tuning functions and sampling match statistics of external events

- $\dots$ connectivity: functional organization to minimize wiring

- $\dots$ development: functional organization can be constructed with simple developmental rules

- $\dots$ further usefulness: optimize ease and efficiency of ``read-out'' by subsequent neurons.


Multiplicative gain modulation appears related to the latter issue.

# **Abstract:**

A recurrent networks may be viewed as an infinite cascade of feedforward networks.  How can we analyze the behaviour and convergence?   

A **Linear approximation** provides the crucial mathematical tools: eigenvectors and eigenvalues of the connectivity matrix.

The solution of this approximation provides the time-development of activity ${\bf v}(t)$, given input $\bf h$ and initial condition ${\bf v}(0)$, as a **linear combination of eigenvectors** ${\bf e}_\mu$, with time-varying coefficients $c_\mu(t)$.  

The **eigenvalues** $\lambda_\mu$ determine the degree of amplification: development in direction ${\bf e}_\mu$ is stable if $0<\lambda_\mu<1$, but unstable if  $1<\lambda_\mu$.

**Linear networks** can selectively amplify the dominant eigenvector ${\bf e}_1$, or linear combinations of degenerate eigenvectors ${\bf e}_{1,2}$.

**Non-linear networks** are biologically (slightly) more realistic, but must be analyzed in simulation.  They are also able to selectively amplify eigenvectors. In addition, they are able to disproportionately amplify the largest input ('winner-take-all'), and to multiplicatively scale inputs ('gain modulation').

# **1. Recurrent networks**

Consider a two-layer network with input units $[u_1, u_2, \ldots ]$ driving output units $[v_1, v_2, \ldots ]$, which are also recurrently connected among themselves.  The dynamic equations of output activity $v_m$ is

<br>

$$
\tau \frac{d v_m}{dt} = - v_m +   F \left( \sum_n \, W_{mn} \, u_n + \sum_{m'} \, M_{mm'} \, v_{m'} \right)
$$

<br>

where $W_{mn}$ is the synaptic weight of the feedforward connection from $u_n$ to $v_m$, and $M_{mm'}$ is the synaptic weight of the recurrent connection from $v_{m'}$ to $v_m$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec03_Many2many_rec.png" width="600">

As always, it is convenient to collect the dynamic equations of all output units into a single vector equation

$$
\tau \frac{d  \bf v}{dt} = - {\bf v} + {\bf F} \left({\bf W} \cdot {\bf u} + {\bf M} \cdot {\bf v} \right)
$$

where $\bf W$ and $\bf M$ are the feedforward and recurrent synaptic weight matrices, respectively.  This is the dynamic equation of the entire network.

(Slide-1)

# Feedforward inputs combined


As we are intested mainly in the dynamic of target units $\bf v$, we combine the feedforward inputs into a vector

<br>

$$
\tau \frac{d  \bf v}{dt} = - {\bf v} +  {\bf F} \left({\bf h} + {\bf M} \cdot {\bf v} \right), \qquad\qquad {\bf h} \equiv {\bf W} \cdot {\bf u}
$$

<br>

where $\bf h$ is the (combined) feedforward inputs and $\bf M$ is the recurrent synaptic weight matrix. For mathematical convenience, we consider only connection matrices that are **symmetric** and satisfy ${\bf M} = {\bf M}^T$, which is biologically plausible.

<br>

# Steady-states in nature

We speak of a steady-state when the rate of increments is exactly balanced by the rate of decrements.

<br>

This definition covers both equilibrium steady-states, where increments and decrements are zero

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec03_Steadystate0.png" width="300">

(polished metal sculptures lying on grass)

and non-equilibrium steady-states, where increments and decrements cancel out.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec03_Steadystate1.jpg" width="300">

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec03_Steadystate2.jpg" width="300">

(sand pile and weir)

<br>

In a neural network, the definition of steady-state is that excitatory synaptic inputs are exactly balanced by inhibitory synaptic inputs. In other words, increments and decrements cancel out.

(Slide-3)

# Steady-state condition

At a steady-state with rate vector $\bf v_{ss}$, each component $v_o$ of this vector satisfy

<br>

$$
0 \stackrel{!}{=} \tau \frac{dv_o}{dt} = {-v_o}+ F\left(  h_o + \sum_{o'} \, M_{oo'} \cdot v_{o'} \right)
$$

<br>

In other words, for each population $o$, the total effect of excitatory and inhibitory inputs
must exactly balance the effect of relaxation, so that the time-derivative $\frac{ d v_o}{dt}$ vanishes.

<br>

Determining the existence and nature of such vectors requires more advanced methods, involving the eigenvectors and eigenvalues of recurrent connectivity matrix $\bf M$.

(Slide-4)

# Linear approximation

To obtain a general idea of the dynamic possibilities, we consider a drastically simplified system, in which the transfer function has been replaced by the identity function:

<br>

$$
{\bf v}_{out} = {\bf F}\left({\bf  v}_{in}\right) \qquad \rightarrow \qquad {\bf v}_{out} = {\bf v}_{in}
$$

<br>

The dynamic equation then simplifies to:

<br>

$$
\tau \frac{d \bf v}{dt} = - {\bf v} + {\bf F} \left( {\bf h} + {\bf M} \cdot {\bf v} \right)
\qquad \rightarrow \qquad
\tau \frac{d \bf v}{dt} = - {\bf v} + {\bf h} + {\bf M} \cdot {\bf v}  
$$

<br>

For this simplification, we seek the  time-evolution of the output vector ${\bf v}(t)$, given a constant input vector ${\bf h}$ and an initial output vector ${\bf v}(0)$.

<br>

While the linear approximation is not plausible biologically (for example, it admits both positive and negative values 'firing rates'), it does allow considerable progress.

# Activity patterns in vector space

An $n$-dimensional pattern of activity $\{ x_1, x_2, \ldots , x_n\}$ can be represented as a point in vector space $\mathfrak{R}^n$. Different patterns correspond to different points in vector space.  Similar patterns map to nearby points, dissimilar patterns to distant points.


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec03_ActivityPatternVector.png" width="600">

```python

from inspect import signature
# @title Activity patterns in vector space {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.mplot3d import Axes3D


def Sine(t, lambd, phi):
    xi = t
    yi = 1 + np.sin(2 * np.pi * t / lambd + phi)
    return yi, xi

def Square(t, lambd, phi):
    xi = t
    yi = 1 + np.sign(np.sin(2 * np.pi * t / lambd + phi))
    return yi, xi

def Cycloid(t, theta):

    xi = 2*theta*t + np.sin( 2*theta*t );
    yi = (1 - np.cos(2*theta*t));
    xi = xi * np.max(t) / np.max(xi);
    return yi, xi

def Gauss(t, mu, sigma):
    xi = t
    yi = np.exp(-((t - mu) ** 2) / (2 * sigma ** 2))
    return yi, xi

def ReduceTo3D( Si, Qi, Ci, Gi):

    # combine patterns
    X = np.vstack((Si, Qi, Ci, Gi))

    # Center the data (zero mean)
    X_centered = X - np.mean(X, axis=0)

    X_transpose = X_centered.T

    # Compute SVD
    U, S, Vt = np.linalg.svd(X_centered, full_matrices=False)

    # Principal components (eigenvectors)
    PCs = Vt[0:3,:]

    Sk = Vt @ Si
    Qk = Vt @ Qi
    Ck = Vt @ Ci
    Gk = Vt @ Gi

    Sk  = Sk[0:3]
    Qk  = Qk[0:3]
    Ck  = Ck[0:3]
    Gk  = Gk[0:3]

    return Sk, Qk, Ck, Gk;


def plot_patterns_in_vector_space():
    """(Slide-6)"""

    # Define parameters
    ti = np.linspace(0, 8, 400)
    lambda_ = 2
    phi = 1
    theta = 1
    mu = 4
    sigma = 2


    # Create the plot
    fig  = plt.figure(figsize=(12, 4))

    axs1 = fig.add_subplot(121)  # Specify 2D projection

    # first subplot

    # Sine wave
    ySi, xi = Sine(ti, lambda_, phi)
    axs1.plot(xi, ySi, 'k', linewidth=2)
    axs1.plot(xi[::10], ySi[::10], 'ko', markersize=10, markerfacecolor='k')

    # Square wave
    yQi, xi = Square(ti, lambda_, phi)
    axs1.plot(xi, yQi, 'r', linewidth=2)
    axs1.plot(xi[::10], yQi[::10], 'ro', markersize=10, markerfacecolor='r')

    # Cycloid
    yCi, xi = Cycloid(ti, theta)
    axs1.plot(xi, yCi, 'b', linewidth=2)
    axs1.plot(xi[::10], yCi[::10], 'bo', markersize=10, markerfacecolor='b')

    # Gaussian curve
    yGi, xi = Gauss(ti, mu, sigma)
    axs1.plot(xi, yGi, 'm', linewidth=2)
    axs1.plot(xi[::10], yGi[::10], 'mo', markersize=10, markerfacecolor='m')

    axs1.set_xlabel('units i')
    axs1.set_box_aspect(1/1)
    axs1.set_ylabel('activity')
    axs1.set_title('Four activity patterns (40 dim)', fontsize=14);

    # second subplot
    axs2 = fig.add_subplot(122, projection='3d')  # Specify 3D projection

    (Sk, Qk, Ck, Gk) = ReduceTo3D( ySi, yQi, yCi, yGi)

    axs2.plot( (0,Sk[0]), (0,Sk[1]), (0,Sk[2]), 'k-', linewidth=2)
    axs2.plot( (0,Qk[0]), (0,Qk[1]), (0,Qk[2]), 'r-', linewidth=2)
    axs2.plot( (0,Ck[0]), (0,Ck[1]), (0,Ck[2]), 'b-', linewidth=2)
    axs2.plot( (0,Gk[0]), (0,Gk[1]), (0,Gk[2]), 'm-', linewidth=2)


    axs2.plot( Sk[0], Sk[1], Sk[2], 'ko', markersize=10, markerfacecolor='k', label='sin')
    axs2.plot( Qk[0], Qk[1], Qk[2], 'ro', markersize=10, markerfacecolor='r', label='square')
    axs2.plot( Ck[0], Ck[1], Ck[2], 'bo', markersize=10, markerfacecolor='b', label='cycloid')
    axs2.plot( Gk[0], Gk[1], Gk[2], 'mo', markersize=10, markerfacecolor='m', label='gauss')

    axs2.view_init(elev=40, azim=-45)  # 30° elevation, 45° azimuth

    axs2.set_xlabel('PC1')
    axs2.set_ylabel('PC2')
    axs2.set_zlabel('PC3')
    axs2.set_title('Four vectors (only 3 dims shown)', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_patterns_in_vector_space )

```

# Five-dimensional linear example

To illustrate these ideas, we consider the dynamic of a 5D recurrent network as an example.

$$
\tau \frac{d{\bf v}}{dt} = -{\bf v} + {\bf M} \cdot {\bf v} + {\bf h}
$$

We choose connetivity matrix $\bf M$, feedforward inputs $\bf h$, and initial activities ${\bf v}_0$ as follows:


$$
{\bf M} = \left(\begin{array}{ccccc} 0.1 & 0 & 0 & 0 & 0 \\ 0 & 0.335 & 0.247487373 & 0.215 & 0 \\ 0 & 0.247487373 & 0.55 & 0.247487373 & 0 \\ 0 & 0.215 & 0.247487373 & 0.335 & 0 \\ 0 & 0 & 0 & 0 & 0.1 \end{array}\right), \qquad {\bf h} = \left( \begin{array}{c} 0.2 \\ 0.6 \\ 0.2 \\ 0.6 \\ 0.2 \end{array} \right) , \qquad {\bf v_0} = \left( \begin{array}{c} 0.4 \\ 0.4 \\ 0.4 \\ 0.4 \\ 0.4 \end{array} \right)
$$

Note that we have chosen the matrix such that $v_2$, $v_3$, and $v_4$ communicate with (and affect) each other, whereas $v_1$ and $v_5$ affect only themselves (respectively).

(Slide-7)


```python
# @title Vectors of five-dimensional system {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.mplot3d import Axes3D

def DefineLinearSystemA():
    h  = (0.2, 0.6, 0.4, 0.6, 0.2)
    v0 = (0.4, 0.4, 0.4, 0.4, 0.4)

    M = np.array([[0.1, 0, 0, 0, 0],
                  [0, 0.335, 0.247487373, 0.215, 0],
                  [0, 0.247487373, 0.55, 0.247487373, 0],
                  [0, 0.215, 0.247487373, 0.335, 0],
                  [0, 0, 0, 0, 0.1]])

    tau = 0.5;

    return h, v0, M, tau

def EVanalysis( M ):
    # matrix is assumed to be symmetric
    # so that eigenvectors and eigenvalues are real

    (Eigvals, Eigvecs) = np.linalg.eig(M)

    # ignore spurious complex parts
    Eigvals = np.real( Eigvals )
    Eigvecs = np.real( Eigvecs )

    # choose sign of eigenvectors for visualization
    Esign = np.sign( np.mean(Eigvecs,axis=0) );
    for k in range( len(Esign) ):
         Eigvecs[:,k] = Esign[k] * Eigvecs[:,k]

    # sort eigenvectors in descending order of eigenvalues
    sidx = np.argsort( Eigvals )
    sidx = sidx[::-1]
    Eigvals = Eigvals[sidx]
    Eigvecs = Eigvecs[:,sidx]


    # extract first two eigenvectors
    E1 = Eigvecs[:,0]
    E2 = Eigvecs[:,1]
    E3 = Eigvecs[:,2]

    return Eigvecs, Eigvals, E1, E2, E3



def plot_5D_vectors( toggle=0):
    """(Slide-8)"""

    # Define parameters
    (h, v0, M, tau) = DefineLinearSystemA()

    # eigenvectors and eigenvalues
    (Eigvecs, Eigvals, E1, E2, E3) = EVanalysis( M )

    # scale with eigenvalue
    E1 = Eigvals[0] * E1
    E2 = Eigvals[1] * E2

    # Create the plot
    fig  = plt.figure(figsize=(12, 4))

    axs1 = fig.add_subplot(121)  # Specify 2D projection

    # first subplot with patterns
    x = np.array([1,2,3,4,5]);

    axs1.plot(x, h, 'k-', linewidth=2, label='h' )
    axs1.plot(x, h, 'ko', markersize=10, markerfacecolor='k')

    axs1.plot(x, v0, 'r-', linewidth=2, label='v0' )
    axs1.plot(x, v0, 'ro', markersize=10, markerfacecolor='r')

    if toggle:
        axs1.plot(x, E1, 'b:', linewidth=2, label='Eig 1' )
        axs1.plot(x, E1, 'bo', markersize=10, markerfacecolor='b')
        axs1.plot(x, E2, 'g:', linewidth=2, label='Eig 2' )
        axs1.plot(x, E2, 'go', markersize=10, markerfacecolor='g')

    axs1.legend(loc='lower center', fontsize=14)

    axs1.set_xlabel('units i')
    axs1.set_box_aspect(1/1)
    axs1.set_ylabel('activity')
    axs1.set_title('Patterns', fontsize=14);

    # second subplot
    axs2 = fig.add_subplot(122, projection='3d')  # Specify 3D projection

    axs2.plot((0,h[1]), (0,h[2]), (0,h[3]), 'k-', linewidth=2, label='h' )
    axs2.plot(h[1], h[2], h[3], 'ko', markersize=10, markerfacecolor='k')

    axs2.plot((0,v0[1]), (0,v0[2]), (0,v0[3]), 'r-', linewidth=2, label='v0' )
    axs2.plot(v0[1], v0[2], v0[3], 'ro', markersize=10, markerfacecolor='r')

    if toggle:
        axs2.plot((0,E1[1]), (0,E1[2]), (0,E1[3]), 'b:', linewidth=2, label='Eig 1' )
        axs2.plot( E1[1], E1[2], E1[3], 'bo', markersize=10, markerfacecolor='b')
        axs2.plot((0,E2[1]), (0,E2[2]), (0,E2[3]), 'g:', linewidth=2, label='Eig 2' )
        axs2.plot( E2[1], E2[2], E2[3], 'go', markersize=10, markerfacecolor='g')

    axs2.view_init(elev=60, azim=165)  # 30° elevation, 45° azimuth

    axs2.set_xlabel('v[2]')
    axs2.set_ylabel('v[3]')
    axs2.set_zlabel('v[4]')
    axs2.set_title('Vector space', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_5D_vectors, toggle=(0,1,1) )

```

# Dynamic of five-dimensional system

To simulate the dynamic of

<br>

$$\begin{eqnarray}
\tau \frac{d\bf v}{dt} = -{\bf v} + {\bf h} + {\bf M} \cdot {\bf v}
\end{eqnarray}$$

<br>

we iteratively integrate the activity vector in small time increments $\Delta t$. For every increment, we obtain the steady-state potential ${\bf v}_i^\infty$ and compute the final state ${\bf v}_{i+1}$ as an exponential relaxation from the initial state ${\bf v}_i$ (see WS Lectures 3 and 4)

<br>

$$\begin{eqnarray}
{\bf v}_i^\infty = {\bf h} + {\bf M} \cdot {\bf v}_i
\end{eqnarray}$$

<br>

$$\begin{eqnarray}
{\bf v}_{i+1} = {\bf v}_i^\infty + \left[{\bf v}_i - {\bf v}_i^\infty\right] \exp\left(-\Delta t / \tau \right)
\end{eqnarray}$$

For $i=0$, we use our initial condition ${\bf v}_0 = {\bf v}(0)$. The result is a series of actity patterns ${\bf v}_i$ at times $t_i = i \Delta t$. We terminate the iteratation at time $t_N=N\Delta t$

(Slide-9)

```python
# @title Dynamic of five-dimensional system {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter("ignore")
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.mplot3d import Axes3D

def IterationStep(v_ini, h, M, tau, dt):

    v_infty = h + M @ v_ini
    v_next  = v_infty + (v_ini - v_infty) * np.exp(-dt / tau)

    return v_next

def plot_5D_dynamic( log10dt=-2, Tend=2, toggle=0):
    """(Slide-10)"""

    # Define linear system
    (h, v0, M, tau) = DefineLinearSystemA()
    Ndim = len(h)

    # eigenvectors and eigenvalues
    (Eigvecs, Eigvals, E1, E2, E3) = EVanalysis( M )

    # scale with eigenvalue
    E1 = Eigvals[0] * E1
    E2 = Eigvals[1] * E2

    # time axis
    dt = 10**log10dt
    N  = round( Tend / dt )
    t_i = np.linspace(0, N, N+1, endpoint=True)
    t_i = dt * t_i

    # iteration
    v_ki = np.zeros((Ndim,N+1))

    v_ki[:,0] = v0;

    for i in range(1,N+1):
        v_ki[:,i] = IterationStep(v_ki[:,i-1], h, M, tau, dt )


    # decimate for plotting
    v_kj = v_ki[:,::10]
    t_j = t_i[::10]
    J   = len(t_j);

    # range of hues
    clr = np.zeros((J, 3))
    for j in range( 0, J, 1 ):
        clr[j,:] = [1, 0.8*(1-j/J), 0.8*(1-j/J)]

    # Create the plot
    fig  = plt.figure(figsize=(12, 4))

    axs1 = fig.add_subplot(121)  # Specify 2D projection

    # first subplot with patterns
    x = np.array([1,2,3,4,5]);

    axs1.plot(x, h, 'k:', linewidth=2, label='h' )
    axs1.plot(x, h, 'ko', markersize=10, markerfacecolor='k')

    axs1.plot(x, v0, 'r:', linewidth=2, label='v' )
    axs1.plot(x, v0, 'ro', markersize=10, markerfacecolor='r')

    for j in range( 1,len(t_j), 1 ):
        axs1.plot(x, v_kj[:,j], '-', linewidth=2, color=clr[j,:] )

    if toggle:
        axs1.plot(x, E1, 'b:', linewidth=2, label='Eig 1' )
        axs1.plot(x, E1, 'bo', markersize=10, markerfacecolor='b')
        axs1.plot(x, E2, 'g:', linewidth=2, label='Eig 2' )
        axs1.plot(x, E2, 'go', markersize=10, markerfacecolor='g')


    axs1.legend(loc='upper left', fontsize=14)

    axs1.set_xlabel('units i')
    axs1.set_box_aspect(1/1)
    axs1.set_ylabel('activity')
    axs1.set_title('Patterns', fontsize=14);

    # second subplot
    axs2 = fig.add_subplot(122, projection='3d')  # Specify 3D projection

    axs2.plot((0,h[1]), (0,h[2]), (0,h[3]), 'k:', linewidth=2, label='h' )
    axs2.plot(h[1], h[2], h[3], 'ko', markersize=10, markerfacecolor='k')

    axs2.plot((0,v0[1]), (0,v0[2]), (0,v0[3]), 'r:', linewidth=2, label='v' )
    axs2.plot(v0[1], v0[2], v0[3], 'ro', markersize=10, markerfacecolor='r')

    for j in range( 1,len(t_j), 1 ):
        #axs2.plot((0,v_kj[1,j]), (0,v_kj[2,j]), (0,v_kj[3,j]), '-', linewidth=2, color=clr[j,:] )
        axs2.plot(v_kj[1,j], v_kj[2,j], v_kj[3,j], 'o', color=clr[j,:], markersize=10, markerfacecolor=clr[j,:] )

    if toggle:
        axs2.plot((0,E1[1]), (0,E1[2]), (0,E1[3]), 'b:', linewidth=2, label='Eig 1' )
        axs2.plot( E1[1], E1[2], E1[3], 'bo', markersize=10, markerfacecolor='b')
        axs2.plot((0,E2[1]), (0,E2[2]), (0,E2[3]), 'g:', linewidth=2, label='Eig 2' )
        axs2.plot( E2[1], E2[2], E2[3], 'go', markersize=10, markerfacecolor='g')


    axs2.view_init(elev=60, azim=165)  # 60° elevation, 45° azimuth

    axs2.set_xlabel('v[2]')
    axs2.set_ylabel('v[3]')
    axs2.set_zlabel('v[4]')
    axs2.set_title('Vector space', fontsize=14);

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_5D_dynamic, log10dt=(-4,-1,1), Tend=(1,4,1), toggle=(0,1,1) )

```

# Summary so far

- Recurrent networks have richer dynamics than feedforward networks, but are more difficult to analyze.

- In *linearized* networks, the activation function ${\bf F}({\bf M \cdot \bf u + \bf h})$ is replaced by $\bf M\cdot \bf u + \bf h$.

- We will see later that linearized networks capture many qualitative aspects of the non-linear network.

- Our five-dimensional example behaves strangely. It develops towards a steady-state with no obvious relation to either input $\bf h$ or initial condition ${\bf v}(0)$.

- To understand this surprising behaviour, we must analyse the eigenvectors/eigenvalues of connection matrix $\bf M$.

# **2. Eigenvectors and eigenvalues**

We will use of the *'eigenvectors'* and *'eigenvalues'* of the connectivity matrix ('eigen' is German for 'one's own').  In a system with $N$ units, the connectivity matrix is of size $N\times N$ and has $N$ eigenvectors ${\bf e}_\mu$ with associated eigenvalues $\lambda_\mu$:


$$\begin{eqnarray}
{\bf M} \cdot {\bf e}_\mu = \lambda_\mu \, {\bf e}_\mu, \qquad\qquad \mu = 1, 2, \ldots, N
\end{eqnarray}$$

When being multiplied with a matrix, the eigenvectors of this matrix maintain their direction (i.e., are **not rotated**) but are scaled by the associated eigenvalues. All other vectors change their direction (are rotated) and are scaled by the eigenvalue (see 3Blue1Brown for a beautiful visualization).

<br>

Eigenvectors simplify matrix equations enormously. For eigenvectors, matrix multiplication is equivalent to scaling. For eigenvectors, matrices dissolve in smoke and 'go puff'.

<br>

The sign of eigenvectors is ambiguous.  If $\bf e$ is an eigenvector, then $-\bf e$ is one, too. Eigenvectors have unit length. Eigenvectors of **symmetric** matrices (biologically plausible!) are orthogonal:  

$$\begin{eqnarray}
{\bf e}_\mu \cdot {\bf e}_\nu = \delta_{\mu\nu} =
\left\{\begin{array}{ccc}
1 & \rm{if} & \mu=\nu
\\
0 & \rm{if} & \mu\neq\nu   
\end{array} \right.
\end{eqnarray}$$

Note that greek letters $\mu$ and $\nu$ index eigenvectors, roman letters $i$, $j$, $n$ etc index network units.

(Slide-12)

# Eigenvector coordinates


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/SSLec03_Eigenvectors.png" width="300">

With the orthonormal basis (coordinate system) provided by the normalized eigenvectors, we can write any $N$-dimensional vector $\bf x$ as a linear combination of eigenvectors

$$
{\bf x} = \sum_{\mu=1}^N \, a_\mu \, {\bf e}_\mu
$$

where $a_\mu$ are linear coefficients and the sum goes over eigenvector indes $\mu \in \{ 1, 2, \ldots , N\}$.

It follows that we can also write a time-varying, $N$-dimensional vector ${\bf v}(t)$ as a linear expansion of eigenvectors

$$
{\bf v}(t) = \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu
$$

where $c_\mu(t)$ are time-varying linear coefficients.

(Slide-15)

# Dynamic equation in eigenvector coordinates

Substituting this linear combination into our dynamic equation

$$
\tau \frac{d \bf v}{dt} = - {\bf v} + {\bf h} + {\bf M} \cdot {\bf v}, \qquad\qquad {\bf v}(t) = \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu
$$

leads to

$$
\tau\sum_{\mu=1}^N \,  \frac{d c_\mu(t) }{dt}  \, {\bf e}_\mu = - \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu + {\bf h} +  \sum_{\mu=1}^N \, c_\mu(t) \, \lambda_\mu \cdot  {\bf e}_\mu
$$

which in turn (next slide) leads to separate equations for the time-varying linear coefficients $c_\nu(t)$:

$$
\tau \, \frac{d c_\nu(t) }{dt}  = -  \left(1- \lambda_\nu \right) c_\nu(t)  +  {\bf h} \cdot {\bf e}_\nu  \qquad\qquad \forall \nu
$$

<br>

If the time-varying coefficients $c_\nu(t)$ satisfy these equations for all $\nu$, then the time-dependent vector ${\bf v}(t)$ satisfies the dynamic equation of the network.

<br>

In other words, we can solve our dynamic equation by transforming into eigenvector coordinates.

(Slide-16)

# Details of derivation

$$
\tau \frac{d \bf v}{dt} = - {\bf v} + {\bf h}+ {\bf M} \cdot {\bf v}, \qquad\qquad {\bf v}(t) = \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu
$$

$$
\tau \frac{d }{dt} \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu = - \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu + {\bf h}+ {\bf M} \cdot \left( \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu \right)  
\\
\\
\tau\sum_{\mu=1}^N \,  \frac{d c_\mu(t) }{dt}  \, {\bf e}_\mu = - \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu + {\bf h}+  \sum_{\mu=1}^N \, c_\mu(t) \, \underbrace{{\bf M} \cdot  {\bf e}_\mu}
$$

<br>

Using the eigenvector property ${\bf M} \cdot  {\bf e}_\mu = \lambda_\mu \,  {\bf e}_\mu$, we get

<br>

$$
\tau\sum_{\mu=1}^N \,  \frac{d c_\mu(t) }{dt}  \, {\bf e}_\mu = - \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu + {\bf h}+  \sum_{\mu=1}^N \, c_\mu(t) \, \lambda_\mu \,  {\bf e}_\mu
$$

To get rid of the sum, we can take the dot-product with any eigenvector $\bf e_\nu$

$$
 \tau \, \sum_{\mu=1}^N \,  \frac{d c_\mu(t) }{dt}  \, {\bf e}_\mu = \left. - \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu + {\bf h}+  \sum_{\mu=1}^N \, c_\mu(t) \, \lambda_\mu \,  {\bf e}_\mu \qquad \right| \,\, \cdot {\bf e}_\nu
$$

so that, because of the orthogonality of eigenvectors ${\bf e}_\mu \cdot {\bf e}_\nu = 0$ for $\nu \neq \mu$, all terms of the sum but one disappear, leaving us with

$$\begin{eqnarray}
\tau \, \frac{d c_\nu(t) }{dt}  \, \underbrace{{\bf e}_\nu \cdot {\bf e}_\nu}_{1}  &=& -  c_\nu(t) \, \underbrace{{\bf e}_\nu \cdot {\bf e}_\nu}_{1} + {\bf h} \cdot {\bf e}_\nu +  c_\nu(t) \, \lambda_\nu \,  \underbrace{{\bf e}_\nu \cdot {\bf e}_\nu}_{1}
\\
\\
\tau \, \frac{d c_\nu(t) }{dt}  &=& -  \left( 1 - \lambda_\nu\right) c_\nu(t)  +  {\bf h} \cdot {\bf e}_\nu \qquad\qquad \forall \nu
\end{eqnarray}$$

(Slide-17)

# Time-dependence of linear coefficients

The one remaining issue is the time-dependence of the linear coefficients.
For constant input $\bf h$, the dynamic equation is

$$
\tau \, \frac{d c_\mu(t) }{dt}  = -  \left(1- \lambda_\mu \right) c_\mu(t)  +  {\bf h} \cdot {\bf e}_\mu
$$

If $\lambda_\mu\leq 1$, this equation has the steady-state solution

$$
c_\mu^\infty = \frac{{\bf h} \cdot {\bf e}_\mu}{1-\lambda_\mu}
$$

where the time derivative of the linear coefficent vanishes. The initial value of the linear coefficient

$$
c_\mu(0) = {\bf v}_0 \cdot {\bf e}_\mu \qquad \Leftrightarrow \qquad {\bf v}_0 = \sum_{\mu=1}^N \, c_\mu(0) \, {\bf e}_\mu
$$

follows from the projection of the initial condition onto eigenvector ${\bf e}_\mu$.

The time-varying solution of the linear-coefficient is a superposition of two exponential relaxations

$$
c_\mu(t) = \underbrace{\frac{{\bf h} \cdot {\bf e}_\mu}{1-\lambda_\mu}}_{c_\mu^\infty} \, \left[1 - \exp\left(-\frac{1-\lambda_\mu}{\tau} t \right) \right] + \underbrace{{\bf v}_0 \cdot {\bf e}_\mu}_{c_\mu(0)} \, \exp\left(-\frac{1-\lambda_\mu}{\tau} t \right)
$$

The first term is the **approach** to the steady-state, and the second term is the **decay** of the initial condition.

If $\lambda_\mu\leq 1$, the exponential terms are **negative** and $c_\mu(t)$ asymptotically approaches the steady-state value.

<br>

In contrast, when $\lambda_\mu>1$, the exponential terms are **positive** and $c_\mu(t)$ grows beyond all bounds.

(Slide-18)





# Summary of case $\lambda_\mu < 1$

For eigenvectors with eigenvalues $0 < \lambda_\mu < 1$, coefficients relax exponentially from an initial value ${\bf v}_0 \cdot {\bf e}_\mu$ to a steady-state value $\frac{{\bf h} \cdot {\bf} e_\mu}{1-\lambda_\mu}$:



$$
\lim_{t\to\infty} c_\mu(t) =  \frac{{\bf h} \cdot {\bf e}_\mu}{1- \lambda_\mu}, \qquad\qquad \lim_{t\to 0} c_\mu(t) =  {\bf v}_0 \cdot {\bf e}_\mu
$$

The factor $\frac{1}{1-\lambda_\mu}$ determines how much the input projection onto $\bf e_\nu$ is *amplified* by the recurrent dynamic.  The closer $\lambda_\mu$ is to unity, the larger the amplification.

Note that

$$
\lim_{t\to 0} \left[1-\exp\left(- t \right)\right] = 0, \qquad\qquad \lim_{t\to 0} \exp\left(- t \right) = 1
$$

$$
\lim_{t\to\infty} \left[1-\exp\left(- t \right)\right] = 1, \qquad\qquad \lim_{t\to\infty} \exp\left(- t \right) = 0
$$

(Slide-19)

```python
# @title Dynamic of five-dimensional system with steady-state {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.mplot3d import Axes3D

def IterationStep(v_ini, h, M, tau, dt):

    v_infty = h + M @ v_ini
    v_next = v_infty + (v_ini - v_infty) * np.exp(-dt / tau)

    return v_next

def DynamicLinearCoefficients( ti, tau, h, v0, Eigvecs, Eigvals ):

    Ndim = len(h);
    N    = len(ti);

    # steady-state linear coefficients
    Ckinfty = np.zeros(Ndim)
    for k in range(0,Ndim):
        Ckinfty[k] = np.dot( h, Eigvecs[:,k] ) / (1-Eigvals[k])

    # initial coefficients
    Ck0 = np.zeros(Ndim)
    for k in range(0,Ndim):
        Ck0[k] = np.dot( v0, Eigvecs[:,k] )

    # analytical solution for coefficients
    Cki = np.zeros((Ndim,N))
    for k in range(0,Ndim):
        Eapp_i = 1 - np.exp(-(1-Eigvals[k])*ti/tau)
        Edec_i = np.exp(-(1-Eigvals[k])*ti/tau)
        Cki[k,:] = Ckinfty[k] * Eapp_i + Ck0[k] * Edec_i

    return Ck0, Ckinfty, Cki

def plot_5D_dynamic_with_SS( log10dt=-2, Tend=10):
    """(Slide-21)"""

    # Define linear system
    (H, V0, M, tau) = DefineLinearSystemA()

    Ndim = len(H);

    # eigenvectors and eigenvalues
    (Eigvecs, Eigvals, E1, E2, E3) = EVanalysis( M )

    # scaled eigenvectors
    E1 = Eigvals[0] * Eigvecs[:,0]
    E2 = Eigvals[1] * Eigvecs[:,1]
    E3 = Eigvals[2] * Eigvecs[:,2]

    # time axis
    dt = 10**log10dt
    N  = round( Tend / dt )
    t_i = np.linspace(0, N, N+1, endpoint=True)
    t_i = dt * t_i

    # linear coefficients
    (Ck0, Ckinfty, Cki) = DynamicLinearCoefficients( t_i, tau, H, V0, Eigvecs, Eigvals )

    # steady-state and approximate steady-state
    Vinfty = Eigvecs @ Ckinfty
    Vinftyapprox = Eigvecs[:,0] * Ckinfty[0]

    # sanity checks
    Vcheck = H + M @ Vinfty
    Vagain = IterationStep(Vinfty, H, M, tau, dt)

    # dynamic solution
    V_ki = np.zeros((Ndim,N+1))
    Valt_ki = np.zeros((Ndim,N+1))
    Valt_ki[:,0] = V0
    for i in range(1,N+1):
         V_ki[:,i] = Eigvecs @ Cki[:,i]
         Valt_ki[:,i] = IterationStep(Valt_ki[:,i-1], H, M, tau, dt)

    # decimate for plotting
    V_kj = V_ki[:,::(10**(-log10dt))]
    Valt_kj = Valt_ki[:,::(10**(-log10dt))]
    t_j = t_i[::(10**(-log10dt))]
    J   = len(t_j);

    # range of hues
    clr = np.zeros((J, 3))
    for j in range( 0, J, 1 ):
        clr[j,:] = [1, 0.8*(1-j/J), 0.8*(1-j/J)]

    # Create the plot
    fig  = plt.figure(figsize=(12, 4))

    axs1 = fig.add_subplot(121)  # Specify 2D projection

    # first subplot with patterns
    x = np.linspace(1,Ndim,Ndim,endpoint='true');

    axs1.plot(x, H, 'k:', linewidth=2, label='h' )
    axs1.plot(x, H, 'ko', markersize=10, markerfacecolor='k')

    axs1.plot(x, V0, 'r:', linewidth=2, label='v' )
    axs1.plot(x, V0, 'ro', markersize=10, markerfacecolor='r')

    for j in range( 1,len(t_j), 1 ):
        axs1.plot(x, V_kj[:,j], '-', linewidth=2, color=clr[j,:] )
        axs1.plot(x, Valt_kj[:,j], ':', linewidth=2, color=clr[j,:] )

    axs1.plot( x, Vinfty, 'g-', linewidth=2, label='Vinfty' )
    axs1.plot( x, Vinfty, 'go', markersize=10, markerfacecolor='g')
    axs1.plot( x, Vinftyapprox, 'g:', linewidth=2, label='Vapprox' )

    axs1.legend(loc='upper left', fontsize=14)

    axs1.set_xlabel('units i')
    axs1.set_box_aspect(1/1)
    axs1.set_ylabel('activity')
    axs1.set_title('Patterns', fontsize=14);

    # second subplot with vector space
    axs2 = fig.add_subplot(122, projection='3d')  # Specify 3D projection

    axs2.plot((0,H[1]), (0,H[2]), (0,H[3]), 'k:', linewidth=2, label='h' )
    axs2.plot(H[1], H[2], H[3], 'ko', markersize=10, markerfacecolor='k')

    axs2.plot((0,V0[1]), (0,V0[2]), (0,V0[3]), 'r:', linewidth=2, label='v' )
    axs2.plot(V0[1], V0[2], V0[3], 'ro', markersize=10, markerfacecolor='r')

    axs2.plot((0,Vinfty[1]), (0,Vinfty[2]), (0,Vinfty[3]), 'g-', linewidth=2, label='Vinfty' )

    for j in range( 1,len(t_j), 1 ):
        #axs2.plot((0,v_k[1,k]), (0,v_k[2,k]), (0,v_k[3,k]), '-', linewidth=2, color=clr[k,:] )
        axs2.plot(V_kj[1,j], V_kj[2,j], V_kj[3,j], 'o', color=clr[j,:], markersize=10, markerfacecolor=clr[j,:] )


    axs2.plot( Vinfty[1], Vinfty[2], Vinfty[3], 'go', markersize=10, markerfacecolor='g')

    axs2.view_init(elev=60, azim=165)  # 60° elevation, 45° azimuth

    axs2.set_xlabel('v[2]')
    axs2.set_ylabel('v[3]')
    axs2.set_zlabel('v[4]')
    axs2.set_title('Vector space', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_5D_dynamic_with_SS, log10dt=(-3,-1,1), Tend=(4,20,1) )

```

We have considered linear networks with **symmetric** connectivity ${\bf M}$

$$
\tau \frac{d {\bf v}}{dt} = - {\bf v} + {\bf h}+ {\bf M} \cdot {\bf v},

and its orthonormal eigenvectors and eigenvalues

$$

# Selective amplification of one pattern

In the last section, we have already encountered one interesting capability, namely, **selective amplification**.

In our example, the eigenvalue of the dominant eigenvector was close to unity, whereas the eigenvalues of all other eigenvectors were much smaller. In such a situation, the solution will be dominated by the dominant eigenvector:

$$
{\bf v}(t) = \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu \approx c_1(t) \, {\bf e}_1
$$

and the steady-state activity will be

$$
{\bf v}_{ss} \approx \frac{{\bf h} \cdot {\bf e}_1}{1 - \lambda} \, {\bf e}_1
$$
The steady state response is dominated by the projection of the input onto the dominant eigenvector.  This projection is amplified by a (potentially large) factor $\frac{1}{1-\lambda_1}$.

The dominant eigenvector corresponds to one particular pattern of activity.  Accordingly, the dynamic selectively amplifies this particular pattern. More precisely, the dynamic amplifies any input component that is aligned with (projects onto) this pattern.

# Selective amplification of a family of patterns

Suppose that two eigenvalues of the connection matrix are close to unity and all others are near zero.  When two eigenvalues $\lambda_1=\lambda_2=\lambda$ are identical, the associated eigenvectors are called **degenerate eigenvectors**.

Then time-dependent activity will be

$$
{\bf v}(t) = \sum_{\mu=1}^N \, c_\mu(t) \, {\bf e}_\mu \approx c_1(t) \, {\bf e}_1 + c_2(t) \, {\bf e}_2
$$

and the steady-state activity will be

$$
{\bf v}_{ss} \approx \frac{({\bf h} \cdot {\bf e}_1) \, {\bf e}_1 + ({\bf h} \cdot {\bf e}_2) \, {\bf e}_2}{1 - \lambda} = \alpha\, {\bf e}_1 + \beta\, {\bf e}_2
$$

Thus, the network amplifies the projection of the input onto the plane defined by $\bf e_1$ and $\bf e_2$.  

In other words, the dynamic is able to amplify any activity patterns within the plane defined by $\bf e_1$ and $\bf e_2$. Which particular member of this family of patterns is amplifed depends on the input, specifically, the projection of the input onto the plane of $\bf e_1$ and $\bf e_2$.

# Connectivity with degenerate eigenvectors

Consider a recurrent network with $N=20$ units, responsive to inputs ranging from $-\pi$ to $\pi$. The preferred input $\theta_i$ of unit $i$ is

$$
\theta_i = -\pi + 2 \pi (i-1) / N
$$

and the connection strength from unit $i$ to unit $j$ is

$$
{\bf M}: \qquad m_{ij} =  \frac{\lambda}{\pi^2} \, \cos\left[\theta_i - \theta_j \right]
$$

In other words, units with similar preference are positively connected, but units with dissimilar preferences negatively.

This recurrent matrix has two dominant eigenvectors

$$
{\bf e}_1 = \pm \cos\left[2 \pi (i-1) / N\right], \qquad {\bf e}_2 = \pm \sin\left[2 \pi (i-1) / N\right],
$$

with degenerate eigenvalues $\lambda_1=\lambda_2=\lambda$. All linear combinations

$$
{\bf e}_\mathit{alt} = \pm \sqrt{\alpha} \, {\bf e}_1 \pm \sqrt{1-\alpha} \, {\bf e}_2 , \qquad ||{\bf e}_\mathit{alt}||=1 \qquad \forall \alpha
$$

are eigenvectors as well.

# Case $\lambda_\mu=1$: input integration

Suppose a recurrent weight matrix has one eigenvalue $\lambda_1=1$ and all other eigenvalues $\lambda_{\mu\neq 1} < 1$.  In this case, the general dynamic equation for coefficients

<br>

$$
\tau \, \frac{d c_\mu(t) }{dt}  = -  \left(1- \lambda_\mu \right) c_\mu(t)  +  {\bf h} \cdot {\bf e}_\mu
$$

<br>

simplifies for $\mu=1$ to

<br>

$$
\tau \, \frac{d c_1(t)}{dt} = {\bf h} \cdot {\bf e}_1
$$

<br>

or, for a time-dependent input ${\bf h}(t)$, to

$$
c_1(t) = c_1(0) + \frac{1}{\tau} \, \int_0^t \, {\bf e}_1 \cdot {\bf h}(t') \, dt'
$$

which is the time-integral of the input projected onto eigenvector ${\bf e}_1$.

# **4. Interesting capabilities of non-linear recurrent networks**


Linear models do not adequately describe networks of biological neurons, firstly because they disregard the activation function of neuronal populations and secondly because they allow negative firing rates.

We conclude this lecture by considering a few nonlinear networks in simulation, gaining biological realism but sacrificing analytical tractability. The next lectures will discuss other non-linear examples in more detail, together with some tools for formal analysis.

Similar to linear networks, the behaviour of nonlinear networks is also governed by the connectivity matrix $\bf M$ (and its eigenvector and eigenvalues).

The nonlinearity we consider is the square root of a simple rectification

$$
\tau \frac{d \bf v}{dt} = - {\bf v} + {\bf F} \left[ {\bf h}+ {\bf M} \cdot {\bf v} \right], \qquad\qquad {\bf F}\left[{\bf x}\right] \equiv \sqrt{\left[ {\bf x} \right]_+}
$$

# Summary winner-take-all

- The network selectively amplifies the linear combination that best matches the input.

- Due to the non-linearity, this does not reflect all inputs components proportionately.

- Larger input components are amplified disproportionately.

- This results in a **winner-take-all** dynamic.

# Multiplicative gain modulation

An additive offset to the input translates into a multiplicative scaling of the output.  This is because the shape of the response profile is fixed by the connectivity, so that additional input cannot broaden the base of the output.

In the code below, A is the amplitude of the 'bump' in the input pattern, whereas B is the additive offset between input patterns.

# Summary gain modulation


<br>

- Gain modulation is the multiplicative scaling of a population response by an additional input.

- It is widely observed in sensory populations (e.g., transformation of EYE to HEAD coordinates).

- In a recurrent network, an additive input results in a multiplicative scaling.

- The multiplication is performed by the network, not by individual neurons.

(Slide-46)


# **Overall summary**

<br>

- **Why is the qualitative behaviour of recurrent networks -- both linear and non-linear -- shaped by eigenvectors of the connectivity matrix?**

<br>

- **In general, patterns of activity change on every pass through a recurrent network (every reverberation). In general, patterns are rotated in vector space.**

- **The exception are patterns of activity that remain unchanged (not rotated). By definition, the eigenvectors of the connectivity matrix are such patterns.**

$$
\lambda_i \, \bf e_i = \bf M \cdot \bf e_i
$$

<br>

- **Although not rotated, eigenvectors are scaled on every pass through the network.  The eigenvectors with the largest eigenvalues scale most rapidly and thus are 'dominant'.**

(Slide-47)





- **We have approximated neural networks with linear systems of the type**

$$
\tau \frac{d \bf v}{dt} = - {\bf v} +  {\bf h}+ {\bf M} \cdot {\bf v}
$$

- **Here we can solve the dynamic analytically in terms of a linear combination of eigenvectors.**

- **The recurrent dynamic is stable when all eigenvalues $\lambda$ are smaller than unity.**

- **Selective amplification results when the dominant eigenvalues are just below unity.**

- **Specifically, the linear combination of degenerate eigenvectors that best matches the input (projection of input onto degenerate plane) is amplifed by a factor of $\frac{1}{1-\lambda}$.**

- **The dynamic of non-linear recurrent networks is qualitatively similar.**

$$
\tau \frac{d \bf v}{dt} = - {\bf v} + {\bf F} \left[ {\bf h}+ {\bf M} \cdot {\bf v} \right], \qquad\qquad {\bf F}\left[{\bf x}\right] \equiv \left[ {\bf x} - {\bf \gamma} \right]_+
$$

- **Such networks can perform *'selective amplification', 'winner-take-all',*  and *'multiplicative-gain-modulation'*.**

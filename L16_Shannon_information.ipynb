# **Abstract:**

Today we learn the basic concepts of Shannon information. As these are a bit peculiar, we'll take it slowly. In Shannon's sense, information is about statistics rather than 'meaning'. Shannon considers not only events that happened, but all the events that could have happened (but didn't).  Also, Shannon cares not at all about the nature of events, only about their frequency.

In short, Shannon takes an extremely narrow view, which at the same time is extremely general. In the remaining three lectures, we'll try to understand the value of Shannon's point of view.

# **2. Shannon information and and Shannon entropy of discrete random events**

We wish to quantify the informativeness of **discrete random events**.  

<br>

* The informativeness of individual events should reflect their rarity or surprise value.

$\qquad\qquad\Rightarrow$ measure should decrease monotonically with probability!

<br>

* For multiple *independent* events, the joint informativeness should equal the sum of individual informativeness(es).

$\qquad\qquad\Rightarrow$ measure should be logarithmic!

<br>

* Intuitive definition in terms of halving possibilities.

$\qquad\qquad\Rightarrow$ Choose $\log_2$.

(Slide-10)


# 'Shannon information' of individual random events

The 'information' (Shannon information content or SIC) of an individual random event $x$ *decreases* with the binary logarithm of its *probability*.

<br>

It is defined as

<br>

$\begin{eqnarray}
h(x) = \log_2 \frac{1}{P(x)}=-\log_2 P(x)
\end{eqnarray}$

<br>

where $P(x)$ is the probability of observing outcome $x$.

Its unit is called 'bits'.

(Slide-11)

# Example: ordinary coin

<br>

When you throw an ordinary coin, the probability of each outcome is $\frac{1}{2}$.  Thus, the SIC of each outcome is

<br>

$\begin{eqnarray}
h = -\log_2 \frac{1}{2} = \log_2 2 = 1
\end{eqnarray}$

<br>

With each throw, you gain $1\,bit$ of 'Shannon information'.

(Slide-12)


# Example: $N$-sided dice

<br>

When you throw an $N$-sided die, the probability of each outcome is $\frac{1}{N}$, so that the SIC of each outcome

<br>

$\begin{eqnarray}
h = -\log_2 \frac{1}{N} = \log_2 N
\end{eqnarray}$

<br>

With each throw, you gain $\log_2 N\,bits$ of information.

# 'Shannon entropy' of random variable or process

<br>

The 'entropy' of a random variable $X$ is defined as the average Shannon information content of all possible outcomes $x$.

<br>

$\begin{eqnarray}
H(X) = \sum_x \, P(x) \, h(x) \qquad\qquad \textit{(probabilistic average)}
\end{eqnarray}$

<br>

where each outcome $x$ contributes entropy $P(x) \, h(x)$.  Since


$\begin{eqnarray}
h(x) = - \log_2 P(x) = \log_2 \frac{1}{P(x)}
\end{eqnarray}$

<br>

we have


$\begin{eqnarray}
H(X)  = - \sum_x \, P(x) \, \log_2 P(x) = \sum_x \, P(x) \, \log_2 \frac{1}{P(x)}
\end{eqnarray}$

<br>

The unit of Shannon entropy is `bits'.

(Slide-15)

# Example: ordinary coin

<br>

The entropy of the random variable realized by throwing an ordinary coin is

$\begin{eqnarray}
H = \sum_x \, P(x) \, h(x) = \sum_x \, P(x) \, \log_2 \frac{1}{P(x)}
\end{eqnarray}$

<br>

The entropy contribution of each outcome is

$\begin{eqnarray}
P(x) \, h(x)  =   \frac{1}{2} \, \log_2 2
\end{eqnarray}$

<br>

Both contributions together add up to

$\begin{eqnarray}
H =   \sum_x \, \frac{1}{2} \, \log_2 2 = 2 \, \frac{1}{2} \, \log_2 2 = 1
\end{eqnarray}$

<br>

*The entropy is the average information of individual outcomes.  Since every outcome is equally informative, the average information equals the individual information.  Thus, the entropy is identical to the information of each individual outcome.*

(Slide-16)

# Example: $N$-sided dice

<br>

The entropy of the random variable realized by throwing an $N$-sided die is

$\begin{eqnarray}
H = \sum_x \, P(x) \, h(x) = \sum_x \, P(x) \, \log_2 \frac{1}{P(x)}
\end{eqnarray}$

<br>

The entropy contribution of each outcome is

$\begin{eqnarray}
P(x) \, h(x)  =   \frac{1}{N} \, \log_2 N
\end{eqnarray}$

<br>

Taken together, the contributions from $N$ outcomes add up to

$\begin{eqnarray}
H =   \sum_x \, \frac{1}{N} \, \log_2 N = N \, \frac{1}{N} \, \log_2 N = \log_2 N
\end{eqnarray}$

(Slide-17)


# Certainty, uncertainty, and impossibility

Neither certain nor impossible outcomes contribute to the entropy:


$\begin{eqnarray}
P(x)=1\qquad\Rightarrow\qquad 1 \cdot \log_2 \frac{1}{1} = 0
\end{eqnarray}$

$\begin{eqnarray}
P(x)=0\qquad\Rightarrow\qquad 0 \cdot \log_2 \frac{1}{0} = 0
\end{eqnarray}$

In contrast, any uncertain outcome contributes positively to the entropy:


$\begin{eqnarray}
0 < P(x) < 1 \qquad\Rightarrow\qquad P \cdot \log_2 \frac{1}{P} > 0
\end{eqnarray}$

```python
from inspect import signature
# @title Entropy contribution and probability {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

# get log probability and entropy, taking care of singularities
def GetContribution( P_i ):

    L_i = np.zeros_like(P_i);
    H_i = np.zeros_like(P_i);

    ix = np.where((0<P_i) & (P_i<1))

    ix0 = np.where(0==P_i);

    ix1 = np.where(P_i==1);

    L_i[ix] = -np.log2(P_i[ix]);
    H_i[ix] = P_i[ix] * L_i[ix];

    if not np.any(ix0):
        H_i[ix0] = 0;
        L_i[ix0] = 7;

    if not np.any(ix1):
        H_i[ix1] = 0;
        L_i[ix1] = 0;

    return L_i, H_i;

def plot_entropy_contribution(p0=0.5):
    """(Slide-19)"""

    Nres = 100

    # range of probabilities
    p_i = np.linspace(0, 1, Nres, endpoint=True );
    (l_i, h_i) = GetContribution( p_i )

    # selected point
    P0 = np.full((1,2),p0)
    (L0, H0) = GetContribution( P0 )

    # Plot results
    fig, axs = plt.subplots( 1,2, figsize=(8,4))

    # probability and negative logarithm

    axs[0].plot( p_i, p_i, color=(0.75,0.25,0.25), linewidth=2, label='probability')
    axs[0].plot( p_i, l_i, color=(0.25,0.25,0.75), linewidth=2, label='negative log')
    axs[0].plot( P0, P0, 'o', color=(0.75,0.25,0.25), markersize=5)
    axs[0].plot( P0, L0, 'o', color=(0.25,0.25,0.75), markersize=5)
    axs[0].legend(loc='upper right')
    axs[0].set_xlabel('probability')
    axs[0].set_box_aspect(2/1)
    axs[0].set_ylabel('p,  -log p')
    axs[0].set_title('Probability and Log', fontsize=14);

    # contribution

    axs[1].plot( p_i, h_i, color=(0.25,0.75,0.25), linewidth=2, label='contribution')
    axs[1].plot( P0, H0, 'o', color=(0.25,0.75,0.25), markersize=5)
    axs[1].legend(loc='lower center')
    axs[1].set_xlabel('probability')
    axs[1].set_box_aspect(1/1)
    axs[1].set_ylabel('-p log p')
    axs[1].set_title('Contribution', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_entropy_contribution, p0 = (0.0, 1.0, 0.01 ) )
```

# Equiprobable ensembles

<br>

Ensemble entropy is maximal when all $N$ outcomes are equally probable.   Thus, entropy has the upper bound

$\begin{eqnarray}
H(X) \leq \log_2 N
\end{eqnarray}$

$\begin{eqnarray}
H(X) = \log_2 N \qquad {\rm iff} \qquad P(x) = \frac{1}{N} \,\, {\rm for} \,\, {\rm all} \,\, x
\end{eqnarray}$

<br>

since, in the case of equiprobability, we have

$\begin{eqnarray}
H(X) = \sum_x \, P(x) \, \log_2 \frac{1}{P(x)} =  -\sum_x \, \frac{1}{N} \, \log_2 \frac{1}{N}  = \log_2 N
\end{eqnarray}$

(Slide-21)

# Example: $2$ outcomes

<br>

Consider a random variable with two outcomes, $(+)$ and $(-)$, with probabilities:

$\begin{eqnarray}
P_+, \,\, P_- \qquad\qquad\qquad\qquad P_- = 1 - P_+
\end{eqnarray}$

<br>

The entropy is

$\begin{eqnarray}
H  &=& \sum_x \, P(x) \, \log_2 \frac{1}{P(x)}
\\
&=& P_+\, \log_2 \, \frac{1}{P_+} + (1-P_+) \, \log_2 \, \frac{1}{1-P_+}
\end{eqnarray}$

<br>

The entropy is maximal when $P_+=\frac{1}{2}$,  in other words, when the outcomes are equiprobable!

**Proof:**

$\begin{eqnarray}
\qquad\qquad 0
&\stackrel{!}{=}& \partial_{P_+} \left[ -P_+\, \log_2 \, P_+ - (1-P_+) \, \log_2(1-P_+)\right] =
\\
&=& -\log_2 P_+ +\log_2 (1-P_+) -\frac{P_+}{P_+} + \frac{1-P_+}{1-P_+} =
\\
&=& \log_2\frac{1+P_+}{P_+}
\end{eqnarray}$


$\begin{eqnarray}
\qquad\qquad \Leftrightarrow
\qquad
1
&=&\frac{1-P_+}{P_+}
\qquad\Leftrightarrow\qquad
P_+ = \frac{1}{2}
\end{eqnarray}$

(Slide-22)


```python
from inspect import signature
# @title Entropy of binary variable {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

# get log probability and entropy, taking care of singularities
def GetContribution( P_i ):

    L_i = np.zeros_like(P_i);
    H_i = np.zeros_like(P_i);

    ix = np.where((0<P_i) & (P_i<1))

    ix0 = np.where(0==P_i);

    ix1 = np.where(P_i==1);

    L_i[ix] = -np.log2(P_i[ix]);
    H_i[ix] = P_i[ix] * L_i[ix];

    if not np.any(ix0):
        H_i[ix0] = 0;
        L_i[ix0] = 7;

    if not np.any(ix1):
        H_i[ix1] = 0;
        L_i[ix1] = 0;

    return L_i, H_i;

def plot_entropy_binary_variable(p0=0.5):
    """(Slide-23)"""

    Nres = 100

    # range of probabilities
    p_i = np.linspace(0, 1, Nres, endpoint=True );
    (l_i, h_i) = GetContribution( p_i )
    pp_i = 1 - p_i;
    (lp_i, hp_i) = GetContribution( pp_i )

    # selected point
    P0 = np.full((1,2),p0)
    Pp0 = 1 - P0;
    (L0, H0) = GetContribution( P0 )
    (Lp0, Hp0) = GetContribution( Pp0 )

    # Plot results
    fig, axs = plt.subplots( 1,3, figsize=(12,4))

    # first outcome

    axs[0].plot( p_i, h_i, color=(0.75,0.25,0.25), linewidth=2, label='probability')
    axs[0].plot( P0, H0, 'o', color=(0.75,0.25,0.25), markersize=5)
    axs[0].legend(loc='lower center')
    axs[0].set_xlabel('probability')
    axs[0].set_box_aspect(1/1)
    axs[0].set_ylabel('h')
    axs[0].set_title('First outcome', fontsize=14);

    # second outcome

    axs[1].plot( p_i, hp_i, color=(0.25,0.25,0.75), linewidth=2, label='probability')
    axs[1].plot( P0, Hp0, 'o', color=(0.25,0.25,0.75), markersize=5)
    axs[1].legend(loc='lower center')
    axs[1].set_xlabel('probability')
    axs[1].set_box_aspect(1/1)
    axs[1].set_ylabel('h')
    axs[1].set_title('Second outcome', fontsize=14);

    # both outcomes

    axs[2].plot( p_i, h_i+hp_i, color=(0.25,0.75,0.25), linewidth=2, label='probability')
    axs[2].plot( P0, H0+Hp0, 'o', color=(0.25,0.75,0.25), markersize=5)
    axs[2].legend(loc='lower center')
    axs[2].set_xlabel('probability')
    axs[2].set_box_aspect(1/1)
    axs[2].set_ylabel('h')
    axs[2].set_title('Both outcomes', fontsize=14);


    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_entropy_binary_variable, p0 = (0.0, 1.0, 0.01 ) )
```

# $N$ outcomes (optional}

Of $N$ possible outcomes

$\begin{eqnarray}
P_1, P_2, \ldots , P_N
\end{eqnarray}$

consider two particular outcomes $1$ and $2$.  Compare the case $P_1 = P_2 = P$

$\begin{eqnarray}
H_1 + H_2 = P  \, \log_2 \frac{1}{P } + P  \, \log_2 \frac{1}{P }
\end{eqnarray}$

to the case $P_1 = P+\epsilon$ and $P_2 = P-\epsilon$

$\begin{eqnarray}
H'_1 + H'_2 = (P+\epsilon)  \, \log_2 \frac{1}{P+\epsilon} +  (P-\epsilon)  \, \log_2 \frac{1}{P-\epsilon}
\end{eqnarray}$

by forming the difference of entropies

$\begin{eqnarray}
\Delta H = H'_1 - H_1+ H'_2  - H_2
\end{eqnarray}$

$\begin{eqnarray}
\Delta H &=& P \, \log_2\left(\frac{P+\epsilon}{P} \right) + P \, \log_2\left(\frac{P-\epsilon}{P} \right) + \epsilon \, \log_2 \left(\frac{P-\epsilon}{P+\epsilon} \right)=
\\
 &=& P \, \log_2\left(1+\frac{\epsilon}{P} \right) + P \, \log_2\left(1-\frac{\epsilon}{P} \right) + \epsilon \, \log_2 \left(1 -\frac{2\epsilon}{P+\epsilon} \right) \approx
\\
&\approx& -\frac{2\epsilon^2}{P+\epsilon}
\end{eqnarray}$



where we have used $\log(1+x)\approx x$ for small $x$.

Thus, any *difference* $\epsilon$ in the probabilities *decreases* the ensemble entropy!  It follows that ensemble entropy is maximal when all outcomes are equiprobable.

# Summary 'information' and 'entropy' of discrete random variables


* The 'Shannon information' $h(x)$ of a random event $x$ depends on its probability $P(x)$:

$\begin{eqnarray}
h(x) = - \log_2 \, P(x)
\end{eqnarray}$

* The `Shannon entropy' $H(X)$ of a random variable or process is the average information of its events:

$\begin{eqnarray}
H(X) = - \sum_x \, P(x) \, h(x)= - \sum_x \, P(x) \, \log_2 \, P(x)
\end{eqnarray}$

* The unit of information and entropy is `bits'.
* The maximal entropy of a random variable with $N$ discrete outcomes is

$\begin{eqnarray}
H(X) \leq \log_2 N
\end{eqnarray}$

# Example runs

Given an eight-by-eight grid, the chances for our first shot are $P(y)=1/64$ and $P(n)=63/64$.   

If we are lucky enough to hit the submarine with the first shot, we have gained

$\begin{eqnarray}
h(x) = h_{(1)}(y) = \log_2 \,  64 = 6 \, {\rm bits}
\end{eqnarray}$

of information and finish the run.

Note that we have reduced the number of possibilities $64$-fold, which corresponds to $6$ consecutive $2$-fold reductions.  Hence we have gained $6$ bits.  

If the first shot misses, the chances for the second shot are $P(y)=1/63$ and $P(n)=62/63$.

If the second misses, too, the chances for the third shot are  $P(y)=1/62$ and $P(n)=61/62$.

Even though our first shot missed, we have still gained

$\begin{eqnarray}
h(x) = h_{(1)}(n) = \log_2 \, \frac{64}{63} = 0.0227 \, {\rm bits}
\end{eqnarray}$

The second miss gained us

$\begin{eqnarray}
h_{(2)}(n) = \log_2 \, \frac{63}{62} = 0.0230 \, {\rm bits}
\end{eqnarray}$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec17_Sinking_subs_c.png" width="800">

(Slide-27)

After 32 consecutive misses (firing at a new square each time), the total information gained is

$\begin{eqnarray}
\log_2 \, \frac{64}{63} +  \log_2 \, \frac{63}{62} + \ldots + \log_2 \, \frac{33}{32} = \log_2 \, \frac{64}{32} =
\\
\\
= 0.0227 + 0.0230 + \ldots + 0.0430 = 1 \, {\rm bits}
\end{eqnarray}$

corresponding to a $2$-fold reduction of the remaining possibilities ($2^1=2$).

After 48 unsuccessful shots, the information gained is 2 bits:

$\begin{eqnarray}
\log_2 \, \frac{64}{63} + \ldots + \log_2 \, \frac{17}{16} = \log_2 \, \frac{64}{16}= 2 \, {\rm bits}
\end{eqnarray}$

corresponding to a $4$-fold reduction of the remaining possibilities ($2^2=4$): The unknown location has been narrowed down to one quarter of the original number of possibilities.

What if we hit on the 49th shot, when there were 16 squares let?  The Shannon information gained is

$\begin{eqnarray}
h_{(49)} = \log_2 \, 16 = 4 \, {\rm bits}
\end{eqnarray}$

corresponding to a $16$-fold reduction of the remaining possibilities ($2^4=16$).

# **Total information gained**

<br>

What is the total information gained by $48$ consecutive misses and a hit with the $49^{th}$ shot?

$\begin{eqnarray}
\sum_{i=1}^{48} \, h_{(i)} = 2 \, {\rm bits}, \qquad\qquad h_{(49)} = 4 \, {\rm bits}
\end{eqnarray}$

so that

$\begin{eqnarray}
\sum_{i=1}^{49} \, h_{(i)} = 6 \, {\rm bits}
\end{eqnarray}$

<br>

This result holds regardless of when we happen to hit the submarine!  We always gain $6$ bits, because we always end up with a $64$-fold reduction of possibilities ($2^6=64$).

# **4. Relative entropy or Kullback-Leibler divergence**

<br>

We have defined the entropy of a random variable with distribution $P(X)$ over a set of outcomes $X\in\{x_1, x_2, \ldots , x_n\}$ as

$\begin{eqnarray}
H_P(X) = -\sum_i P(x_i) \log_2 P(x_i)
\end{eqnarray}$

<br>

Suppose we also observe another random variable with distribution $Q(X)$ over the {\it same set} of outcomes $X\equiv\{x_1, x_2, \ldots , x_n\}$, and entropy

$\begin{eqnarray}
H_Q(X) = -\sum_i Q(x_i) \log_2 Q(x_i)
\end{eqnarray}$

<br>

What measure might be suitable for quantifying the similarity or dissimilarity of these distributions?

Evidently, the entropy itself is no such measure!

(Slide-32)


# Example distributions

<br>

Possible outcomes

$\begin{eqnarray}
X \in \{ 1, 2, 3, 4, 5, 6, 7, 8\}
\end{eqnarray}$

<br>

Two distributions

$\begin{eqnarray}
P(X) = \{\frac{1}{16}, \frac{1}{16}, \frac{1}{8},  \frac{1}{4},\frac{1}{4},\frac{1}{8}, \frac{1}{16}, \frac{1}{16}\}
\\
Q(X) = \{\frac{1}{4}, \frac{1}{8}, \frac{1}{16},  \frac{1}{16},\frac{1}{16},\frac{1}{16}, \frac{1}{8}, \frac{1}{4}\}    
\end{eqnarray}$

<br>

Two entropies

$\begin{eqnarray}
H_P = -\frac{2}{8} \log_2 4 - \frac{2}{8} \log_2 8-\frac{4}{8} \log_2 16 = \frac{2}{4}+\frac{3}{4}+\frac{4}{2} = 3\frac{1}{4}
\end{eqnarray}$

$\begin{eqnarray}
H_Q = 3\frac{1}{4}
\end{eqnarray}$

Entropies are the same, even if distributions are different!

(Slide-33)


# Relative entropy or Kullback-Leibler divergence

<br>

A good measure of the difference between two probability distributions (over the same set of outcomes) is the KL-divergence.  Observing an outcome $x$, the difference between the two distributions $P(x)$ and $Q(x)$ is, in terms of Shannon information, is

$\begin{eqnarray}
\log_2 \, \frac{1}{Q(x)} - \log_2 \, \frac{1}{P(x)} = \log_2 \, \frac{P(x)}{Q(x)}
\end{eqnarray}$

<br>

The average difference (assuming the true probability $P(x)$) over all outcomes $x$ is then

$\begin{eqnarray}
D_{KL}\left(P||Q\right) = \sum_x \, P(x) \,  \log_2 \, \frac{P(x)}{Q(x)}
\end{eqnarray}$

<br>

In our example,

$\begin{eqnarray}
D_{KL}\left(P||Q\right) = \frac{2}{16} \log_2 \frac{4}{16}+\frac{2}{16} \log_2 \frac{8}{16}+\frac{2}{8} \log_2 \frac{16}{8} + \frac{2}{4} \log_2 \frac{16}{4} = \frac{14}{16}
\\
D_{KL}\left(Q||P\right) = \frac{2}{16} \log_2 \frac{4}{16}+\frac{2}{16} \log_2 \frac{8}{16}+\frac{2}{8} \log_2 \frac{16}{8} + \frac{2}{4} \log_2 \frac{16}{4} = \frac{14}{16}
\end{eqnarray}$

(Slide-34)


# Gibbs inequality

<br>

Note that, in general,

$\begin{eqnarray}
D_{KL}\left(P||Q\right) \neq D_{KL}\left(Q||P\right)
\end{eqnarray}$

<br>

Importantly, the KL-divergence is positive unless the two distributions are identical:

$\begin{eqnarray}
D_{KL}(P||Q) \geq 0  
\end{eqnarray}$

<br>

$\begin{eqnarray}
D_{KL}(P,Q) =  0 \quad {\rm iff} \quad P(x) = Q(x)
\end{eqnarray}$

<br>

(Slide-34)



# Jensen's inequality (optional)

<br>

The non-negativity of the KL divergence follows from the fact that the negative logarithm is a convex function

<br>

*Convexity:*

$\begin{eqnarray}
\qquad\qquad f''(x)\geq 0 \qquad\qquad \forall x  
\end{eqnarray}$

<br>

*Consequence of convexity:*

$\begin{eqnarray}
x^* \equiv \lambda x_1 + (1-\lambda) x_2, \qquad\qquad 0\leq\lambda\leq 1
\\
\\
\Rightarrow\qquad\qquad f[x^*] \quad \leq \quad \lambda \, f(x_1) + (1-\lambda) \, f(x_2)  
\qquad\qquad
\forall x_1, x_2
\end{eqnarray}$

# Proof that $D_{KL} \geq 0$ (optional)


<br>

The negative logarithm is a convex function, so that

$\begin{eqnarray}
\left\langle{- \log z}\right\rangle \geq -\log \left\langle{z}\right\rangle
\end{eqnarray}$

<br>

For

$\begin{eqnarray}
z \equiv \frac{Q(x)}{P(x)}, \quad \left\langle{z}\right\rangle = \left\langle \frac{Q(x)}{P(x)} \right\rangle
\quad {\rm with} \quad \sum_x\,Q(x) = \sum_x\,P(x) = 1
\end{eqnarray}$

<br>

it follows that

$\begin{eqnarray}
D_{KL}\left(P,Q\right) = \sum_x \, P(x) \, \log\frac{P(x)}{Q(x)} = \left\langle -\log\frac{Q(x)}{P(x)} \right\rangle_P
\\
\geq -\log \left\langle \frac{Q(x)}{P(x)} \right\rangle_P = -\log\left(  \sum_x \, P(x) \,\frac{Q(x)}{P(x)}  \right)= 0
\end{eqnarray}$

```python
from inspect import signature
# @title Kullback-Leibler Divergence {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.stats import poisson
from scipy.stats import norm
from matplotlib.colors import LinearSegmentedColormap

def MakeRedBlueColormap():

    # Define colors and their positions (0 to 1)
    colors = [(0, "white"), (0.5, "yellow"), (1, "red")]  # Red -> Yellow -> White

    # Create the colormap
    ccmap = LinearSegmentedColormap.from_list("custom_gradient", colors)

    # cmap = plt.cm.jet;

    ccmap.set_bad((1.0, 1.0, 1.0, 1.0));

    return ccmap;

def ShowColorArray(axs, V, Vmax, Cmap):
    # show a 2D numpy array V, using a colormap (e.g., plt.cm.viridis)


    cmax = Vmax;

    (nrow,ncol) = np.shape(V);

    pcolor_plot = axs.pcolor(V, cmap=Cmap, vmin=0, vmax=cmax)
    for i in range(0,nrow):
        for j in range(0,ncol):
            axs.text(j+0.3, i+0.4, str(round(V[i,j],2)) )
    axs.set_xlim((0,ncol))
    axs.set_ylim((0,nrow))
    #axs.axis('off')

    return pcolor_plot

# get entropy contributions
def GetKullbackLeibler( P_i, Q_i ):

    LP_i = np.zeros_like(P_i);
    HP_i = np.zeros_like(P_i);

    LQ_i = np.zeros_like(Q_i);
    HQ_i = np.zeros_like(Q_i);

    # entropy of distribution P

    ix = np.where((0<P_i) & (P_i<1))

    LP_i[ix] = -np.log2(P_i[ix]);
    HP_i[ix] = P_i[ix] * LP_i[ix];

    HP = np.sum(HP_i);

    # entropy of distribution Q

    ix = np.where((0<Q_i) & (Q_i<1))

    LQ_i[ix] = -np.log2(Q_i[ix]);
    HQ_i[ix] = Q_i[ix] * LQ_i[ix];

    HQ = np.sum(HQ_i);

    # divergence

    DKL = np.sum( P_i * (LQ_i - LP_i) );

    return HP_i, HQ_i, HP, HQ, DKL

# get log probability and entropy, taking care of singularities
def GetFourDistributions( Noutcome ):

    x_i = range(0,Noutcome);
    P_i = np.random.rand(Noutcome);
    P_i = P_i / np.sum(P_i);

    Pup_i = np.sort( P_i );

    Pdown_i = Pup_i[::-1];

    ix1 = range(0,Noutcome-1,2);
    ix2 = range(1,Noutcome,2);
    ix2 = ix2[::-1];

    arr1 = Pup_i[ix1];
    arr2 = Pup_i[ix2];

    Pmid_i = np.concatenate( (arr1, arr2) );

    Pran_i = P_i;

    return Pup_i, Pmid_i, Pdown_i, Pran_i

def plot_kullback_leibler_divergence( Noutcome=10 ):
    """(Slide-37)"""

    x_i = np.arange(1,Noutcome+1);

    # Get four distributions with an identical probability set
    PA_i, PB_i, PC_i, PD_i = GetFourDistributions( Noutcome )

    # entropies and divergence
    (HA_i, HB_i, HA, HB, DKL_AB) = GetKullbackLeibler( PA_i, PB_i );
    (HA_i, HC_i, HA, HC, DKL_AC) = GetKullbackLeibler( PA_i, PC_i );
    (HA_i, HD_i, HA, HD, DKL_AD) = GetKullbackLeibler( PA_i, PD_i );
    (HB_i, HA_i, HB, HB, DKL_BA) = GetKullbackLeibler( PB_i, PA_i );
    (HB_i, HC_i, HB, HC, DKL_BC) = GetKullbackLeibler( PB_i, PC_i );
    (HB_i, HD_i, HB, HD, DKL_BD) = GetKullbackLeibler( PB_i, PD_i );
    (HC_i, HA_i, HC, HA, DKL_CA) = GetKullbackLeibler( PC_i, PA_i );
    (HC_i, HB_i, HC, HB, DKL_CB) = GetKullbackLeibler( PC_i, PB_i );
    (HC_i, HD_i, HC, HD, DKL_CD) = GetKullbackLeibler( PC_i, PD_i );
    (HD_i, HA_i, HD, HA, DKL_DA) = GetKullbackLeibler( PD_i, PA_i );
    (HD_i, HB_i, HD, HB, DKL_DB) = GetKullbackLeibler( PD_i, PB_i );
    (HD_i, HC_i, HD, HC, DKL_DC) = GetKullbackLeibler( PD_i, PC_i );


    Hentropy = np.array( [HA, HB, HC, HD] );

    Divergence = np.zeros( (4,4) );

    Divergence[0,:] = [0, DKL_AB, DKL_AC, DKL_AD];
    Divergence[1,:] = [DKL_BA, 0, DKL_BC, DKL_BD];
    Divergence[2,:] = [DKL_CA, DKL_CB, 0, DKL_CD];
    Divergence[3,:] = [DKL_DA, DKL_DB, DKL_DC, 0];

    # Plot results
    fig, axs = plt.subplots( 1,3, figsize=(12,4))

    # distributions

    axs[0].plot( x_i, PA_i, color=(0.75,0.25,0.25), label='A')
    axs[0].plot( x_i, PB_i, color=(0.25,0.75,0.25), label='B')
    axs[0].plot( x_i, PC_i, color=(0.25,0.25,0.75), label='C')
    axs[0].plot( x_i, PD_i, color=(0.5,0.5,0.5), label='D')
    axs[0].plot( x_i, PA_i, 'o', color=(0.75,0.25,0.25), markersize=7)
    axs[0].plot( x_i, PB_i, 'o', color=(0.25,0.75,0.25), markersize=7)
    axs[0].plot( x_i, PC_i, 'o', color=(0.25,0.25,0.75), markersize=7)
    axs[0].plot( x_i, PD_i, 'o', color=(0.5,0.5,0.5), markersize=7)
    axs[0].legend(loc='lower center')
    axs[0].set_ylabel('probability')
    axs[0].set_xlabel('outcome')
    axs[0].set_box_aspect(1/1)
    axs[0].set_title('Four distributions', fontsize=14);

    # entropies

    axs[1].bar( [1], Hentropy[0], color=(0.75,0.25,0.25) )
    axs[1].bar( [2], Hentropy[1], color=(0.25,0.75,0.25) )
    axs[1].bar( [3], Hentropy[2], color=(0.25,0.25,0.75) )
    axs[1].bar( [4], Hentropy[3], color=(0.5,0.5,0.5) )
    axs[1].set_xticks([1,2,3,4])
    axs[1].set_xticklabels(['A','B','C','D'])
    axs[1].set_box_aspect(1/1)
    axs[1].set_ylabel('Entropy H')
    axs[1].set_title('Four entropies', fontsize=14);

    # divergences

    Ccmap = MakeRedBlueColormap()

    pcp = ShowColorArray(axs[2], Divergence, np.max(Divergence), Ccmap)
    axs[2].set_xticks([0.5,1.5,2.5,3.5])
    axs[2].set_xticklabels(['A','B','C','D'])
    axs[2].set_yticks([0.5,1.5,2.5,3.5])
    axs[2].set_yticklabels(['A','B','C','D'])
    axs[2].set_box_aspect(1/1)
    axs[2].set_title('Pairwise divergence', fontsize=14);

    fig.colorbar(pcp, ax=axs[2], orientation='vertical' );

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_kullback_leibler_divergence, Noutcome = (5, 100, 1 ) )
```

# **Joint probability**

In general, when jointly observing two ensembles, we establish the probabilities of joint outcomes $x,y$:

<br>

$\begin{eqnarray}
\begin{array}{c|ccccccc}
 & & x=1 & x=2 & x=3 & x=4 & \ldots\\
 \\
 \hline
 \\
 y=1 & & P(1,1) & P(1,2) & P(1,3) & P(1,4) & \ldots \\
 \\
 y=2 & & P(2,1)  & P(2,2) & P(2,3) & P(2,4) & \ldots \\
 \\
 y=3 & & P(3,1)  & P(3,2) & P(3,3) & P(3,4) & \ldots \\
 \\
 y=4 & & P(4,1)  & P(4,2) & P(4,3) & P(4,4) & \ldots \\
 \\
 \vdots & & \vdots & \vdots & \vdots & \vdots & \ddots \\
\end{array}
\end{eqnarray}$

(Slide-41)


# **Joint entropy**

<br>

The entropy of a joint ensemble is called 'joint entropy':

<br>

$\begin{eqnarray}
H(X,Y) &=& -\sum_{x,y} P(x,y) \log_2 P(x,y) =
\\
&=& -\sum_{x} \left[ \sum_{y} P(x,y) \log_2 P(x,y) \right] =
\\
&=& -\sum_{y} \left[ \sum_{x} P(x,y) \log_2 P(x,y) \right]
\end{eqnarray}$

where the sums go over all ordered pairs.

(Slide-42)

# **Marginal entropies**

The marginal entropies are

<br>

$\begin{eqnarray}
H(X) &=& -\sum_{x} P(x) \log_2 P(x) =
\\
&=& -\sum_{x} \left[ \sum_{y} P(x,y)\right] \log_2 P(x) = -\sum_{x,y} P(x,y) \log_2 P(x)
\end{eqnarray}$

<br>

$\begin{eqnarray}
H(Y) &=& -\sum_{y} P(y) \log_2 P(y) =
\\
&=& -\sum_{y} \left[ \sum_{x} P(x,y)\right] \log_2 P(y) = -\sum_{x,y} P(x,y) \log_2 P(y)
\end{eqnarray}$

<br>

Taken together, the two marginal entropies are

$\begin{eqnarray}
H(X) + H(Y) = -\sum_{x,y} P(x,y) \log_2 \left[P(x) P(y)\right]
\end{eqnarray}$

(Slide-43)


# **Comparison**

How do joint and marginal entropies compare?  Consider the difference between the sum of the marginal entropies and the joint entropy:

<br>

$\begin{eqnarray}
H(X) + H(Y)  &-& H(X,Y) =
\\
&=& - \sum_{x,y} P(x,y) \log_2 \left[P(x) P(y)\right]
\\
&&\quad + \sum_{x,y} P(x,y) \log_2 P(x,y)
\\
&=& \sum_{x,y} P(x,y) \log_2 \frac{P(x,y)}{P(x) P(y)} =
\\
&=& D_{KL}\left[P(x,y)||P(x)P(y)\right] \quad \geq \quad 0
\end{eqnarray}$

<br>

The joint entropy *cannot be larger* than the sum of individual entropies

$\begin{eqnarray}
\qquad\qquad\qquad H(X,Y) \leq H(X) + H(Y)
\end{eqnarray}$

*In terms of Shannon information, the whole is **smaller** than the sum of its parts!*

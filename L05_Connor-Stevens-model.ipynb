# **1. Discovery of action potential mechanism**

Hodgkin and Huxley demonstrated that the action potential ("spike") is the consequence of voltage-dependent $Na^+$ and $K^+$ conductances.

How did they arrive at this conclusion?

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec05_Fig03.png" width="500">

 "Hodgkin and Huxley made the first intracellular recording of an action potential, demonstrating directly that the action potential exceeds zero mV. Hodgkin and Huxley then showed that step depolarizations of the squid axon trigger an inward current followed by an outward current. Using ionic substitution, they demonstrated that this net current could be separated into two distinct components, a rapid inward current carried by Na+ ions, and a more slowly activating outward current carried by K+ ions. From experiments using ingenious voltage-clamp protocols, they concluded that these two currents result from independent permeability mechanisms for Na+ and K+ with conductances changing as a function of time and membrane potential. This was a stunning conceptual breakthrough, later termed the 'ionic hypothesis,' a unifying framework for the field that triggered the search for the underlying molecular structures."

[Michael Haeusser, 2000, Nature Neuroscience](https://www.nature.com/articles/nn1100_1165)

# Universality of sodium-potassium action potential

# Voltage clamp

Hodgkin, Huxley, and Katz applied the "voltage clamp" technique, which continues to be an extremely useful tool in electrophysiology.  

Normally, a current is applied to a cell and the resulting voltage change is recorded.  The applied current distributes itself across membrane resistance, membrane capacitance, and laterally to distant membrane regions (e.g., along the axon).   

With the voltage-clamp, a voltage is applied and the resulting current is measured.  To maintain the desired voltage, a high-frequency amplifier rapidly adjusts the required current.  A step change in voltage reveals currents through membrane conductances, as capacitative currents are minimized.

# Sodium substitution experiment

HH suspected that the inward current was carried by $Na^+$ ions. To confirm this hypothesis, they performed a sodium-substitution experiment, in which they substituted choline $(CH_3)_3[N^+]CH_2CH_2OH$ for sodium in the outside medium, reducing the sodium concentration approximately ten-fold.

After correcting for capacitative and leak currents, HH termed the remaining ionic currents $I_i = I_{Na} + I_K$ in sea water, and $I'_i = I'_{Na} + I'_K$ in choline water. Assuming $I_K=I'_K$ and proportionality $I'_{Na} = k I_{Na}$, they plotted $I'_{Na}$ against $I_{Na}$ to estimate proportionality constant $k$ and to obtain the relation
$\begin{eqnarray}
I_i-I'_i = I_{Na} - I'_{Na} = I_{Na} (1-k)
\end{eqnarray}$

This permitted HH to estimate
$\begin{eqnarray}
I_{Na}= \frac{I_i-I'_i}{1-k}, \qquad\qquad I'_{Na}= \frac{k (I_i-I'_i)}{1-k}, \qquad\qquad I_K = I'_K = I_i - I_{Na} = \frac{I'_i-k I_i}{1-k}
\end{eqnarray}$

The measured and deduced currents are illustrated below. Note that the crossing of $I_{Na}$ and $I'_{Na}$ around $2 ms$ is an artefact and reflects partial violations of assumptions.

```python
# @title Voltage-clamp experiments {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Voltage-clamp experiments demonstrating asymmetric condutance

# Define auxiliary functions

def HH_alpha_M( V ):
    #    alpha = 0.1 * (V+40) / (1 - exp( -0.1*(V + 40) ) );
    va = 0.1 * ( V + 40 );
    alpha = va / ( 1 -  np.exp( -va ) );  # CAUTION: this fails when V equals exactly -40 mV !!!!
    return alpha

def HH_beta_M( V ):
    #    beta  = 4 * exp( -0.0556*(V+65) );
    vb = 0.0556 * ( V + 65 );
    beta = 4 *  np.exp( -vb );
    return beta

def HH_tau( alpha , beta ):
    tau = 1 / (alpha + beta );
    return tau

def HH_ss( alpha , beta ):
    ss = alpha / (alpha + beta );
    return ss

def HH_alpha_H( V ):
    # alpha = 0.07 * exp( -0.05 * (V + 65) )
    va = 0.05 * ( V + 65 );
    alpha = 0.07 *  np.exp( -va );
    return alpha

def HH_beta_H( V ):
    #    beta  = 1 / ( 1 + exp(-0.1*(V + 35) ) )
    vb = 0.1 * ( V + 35 );
    beta = 1 / ( 1 + np.exp( -vb ) );
    return beta

def HH_alpha_N( V ):
    # alpha = 0.01 * (V+55) / (1 - exp(-0.1 * (V+55) ) )
    va = 0.1 * ( V + 55 );                      # CAUTION: this fails when V equals exactly -55 mV !!!!
    if va!=0:
        alpha = 0.1 * va / ( 1 - np.exp( -va ) );
    else:
        alpha = 0.1;
    return alpha

def HH_beta_N( V ):
    #   beta  = 0.125 * exp( -0.0125*(V+65) );
    vb = 0.0125 * ( V + 65 );
    beta = 0.125 * np.exp( -vb );
    return beta

# Define an interactive plot
def plot_voltage_clamp_experiment(V_m=-1, CNa_out = 400):
    """(Slide-8)"""

    V0_m = -65                                 # [mV] initial value of membrane potential
    #V_m  = -65                                 # [mV] step change of membrane potential

    # parameters
    CNa_base = 400; # mM   baseline under sea water conditions
    CNa_in = 50;  # mM internal concentration
    cm = 10;      # nF/mm^2
    gL = 3;       # uS/mm^2  (10e-6 S / mm)
    gNa = 1200;   # uS/mm^2
    gK = 360;     # uS/mm^2
    EL = -54.402; # mV
    ENa_chol = 24 * np.log( CNa_out  / CNa_in );     # mV
    ENa_base = 24 * np.log( CNa_base / CNa_in );     # mV  baseline under sea water conditions
    EK       = -77;     # mV

    dt       = 0.1;                   # time step, in ms (0.1 ms)
    T        = 12;                    # total time, in ms (15 ms)
    Ni       = round(T/dt) + 1;       # number of time steps
    ti       = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # allocate vectors for variables to be plotted (membrane potential, steady-state potential, activation variables, currents)

    Imi_chol  = np.zeros_like(ti)
    Vmi       = np.zeros_like(ti)
    iLi       = np.zeros_like(ti)
    iNai_chol = np.zeros_like(ti)
    iKi       = np.zeros_like(ti)

    Imi_base  = np.zeros_like(ti)     # duplicate variables for baseline under sea water conditions
    iNai_base = np.zeros_like(ti)

    # assign initial values where necessary

    Vmi[0]     = V0_m                     # membrane potential

    Imi_chol[0]  = 0                      # membrane current in choline water
    Imi_base[0]  = 0                      # initialize baseline in sea water

    alphaMi = HH_alpha_M( Vmi[0] )        # initial rates [1/ms]
    betaMi  = HH_beta_M( Vmi[0] )
    alphaHi = HH_alpha_H( Vmi[0] )
    betaHi  = HH_beta_H( Vmi[0] )
    alphaNi = HH_alpha_N( Vmi[0] )
    betaNi  = HH_beta_N( Vmi[0] )

    ssMi    = HH_ss( alphaMi , betaMi )   # initial steady-states [fractions]
    ssHi    = HH_ss( alphaHi , betaHi )
    ssNi    = HH_ss( alphaMi , betaMi )

    Mi      = ssMi                        # initial activation variables [fractions]
    Hi      = ssHi
    Ni      = ssNi


    # iterate over successive time steps, starting at initial value of potential

    for i in range( len(ti) ):

        if i>0:                      # voltage-clamp keeps voltage constant!
            Vmi[i] = V_m

                    # activation variables       [mV]

        alphaMi    = HH_alpha_M( Vmi[i] )       # rates depend on V
        betaMi     = HH_beta_M( Vmi[i] )        # remain constant under
        alphaHi    = HH_alpha_H( Vmi[i] )       # voltage clamp!
        betaHi     = HH_beta_H( Vmi[i] )
        alphaNi    = HH_alpha_N( Vmi[i] )
        betaNi     = HH_beta_N( Vmi[i] )

        tauMi      = HH_tau( alphaMi , betaMi )
        ssMi       = HH_ss( alphaMi , betaMi )
        tauHi      = HH_tau( alphaHi , betaHi )
        ssHi       = HH_ss( alphaHi , betaHi )
        tauNi      = HH_tau( alphaNi , betaNi )
        ssNi       = HH_ss( alphaMi , betaMi )

        gNai       = gNa * Hi * Mi ** 3               # conductances [uS/mm^2]
        gKi        = gK * Ni ** 4

        DFLi       = -Vmi[i] + EL                     # driving forces [mV]
        DFNai_chol = -Vmi[i] + ENa_chol                # sodium under choline
        DFNai_base = -Vmi[i] + ENa_base                # sodium under sea water conditions
        DFKi       = -Vmi[i] + EK

        iLi[i]       = gL      * DFLi                   # currents [nA/mm^2 =  mV * uS/mm^2]
        iNai_chol[i] = gNai    * DFNai_chol
        iNai_base[i] = gNai    * DFNai_base
        iKi[i]       = gKi     * DFKi

        Imi_chol[i]  = iNai_chol[i] + iKi[i] + iLi[i]            # membrane current
        Imi_base[i]  = iNai_base[i] + iKi[i] + iLi[i]

        if i < len(ti)-1:
             # update state variables at time i+1

            Mi = ssMi + (Mi - ssMi) * np.exp(-dt / tauMi)

            Hi = ssHi + (Hi - ssHi) * np.exp(-dt / tauHi)

            Ni = ssNi + (Ni - ssNi) * np.exp(-dt / tauNi)


    # Create a figure and a grid of subplots (2 row, 1 column)
    fig, axs = plt.subplots(2, 2, figsize=(8, 8))

    # First subplot
    axs[0,0].plot(ti, Vmi, 'k');
    axs[0,0].plot([0,T], [V0_m,V0_m], 'k:');
    axs[0,0].set_title('Potential');
    axs[0,0].set_ylim(-130,0);
    axs[0,0].set_xlabel('time [ms]');

    axs[0,0].set_ylabel('[mV]');

    # Second subplot
    axs[1,0].plot([0,T], [0,0], 'k:');
    axs[1,0].plot(ti, iKi/1000, 'b');
    axs[1,0].plot(ti, iNai_chol/1000, 'r');
    axs[1,0].plot(ti, Imi_chol/1000, 'k');
    axs[1,0].set_title('Currents in choline');
    axs[1,0].set_ylim(-25,15);
    axs[1,0].set_xlabel('time [ms]');
    axs[1,0].set_ylabel('uA');

    # Third subplot
    axs[0,1].plot(Imi_base/1000, Imi_chol/1000, 'k');
    axs[0,1].set_title('Membrane current ratio');
    axs[0,1].set_xlim(-25,15);
    axs[0,1].set_ylim(-25,15);
    axs[0,1].set_xlabel('Sea water [uA]');
    axs[0,1].set_ylabel('Choline water [uA]');

    # Third subplot
    axs[1,1].plot(iNai_base/1000, iNai_chol/1000, 'r');
    axs[1,1].set_title('Sodium current ratio');
    axs[1,1].set_xlim(0,15);
    axs[1,1].set_ylim(0,15);
    axs[1,1].set_xlabel('Sea water [uA]');
    axs[1,1].set_ylabel('Choline water [uA]');

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_voltage_clamp_experiment, V_m=(-127, +44, 9), CNa_out=(50, 400, 350))

```

# Summary

* The membrane conductance increases during an AP.

* Voltage clamp reveals the voltage- and time-dependence of the additional conductances.

* Voltage-dependence: conductance increase is asymmetric and occurs for depolarization, not hyperpolarization.

* Time-dependence: a transient conductance carries and inward current, a persistent conductance an outward current.

* Sodium substitution reveals that the inward current is carried by sodium.

* The outward current is carried by potassium.

# **2. Determination of opening and closing rates of $m$, $h$, and $n$**

To infer the maximal conductance for $Na^+$ and $K^+$, H-H measured currents in response to a voltage step from $-65 mV$ to $V_m$.  They established steady-state conductance for persistent $K^+$, but peak conductance for transient $Na^+$.  The slopes of the linear regimes at positive voltages reveals the maximal conductance values:

$\begin{eqnarray}
\qquad\qquad g_K = \frac{I_K}{V_m}, \qquad\qquad g_{Na} = \frac{I_{Na}}{V_m}
\end{eqnarray}$

The curvy parts at negative voltages reflect the respective voltage dependencies.

# Time course of activation variables

# Historical significance of HHM

* Summarizing experimental observations in a quantitative model.

* First complete description of the excitability of a single cell.

* Linking the microscopic level of ion channels to the macroscopic level of currents and action potentials.

* Reproduce and explain a remarkable range of data from squid axon: shape and propagation of the action potential, sharp threshold, refractory period, refractoryness, and subthreshold oscillations.

* With minor parameter changes, HHM can describe many channel types. Even today, most biophysical spiking models are still based on the HHM.

(Slide-17)


 # What Hodgkin and Huxley got right

* First, they chose the right model system. The squid axon offered technical advantages due to its size and relative simplicity, with only two types of voltage-gated conductances.

* Second, they adapted the right experimental method (voltage clamp), disabling the feedback loops between membrane potential and conductance.

* Third, they chose the right level of detail for the model and adapted concepts from electronics, reaction kinetics, and statistical physics to obtain a simple and transparent description.

* Fourth, the unprecedented degree of completeness and predictiveness of their model influenced biophysical and modeling work for decades.

* Fifth, the collaboration exemplified a balance between experiment and theory that has rarely been matched.

The HHM is a prototypical example for **"emergence"** in biology (collective behaviour of dynamical systems).

# Orthodromic initiation, ephaptic coupling, and mutual annihilation

In principle, axons can propagate APs in both directions.  The 'normal' direction, away from the soma, is called **orthodromic**, whereas the opposite direction, towards the soma, is called **antidromic**.  

In the spike-initiation zone, voltage-dependent channels are distributed unevenly, ensuring **orthodromic initiation.**

Direct electrical interaction between adjacent nerve fibers has occasionally been observed and is termed **ephaptic coupling**.

In experimental situations, APs can be evoked simultaneously at both ends of an axon.  In such situations, the colliding APs **mutually annihilate** each other and leave the axonal membrane refractory.

# Conduction velocity

In invertebrate axons, typical velocities are on the order of $10 m/s$.   

In thicker axons, the ratio of longitudinal resistance to membrane resistance is more favourable, so that depolarisation propagates further and AP conduction is faster.  See also Lecture 7 (cable theory).

# Summary

* Axons are symmetric and able to conduct APs in both directions (ortho- and antidromic).

* Under physiological conditions, AP conduction is orthodromic (from soma to synapse).

* Conduction velocity increases with axon diameter and (much more) with myelinization.

* Two mechanism ensure that axon conduction proceeds orthodromically:



1.   Forward membrane is excitable, backward membrane is refractory.
2.   Depolarization penetrates further in the forward than backward direction.

# **4. Connor-Stevens model of vertebrate action potential**

The Hodgkin-Huxley model was developed to describe the action potential of the squid giant axon.  

In vertebrate neurons, the Connor-Stevens model (CSM) provides a more accurate and useful description.  It also relies on active $Na^+$ and $K^+$ conductances, but with somewhat faster kinetics, so that the resulting action potentials are shorter.  

In addition, the CSM includes a second active $K^+$ conductance called the A-current, which is transient (i.e., activates and deactivates similar to the $Na^+$ conductance).

(Slide-29)


# Conner-Stevens model

The CSM is a single-compartment model with three active conductances, $g_{Na^+}$, $g_{K^+}$, and $g_A$:

$\begin{eqnarray}
\qquad\qquad c_m \, \frac{dV(t)}{dt} = \, -i_m(t) \, + \, I_e(t)
\qquad\qquad
i_m = i_L + i_{Na} + i_K + i_A
\\
\\
\end{eqnarray}$

$\begin{eqnarray}
\qquad\qquad i_L = g_L\,(V-E_L)
\qquad\qquad\qquad
i_{Na} = \bar g_{Na^+} \, m^3\,h \, (V-E_{Na^+})
\\
\\
\qquad\qquad i_K = \bar g_{K+} \, n^4 \, (V-E_{K^+})
\qquad\qquad
i_A = \bar g_A \, a^3 \, b \, (V-E_A)\qquad
\\
\\
\end{eqnarray}$

$\begin{eqnarray}
\qquad\qquad g_L = 0.003\,  mS/mm^2
\qquad & \qquad
E_L = -17 \, mV
\\
\qquad\qquad \bar g_{Na^+} = 1.2\,  mS/mm^2
\qquad & \qquad
E_{Na^+} = 55 \, mV
\\
\qquad\qquad \bar g_{K^+} = 0.2\,  mS/mm^2
\qquad & \qquad
E_{K^+} = -72 \, mV
\\
\qquad\qquad \bar g_A = 0.477\,  mS/mm^2
\qquad & \qquad
E_A = -75 \, mV
\\
\end{eqnarray}$

(Slide-30)

# Five activation variables

All five activation variables relax exponentially towards voltage-dependent steady-states with voltage-dependent time-constants.

$\begin{eqnarray}
\qquad\qquad \frac{d m}{d t} = -\frac{1}{\tau_m} \, \left[ m - m_\infty\right]
\qquad\qquad
\frac{d h}{d t} = -\frac{1}{\tau_h} \, \left[ h - h_\infty\right]
\end{eqnarray}$

$\begin{eqnarray}
\qquad\qquad\qquad\qquad \frac{d n}{d t} = -\frac{1}{\tau_n} \, \left[ n - n_\infty\right]
\end{eqnarray}$

$\begin{eqnarray}
\qquad\qquad \frac{d a}{d t} = -\frac{1}{\tau_a} \, \left[ a - a_\infty\right]
\qquad\qquad
\frac{d b}{d t} = -\frac{1}{\tau_b} \, \left[ b - b_\infty\right]
\end{eqnarray}$

(Slide-31)

```python
# @title Connor-Stevens model of vertebrate action potential {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Connor-Stevens model of vertebrate action potential
# Define auxiliary functions

def CS_alpha_M( V ):
    #    alpha = 3.8 * 0.1 * (V+29.7) / (1 - exp( -0.1*(V + 29.7) ) );
    va = 0.1 * ( V + 29.7 );
    alpha = 3.8 * va / ( 1 -  np.exp( -va ) );  # CAUTION: this fails when V equals exactly -29.7 mV !!!!
    return alpha

def CS_beta_M( V ):
    #    beta  = 15.2 * exp( -0.0556*(V+54.7) );
    vb = 0.0556 * ( V + 54.7 );
    beta = 15.2 *  np.exp( -vb );
    return beta

def CS_tau( alpha , beta ):
    tau = 1 / (alpha + beta );
    return tau

def CS_ss( alpha , beta ):
    ss = alpha / (alpha + beta );
    return ss

def CS_alpha_H( V ):
    # alpha = 0.266 * exp( -0.05 * (V + 48) )
    va = 0.05 * ( V + 48 );
    alpha = 0.266 *  np.exp( -va );
    return alpha

def CS_beta_H( V ):
    #    beta  = 3.8 / ( 1 + exp(-0.1*(V + 18) ) )
    vb = 0.1 * ( V + 18 );
    beta = 3.8 / ( 1 + np.exp( -vb ) );
    return beta

def CS_alpha_N( V ):
    # alpha = 0.2 * 0.1 * (V+45.7) / (1 - exp(-0.1 * (V+45.7) ) )
    va = 0.1 * ( V + 55 );                      # CAUTION: this fails when V equals exactly -45.7 mV !!!!
    alpha = 0.2 * va / ( 1 - np.exp( -va ) );
    return alpha

def CS_beta_N( V ):
    #   beta  = 0.25 * exp( -0.0125*(V+55.7) );
    vb = 0.0125 * ( V + 65 );
    beta = 0.25 * np.exp( -vb );
    return beta

def CS_ss_A( V ):
    va = 0.0314 * ( V + 94.22 );
    vb = 0.0346 * ( V + 1.17 );
    vc = 0.0497 * ( V + 55.96 );
    ss = ( 0.0761 * np.exp( va ) / (1 + np.exp( vb ))) ** 0.333333333;
    return ss

def CS_tau_A( V ):
    vc = 0.0497 * ( V + 55.96 );
    tau = 0.3632 + 1.158 / ( 1 + np.exp( vc ) );
    return tau

def CS_ss_B( V ):
    va = 0.0688 * ( V + 53.3 );
    ss = ( 1 / (1 + np.exp( va ))) ** 4;
    return ss

def CS_tau_B( V ):
    va = 0.0624 * ( V + 50 );
    tau = 1.24 + 2.678 / ( 1 + np.exp( va ) );
    return tau

# Define an interactive plot
def plot_vertebrate_action_potential(V0_m = -85, i_e=80, gA = 477):
    """(Slide-36)"""

    #i_e  = 82                                  # [nA] constant electrode current
    #V0_m = -65                                 # [mV] initial value of membrane potential
    #gA   = 477;                                # uS/mm^2

    # parameters
    cm  = 10;      # nF/mm^2
    gL  = 3;       # uS/mm^2  (10e-6 S / mm)
    gNa = 1200;    # uS/mm^2
    gK  = 200;     # uS/mm^2
    #gA  = 477;     # uS/mm^2
    EL  = -17;     # mV
    ENa = 55;      # mV
    EK  = -72;     # mV
    EA  = -75;     # mV

    dt  = 0.02;                  # time step, in ms (0.02 ms)
    T   = 60;                   # total time, in ms (15 ms)
    Ni  = round(T/dt) + 1;       # number of time steps
    ti  = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # allocate vectors for varaibles to be plotted (membrane potential, steady-state potential, activation variables, currents)

    Vmi = np.zeros_like(ti)
    Vinftyi = np.zeros_like(ti)

    Mi   = np.zeros_like(ti)
    Hi   = np.zeros_like(ti)
    Ni   = np.zeros_like(ti)
    Ai   = np.zeros_like(ti)
    Bi   = np.zeros_like(ti)

    iLi  = np.zeros_like(ti)
    iNai = np.zeros_like(ti)
    iKi  = np.zeros_like(ti)
    iAi  = np.zeros_like(ti)

    # assign initial values where necessary

    Vmi[0]     = V0_m                     # membrane potential

    alphaMi = CS_alpha_M( Vmi[0] )        # initial rates [1/ms]
    betaMi  = CS_beta_M( Vmi[0] )
    alphaHi = CS_alpha_H( Vmi[0] )
    betaHi  = CS_beta_H( Vmi[0] )
    alphaNi = CS_alpha_N( Vmi[0] )
    betaNi  = CS_beta_N( Vmi[0] )
    ssAi    = CS_ss_A( Vmi[0] )
    tauAi   = CS_tau_A( Vmi[0] )
    ssBi    = CS_ss_B( Vmi[0] )
    tauBi   = CS_tau_B( Vmi[0] )

    ssMi    = CS_ss( alphaMi , betaMi )   # initial steady-states [fractions]
    ssHi    = CS_ss( alphaHi , betaHi )
    ssNi    = CS_ss( alphaMi , betaMi )

    Mi[0]   = ssMi                        # initial activation variables [fractions]
    Hi[0]   = ssHi
    Ni[0]   = ssNi
    Ai[0]   = ssAi
    Bi[0]   = ssBi


    # iterate over successive time steps, starting at resting potential (reversal
    # potential of leak)

    for i in range( len(ti) ):

        gNai       = gNa * Hi[i] * Mi[i] ** 3         # conductances [uS/mm^2]
        gKi        = gK * Ni[i] ** 4
        gAi        = gA * Bi[i] * Ai[i] ** 3

        DFLi       = -Vmi[i] + EL                     # driving forces [mV]
        DFNai      = -Vmi[i] + ENa
        DFKi       = -Vmi[i] + EK
        DFAi       = -Vmi[i] + EA

        iLi[i]     = gL      * DFLi                   # currents [nA/mm^2 =  mV * uS/mm^2]
        iNai[i]    = gNai    * DFNai
        iKi[i]     = gKi     * DFKi
        iAi[i]     = gAi     * DFAi

        # current steady-states and time-constants       [mV]

        Vinftyi[i] = (gL*EL+gNai*ENa+gKi*EK+gAi*EA+i_e)/(gL+gNai+gKi+gAi);
        taueffi    = cm/(gL+gNai+gKi+gAi);                # [ms = (nF/mm^2) / (uS/mm^2)]

        alphaMi    = CS_alpha_M( Vmi[i] )             # rates
        betaMi     = CS_beta_M( Vmi[i] )
        alphaHi    = CS_alpha_H( Vmi[i] )
        betaHi     = CS_beta_H( Vmi[i] )
        alphaNi    = CS_alpha_N( Vmi[i] )
        betaNi     = CS_beta_N( Vmi[i] )

        tauMi      = CS_tau( alphaMi , betaMi )
        ssMi       = CS_ss( alphaMi , betaMi )
        tauHi      = CS_tau( alphaHi , betaHi )
        ssHi       = CS_ss( alphaHi , betaHi )
        tauNi      = CS_tau( alphaNi , betaNi )
        ssNi       = CS_ss( alphaMi , betaMi )

        ssAi       = CS_ss_A( Vmi[i] )
        tauAi      = CS_tau_A( Vmi[i] )
        ssBi       = CS_ss_B( Vmi[i] )
        tauBi      = CS_tau_B( Vmi[i] )

        if i < len(ti)-1:
             # update state variables at time i+1
            Vmi[i + 1] = Vinftyi[i] + (Vmi[i] - Vinftyi[i]) * np.exp(-dt / taueffi)

            Mi[i + 1] = ssMi + (Mi[i] - ssMi) * np.exp(-dt / tauMi)

            Hi[i + 1] = ssHi + (Hi[i] - ssHi) * np.exp(-dt / tauHi)

            Ni[i + 1] = ssNi + (Ni[i] - ssNi) * np.exp(-dt / tauNi)

            Ai[i + 1] = ssAi + (Ai[i] - ssAi) * np.exp(-dt / tauAi)

            Bi[i + 1] = ssBi + (Bi[i] - ssBi) * np.exp(-dt / tauBi)


    # Create a figure and a grid of subplots (2 row, 2 columns)
    fig, axs = plt.subplots(2, 2, figsize=(8, 6))

    # First subplot
    axs[0,0].plot(ti, Vmi, 'k', ti, Vinftyi, 'k--');
    axs[0,0].set_title('Potentials');
    axs[0,0].set_xlabel('time [ms]');
    axs[0,0].set_ylabel('potential [mV]');

    # Second subplot
    axs[1,0].plot(ti, Mi, 'r', ti, Hi, 'r--', ti, Ni, 'b');
    axs[1,0].set_title('Activation m, h, n');
    axs[1,0].set_xlabel('time [ms]');
    axs[1,0].set_ylabel('fraction');

    # Third subplot
    axs[0,1].plot(ti, iNai/1000, 'r', ti, iKi/1000, 'b', ti, iAi/1000, 'c');
    axs[0,1].set_title('Currents Na, K, A');
    axs[0,1].set_xlabel('time [ms]');
    axs[0,1].set_ylabel('current [uA]');

    # Fourth subplot
    axs[1,1].plot(ti, Ai, 'c', ti, Bi, 'c--');
    axs[1,1].set_title('Activation a, b');
    axs[1,1].set_xlabel('time [ms]');
    axs[1,1].set_ylabel('current [uA]');

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_vertebrate_action_potential, V0_m=(-85, -65, 20), i_e=(-130, 200, 10), gA=(0, 477, 477))

```

# Function of A-current

The A-current delays the first spike and qualitatively changes the threshold behavior from a Type I to a Type II neuron.

D&A Figure 6.1: Firing of action potentials in the Connor-Stevens model.


**Type I neuron** (with A-current)

Output firing increases continuously from zero as input current exceeds the firing threshold.

**Type II neuron** (without A-current)

Output firing increases discontinuously as input current exceeds the firing threshold.

(Slide-37)

# Summary of Connor-Stevens model

* Vertebrate APs involve an additional $K^+$-conductance, the A-current.

* The A-current slows down the $Na^+$ depolarization and delays firing.

* The A-current ensures that firing rates increases continuously with input current (type I neuron).

# **What is a linear systems model good for?**


A model helps us understand and predict the responses of a system to arbitrary stimuli.  Predictions can inform new experiments and further deepen understanding.

A good model is simple, appropriate for the system in question, constrained by few observations, and accurately predicts many additional observations.

For many sensory neurons, a linear systems model is all of this.  Not all types of sensory neurons respond like linear systems, but many important types do.

# **Conditions for linearity**


- When the responses of a system satisfy both *homogeneity* and *additivity*, the system is said to be **linear**.

- Together, *homogeneity* and *additivity* guarantee the **"principle of linear superposition"**.

- When responses additionally satisfy *stationarity*, the system is considered both **linear and stationary**.

- Not all linear systems are stationary, but many are.

- Linear-system models of neuronal responses often combine all three properties: **homogeneity, additivity, stationarity**.

(Slide-9)

# **Space-time method for predicting neuronal responses**

Given linearity, we can measure responses to some simple stimuli in order to predict responses to arbitrarily complex stimuli.


This because complex stimuli can always be considered a 'superposition' of simple stimuli.  For example, any wave-form can be considered a 'superposition' of scaled and shifted standard 'impulses'.

For simplicity, consider discrete timepoints $t_i = i \cdot \Delta t$ and construct waveform $S(t_i)$ as a superposition of standard impulses $s_k(t_k)$

$\begin{eqnarray}
S(t_i) = \sum a_k s_k(t_i- k \cdot \Delta t), \qquad\qquad s_k(t_k) = \left\{\begin{array}{cc} 1 & t_k=0 \\ 0 & t_k \neq 0 \end{array} \right.
\end{eqnarray}$

where $a_k$ are scaling factors and $k \cdot \Delta t$ are time shifts.



<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_03b.png" width="600">

(Slide-10)

# **Superposition of impulse responses**

If we know the response $R_i(t_i)$ to a standard impulse $s_0(0)$ at time $t_k = 0$, we can predict the response to the waveform as the superposition (sum) of the scaled and shifted responses:

$\begin{eqnarray}
R(t_i) = \sum_k \, a_k \, R_k(t_i - \tau_k), \qquad\qquad \tau_k = k \cdot \Delta t
\end{eqnarray}$

where $a_k$ and $\tau_k$ are the same scaling factors and time-shifts.

In practice, the method is not always robustly accurate, in part because 'sharp' standard impulses can be difficult to produce.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_03a.png" width="600">

(Slide-11)

# **Decomposition into Fourier series**

Any waveform $S(t)$ with $0\leq t \leq T$ can be approximated by a Fourier series $S_K(t)$

$\begin{eqnarray}
S(t) \approx S_K(t) = a_0 + \sum_{k=1}^K \left[a_k \cos\left( 2\pi \frac{k}{T} t\right) + b_k \cdot \sin\left(2\pi \frac{k}{T} t \right) \right]
\end{eqnarray}$

where $a_k$ and $b_k$ are scaling factors and $K$ is the number of harmonics. The approximation becomes exact when $K\to \infty$.  $T$ is the duration (period) of the waveform.

Fourier decomposition of square wave function:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_03c.png" width="600">

Fourier decomposition of sawtooth function:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_03d.png" width="600">

The following Fourier series produces a perfect 'impulse function' for $-\pi\leq t \leq +\pi$:

$\begin{eqnarray}
\delta(t) = \frac{1}{2\pi} + \frac{2}{2\pi} \sum_{k=1}^K \cos\left( k t\right)  
\end{eqnarray}$

This function is called **Dirac's delta** function and is defined by

$\begin{eqnarray}
\delta(t) = 0, \qquad \forall t\neq 0, \qquad \int_{-\pi}^{+\pi} \delta(t) dt = 1
\end{eqnarray}$

# **2. Waveforms in time domain and frequency domain (e.g. auditory simuli)**

To understand the response of, say, auditory neurons, we can expose an experimental animal to various kinds of test waveforms and observe the neuron's response in each case.   Popular choices for test waveforms are harmonic functions (e.g., pure tones), short pulses of positive or negative pressure (e.g., narrow Gauss functions), and Gabor functions (also known as 'clicks' or 'pings').  Gabor functions are quite common in sensory biology (e.g., some bats and dolphins emit Gabor function waveforms in the context of echolocation).

We consider and compare the following waveforms and their Fourier decompositions.

- Harmonic function
- Delta function
- Gaussian function
- Gabor function
- White noise

(Slide-17)


```python
from inspect import signature
# @title Fourier function pair: Exponential {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def ExponentialTimedomain(ti, alpha):
    yi = 0.5 * alpha * np.exp(-alpha * np.abs(ti))
    return yi

def ExponentialFreqdomain(ok, alpha):
    yk = alpha**2 / (alpha**2 + ok**2)
    return yk

def MakeAxes(Fs, Nt):
    """
    Create time, frequency, and circular frequency axes.

    Parameters:
        Fs (float): Sampling frequency
        Nt (int): Number of samples

    Returns:
        ti (numpy array): Time points
        fk (numpy array): Frequency points
        ok (numpy array): Circular frequency points
    """
    dt = 1 / Fs  # Time interval
    ti = np.arange(-Nt/2, Nt/2 + 1) * dt  # Time points from -Nt/2 to Nt/2
    fk = Fs * np.arange(-Nt/2, Nt/2) / Nt  # Frequency points
    ok = 2 * np.pi * fk  # Circular frequency points

    return ti, fk, ok

def ShowFourierPair(Fs, Nt, Tfunc, Ffunc, Params):
    """
    Visualize Fourier transform pairs in time and frequency domains.

    Parameters:
        Fs (float): Sampling frequency
        Nt (int): Number of samples
        Tfunc (callable): Function defining time domain signal
        Ffunc (callable): Function defining frequency domain signal
        Params (tuple): Parameters for the functions
    """
    dt = 1 / Fs  # Time interval
    ti, fk, ok = MakeAxes(Fs, Nt)

    #Tlim = dt * np.array([-Nt/2, Nt/2])  # Time axis limits
    Tlim = np.array([-2.5, 2.5])  # Time axis limits
    Flim = 4 * Tlim  # Frequency axis limits

    # Analytical functions in time and frequency domains
    FuncAi = Tfunc(ti, Params)
    FuncAk = Ffunc(ok, Params)

    # Fourier transform of the time-domain function
    FuncFk = np.fft.ifftshift(np.fft.fft(FuncAi, Nt)) * dt
    FuncFk = np.abs(FuncFk)

    # Plot results
    fig, axs = plt.subplots( 1, 2, figsize=(8, 4))

    # Time domain plot
    axs[0].plot(ti, FuncAi, 'k-', linewidth=0.5, label='Analytical')
    axs[0].plot(ti, FuncAi, 'r.', markersize=1, label='Samples')
    axs[0].set_xlim(Tlim)
    axs[0].set_xlabel('time [s]')
    axs[0].set_ylabel('amplitude')
    axs[0].set_box_aspect(1/1)
    axs[0].set_title('Time domain', fontsize=14);

    # Frequency domain plot
    axs[1].plot(fk, FuncAk, 'k-', linewidth=0.5, label='Analytical')
    axs[1].plot(fk, FuncFk, 'b.', markersize=4, label='FFT')
    axs[1].set_xlim(Flim)
    axs[1].set_ylim([0, 1.05])
    axs[1].set_xlabel('freq [Hz]')
    axs[1].set_ylabel('amplitude (absolute value)')
    axs[1].set_box_aspect(1/1)
    axs[1].set_title('Frequency domain', fontsize=14);

    plt.tight_layout()
    plt.show()

def plot_show_exponential_function_pair( Alpha = 5):
    """(Slide-18)"""

    """
    Fourier function pair: Exponential
    """

    # Time axis
    Fs = 1000  # Sampling frequency
    Nt = 10000  # Length of signal

    # Function handles and parameters
    # Exponential function

    Tfunc = ExponentialTimedomain
    Ffunc = ExponentialFreqdomain
    Params = Alpha

    ShowFourierPair(Fs, Nt, Tfunc, Ffunc, Params)

# Creating sliders for concentrations and charge
interact(plot_show_exponential_function_pair, Alpha = (1,10,1) )


```

# **Harmonic functions**

Consider a pure tone. Sound pressure increase and decreases regularly, folling a harmonic function.

Harmonic functions are parameterized by amplitude $A$, frequency $\omega$, and phase $\phi$

$\begin{eqnarray}
S_\mathit{harmonic}(t) = A \cos\left( \omega t + \phi \right)
\end{eqnarray}$

It's Fourier decomposition is the sum of two harmonics with the same frequency (positive and negative).

(Slide-19)

```python
from inspect import signature
# @title Fourier function pair: Harmonic {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def CosineTimedomain(ti, Param):
    lambda_, phi = Param
    yi = np.cos(2 * np.pi * ti / lambda_ + phi)
    return yi

def CosineFreqdomain(ok, Param):
    lambda_, phi = Param

    distpos = np.abs(ok - 2 * np.pi / lambda_)
    distneg = np.abs(ok + 2 * np.pi / lambda_)

    ixpos = np.where(distpos == np.min(distpos))[0]
    ixneg = np.where(distneg == np.min(distneg))[0]

    yk = np.zeros_like(ok)
    yk[ixpos[0]] = np.pi
    yk[ixneg[0]] = np.pi

    return yk

def plot_show_harmonic_function_pair( Lambda = 0.2):
    """(Slide-X)"""

    """
    Fourier function pair: Harmonic
    """

    # Time axis
    Fs = 1000  # Sampling frequency
    Nt = 10000  # Length of signal

    # Function handles and parameters
    # Exponential function

    Tfunc = CosineTimedomain
    Ffunc = CosineFreqdomain

    Phi = 0.0;
    Params = (Lambda, Phi)

    ShowFourierPair(Fs, Nt, Tfunc, Ffunc, Params)

# Creating sliders for concentrations and charge
interact(plot_show_harmonic_function_pair, Lambda = (0.05,0.5,0.05) )


```

# **Dirac's delta function**

Consider a dry clap in an anechoic chamber. Sound pressure increases sharply and equally sharply returns to zero.

This waveform is approximated by **Dirac's delta function**, defined as the limit of a function that becomes both infinitely narrow and infinitely high, while maintaining unit area.

For example, we can define it in terms of the limit of a Gaussian function

$\begin{eqnarray}
\delta(t) = \lim_{\sigma\to 0} \frac{1}{\sqrt{2\pi} \sigma} \exp\left(-\frac{t^2}{2\sigma^2}\right), \qquad\qquad \int_{-\infty}^{+\infty} \delta(t) dt = 1
\end{eqnarray}$

As illustrated graphically further above, the **Dirac delta function** is the sum of cosine functions of all frequencies. Accordingly, in the limit of $\sigma\to 0$, the Fourier transform is unity at all frequencies.

(Slide-20)

```python
from inspect import signature
# @title Fourier function pair: Dirac's Delta function {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def GaussianTimedomain(ti, sigma):
    yi = np.exp(-ti**2 / (2 * sigma**2)) / (np.sqrt(2 * np.pi) * sigma)
    return yi

def GaussianFreqdomain(ok, sigma):
    yk = np.exp(-sigma**2 * ok**2 / 2)
    return yk

def plot_show_delta_function_pair( Sigma = 0.01):
    """(Slide-X)"""

    """
    Fourier function pair: Dirac's Delta function
    """

    # Time axis
    Fs = 1000  # Sampling frequency
    Nt = 10000  # Length of signal

    # Function handles and parameters
    # Exponential function

    Tfunc = GaussianTimedomain
    Ffunc = GaussianFreqdomain

    Params = Sigma;

    ShowFourierPair(Fs, Nt, Tfunc, Ffunc, Params)

# Creating sliders for concentrations and charge
interact(plot_show_delta_function_pair, Sigma = (0.005,0.05,0.005) )


```

# **Gaussian function: a low-pass filter**

Consider a smooth increase of sound pressure, followed by an equally smooth decrease.  Sound pressure follows a Gaussian function


$\begin{eqnarray}
S_\mathit{Gaussian}(t) = \frac{A}{\sqrt{2\pi} \sigma} \exp\left(-\frac{x^2}{2\sigma^2}\right), \qquad\qquad \int_{-\infty}^{+\infty} S_\mathit{Gaussian}(t) dt = A
\end{eqnarray}$

Gaussian functions are parameterized by amplitude $A$ and variance $\sigma^2$ (standard deviation $\sigma$).


It's Fourier decomposition is the sum of harmonics of different frequences, but with a Gaussian distribution of amplitude. In the frequency domain, the variance is $\frac{1}{\sigma^2}$ is thus the exact inverse of the variance in the time domain, $\sigma^2$.

Gaussian functions are **low-pass filters** in the frequency domain.

(Slide-21)

```python
from inspect import signature
# @title Fourier function pair: Gaussian {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def GaussianTimedomain(ti, sigma):
    yi = np.exp(-ti**2 / (2 * sigma**2)) / (np.sqrt(2 * np.pi) * sigma)
    return yi

def GaussianFreqdomain(ok, sigma):
    yk = np.exp(-sigma**2 * ok**2 / 2)
    return yk

def plot_show_gaussian_function_pair( Sigma = 1.0):
    """(Slide-X)"""

    """
    Fourier function pair: Gaussian
    """

    # Time axis
    Fs = 1000  # Sampling frequency
    Nt = 10000  # Length of signal

    # Function handles and parameters
    # Exponential function

    Tfunc = GaussianTimedomain
    Ffunc = GaussianFreqdomain

    Params = Sigma;

    ShowFourierPair(Fs, Nt, Tfunc, Ffunc, Params)

# Creating sliders for concentrations and charge
interact(plot_show_gaussian_function_pair, Sigma = (0.01,1,0.01) )


```

# **Gabor function: a bandpass filter**

Consider a pure tone that grows smoothly louder and equally smoothly softer. Sound pressure follows a harmonic with a Gaussian amplitude modulation:

$\begin{eqnarray}
S_\mathit{Gabor} = \frac{A}{\sqrt{2\pi} \sigma} \exp\left(-\frac{t^2}{2\sigma^2} \right) \cos\left( \omega t + \phi \right)
\end{eqnarray}$

Gabor functions are parametrized by amplitude $A$, variance $\sigma^2$, frequency $\omega$, and phase $\phi$.

The Fourier decomposition is a broad spectrum of harmonics, which are centered around two amplitude peaks, one positive and one negative.

Gabor functions are **bandpass filters** in the frequency domain.

(Slide-22)

```python
from inspect import signature
# @title Fourier function pair: Gabor {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def GaborTimedomain(ti, Param):
    sigma, omega = Param
    yi = 2 * np.cos(omega * ti) * np.exp(-ti**2 / (2 * sigma**2)) / (np.sqrt(2 * np.pi) * sigma)
    return yi

def GaborFreqdomain(ok, Param):
    sigma, omega = Param
    yk = (np.exp(-sigma**2 * (ok - omega)**2 / 2) +
          np.exp(-sigma**2 * (ok + omega)**2 / 2))
    return yk

def plot_show_gabor_function_pair( Sigma = 0.2, Lambda=0.2):
    """(Slide-X)"""

    """
    Fourier function pair: Gabor
    """

    # Time axis
    Fs = 1000  # Sampling frequency
    Nt = 10000  # Length of signal

    # Function handles and parameters
    # Exponential function

    Tfunc = GaborTimedomain
    Ffunc = GaborFreqdomain

    Omega = 2*np.pi/Lambda;
    Params = (Sigma, Omega);

    ShowFourierPair(Fs, Nt, Tfunc, Ffunc, Params)

# Creating sliders for concentrations and charge
interact(plot_show_gabor_function_pair, Sigma = (0.05,1.0,0.05), Lambda=(0.1,2.0,0.1) )


```

# Math for white noise (optional)

Finally, we construct a white noise stimulus $\xi(t)$ from a series of discrete values $\xi_m$, each lasting for a brief interval $\Delta t$.  We choose each $\xi_m$ independently from a probability distribution with mean $0$ and variance $\sigma_\xi^2$.

By construction, the average is zero and the variance is $\sigma_\xi^2$

$\begin{eqnarray}
\left\langle{\xi}\right\rangle = \frac{1}{T}  \int_0^T  \xi(t)  dt = 0, \qquad\qquad \left\langle{\xi^2}\right\rangle = \frac{1}{T}  \int_0^T  \xi^2(t)  dt = \sigma_\xi^2
\end{eqnarray}$

As discussed further below, the auto-correlation of white noise is a delta function:

\begin{eqnarray}
Q_{\xi\xi}(\tau) = \frac{1}{T} \, \int_0^T \, \xi(t)\,\xi(t+\tau) \, dt = \sigma_\xi^2 \, \delta(\tau)
\end{eqnarray}

For $\tau=0$, the average product is $\sigma_s^2$. However, for all $\tau\neq 0$, the individual products are equally likely to be positive and negative, so that the average product $s(t)\,s(t+\tau)$ becomes zero.

*Advanced:* For a stationary process $x(t)$ with autocorrelation $R_x(\tau)$ and Fourier transform ${\tilde X}(\omega)$, the power spectral density

$\begin{eqnarray}
S_x(\omega) \equiv \frac{1}{T} \left| {\tilde X}(\omega)^2 \right| = \int_{-\infty}^{+\infty} R_x(\tau) e^{-i\omega \tau} d\tau
\end{eqnarray}$

equals the Fourier transform of the autocorrelation (Wiener-Khinchin).

For white noise, $R_\xi(\tau) = \sigma_\xi^2 \delta(\tau)$ implies

$\begin{eqnarray}
S_\xi(\omega) = \int_{-\infty}^{+\infty} R_\xi(\tau) e^{-i\omega \tau} d\tau = \int_{-\infty}^{+\infty} \sigma_\xi^2 \delta(\tau) e^{-i\omega \tau} d\tau = \sigma_\xi^2
\end{eqnarray}$

(Slide-25)

# **3. Correlation and convolution**

We wish to remind ourselves of the definition and meaning of 'correlation' and 'convolution'.  We begin by considering sequences of $n$ values $X_i$ (measurements) with approximately zero mean and unit variance

$\begin{eqnarray}
X_i, \qquad\qquad \frac{1}{n} \sum_{i=1}^n \, X_i  \approx 0, \qquad\qquad  \frac{1}{n}  \sum_{i=1}^n \, X_i^2  \approx 1
\end{eqnarray}$

without loss of generality. Basically, these conditions ensure that values $X_i$ equal the deviations from the mean (0), in units of standard deviations (1).

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_corr_1.png" width="800">

(Slide-27)

# **Average product of two sequences**

Given two sequences $X_i$ and $Y_i$, the 'correlation' is the average product of corresponding values:

$\begin{eqnarray}
C_{XY} =  \frac{1}{n} \sum_i \, X_i  \, Y_i
\end{eqnarray}$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_corr_2a.png" width="800">

(Slide-28)



# **Pearson's correlation**

In statistics, this average product is known as Pearson's correlation

$\begin{eqnarray}
C_{XY} =  \frac{1}{n} \sum_{i=1}^n \, X_i  \, Y_i , \qquad\qquad -1\leq C_{XY} \leq 1
\end{eqnarray}$

Its value lies between $-1$ and $+1$ (because of our zero-mean and unit-variance condition). In our example, we find a positive correlation of $C=0.69$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec10_corr_2b.png" width="800">

Correlation quantifies how 'similar' or 'anti-similar', 'matched' or 'anti-matched', the sequences $X_i$ and $Y_i$ are.

The familiar formula for Pearson's correlation coefficient takes into account that the zero-mean and unit-variance **may not hold**:

$\begin{eqnarray}
C_{XY} =  \frac{\frac{1}{n} \sum_{i=1}^n \, \left(X_i - \bar X\right)  \, \left(Y_i- \bar Y \right)}{\sqrt{\frac{1}{n} \sum_{i=1}^n \, \left(X_i - \bar X\right)^2} \sqrt{\frac{1}{n} \sum_{i=1}^n \, \left(Y_i - \bar Y\right)^2}} , \qquad\qquad -1\leq C_{XY} \leq 1
\end{eqnarray}$

where $\bar X$ and $\bar Y$ are the of average $X_i$ and $Y_i$, respectively.



(Slide-29)

\end{eqnarray}$

# **Filtering property of Dirac's delta function**


The  "filtering property" is an important feature of a delta function.  Basically, the integral over the product $f(t) \cdot \delta(t)$ disappears for any continuous $f(t)$:

$\begin{eqnarray}
\int \delta(t-t_0) \, f(t) \, dt &=&
 \lim_{\epsilon \rightarrow 0} \int_{t_0-\epsilon}^{t_0+\epsilon} \delta(t-t_0) \, f(t) \, dx  =
\\
\\
&=&  f(t_0) \, \lim_{\epsilon \rightarrow 0} \, \int_{t_0-\epsilon}^{t_0+\epsilon} \delta(t-t_0)\, dt =
\\
\\
&=& f(t_0)
\end{eqnarray}$

This disappearing integral is often handy, for example when computing cross-correlations.

(Slide-34)

# **Cross-correlation with Dirac's delta function**

Cross-correlating the delta function with any function $X(t)$ yields the other function:

$\begin{eqnarray}
C_{X\delta}(\tau) = \int_{-\infty}^{+\infty} \delta(t-\tau) \, f(t) \, dt = f(\tau)
\end{eqnarray}$

(Slide-35)


# **Definition of convolution, aka reverse correlation**

The convolution of two integrable function $f(t)$ and $g(t)$ is defined as

$\begin{eqnarray}
C(\tau) = \int_{-\infty}^{+\infty} f(t) \, g(\tau-t) \, dt
\end{eqnarray}$

where $t$ is a dummy variable for integration and $\tau$ is the time-shift for which the convolution integral is evaluated. Note that for $t'=\tau-t$, we can write equivalently

$\begin{eqnarray}
C(\tau) = \int_{+\infty}^{-\infty} f(\tau - t') \, g(t') \, (-dt') = \int_{-\infty}^{+\infty} f(\tau - t') \, g(t') \, dt'
\end{eqnarray}$

Intuitively, a convolution measures the 'overlap' between $f(t)$ and a shifted and flipped version of $g(t)$. As $g(t)$ slides over $f(t)$, the convolution measures the 'weighed overlap'.

Because one function is flipped in time, we may also speak of a *reverse* correlation.

# **Relevance of correlation and convolution**


- The cross-correlation of time-varying stimuli \& responses describes their correlation as a function of time-shift.

- It measures the effect of earlier stimuli on later responses.

- A convolution is the cross-correlation of two integrable functions $f(t)$ and $g(t)$ with time-shift $\tau$, after one function has been flipped. It maps these functions onto a third function $C(\tau)$, where $\tau$ is the time-shift.

- It measures how well (or poorly) the two functions match, as a function of time-shift.

- Linear-filter models of neural responses are based on convolutions.

(Slide-36)

# **Optimal linear filter**

A linear-filter model provides an estimate $r_{est}(t)$ of a neuron's actual response $r(t)$.

$\begin{eqnarray}
r_{est}(t) = r_0 + \int_0^\infty \, s(t-\tau) \, D(\tau)  \, d\tau
\end{eqnarray}$

The instantaneous error $E(t)$ and the mean square error $MSE$ of this estimate are:

$\begin{eqnarray}
E(t) &=  r_{est}(t) - r(t)
\\
\\
MSE &= \frac{1}{T} \, \int \, E^2(t) \, dt
\end{eqnarray}$

Which filter $D(\tau)$ is **optimal** in the sense that it minimizes the $MSE$?

(Slide-40)


# **Integral equation for optimal linear filter**


The optimal filter $D(\tau)$ satisfies ('functional derivative')

$\begin{eqnarray}
\frac{\delta MSE}{\delta D} = 0
\end{eqnarray}$

This leads to an integral equation (D\&A, Chapter 2, Appendix A) with the stimulus-response correlation $Q_{rs}$ and the stimulus autocorrelation $Q_{ss}$:

$\begin{eqnarray}
\int_0^\infty \, Q_{ss}(\tau-\tau') \, D(\tau') \, d\tau = Q_{rs}(-\tau)
\end{eqnarray}$

where

$\begin{eqnarray}
Q_{rs}(\tau) &=& \frac{1}{T} \, \int \, r(t) \, s(t+\tau) \, dt
\\
Q_{ss}(\tau) &=& \frac{1}{T} \, \int \, s(t) \, s(t+\tau) \, dt
\end{eqnarray}$

Note that $Q_{rs}(-\tau)$ is evaluated at $-\tau$, i.e., for stimuli preceding the response.


# **Solve integral equation for white noise stimulation**

The integral equation is easy to solve for white noise stimuli:

$\begin{eqnarray}
Q_{ss}(\tau) = \sigma_s^2 \, \delta(\tau)
\end{eqnarray}$

simplifies the integral equation

$\begin{eqnarray}
\int_0^\infty \, Q_{ss}(\tau-\tau') \, D(\tau') \, d\tau' &=& Q_{rs}(-\tau)
\\
\sigma^2_s \int_0^\infty \, \delta(\tau-\tau') \, D(\tau') \, d\tau' &=& Q_{rs}(-\tau)
\\
\\
\sigma^2_s\,D(\tau) &=& Q_{rs}(-\tau)
\\
\\
D(\tau) &=& \frac{Q_{rs}(-\tau)}{\sigma^2_s}
\end{eqnarray}$

Thus, we find that the *optimal linear filter*  is the  *reverse correlation* with a white noise stimulus!

(Slide-41)

# **Optimal kernel and preferred stimulus**


- Reverse correlation between response and preceding stimulus yields *response-weighted average stimulus*.

\end{eqnarray}$

# **Summary linear filter model (LFM)**


- A linear-filter model estimates the response to a stimulus $s(t)$ as its convolution with a {\it linear kernel} $D(\tau)$:

$\begin{eqnarray}
\qquad\qquad\qquad r_{est}(t) = r_0 + \int_0^\infty \,  s(t-\tau) \, D(\tau) \, d\tau
\end{eqnarray}$

- A LFM predicts a maximally positive response when the stimulus matches the time-reversed linear kernel.

- A LFM predicts a maximally negative response when the stimulus matches the time-reversed and {\it inverted} linear kernel.

- The optimal linear kernel is the reverse correlation between response and white noise stimulus.

- The preferred stimulus matches the optimal linear kernel.

- Theoretical justification for reverse correlation methods!

(Slide-44)


# **5. A non-linear part of linear-filter models**

Linear-filter models of neuronal responses have evident problems:

- LFM predicts both positive and negative responses.   Neuronal responses are only positive.

- LFM predicts that response amplitude grows linearly with stimulus intensity/amplitude.  Neuronal responses saturate.

Both problems are easy to fix!

# **Static non-linearity**

We introduce a non-linear function $F()$ to map the linear response $L(t)$  onto a predicted response $r_{est}(t)$:

$\begin{eqnarray}
r_{est}(t) = F\left[ L(t) \right] \qquad {\rm where} \qquad L(t) = \int \, s(t-\tau) \, D(\tau)\,  d\tau \\
\end{eqnarray}$

The non-linear function $F()$ is called  a **static non-linearity** because it is evaluated instantaneously at each time $t$, with no explicit dependence on time.  Thus $F()$ is **'memoryless'**.

Pattern matching between $s(t)$ and $D(\tau)$ remains *linear* !

Scaling of response with stimulus amplitude is *not linear* !

(Slide-45)

# **Suitable functions**

Some functions that can be used as a static-nonlinearity are listed below:

$\begin{eqnarray}
F(L) &=& k \, \left[  L \right]_+ \qquad\qquad\qquad\qquad\qquad\qquad {\rm rectifying}
\\
F(L) &=& k \, \left[  L - \Theta \right]_+ \qquad\qquad\qquad\qquad\quad {\rm rectifying+threshold}
\\
\\
F(L) &=& \frac{R_{max}}{1 + \exp\left[-k (L - \Theta) \right]}  \qquad\qquad {\rm logistic}
\\
\\
F(L) &=& R_{max} \, \left[ \tanh\left( k(L-\Theta)\right)  \right]_+ \qquad\qquad {\rm tanh+threshold}
\\
\\
F(L) &=& R_{max} \,\frac{k\, \left[L\right]_+^2}{\Theta^2\,+\,k\, \left[L\right]_+^2}\qquad\qquad\qquad\qquad {\rm Naka-Rushton}
\end{eqnarray}$

```python
from inspect import signature
# @title Static non-linearities {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def RectifyingFunc(Li, Param):
    kappa = Param[0];

    yi = kappa * Li;
    yi[yi < 0] = 0;

    return yi

def ThresholdFunc(Li, Param):
    kappa = Param[0];
    theta = Param[1];

    yi = kappa * (Li - theta);
    yi[yi < 0] = 0;

    return yi

def LogisticFunc(Li, Param):
    kappa = Param[0];
    theta = Param[1];
    Rmax  = Param[2];

    yi = Rmax / (1 + np.exp(-kappa*(Li-theta)));

    return yi

def TanhFunc(Li, Param):
    kappa = Param[0];
    theta = Param[1];
    Rmax  = Param[2];

    yi = Rmax * np.tanh(kappa*(Li-theta));

    yi[yi < 0] = 0;

    return yi

def NakaRushtonFunc(Li, Param):
    kappa  = Param[0];
    theta  = Param[1];
    Rmax   = Param[2];

    yi = Rmax * kappa * Li**2 / ( theta**2 + kappa * Li**2 );

    yi[Li<0] = 0;

    return yi

def plot_static_nonlinearities( Kappa=1.0, Theta=1.0, Rmax = 10 ):
    """(Slide-48)"""

    Li = np.arange(-20, 101) / 10

    # compute functions
    Params = (Kappa, Theta, Rmax);
    RecFunc = RectifyingFunc(Li, Params);
    ThrFunc = ThresholdFunc(Li, Params);
    LogFunc = LogisticFunc(Li, Params);
    TanFunc = TanhFunc(Li, Params);
    NakaFunc = NakaRushtonFunc(Li, Params);


    # Plot results
    fig, axs = plt.subplots( 1, 2, figsize=(8, 4))

    # Plot the first two functions
    axs[0].plot(Li, RecFunc, 'k.', markersize=1, label='rectifying')
    axs[0].plot(Li, ThrFunc, 'r.', markersize=1, label='threshold')
    axs[0].set_xlabel('linear')
    axs[0].set_ylabel('non-linear')
    axs[0].set_xlim([-2, 10])
    axs[0].set_box_aspect(1/1)
    axs[0].set_title('Static non-linearities', fontsize=14);
    axs[0].legend();


    # Plot the remaining three functions
    axs[1].plot(Li, LogFunc, 'b.', markersize=1, label='logistic')
    axs[1].plot(Li, TanFunc, 'c.', markersize=1, label='tanh')
    axs[1].plot(Li, NakaFunc, 'm.', markersize=1, label='naka')
    axs[1].set_xlabel('linear')
    axs[1].set_ylabel('non-linear')
    axs[1].set_xlim([-2, 10])
    axs[1].set_box_aspect(1/1)
    axs[1].set_title('Static non-linearities', fontsize=14);
    axs[1].legend();

    plt.tight_layout()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_static_nonlinearities, Kappa = (0.2,5.0,0.2), Theta=(-10, 10, 1), Rmax=(1, 10, 1) )


```

- The static non-linearity converts the linear convolution result (positive or negative, non-saturating) into a neuronal response (positive and saturating).

- LFMs capture many aspects of the response properties of sensory neurons  (e.g., ``simple cells'' in visual cortex).

- Extending LFMs to include multiple linear filters captures the response properties of even more sensory neurons  (e.g., ``complex cells'' in visual cortex).

- Further extensions are needed to model responses to all stimuli (including natural stimuli).

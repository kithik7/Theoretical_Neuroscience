**Abstract:**

Being alive is being unbalanced, restless, and active. Being dead is the opposite. Living cells maintain multiple imbalances between their insides and outsides. This lecture discusses the most important mechanisms contributing specifically to ionic imbalances: **cell membrane**, **ionic concentration gradients**, and **ion channels**. Ionic imbalances are central to neuronal function because they can generate and modulate electric potential differences. We introduce the **Nernst equation** for the **reversal potential** of an ion, estimate the **resting potential** of a neuron with multiple ionic imbalances, and estimate the **number of ions** crossing the membrane during a potential change.

# **3. Intra- and extracellular ion concentrations**

Extracellular medium resembles sea water (high $Na^+$, low $K^+$, high $Cl^-$).

Intracellular medium reverses these concentrations (low  $Na^+$, high $K^+$, low $Cl^-$).

Active ion pumps maintain the difference, expending metabolic energy.

# Active ion pumps

Concentrations gradients are maintained by active (energy-consuming) ion pumps. The most common is the $Na^+$-$K^+$ exchange pump. One ATP is used to move one ion of each type. This is *charge neutral* and thus does not affect the membrane potential.  Other pumps are *electrogenic* in that they exchange 3 $N^+$ for 2 $K^+$.  In most neurons, the resulting pump current makes only a small difference to the membrane potential.

# Semipermeable membranes

Thanks to ion-selective channel proteins, cell membranes are semi-permeable. Typically, cell membranes are most permeable to potassium ions, $K^+$. Note that 'typically' does NOT mean 'at all times'!

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec01_Fig12.png" width="500">

In combination with a concentration gradient, semi-permeable membranes result in an electric potential difference: *higher* concentration of $K^+$ (inside) results in a *negative* potential (inside), relative to outside.

(Slide-19)

# Electrochemical equilibrium

Both concentration gradients and electrostatic potential influence the flow of ions. Flow is defined as number of ions per unit time. A balance is reached when both influences cancel, so that there is *no net* ion flow.

# Reversal potential and Nernst equation

The **Nernst equation** defines the *electrochemical equilibrium* between a concentration gradient and an electrical potential, for one particular ion.  

The **reversal potential** is the electrostatic potential that exactly balances a concentration gradient, and *vice versa*.


The Nernst equation links reversal potential and concentration gradient:
$$\begin{eqnarray}
E_{rev} = \frac{RT}{zF} \ln \frac{[X]_{out}}{[X]_{in}}
\end{eqnarray}$$


$$\begin{eqnarray}
z && {\rm positive \,\, or \,\, negative \,\, elementary \,\, charge}
\\
R &=& 8.31 \, {\rm Joule / mol}\, K^\circ \quad {\rm  gas \,\, constant}
\\
T &\approx& 310 \, K \approx 37^\circ C \quad {\rm  body \,\, temperature}
\\
T &\approx& 283 \, K \approx 10^\circ C \quad {\rm  ocean \,\, temperature}
\\
F &=& 96480 \, {\rm Coulomb / Mole}\\
\end{eqnarray}$$

(Slide-25)

# Working out the units

Gas constant and body temperature
$$\begin{eqnarray}
RT &\approx 2576 \, {\rm Joule / Mole} \approx 0.6 {\rm kCal / Mole}
\end{eqnarray}$$

Energy corresponds to both mechanical work and electrical work:
$$\begin{eqnarray}
2576 \, {\rm Joule / Mole} = 2576 \, {\rm Newton \cdot m / Mole} = 2576 \, {\rm Coulomb \cdot Volt / Mole}
\end{eqnarray}$$

Here, units of electrical work are more useful:
$$\begin{eqnarray}
\frac{RT}{F} &\approx 0.027 \, \frac{\rm Joule}{\rm Coulomb} = 0.027 \, \frac{\rm Coulomb \cdot Volt}{\rm Coulomb} = 27 \, {\rm mV}
\end{eqnarray}$$

The Nernst equation links electrostatic potential and concentration gradient:
$$\begin{eqnarray}
E_{rev} = \frac{27\,{\rm mV}}{z} \ln \frac{[X]_{out}}{[X]_{in}} \qquad \mathrm{at} \quad 37^\circ C
\end{eqnarray}$$
$$\begin{eqnarray}
E_{rev} = \frac{24\,{\rm mV}}{z} \ln \frac{[X]_{out}}{[X]_{in}} \qquad \mathrm{at} \quad 10^\circ C
\end{eqnarray}$$

(Slide-26)

# Electrochemical balance of $K^+$ (potassium / Kalium)

Potassium concentrations are $[K^+]_\mathit{in}=400{\rm mM}$ and $[K^+]_\mathit{out}=20{\rm mM}$ in the squid giant axon, resulting in a net potassium **outflow** when the potential difference is zero. The abbreviation $\rm mM$ stands for ${\rm milli Molar} = 10^{-3} {\rm Mole} / {\rm liter}$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec01_Fig17.png" width="300">


This net outflow vanishes for a potential difference of
$$\begin{eqnarray}
E_{K^+} = \frac{24\,{\rm mV}}{+1} \ln \frac{20 {\rm mM}}{400 {\rm mM}}= -72 {\rm mV}\qquad \mathrm{at} \quad 10^\circ C
\end{eqnarray}$$
This is the **reversal potential** of $K^+$.

(Slide-27)

# Electrochemical balance of $Na^+$ (sodium / Natrium)

Sodium concentrations are $[Na^+]_\mathit{in}=50{\rm mM}$ and $[Na^+]_\mathit{out}=440{\rm mM}$ in the squid giant axon, resulting in a net sodium **inflow** when the potential difference is zero.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec01_Fig19.png" width="300">


This net inflow vanishes for a potential difference of
$$\begin{eqnarray}
E_{Na^+} = \frac{24\,{\rm mV}}{+1} \ln \frac{440 {\rm mM}}{50 {\rm mM}}= +52 {\rm mV}\qquad \mathrm{at} \quad 10^\circ C
\end{eqnarray}$$
This is the **reversal potential** of $Na^+$.

(Slide-28)

# Electrochemical balance of $Cl^-$ (chlorine / Chlor) and $Ca^{2+}$ (calcium / Calzium)

Chlorine concentrations are $[Cl^-]_\mathit{in}=40{\rm mM}$ and $[Cl^-]_\mathit{out}=450{\rm mM}$ in the squid giant axon, resulting in a net chlorine **inflow** when the potential difference is zero. This net inflow vanishes for a potential difference of
$$\begin{eqnarray}
E_{Cl^-} = \frac{24\,{\rm mV}}{-1} \ln \frac{450 {\rm mM}}{40 {\rm mM}}= -58 {\rm mV}\qquad \mathrm{at} \quad 10^\circ C
\end{eqnarray}$$
This is the **reversal potential** of $Cl^-$.

Calcium concentrations are $[Ca^{2+}]_\mathit{in}=2{\rm mM}$ and $[Ca^{2+}]_\mathit{out}=0.0002{\rm mM}=0.2\mu{\rm M}$ in the squid giant axon, resulting in a net calcium **inflow** when the potential difference is zero. This net inflow vanishes for a potential difference of
$$\begin{eqnarray}
E_{Ca^{2+}} = \frac{24\,{\rm mV}}{+2} \ln \frac{2 {\rm mM}}{0.2 \mu{\rm M}}= +111 {\rm mV}\qquad \mathrm{at} \quad 10^\circ C
\end{eqnarray}$$
This is the **reversal potential** of $Ca^{2+}$.

```python
# @title Reversal potential {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# See how the reversal potential changes with ion concentrations (in and out) and ion charge !

# Define an interactive plot
def plot_current_and_potential(Xout=400, Xin=50, zX = 1):
    """ Slide-30 """
    #Xout = 400                                    # [mM] outside concentration
    #Xin = 50                                      # [mM] inside concentration
    #zX  = 1                                       # [1] ion charge
    RT = 24                                        # [mV] gas constant times temperature (10 deg C)
    EX = RT * np.log( Xout / Xin ) / zX            # reversal potential from Nernst equation
    Rmem = 100                                     # [mOhm] membrane resistance (times square centimeter)
    Emem = np.linspace(-100, +100, 200)            # [mV] sample range of possible membrane potentials
    Imem = ( Emem - EX ) / Rmem                    # [nA = mV / MOhm] membrane outward current from Ohm's Law
    plt.plot(Emem, Imem, label='current')
    plt.plot([-100,+100], [0,0], label='no net flow')
    plt.plot([EX,EX], [min(Imem),max(Imem)], label='reversal potential')
    plt.xlabel('Inside potential E_in')
    plt.ylabel('Net outward current [nA]')
    plt.title('Membrane current and potential')
    plt.legend()
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_current_and_potential, Xout=(1, 500, 1), Xin=(1, 500.0, 1), zX=(-2, +2, 1))
```

# **6. Resting potential of neuron**

Different ionic species have different concentration gradients and different resting potentials. The actual membrane potential reflects the difference of charge (in and out), independent of the ionic species, at any particular moment.

The **resting potential** is defined by 'no net flux of charge'.  In other words, the net fluxes of different ionic species cancel in terms of charge.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec01_Fig14.png" width="200">

We will use the Nernst equation to estimate the resting potential.  

Textbooks typically use the the Goldman-Hodgkin-Katz equation (see further below), which is more precise but uses different physical assumptions.

(Slide-34)

# Electrochemical driving force

If the membrane potential is not at equilibrium with the concentration gradient of an ionic species $X$, the difference provides an **electrochemical driving force** for this ionic species:

$$\begin{eqnarray}
\Delta V_X = V_{mem} - E_{rev} = V_{mem} -  \frac{27 mV}{z} \,\, \ln \frac{[X]_{out}}{[X]_{in}}
\end{eqnarray}$$

This electrochemical driving force is responsible for a net ion flux, which vanishes only at equilibrium (i.e., at the reversal potential $V_{mem} = E_{rev}$).

(Slide-35)

# Net flux of one ion

The net flux of one ion may be computed as the product of *driving force* and *ionic conductance*, i.e., the conductance of the membrane with its embedded ion channels for the particular ion in question:
$$\begin{eqnarray}
I_X [{\rm n Ampere}] = g_X [\mu {\rm Siemens}] \cdot \Delta V [m{\rm Volt}]
\end{eqnarray}$$

(Slide-36)

# Total flux of all ions

The total ion flux (current) across the membrane is the sum of all individual ion currents. Here we define the total membrane current as an **outward current** of positive charge (positive when the net driving force is positive, i.e., larger inside than outside):

$$\begin{eqnarray}
I_{total} &=& I_{Na^+} + I_{K^+} + I_{Cl^-} =
\\
\\
&=& g_{Na^+} \, \left(V - E_{Na^+} \right) + g_{K^+} \, \left(V - E_{K^+} \right)  + g_{Cl^-} \, \left(V - E_{Cl^-} \right)
\end{eqnarray}$$

(Slide-37)

# Resting potential

The **resting potential** $V=V_{rest}$ is reached when the total membrane current is zero:
$$\begin{eqnarray}
0 = g_{Na^+} \, \left(V_{rest} - E_{Na^+} \right) + g_{K^+} \, \left(V_{rest} - E_{K^+} \right)  + g_{Cl^-} \, \left(V_{rest} - E_{Cl^-} \right)
\end{eqnarray}$$
$$\begin{eqnarray}
\Leftrightarrow
\end{eqnarray}$$
$$\begin{eqnarray}
V_{rest} = \frac{g_{Na^+} \,  E_{Na^+}  + g_{K^+} \, E_{K^+}  + g_{Cl^-} \,  E_{Cl^-} }{g_{Na^+} + g_{K^+}  + g_{Cl^-} }
\end{eqnarray}$$
Thus, the resting potential of the neuron is a weighted average of the reversal potential of all ionic species.  The largest contribution comes from the most permeable ion (largest conductance)!

This is analogous to elections: elections are not decided by the number of party members (ion concentrations in the cell), but by the number of votes (ions passing through the membrane).

(Slide-38)

# Example calculation

Let's calculate the resting potential in a concrete case:
$$\begin{eqnarray}
E_{Na^+} =  27 {\rm mV} \ln \frac{440 {\rm mM}}{50 {\rm mM}} = +59 {\rm mV}
\end{eqnarray}$$
$$\begin{eqnarray}
E_{K^+} =  27 {\rm mV} \ln \frac{20 {\rm mM}}{400 {\rm mM}} =
-81 {\rm mV}
\end{eqnarray}$$
$$\begin{eqnarray}
E_{Cl^-} =  -27 {\rm mV} \ln \frac{450 {\rm mM}}{40 {\rm mM}} =
-65 {\rm mV}
\end{eqnarray}$$

$$\begin{eqnarray}
V_{rest} &=& \frac{g_{Na^+} \,  E_{Na^+}  + g_{K^+} \, E_{K^+}  + g_{Cl^-}\,  E_{Cl^-} }{g_{Na^+} + g_{K^+}  + g_{Cl^-} }=\frac{\frac{g_{Na^+}}{g_{K^+}} \,  E_{Na^+}  +  E_{K^+}  + \frac{g_{Cl^-}}{g_{K^+}} \,  E_{Cl^-} }{\frac{g_{Na^+}}{g_{K^+} } + 1  + \frac{g_{Cl^-}}{g_{K^+}} }
\\
\\
\frac{g_{Na^+}}{g_{K^+} } &\approx& 0.04 \qquad\qquad \qquad\qquad
\frac{g_{Cl^-}}{g_{K^+}}  \approx 0.45
\\
\\
V_{rest} &=& \frac{0.04 \cdot  59 mV  - 1 \cdot 81 mV  - 0.45 \cdot 65 mV }{0.04 + 1 + 0.45 } = -76mV
\end{eqnarray}$$

The **resting potential** of a neuron is a compromise between the **reversal potentials** of all ions to which the membrane is permeable. The influence of each ion in this compromise depends on its permeability. More permeable ions exert more influence than less permeable ones.

The relative permeabilities are from Table 7 of  Hodgkin and Katz (1949) J. Physiol. 108, 37-77, and were measured in the squid.

(Slide-39)


# Goldman-Hodgkin-Katz equation

In textbooks, the resting potential is typically calculated with the Goldman-Hodgkin-Katz equation.  

$$\begin{eqnarray}
V_{rest} = 27 mV \,\, \ln \frac{P_K\,[K^+]_{out} + P_{Na} \, [Na^+]_{out} + P_{Cl}\,[Cl^-]_{in}}{P_K\,[K^+]_{in} + P_{Na} \, [Na^+]_{in} + P_{Cl}\,[Cl^-]_{out}}
\end{eqnarray}$$

This equation models the permeability of membranes to ions more accurately.

For our example calculation, this equation produces
$$\begin{eqnarray}
V_{rest} = 27 mV \,\, \ln \frac{[K^+]_{out} + b \, [Na^+]_{out} + c \, [Cl^-]_{in}}{[K^+]_{in} + b \, [Na^+]_{in} + c \, [Cl^-]_{out}} \\
\end{eqnarray}$$
$$\begin{eqnarray}
V_{rest} = 27 mV \,\, \ln \frac{20 + 0.04 \cdot 440 + 0.45 \cdot 40}{400 + 0.04 \cdot 50 + 0.45 \cdot 450} = -63.4 \, mV\\
\end{eqnarray}$$
where we have used the same *relative permeabilities* as before (relative to $K^+$)
$$\begin{eqnarray}
P_K = 1 \qquad\qquad
\frac{P_{Na}}{P_K} = b \approx  0.04
\qquad\qquad
\frac{P_{Cl}}{P_K} = c \approx  0.45
\end{eqnarray}$$
Our earlier calculation approximated permeabilities $P$ with conductances $g$, which was less accurate.

(Slide-40)

# Inferring relative permeability

If you know the concentration gradients and the resting potential, you can infer the relative permeability of two ionic species. As an example, we consider a neuron dominated by sodium and potassium permeability.

Solving the GHK equation for the permeability ratio $P_{Na} / P_{K}$:
$$\begin{eqnarray}
V_{rest} = 27 {\rm mV} \ln \frac{P_K\,[K^+]_{out} + P_{Na} \, [Na^+]_{out}}{P_K\,[K^+]_{in} + P_{Na} \, [Na^+]_{in}}
\end{eqnarray}$$
produces
$$\begin{eqnarray}
\frac{P_{Na}}{P_K} = \frac{-[K^+]_{out} + [K^+]_{in} \, e^{E_{rest}/27 {\rm mV}}}{[Na^+]_{out} - [Na^+]_{in} \, e^{E_{rest}/27 {\rm mV}}}
\end{eqnarray}$$

Given a resting potential
$$E_{rest}=-77{\rm mV}$$

and ion concentrations
$$[K^+]_{out}=5 {\rm mM},\quad [K^+]_{in}=148{\rm mM}, \quad [Na^+]_{out}=142 {\rm mM}, \quad [Na^+]_{in}=10{\rm mM}$$
you can estimate
$$\begin{eqnarray}
\frac{P_{Na}}{P_K} = \frac{-5 + 148 \, e^{-2.85}}{142 - 10 \, e^{-2.85}} \approx  0.025
\end{eqnarray}$$

(Slide-41)


```python
# @title Resting potential {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# See how relative permeabilities change the resting potential !

# Define an interactive plot
def plot_permeability_and_resting_potential(PNa=0.03, PCl=0.1, PK=1.0):
    """ Slide-42 """
    Naout = 440                                    # [mM] outside Na concentration
    Nain = 50                                      # [mM] inside Na concentration
    Clout = 450                                    # [mM] outside Cl concentration
    Clin = 40                                      # [mM] inside Cl concentration
    Kout = 20                                     # [mM] outside K concentration
    Kin = 400                                       # [mM] inside K concentration
    RT = 24                                        # [mV] gas constant times temperature (10 deg C)
    ENa = RT * np.log( Naout / Nain ) / (+1)       # reversal potential Na from Nernst equation
    ECl = RT * np.log( Clout / Clin ) / (-1)       # reversal potential Cl from Nernst equation
    EK  = RT * np.log( Kout / Kin ) / (+1)         # reversal potential K from Nernst equation
    Vrest = RT * np.log((PNa * Naout + PCl * Clin + PK * Kout) / (PNa * Nain + PCl * Clout + PK * Kin))
    plt.plot([0,1], [Vrest,Vrest], label='resting potential')
    plt.plot([0,1], [ENa,ENa], label='Na reversal potential')
    plt.plot([0,1], [ECl,ECl], label='Cl reversal potential')
    plt.plot([0,1], [EK,EK], label='K reversal potential')
    plt.xlabel('Inside potential')
    plt.title('Permeability and resting potential')
    plt.legend()
    plt.show()

# Creating sliders for relative permeabilities
interact(plot_permeability_and_resting_potential, PNa=(0.01, 1, 0.01), PCl=(0.01, 1, 0.01), PK=(0.01,1,0.01))
```

# **Interim summary**

*  Neurons have **concentration gradients** \& **semi-permeable membranes**, with steady-states maintained by **ion pumps**.
*  **Reversal potential** of one ion: membrane potential at which this ion is at electrochemical equilibrium (net zero flux).
*  **Resting potential** of neuron: membrane potential at which all ion fluxes cancel (net zero flux of charge).
*  Typically, the resting potential of a neuron is approximately $-70 {\rm mV}$ and dominated by $K^+$, for which the resting membrane is most permeable.
*  The **membrane potential** is the actual electrical potential (inside relative to outside). Neurons can can and do rapidly change their membrane potentials (as discussed in the next lectures).

# **7. How many ions cross the membrane? (advanced)**

What fraction of the ions inside a cell must cross the membrane in order to change its potential?   To obtain an order-of-magnitude answer to this question,  we proceed in three steps:

* Estimate the number of ions in a typical cell.
* Estimate the number of ions that cross the membrane to effect a typical potential change.
* Estimate how many times the cell can effect such a change before running out of ions.

(Slide-45)


# Surface and volume of typical cells

We consider two cells with surface areas of $0.01\,{\rm mm}^2$ and  $0.001\,{\rm mm}^2$, respectively.

Assuming their soma to be spherical, we can compute their radii to be

$$\begin{eqnarray}
A = 4\pi r^2  \qquad\qquad \Leftrightarrow \qquad\qquad r = \sqrt{\frac{A}{4\pi}}
\end{eqnarray}$$

as  $r_{large} = 0.03 mm = 30 \mu m$ and  $r_{small}= 0.009 mm = 9 \mu m$,  and their volumes

$$\begin{eqnarray}
V = \frac{4}{3}\pi r^3
\end{eqnarray}$$

as $V_{large} = 1.13\times10^{-4} mm^3$ and $V_{small} = 3\times10^{-6} mm^3$.

(Slide-46)

# Number of $K^+$ ions in typical cells

At a concentration of $400 mM$ ---- or $0.4 \, {\rm Mole} / {\rm liter}$, which corresponds to $0.4\times 10^{-6} \, {\rm Mole} / {\rm mm}^3$ ---- these volumes contain approximately

* $N_{large}^{vol} = 4.5\times10^{-11} {\rm Mole}$ $K^+$

and

* $N_{small}^{vol} = 1.2\times10^{-12} {\rm Mole}$ $K^+$

ions, respectively.

If these amounts do not seem like much, recall Avogadro's number,

* $1\,{\rm Mole}$ contains approximately $N_A = 6.023 \times 10^{23}$ ions

so that

* $N_{large}^{vol} = 4.5\times10^{-11} \times 6.023 \times 10^{23} \approx 27,200 \times 10^{9}K^+$ ions

and

* $N_{small}^{vol} = 1.2\times10^{-12} \times 6.023 \times 10^{23} \approx 735 \times 10^{9}K^+$ ions

(Slide-47)

# Specific membrane capacitance

When (positive or negative) ions cross the membrane, this creates an electrostatic potential difference.  The more ions cross, the larger the potential. This proportionality between charge and potential is called the
**capacitor equation**

$$\begin{eqnarray}
Q[{\rm Coulomb}] = a[{\rm cm}^2] \cdot C[{\rm Farad/cm^2}] \cdot V[{\rm Volt}]
\end{eqnarray}$$

where $Q$ is electrostatic charge (unit 'Coulomb'), $a$ is the surface area, and $V$ (unit 'Volt') is the electrostatic potential.

The proportionality constant is called the **capacitance** $C$ (unit 'Farad', defined as $1F \equiv \frac{1 {\rm Coulomb}}{1 {\rm Volt}}$).

In our case, the proportionality constant is **specific membrane capacitance**, $C_{mem}$ , which takes into account also membrane area. A typical value is
$$\begin{eqnarray}
C_{mem}=1 \, \mu {\rm F}/{\rm cm}^2 = 1 \cdot 10^{3} \cdot \, n{\rm F} / 10^2 \cdot {\rm mm}^2 = 10 \, n{\rm F}/{\rm mm}^2
\end{eqnarray}$$   

(Slide-48)

# Number of $K^+$ ions crossing the membrane

In neurons a typical change of membrane potential ('action potential') is $70{\rm mV}$. The capacitor equation tells us how much charge $Q$ must cross the membrane for such a change.

To calculate how many ions mus cross the membrane, we can use the Faraday constant  $F=96.5 \times 10^{3} \, {\rm Coulomb / Mole}$ to convert charge into Mole, and Avogadro's number $N_A = 6.023 \times 10^{23} / {\rm Mole}$ to convert Mole into number of ions.

$$\begin{eqnarray}
N^{flow} = \frac{a \cdot C_{mem} \cdot V \cdot N_A}{F}
\end{eqnarray}$$

Putting in the numbers gives
$$\begin{eqnarray}
N^{flow}_{large} = \frac{10^{-4} {\rm cm}^2 \cdot 1\times 10^{-6} {\rm Farad}/{\rm cm}^2 \cdot 7\times 10^{-2} {\rm V} \cdot 6.023 \times 10^{23} / {\rm Mole} }{ 96.5 \times 10^3 {\rm Coulomb} / {\rm Mole}} \approx
43.6 \times 10^{6} \, {\rm ions}
\end{eqnarray}$$

$$\begin{eqnarray}
N^{flow}_{small} = \frac{10^{-5} {\rm cm}^2 \cdot 1\times 10^{-6} {\rm Farad}/{\rm cm}^2 \cdot 7\times 10^{-2} {\rm V} \cdot 6.023 \times 10^{23} / {\rm Mole} }{ 96.5 \times 10^3 {\rm Coulomb} / {\rm Mole}} \approx
4.36 \times 10^{6} \, {\rm ions}
\end{eqnarray}$$

(Slide-49)

# How often can the potential change?

Comparing the number of ions in the cell volume and the number that must cross the membrane to change the potential, we obtain how many times the potential can change:

$$\begin{eqnarray}
N^{change}_{large} = \frac{N_{large}^{vol}}{N_{large}^{flow}} = \frac{27,200 \times 10^9}{43.6\times 10^6} \approx 623,000
\end{eqnarray}$$

$$\begin{eqnarray}
N^{change}_{small} = \frac{N_{small}^{vol}}{N_{small}^{flow}} = \frac{735 \times 10^9}{4.36\times 10^6} \approx 169,000
\end{eqnarray}$$

Thus we reach a comforting conclusion: our neurons can change potential several hundred thousand times, without troubling ion pumps to restore the concentration gradients.

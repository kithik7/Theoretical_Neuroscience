# Expectation maximisation

In this exercise, you are provided with a set of observations that stem from multiple
causes/origins. Of course, you do not know which particular cause produced each observation. Your task is to identify the number and nature of these causes/origins. To
solve this task, you will first try ‘by hand’ and then apply the formalism of “expectation
maximisation” (EM).


EM uses a generative model, which predicts observations from estimated causes/origins.
The ‘expectation’ step assigns ‘cause probabilities’ to actual observations. The ‘maximisation’ step averages actual observations with ‘cause probability’ weights. Alternating
these steps incrementally improves the estimate of causes/origins.


EM is an iterative method which can be used to find the the maximum likelihood (or
maximum a posteriori) estimates of parameters (in our case, parameters describing the
Poisson distributions). It is extremely useful when we don’t have full knowledge of the
data, that is, some missing or latent variable. In our case, the missing variable is the
source of the data points (which cause produced this observation?). I highly recommend
watching some YouTube videos on this.

## Observations and causes

Consider a (large) set of scalar observations $\{u_{obs}\}$ (which will be supplied).  Assume that several distinct processes (causes $X$) independently contribute observations, with each $X$ supplying a fraction $\gamma_X$ of the total.  Assume further that each $X$ is Poisson in the sense that it contributes observations distributed with Poisson distribution of mean and variance $\lambda$ (consider $\lambda > 0$).

For example, if there are three distinct causes $A$, $B$, and $C$, the generative parameters would be

\\

$$
G = \{ \gamma_A, \lambda_A, \gamma_B,\lambda_B, \gamma_C, \lambda_C \}, \qquad\qquad \gamma_A+\gamma_B+\gamma_C = 1
$$

\\

with three generative distributions

\\

$$
p(u | A)  =\frac{\lambda_A ^{u} e^{-\lambda_A}}{u!} , \qquad \ldots
$$

For each observation $u_{obs}$ we seek the conditional probability of specific causes.   This amounts to inverting the generative distributions and to obtaining recognition distributions

\\

$$
P(A|u_{obs}), \qquad P(B|u_{obs}), \qquad P(C|u_{obs})
$$

\\

which must sum to unity.   According to Bayes rule, $P(X|u)$ is proportional to $p(u|X)$, so that

\\

$$
P(A| u_{obs} ) = \frac{\gamma_A \, p(u_{obs}|A)}{\gamma_A \, p(u_{obs}|A)+\gamma_B \, p(u_{obs}|B)+\gamma_C \, p(u_{obs}|C)}, \qquad \ldots
$$

\\

This is the 'expectation' step of the EM algorithm.

We are now in a position to separately evaluate the subset of observations contributed by each cause.  Specifically, we may use the $P(A|u_{obs}) \ldots$, to compute cause-probability weighted averages over all observations $u_{obs}$, to recompute the generative parameters

\\

$$
\gamma_A = \langle P(A|u_{obs}) \rangle, \ldots
\\
\mu_A = \frac{1}{\gamma_A} \, \langle u_{obs} \, P(A|u_{obs} )\rangle,  \ldots
\\
\lambda_A = \mu_A ,  \ldots
$$

\\

This is the 'maximisation' step of an EM algorithm.

If you have a sufficient number of observations, you can alternate 'expectation' and 'maximisation' steps to optimise the generative model (such as it is, given your assumptions).

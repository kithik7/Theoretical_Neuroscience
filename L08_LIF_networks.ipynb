# **1. Spike train statistics**

Neuronal activity consists of sequences of spikes, which are also called 'spike trains'. We will use spike trains to review some concepts from statistics:

- Mean, variance, standard deviation.

- Coefficient of variation and Fano factor.

- Probability distributions and probability densities.

- Linear hazard distribution, exponential distribution, and Cauchy distribution.


**Statistical distributions are useful tools and essential throughout the entire course (winter and summer).**

# Spike train

A **spike train** is a sequence of spikes in time




<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_Spike_train_a.png" width="500">

$\begin{eqnarray}
\{ t_{1}, t_{2}, t_{3}, \ldots, t_{N}\} \qquad\qquad\qquad t_{i} < t_{i+1}
\end{eqnarray}$

Spike trains tend to be irregular and must be described in statistical terms. They are many different kinds of irregularity (just like unhappiness)!

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_Spike_train_b.png" width="500">

First-order measures of firing are average spike rate $\lambda$, measured in $[Hz]$, average inter-spike-interval $\left\langle{t_\mathit{isi}}\right\rangle$, measured in $[s]$, or average spike count $n(\Delta t)$ in time intervals of size $\Delta t$ (not shown here, but see further below).

(Slide-2)

# Average spike rate and inter-spike-interval


Spike rate $\lambda$ is defined as the total number (total count) of spikes $N$ observed in an interval of duration $T$.

$\begin{eqnarray}
\qquad\qquad\lambda \equiv \frac{N}{T}
\end{eqnarray}$


Conversely, the average duration of inter-spike-intervals is defined as

$\begin{eqnarray}
\qquad\qquad \left\langle{t_\mathit{isi}}\right\rangle \equiv \frac{T}{N}
\end{eqnarray}$


From these definitions, it follows that $\lambda = \frac{1}{\left\langle{t_\mathit{isi}}\right\rangle}$

Neither spike rate nor inter-spike-interval (nor binned spike count) tell us anything about the degree of regularity or irregularity.  The illustration shows three spike trains with identical rates but different degrees of irregularity.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_Spike_train_c.png" width="500">

(Slide-3)


# Coefficient of variation, $C_V$

To quantify the degree of irregularity, we can compute the ratio of standard deviation to mean of the inter-spike-interval.

$\begin{eqnarray}
\left\langle{t_\mathit{isi}}\right\rangle = \frac{1}{N} \sum \, t_\mathit{isi}, \qquad \left\langle{t_\mathit{isi}^2}\right\rangle = \frac{1}{N} \sum \, t_\mathit{isi}^2,\qquad \sigma_\mathit{isi} = \sqrt{\left\langle{t_\mathit{isi}^2}\right\rangle-\left\langle{t_\mathit{isi}}^2\right\rangle}
\end{eqnarray}$

This is also called the **coefficient of variation**:

$\begin{eqnarray}
\qquad\qquad C_v \equiv \frac{\sigma_\mathit{isi}}{\left\langle{t_\mathit{isi}}\right\rangle} = \frac{\sqrt{\left\langle{t_\mathit{isi}^2}\right\rangle-\left\langle{t_\mathit{isi}}^2\right\rangle}}{\left\langle{t_\mathit{isi}}\right\rangle}
\end{eqnarray}$

(Slide-4)

# Binned spike count and Fano factor

Another approach is to count spikes in a **'sliding window'** of fixed width $\Delta t$ (e.g. $\Delta t = 50 ms$) and to establish the spike count $n_i$ for every window position $i$.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_Spike_train_e.png" width="300">

Given a list of counts $\{n_1, n_2, \ldots , n_K\}$ for $K$ window positions, one obtains the mean and variance of the **'binned count'**
$\begin{eqnarray}
\left\langle n \right\rangle = \frac{1}{K} \sum_{i=1}^K n_i, \qquad \left\langle \left( n - \left\langle n \right\rangle\right)^2\right\rangle = \left\langle n^2\right\rangle - \left\langle n \right\rangle^2 = \frac{1}{K} \sum_{i=1}^K n_i^2 - \left\langle n \right\rangle^2
\end{eqnarray}$

To quantify the degree of irregularity, we can use the ratio of variance to mean, which is also called the **Fano factor**:

$\begin{eqnarray}
\qquad\qquad \mathit{FF}(\Delta t) \equiv \frac{\left\langle n^2\right\rangle - \left\langle n \right\rangle^2}{\left\langle n\right\rangle}
\end{eqnarray}$

As the Fano factor depends on the chosen bin width $\Delta t$, it quantifies the degree of irregularity at a **particular time scale**.

(Slide-5)




# Distribution of inter-spike-intervals

An even more comprehensive characterization of irregularity is offered by the  **distribution** $P\left(t_{isi}\right)$ of inter-spike-intervals $t_{isi} = t_{i+1} - t_{i}$

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_Spike_train_b.png" width="300">

To establish this distribution, we consider the ensemble of inter-spike-intervals, sorted from the shortest to the longest interval:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_TISI_ensemble_a.png" width="500">

Next, we divide the interval duration axis into **non-overlapping** bins $\Delta t$ (not sliding bins !!!)

$\begin{eqnarray}
[0,\Delta t] \qquad [\Delta t, 2\Delta t] \qquad \ldots \qquad [i\Delta t, (i+1)\Delta t] \qquad \ldots
\end{eqnarray}$

and count the number $n_i$ of intervals falling into each bin:

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec07_TISI_ensemble_b.png" width="500">

Obviously, when the counts $n_i$ are summed over all bins, we retrieve the ensemble size $N$ (i.e., the total number of intervals)

$\begin{eqnarray}
\qquad\qquad \sum_i n_i = N
\end{eqnarray}$

(Slide-6)



# Count, fraction, and density

To be independent of  interval number $N$, we can convert interval **counts** into interval **fractions**

$\begin{eqnarray}
f_i = \frac{n_i}{N} = \frac{n_i}{\sum_i n_i} \qquad\qquad \sum_i f_i = 1
\end{eqnarray}$

which sum to unity.

To be independent of bin width $\Delta t$, we can convert interval **fractions** into interval **densities**

$\begin{eqnarray}
p_i = \frac{f_i}{\Delta t} = \frac{n_i}{N \Delta t}, \qquad\qquad \sum_i p_i \Delta t = 1
\end{eqnarray}$

Densities **integrate** to unity, but do **not sum** to any value of interest.

The following code illustrates these concepts with some examples.

(Slide-7)

```python
# @title Observe irregular spike train {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Illustrate counts, fractions, and densities
# for a distribution with linearly increasing hazard

# Rho(t) = (pi * t)/(2 * a^2);
# S(t)   = exp( -int Rho(t) dt ) = exp( - (pi * t^2)/(4 * a^2) )
# P(t)   = Rho(t) * S(t) = (pi * t)/(2 * a^2) * exp( -pi * t^2 / (4 * a^2) );

# ti = sqrt( -(4 * a^2 / pi ) * log( S(t) ) )

# int P(t) dt = 1
# int t P(t) dt = a
# int t^2 P(t) dt = 4 * a^2 / pi

def Samplelinearhazard( Nsample, m ):
    uk = np.random.rand(Nsample);
    tk = np.sqrt( -4 * m**2 * np.log( uk ) / np.pi );
    return tk

# Define an interactive plot
def plot_make_random_spiketrain(Nspike = 100):
    """(Slide-8)"""

    # parameters
    m_isi = 30;      # 30 ms

    # sample linear hazard distribution
    tk = Samplelinearhazard(Nspike, m_isi);

    ti = np.cumsum( tk );
    Ni = len( ti );

    T  = ti[Ni-1];

    # Create a figure and a grid of subplots (1 rows, 1 columns)
    fig, axs = plt.subplots(1, figsize=(9, 3))

    # First subplot
    plt.subplot(1, 1, 1)  # (nrows, ncols, plot_index)
    for i in range(0,Ni):
        plt.plot((ti[i],ti[i]), (0,1), color='black', linewidth=0.5);
    plt.title('Linear hazard spike train, Rate = 33 Hz', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()


# Creating sliders for concentrations and charge
interact(plot_make_random_spiketrain, Nspike=(100, 1000, 100))

```

```python
# @title Count, fraction, and density of inter-spike-intervals {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Illustrate counts, fractions, and densities
# for a distribution with linearly increasing hazard

# Linear hazard     Rho(t) = (pi * t)/(2 * a^2);
# Survivor fraction S(t)   = exp( -int Rho(t) dt ) = exp( - (pi * t^2)/(4 * a^2) )
# Density           P(t)   = Rho(t) * S(t) = (pi * t)/(2 * a^2) * exp( -pi * t^2 / (4 * a^2) );

# ti = sqrt( -(4 * a^2 / pi ) * log( S(t) ) )

# int P(t) dt = 1
# int t P(t) dt = a
# int t^2 P(t) dt = 4 * a^2 / pi

def Linearhazard(  ti, m ):
    Pi = np.pi * ti / (2 * m**2) * np.exp( -np.pi * ti**2 / (4 * m**2) );
    return Pi

def Samplelinearhazard( Nsample, m ):
    uk = np.random.rand(Nsample);
    tk = np.sqrt( -4 * m**2 * np.log( uk ) / np.pi );
    return tk

def SortIntoBins( tbin, Tmin, Tmax, tlist ):

    Nmin  = round( Tmin / tbin);         # first bin
    Nmax  = round( Tmax / tbin);         # last bin
    Nbin  = Nmax - Nmin + 1;             # number of bins

    t_i   = tbin * np.linspace(Nmin, Nmax, Nbin);       # left edges
    ix    = np.digitize( tlist, t_i );

    n_i   = np.zeros_like(t_i);
    for i in range(0, Nbin):
        n_i[i] = np.sum( ix == i+1 );

    return t_i, n_i

# Define an interactive plot
def plot_count_fraction_density(binsize = 5, Nsample = 100):
    """(Slide-9)"""

    # parameters
    m_isi = 30;      # 30 ms

    T       = 100;                                       # total interbal range, in ms (15 ms)
    Nbin    = round(T/binsize);                          # number bins
    tbin_i  = binsize * np.linspace(0, Nbin, Nbin+1);    # time vector with left edges

    # get analytical density
    ti      = np.linspace(0, T, 1001);
    Pi      = Linearhazard(  ti, m_isi );

    # sample linear hazard distribution
    tk = Samplelinearhazard(Nsample, m_isi);

    # sort intervals into bins and obtain counts
    (t_i, n_i) = SortIntoBins( binsize, 0, T, tk );

    # convert counts to fractions and to densities
    f_i = n_i / Nsample;
    p_i = f_i / binsize;

    # Create a figure and a grid of subplots (2 rows, 1 columns)
    fig, axs = plt.subplots(1, 3, figsize=(9, 3))

    # First subplot
    plt.subplot(1, 3, 1)  # (nrows, ncols, plot_index)
    plt.step(t_i, n_i, where='post', color='blue')
    plt.title('Count', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)

    # Second subplot
    plt.subplot(1, 3, 2)
    plt.step(t_i, f_i, where='post', color='green')
    plt.title('Fraction', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)

    # Third subplot
    plt.subplot(1, 3, 3)
    plt.plot(ti, Pi, color='black', linestyle='--')
    plt.step(t_i, p_i, where='post', color='red')
    plt.title('Density [1/ms]', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)
    plt.text(65, 0.025, 'fraction', fontsize=10, color='red')
    plt.text(60, 0.020, 'per binwidth', fontsize=10, color='red')

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_count_fraction_density, binsize=(1, 10, 1), Nsample=(100, 10000, 100))

```

# Different densities

To finish our little detour into statistics, we compare three different densities for inter-spike-intervals:

- Linear hazard density (which was used for the above example)

$\begin{eqnarray}
\qquad P(t) = \frac{\pi t}{2 \mu^2} \exp\left(-\frac{\pi t^2}{4\mu^2} \right), \qquad \int_0^\infty P(t) dt = 1, \qquad \left\langle t \right\rangle = \mu, \qquad \left\langle t^2 \right\rangle = \frac{4 \mu^2}{\pi}
\end{eqnarray}$

- Exponential or constant hazard density (which underlies **Poisson spikes**)

$\begin{eqnarray}
\qquad P(t) = \lambda \exp\left(-\lambda t\right), \qquad \int_0^\infty P(t) dt = 1, \qquad \left\langle t \right\rangle = \frac{1}{\lambda}, \qquad  \left\langle t^2 \right\rangle =\frac{2}{\lambda^2}
\end{eqnarray}$

- Cauchy density (caution: this one is pathological!)

$\begin{eqnarray}
\qquad P(t) = \frac{2}{\pi \gamma \left[1 + \left(\frac{t}{\gamma}\right)^2 \right]}, \qquad \int_0^\infty P(t) dt = n.d., \qquad \left\langle t \right\rangle = n.d., \qquad \left\langle t^2 \right\rangle = n.d.
\\
\\
\end{eqnarray}$

Here, $\int_0^\infty P(t) dt $ represents the **normalization condition**, $\left\langle t \right\rangle = \int_0^\infty t P(t) dt$ the **first moment**, and $\left\langle t^2 \right\rangle = \int_0^\infty t^2 P(t) dt$ the **second moment** of the density.

(Slide-10)

# Spiking of cortical neurons

The spike trains of **cortical neurons** often resemble **Poisson events**, which are defined by independence and constant hazard and therefore exhibit an **exponential distribution** of inter-spike-intervals.

In this case, the mean and variance of inter-spike-intervals are identical and the coefficient of variation is unity:

$\begin{eqnarray}
C_V^\mathit{Poisson} = \frac{\sqrt{\left\langle t^2 \right\rangle - \left\langle t \right\rangle^2}}{\left\langle t \right\rangle}
= \frac{\sqrt{\frac{2}{\lambda^2} - \frac{1}{\lambda^2}}}{\frac{1}{\lambda}} = 1
\end{eqnarray}$

Thus, the typical irregularity of cortex corresponds to a $C_v$ of unity.

(Slide-11)

```python
# @title Compare different densities {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import cauchy
from scipy.stats import expon
from ipywidgets import interact

# Illustrate and compare three densities

# Linear hazard
# Plin(t)   = Rho(t) * S(t) = (pi * t)/(2 * a^2) * exp( -pi * t^2 / (4 * a^2) );

# Exponential or constant hazard
# Pexp(t)   = lambda * exp( -lambda*t )

# Cauchy
# Pcauchy(t) = 2 / (pi * gamma) / (1 + (t/gamma)^2)

def Linearhazard(  ti, m ):
    Pi = np.pi * ti / (2 * m**2) * np.exp( -np.pi * ti**2 / (4 * m**2) );
    return Pi

def ExponentialDist(  ti, Lambda):
    Pi = Lambda * np.exp( -Lambda*ti );
    return Pi

def CauchyDist(  ti, gamma):
    Pi = 2 / (np.pi * gamma) / (1 + (ti/gamma)**2);
    return Pi

def Samplelinearhazard( Nsample, m ):
    uk = np.random.rand(Nsample);
    samples = np.sqrt( -4 * m**2 * np.log( uk ) / np.pi );
    return samples

def Sampleexponential( Nsample, Lambda ):
    #variant 1:
    #uk = np.random.rand(Nsample);
    #samples = -np.log( uk ) / Lambda;
    #variant 2:
    # beta = 1 / Lambda;
    #samples = np.random.exponential(scale=beta, size=Nsample);
    # variant 3:
    beta = 1 / Lambda;
    samples = expon.rvs(scale=beta, size=Nsample);
    return samples

def Samplecauchy( Nsample, gamma ):
    # get positive and negative samples
    samples = cauchy.rvs(loc=0, scale=gamma, size=3*Nsample);
    # restrict to positive samples only
    ixpos   = (0 < samples);
    samples = samples[ixpos];
    # take first Nsample samples
    samples = samples[0:Nsample];
    return samples


# Define an interactive plot
def plot_compare_three_densities(binsize = 5, Nsample = 100):
    """(Slide-12)"""
    # parameters
    mean_isi = 30;      # 30 ms

    Lambda   = 1 / mean_isi;

    # set size and number of bins
    T       = 100;                                       # total interbal range, in ms (15 ms)
    Nbin    = round(T/binsize);                          # number bins
    tbin_i  = binsize * np.linspace(0, Nbin, Nbin+1);    # time vector with left edges

    # get three analytical densities
    ti    = np.linspace(0, T, 1001);
    Plini = Linearhazard(  ti, mean_isi );
    Pexpi = ExponentialDist(  ti, 1/mean_isi );
    Pcaui = CauchyDist(  ti, mean_isi );

    # sample three distributions
    tlin_k = Samplelinearhazard(Nsample, mean_isi);
    texp_k = Sampleexponential(Nsample, 1/mean_isi);
    tcau_k = Samplecauchy(Nsample, mean_isi);

    # compute mean and variance
    Elin   = np.mean( tlin_k );
    Eexp   = np.mean( texp_k );
    Ecau   = np.mean( tcau_k );

    Vlin   = np.var( tlin_k );
    Vexp   = np.var( texp_k );
    Vcau   = np.var( tcau_k );

    # coefficient of variation
    CVlin   = np.sqrt(Vlin) / Elin;
    CVexp   = np.sqrt(Vexp) / Eexp;
    CVcau   = np.sqrt(Vcau) / Ecau;

    # sort intervals into bins
    (tbin_i, Nlin_i) = SortIntoBins( binsize, 0, T, tlin_k );
    (tbin_i, Nexp_i) = SortIntoBins( binsize, 0, T, texp_k );
    (tbin_i, Ncau_i) = SortIntoBins( binsize, 0, T, tcau_k );

    # convert counts to fractions and to densities
    Plin_i = Nlin_i / (Nsample * binsize);
    Pexp_i = Nexp_i / (Nsample * binsize);
    Pcau_i = Ncau_i / (Nsample * binsize);

    # Create a figure and a grid of subplots (2 rows, 1 columns)
    fig, axs = plt.subplots(1, 3, figsize=(9, 3))

    # First subplot
    plt.subplot(1, 3, 1)  # (nrows, ncols, plot_index)
    plt.plot(ti, Plini, color='black', linestyle='--')
    plt.step(tbin_i, Plin_i, where='post', color='blue')
    plt.title('Linear hazard', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)
    plt.ylabel('density', fontsize=14)
    plt.ylim([0.0,0.03])
    str1 = 'CV=';
    str2 = f'{CVlin:.2f}';
    plt.text(60, 0.025, str1 + str2, fontsize=12, color='black')
    plt.text(55, 0.020, 'more regular', fontsize=10, color='black')

    # Second subplot
    plt.subplot(1, 3, 2)
    plt.plot(ti, Pexpi, color='black', linestyle='--')
    plt.step(tbin_i, Pexp_i, where='post', color='green')
    plt.title('Exponential', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)
    plt.ylim([0.0,0.04])
    str2 = f'{CVexp:.2f}';
    plt.text(60, 0.030, str1 + str2, fontsize=12, color='green')
    plt.text(55, 0.025, 'less regular', fontsize=10, color='green')


    # Third subplot
    plt.subplot(1, 3, 3)
    plt.plot(ti, Pcaui, color='black', linestyle='--')
    plt.step(tbin_i, Pcau_i, where='post', color='red')
    plt.title('Cauchy', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)
    plt.ylim([0.0,0.03])
    str2 = f'{CVcau:.2f}';
    plt.text(50, 0.025, str1 + str2, fontsize=12, color='red')
    plt.text(45, 0.020, 'least regular', fontsize=10, color='red')


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_compare_three_densities, binsize=(1, 10, 1), Nsample=(100, 10000, 100))

```

# Summary of spike statistics

- Spike trains are best described in statistical terms.  First-order measures are average rate $\lambda \, [Hz]$ and average interval $\left\langle{t_\mathit{isi}}\right\rangle \, [s]$

- A second-order measure of irregularity is the coefficient of variation, which is based on inter-spike-intervals:

$\begin{eqnarray}
\qquad\qquad C_v \equiv \frac{\sqrt{\left\langle{t_\mathit{isi}^2}\right\rangle-\left\langle{t_\mathit{isi}}\right\rangle^2}}{\left\langle{t_\mathit{isi}}\right\rangle} = \frac{\sigma_\mathit{isi}}{\mu_\mathit{isi}}
\end{eqnarray}$

- Another second-order measure is the Fano factor, which is based on binned spike counts. It quantifies irregularity at the time-scale of the bin size $\Delta t$:

$\begin{eqnarray}
\qquad\qquad
\mathit{FF}(\Delta t) = \frac{\left\langle n^2\right\rangle - \left\langle n \right\rangle^2}{\left\langle n\right\rangle}
\end{eqnarray}$

- The shape of the interval density $P(t_\mathit{isi})$, measured in $[Hz]$, characterizes irregularity even better:

$\begin{eqnarray}
\qquad\qquad P(t_\mathit{isi}) =  \frac{\mathit{Prob}[t \leq t_\mathit{isi} \leq t+dt ]}{dt}
\end{eqnarray}$

- Spike trains often resemble Poisson events, which are defined by independence and constant hazard.  In this case, the interval density is exponential and the coefficient of variation is unity

$\begin{eqnarray}
\qquad\qquad P\left(t_\mathit{isi}\right) = \lambda \, \exp\left({-\lambda \, t_\mathit{isi}}\right),\qquad\qquad C_V = 1
\\
\\
\end{eqnarray}$

# Excitatory and inhibitory synapses

For the synapses, we assume the following reversal potentials and maximal conductances

$\begin{eqnarray}
E_{ex} = 0\, mV\qquad E_{in} = -80\,mV\qquad r_m\,g_{ex}=0.025 \qquad r_m\,g_{in} = 0.1
\end{eqnarray}$

and we assume an alpha function time-course for the time-varying component:

$\begin{eqnarray}
P^{ex}(t) =  \frac{\Delta t}{\tau_{s}} \, \exp\left(1- \frac{\Delta t}{\tau_{s}} \right),
\qquad
P^{in}(t) =  \frac{\Delta t}{\tau_{s}} \, \exp\left(1- \frac{\Delta t}{\tau_{s}} \right)
\end{eqnarray}$

Here, $\Delta t = [t-t_i]_+$ is the (positive) time since the last spike and $\tau_s = 10\,ms$ is the time constant.

(Slide-15)


# Dynamic equation of membrane voltages

Below threshold -- in the regime of $V \leq V_{th}$ -- the dynamic equations for the two membrane voltages are

$\begin{eqnarray}
\tau_m \, \frac{dV_1}{dt} = -V_1+E_L - r_m \, g_{ex} \, P^{ex}_{2\to 1}(t) \, \left[ V_1- E_{ex}\right] + r_m \, I_1(t)
\end{eqnarray}$

$\begin{eqnarray}
\tau_m \, \frac{dV_2}{dt} = -V_2+E_L - r_m \, g_{ex} \, P^{ex}_{1\to 2}(t) \, \left[ V_2- E_{ex}\right] + r_m \, I_2(t)
\end{eqnarray}$

where $P^{ex}_{1\to 2}$ represents synaptic transmission from neuron 1 to neuron 2, and $P^{ex}_{2\to 1}$ transmission in the opposite direction.  $I_1(t)$ and $I_2(t)$ are two independent noise currents. (We realize these currents as 'physical' noise and generate them with an Ornstein-Uhlenbeck process.)

Rearranged into standard form:

$\begin{eqnarray}
\qquad\qquad\qquad\qquad\tau_m^\mathit{eff} \, \frac{dV}{dt} = -V + V_\infty^\mathit{eff}
\end{eqnarray}$

$\begin{eqnarray}
V_\infty^\mathit{eff} =  \frac{E_L + r_m \, g_{ex} \,  \sum_k P_k^{ex}(t) \, E_s }{1 + r_m \, g_s \, \sum_k P_k^{ex}(t) },
\qquad
\tau_m^\mathit{eff} = \frac{\tau_m}{1 + r_m \, g_s \, \sum_k P_k^{ex}(t)}
\end{eqnarray}$

where $\sum_k P_k^{ex}(t)$ sums over all past synaptic events (i.e., events before time $t$).

```python
# @title Mutual excitation, short simulation {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Two LIF neurons coupled by mutual excitation

# define Ornstein-Uhlenbeck process for noisy input currents
def OUprocess( Nneuron, ti, mu_n, sigma_n, tau_n ):
    dt    = ti[1] - ti[0];
    # allocate OU time samples
    x_ni  = np.zeros((Nneuron,len(ti)));

    # normally distributed samples (white noise)
    Xi_ni = np.random.normal(0, 1, (Nneuron,len(ti)));

    # OU samples relax exponentially, while being 'bumped' by white noise
    x_ni[:,0] = mu_n + sigma_n * Xi_ni[:,0];
    for i in range(1,len(ti)):
        dx_n      = (-x_ni[:,i-1]+mu_n) * dt / tau_n + sigma_n * np.sqrt(dt/tau_n) * Xi_ni[:,i];
        x_ni[:,i] = x_ni[:,i-1] + dx_n;

    return x_ni

# simulation of coupled neurons
def CoupledNeuronsWithDetail(ti, rmIe, rmgsyn, Esyn ):

    # parameters
    EL        = -70;  # mV
    Vth       = -54;  # mV
    Vreset    = -80;  # mV

    # time constants
    tau_s     = 10;   # ms
    tau_m     = 20;   # ms

    # time axis
    dt        = ti[1]-ti[0];   # time step, in ms (0.1 ms)
    Ni        = len(ti);       # number of time steps

    # number of neurons in network
    Nneuron   = 2;

    # noisy input currents
    mu_n      = rmIe;                # mV
    sigma_n   = 2 * np.sqrt(rmIe);   # mV
    tau_n     = 3;                   # ms
    rmIe_ni   = OUprocess( Nneuron, ti, mu_n, sigma_n, tau_n);

    # allocate vectors for variables to be returned
    Vm_ni     = np.zeros((Nneuron, Ni));
    Psyn_ni   = np.zeros((Nneuron, Ni));

    Vinfty_ni = np.zeros((Nneuron, Ni));

    tspike_nk = np.zeros((Nneuron, Ni)) + ti[len(ti)-1];

    # assign initial values where necessary

    Vm_ni[:,0]     = EL + rmIe_ni[:,0]      # membrane potential
    Vinfty_ni[:,0] = (EL + rmgsyn * Psyn_ni[:,0] * Esyn + rmIe_ni[:,0]) / (1 + rmgsyn * Psyn_ni[:,0]);
    taueff_ni      = tau_m / (1 + rmgsyn * Psyn_ni[:,0]);

    # iterate over successive time steps

    kpost_n = np.zeros(Nneuron, dtype=int);

    for i in range( 1, len(ti) ):

        # current contributions from OTHER neuron
        for n in range( Nneuron ):
            # n of other neuron
            nprime = Nneuron - n - 1;
            # indices to recent spikes
            ix_spike     = (ti[i]-10*tau_s <= tspike_nk[nprime,:]) & (tspike_nk[nprime,:] <= ti[i]);
            # times since recent spikes
            dt_spike     = ( ti[i] - tspike_nk[nprime,ix_spike] ) / tau_s;
            # synaptic conductances from recent spikes
            Pt_spike     = dt_spike * np.exp(1 - dt_spike );
            # total conductance
            Psyn_ni[n,i] = np.sum(Pt_spike);

        # steady-state values
        Vinfty_ni[:,i] = (EL + rmgsyn * Psyn_ni[:,i] * Esyn + rmIe_ni[:,i]) / (1 + rmgsyn * Psyn_ni[:,i]);
        taueff_ni      = tau_m / (1 + rmgsyn * Psyn_ni[:,i]);

        # update state variables at time i
        Vm_ni[:,i] = Vinfty_ni[:,i] + (Vm_ni[:,i-1] - Vinfty_ni[:,i]) * np.exp(-dt / taueff_ni);

        for n in range( Nneuron ):
            if Vm_ni[n,i ] > Vth:
                Vm_ni[n,i ] = Vreset;
                kpost_n[n]  = kpost_n[n] + 1;
                tspike_nk[n,kpost_n[n]-1] = ti[i];

    # truncate spike arrays
    t1spike_k = tspike_nk[0,0:kpost_n[0]];
    t2spike_k = tspike_nk[1,0:kpost_n[1]];

    return Vm_ni, Vinfty_ni, Psyn_ni, t1spike_k, t2spike_k;


# Define an interactive plot
def plot_two_neurons_coupled_excitation(rmgex = 0.10, rmIe=18 ):
    """(Slide-17)"""

    # synaptic reversal potentials
    Eex       = 0;    # mV
    Ein       = -80;  # mV
    Vreset    = -80;  # mV

    Nneuron   = 2;

    #rmgex     = 0.10;  # excitatory strengty
    #rmgin     = 0.25;   # inhibitory strength

    # time axis
    dt        = 0.1;                  # time step, in ms (0.02 ms)
    T         = 500;                   # total time, in ms (15 ms)
    Ni        = round(T/dt) + 1;       # number of time steps
    ti        = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # plot limits
    tmin      = 3*T/4;
    tmax      = T;
    ix_range  = (ti >= tmin) & (ti <= tmax);
    ti_range  = ti[ix_range];

    # run simulation

    (Vm_ni, Vinfty_ni, Psyn_ni, t1spike_k, t2spike_k) = CoupledNeuronsWithDetail(ti, rmIe, rmgex, Eex );

    # Firing rates
    Lambda_n = np.zeros(Nneuron);
    Lambda_n[0] = 1000 * len(t1spike_k) / T;
    Lambda_n[1] = 1000 * len(t2spike_k) / T;

    # Create a figure and a grid of subplots (4 rows, 1 columns)
    fig, axs = plt.subplots(2*Nneuron, 1, figsize=(8, 6))

    # First subplot
    for n in range( Nneuron ):
        str1 = 'Neuron ';
        str2 = f'{n+1:d}';
        str3 = ', firing rate ';
        str4 = f'{Lambda_n[n]:.2f}';
        str5 = ' Hz';
        plt.subplot(2*Nneuron, 1, n+1)  # (nrows, ncols, plot_index)
        if n==0:
            for k in range( len(t1spike_k) ):
                plt.plot((t1spike_k[k],t1spike_k[k]), (Vreset,-25), 'k:');
        if n==1:
            for k in range( len(t2spike_k) ):
                plt.plot((t2spike_k[k],t2spike_k[k]), (Vreset,-25), 'k:');
        plt.plot(ti_range, Vm_ni[n,ix_range], 'r', ti_range, Vinfty_ni[n,ix_range], 'r:');
        plt.xlim(tmin, tmax);
        plt.title(str1 + str2 + str3 + str4 + str5, fontsize=18)
        #plt.xlabel('time [ms]', fontsize=14)
        plt.ylabel('Vm(t)', fontsize=14)

    # Second subplot
    for n in range( Nneuron ):
        str1 = 'Synapse from ';
        str2 = f'{Nneuron-n:d}';
        str3 = ' to ';
        str4 = f'{n+1:d}';

        plt.subplot(2*Nneuron, 1, n+1+Nneuron)  # (nrows, ncols, plot_index)
        nprime = Nneuron - n - 1;
        if nprime==0:
            for k in range( len(t1spike_k) ):
                plt.plot((t1spike_k[k],t1spike_k[k]), (0,1.5), 'k:');
        if nprime==1:
            for k in range( len(t2spike_k) ):
                plt.plot((t2spike_k[k],t2spike_k[k]), (0,1.5), 'k:');
        plt.plot(ti_range, Psyn_ni[n,ix_range], 'b');
        plt.xlim(tmin, tmax);
        plt.title(str1 + str2 + str3 + str4, fontsize=18)
        if n == Nneuron-1:
            plt.xlabel('time [ms]', fontsize=14)
        plt.ylabel('Ps(t)', fontsize=14)


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_two_neurons_coupled_excitation, rmgex = (0.05, 0.5, 0.05), rmIe=(10, 30, 2))

```

```python
# @title Coupled response function {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Two LIF neurons coupled by mutual excitation

# define a compact simulation which outputs only firing rates and spike trains
def CoupledNeurons( ti, rmIe, rmgsyn, Esyn ):

    # parameters
    EL        = -70;  # mV
    Vth       = -54;  # mV
    Vreset    = -80;  # mV

    # time constants
    tau_s     = 10;   # ms
    tau_m     = 20;   # ms

    # time step
    dt        = ti[1]-ti[0];
    Ni        = len(ti);       # number of time steps

    # number of neurons in network
    Nneuron   = 2;

    # noisy input currents
    #rmIe     = 25;                  # mV
    mu_n      = rmIe;                # mV
    sigma_n   = 2 * np.sqrt(rmIe);   # mV
    tau_n     = 3;                   # ms
    rmIe_ni   = OUprocess( Nneuron, ti, mu_n, sigma_n, tau_n);

    # allocate vectors for variables to be plotted (membrane potential, synaptic conductance)
    tspike_nk = np.zeros((Nneuron, Ni)) + ti[Ni-1];

    # assign initial values where necessary

    Vm_ni     = EL + rmIe_ni[:,0]      # membrane potential
    Psyn_ni   = np.zeros(Nneuron);
    Vinfty_ni = (EL + rmgsyn * Psyn_ni * Esyn + rmIe_ni[:,0]) / (1 + rmgsyn * Psyn_ni);
    taueff_ni = tau_m / (1 + rmgsyn * Psyn_ni);

    # iterate over successive time steps

    kpost_n = np.zeros(Nneuron, dtype=int);

    for i in range( 1, len(ti) ):

        # current contributions from OTHER neuron
        for n in range( Nneuron ):
            nprime       = Nneuron - n - 1;         # other neuron
            tprime_k     = tspike_nk[nprime,:];     # all spikes of other neuron
            ix_spike     = (ti[i]-10*tau_s <= tprime_k) & (tprime_k <= ti[i]);   # recent spikes
            dt_spike     = ( ti[i] - tspike_nk[nprime,ix_spike] ) / tau_s;
            Pt_spike     = dt_spike * np.exp(1 - dt_spike );
            Psyn_ni      = np.sum(Pt_spike);

        # steady-state values
        Vinfty_ni = (EL + rmgsyn * Psyn_ni * Esyn + rmIe_ni[:,i]) / (1 + rmgsyn * Psyn_ni);
        taueff_ni = tau_m / (1 + rmgsyn * Psyn_ni);

        # update state variables at time i
        Vm_ni = Vinfty_ni + (Vm_ni - Vinfty_ni) * np.exp(-dt / taueff_ni);

        for n in range( Nneuron ):
            if Vm_ni[n] > Vth:
                Vm_ni[n] = Vreset;
                kpost_n[n]  = kpost_n[n] + 1;
                tspike_nk[n,kpost_n[n]-1] = ti[i];

    # truncate spikes
    N_1 = kpost_n[0];
    N_2 = kpost_n[1];

    t1spike_k = tspike_nk[0,0:N_1];
    t2spike_k = tspike_nk[1,0:N_2];

    # firing rates
    Lambda_1 = N_1 / ti[len(ti)-1];
    Lambda_2 = N_2 / ti[len(ti)-1];

    Lambda   = 500 * (Lambda_1 + Lambda_2);

    return Lambda, t1spike_k, t2spike_k

# Define an interactive plot
def plot_coupled_response_function(rmgex = 0.10, rmgin = 1.0 ):
    """(Slide-19)"""

    # aray of input currents
    rmIe_k = np.linspace(10,40,15);

    Eex  = 0.0;
    Ein  = -70.0;

    # allocate array of output rates
    Lambda_ex_k   = np.zeros(len(rmIe_k));
    Lambda_in_k   = np.zeros(len(rmIe_k));
    Lambda_null_k = np.zeros(len(rmIe_k))


    # time axis
    T = 1000;
    dt        = 1;                   # time step, in ms (0.02 ms)
    Ni        = round(T/dt) + 1;       # number of time steps
    ti        = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # loop over input currents
    for k in range( len(rmIe_k) ):
        (Lambda_ex_k[k], dummy1, dummy2)   = CoupledNeurons( ti, rmIe_k[k], rmgex, Eex );
        (Lambda_in_k[k], dummy1, dummy2)   = CoupledNeurons( ti, rmIe_k[k], rmgin, Ein );
        (Lambda_null_k[k], dummy1, dummy2) = CoupledNeurons( ti, rmIe_k[k], 0, Eex );

    # Create a figure and a grid of subplots (4 rows, 1 columns)
    fig, axs = plt.subplots(1, figsize=(6, 4))

    plt.plot(rmIe_k, Lambda_ex_k, marker='o', color='blue');
    plt.plot(rmIe_k, Lambda_in_k, marker='o', color='red');
    plt.plot(rmIe_k, Lambda_null_k, marker='o', color='green');
    plt.title('Coupled response', fontsize=18)
    plt.xlabel('input rm*Ie [mV]', fontsize=14)
    plt.ylabel('firing rate', fontsize=14)

    plt.text(11, 90, 'excitation', fontsize=12, color='blue')
    plt.text(11, 80, 'no coupling', fontsize=12, color='green')
    plt.text(11, 70, 'inhibition', fontsize=12, color='red')

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_coupled_response_function, rmgex = (0.05, 0.5, 0.05), rmgin = (0.05, 1.0, 0.05))

```

# Synchronous and alternating modes of firing

When the synaptic time constant is sufficiently large, coupling two neurons via excitatory synapses can produce alternate firing, while coupling via inhibitory synapses can produce synchronous firing.

In slightly larger networks of, say, five coupled neurons, excitatory synapses result in desynchronized firing, whereas inhibitory synapses produce synchronous firing.

# Coupled LIF neurons

- Coupling LIF neurons with reciprocal synapses has several effects.

- Excitatory coupling **increases** and inhibitory coupling **decreases** the gain of the response function.

- Excitatory coupling tends to **desynchronize** spikes across the population.

- Inhibitory coupling tends to **synchronize** spikes across the population.

- Post hoc rationalization: spike probability peaks when excitation is maximally **effective** and/or inhibition is maximally **ineffective**.

- **Already the simplest networks show emergent behaviour.**

(Slide-23)


# **3. Operating regime of cortical neurons**

In mammalian cortex, neurons, dendrites, axons and synapses are packed extremely densely and every pyramidal (excitatory) neuron receives a very large number of excitatory and inhibitory inputs from presynaptic neurons.

And what might be the point of receiving both excitation and inhibition from the same presynaptic sources?

# Number of presynaptic inputs

The microanatomy of axonal projections offers one way to estimate the number of presynaptic inputs. The illustration shows **efferent** excitatory axons radiating from pyramidal neurons at one central location (where stain was deposited). The target locations form a patchy pattern over approximately 3 mm of cortex and cover approximately 15,000 functional units ('minicolumns'). In return, **afferent** axons from all 'minicolumns' at the target locations converge back on the central location.

Even though these long-range projections (efferent and afferent) are exclusively excitatory, their effect on the target location is both excitatory and inhibitory. This is because presynaptic sources contact their postsynaptic targets both monosynaptically (via a direct excitatory synapse) and disynaptically (via an inhibitory interneuron).

Thus, pyramidal neurons in mammalian cortex typically receive on the order of **10,000 presynaptic inputs**, of which approximately 80% are excitatory and 20% inhibitory.

# Summation of many excitatory and inhibitory synapses

We wish to investigate the combined effect of large numbers of excitatory and inhibitory synaptic conductances onto a single LIF neuron model.

The dynamic equation for the membrane potential is

$\begin{eqnarray}
 \tau_m \, \frac{dV(t)}{dt} = -V(t) + E_L - r_m \, g_\mathit{ex} \, \sum_j P_j^\mathit{ex}(t) \, [V(t) - E_\mathit{ex}] - r_m \, g_\mathit{in} \, \sum_k P_k^\mathit{in}(t) \, [V(t) - E_\mathit{in}]
\end{eqnarray}$

with $\tau_m=100 ms$, $E_L=-70 ms$, $r_m \, g_\mathit{ex}= 0.05$, and $r_m \, g_\mathit{in} = 0.5$. The sums go over all excitatory presynaptic events $j$ and inhibitory presynaptic events $k$ prior to time $t$.  The firing threshold is $V_\mathit{th} = -54\, ms$ and the reset potential is $V_\mathit{reset} = -60 \, ms$.

(Slide-26)




# Iterative solution of dynamic equation

To iteratively solve for the membrane potential $V(t)$, we must rearrange the dynamic equation into its standard form (as we have done several times before)

$\begin{eqnarray}
\qquad\qquad \tau_\mathit{eff} \frac{dV}{dt} = - V + V^\infty_\mathit{eff}
\end{eqnarray}$

with

$\begin{eqnarray}
\tau_\mathit{eff} = \frac{\tau_m}{1 + r_m \, g_\mathit{ex} \, \sum_j P_j^\mathit{ex}(t) + r_m \, g_\mathit{in} \, \sum_k P_k^\mathit{in}(t)} \qquad\qquad V^\infty_\mathit{eff} = \frac{E_L + r_m \, g_\mathit{ex} \, E_\mathit{ex} \, \sum_j P_j^\mathit{ex}(t) + r_m \, g_\mathit{in} \, E_\mathit{in} \, \sum_k P_k^\mathit{in}(t)}{1 + r_m \, g_\mathit{ex} \, \sum_j P_j^\mathit{ex}(t) + r_m \, g_\mathit{in} \, \sum_k P_k^\mathit{in}(t)}
\end{eqnarray}$

and solve the exponential relaxation for every time step

$\begin{eqnarray}
V(t+\Delta t) &=& V^\infty_\mathit{eff}(t) + \left[V(t) - V^\infty_\mathit{eff}(t) \right] \, \exp\left(-\frac{\Delta t}{\tau_\mathit{eff}} \right), \qquad\mathit{if}\qquad V(t+\Delta t) < V_\mathit{th}
\\
V(t+\Delta t) &=& V_\mathit{reset} \qquad\qquad\qquad\qquad\qquad\qquad\qquad\mathit{otherwise}
\\
\\
\end{eqnarray}$

(Slide-27)


# Synaptic conductances and potentials

For the time-course of synaptic conductances, we assume alpha functions, but with different time-constants for excitatory and inhibitory synapses:

$\begin{eqnarray}
 P_\mathit{syn}(t) \frac{dV(t)}{dt} = \frac{\Delta t}{\tau_\mathit{syn}} \exp\left(1 - \frac{\Delta t}{\tau_\mathit{syn}} \right), \qquad \tau_\mathit{ex} = 1 \, ms, \qquad \tau_\mathit{in} = 2 \, ms, \qquad E_{ex} = 0 \, mV, \qquad E_\mathit{in} = -80 \, mV
\end{eqnarray}$

To compute the total excitatory or inhibitory conductance at time $t$, we must proceed from the list of presynaptic events (up to time $t$). Noticeable contributions will come from events $t_\mathit{ex}$ and $t_\mathit{in}$ in the past 10 time-constants:

$\begin{eqnarray}
t-10 \, \tau_\mathit{ex} < t_\mathit{ex}<= t \qquad \mathit{and} \qquad  t-10 \, \tau_\mathit{in} < t_\mathit{in} <= t
\end{eqnarray}$

Given the lists $t_\mathit{ex}$ and $t_\mathit{in}$, we obtain the total conductances at time $t$ as

$\begin{eqnarray}
P_\mathit{ex}^\mathit{tot}(t) =  \sum_{t_\mathit{ex}} P_\mathit{ex}(t-t_\mathit{ex}) \qquad\qquad\mathit{and}\qquad\qquad P_\mathit{in}^\mathit{tot}(t) =  \sum_{t_\mathit{in}} P_\mathit{in}(t-t_\mathit{in})
\end{eqnarray}$

and the total postsynaptic potentials as

$\begin{eqnarray}
EPSP_\mathit{tot}(t) = - r_m \, g_\mathit{ex} \, P_\mathit{ex}^\mathit{tot}(t) \cdot [E_L - E_\mathit{ex}] \qquad\qquad\mathit{and}\qquad\qquad IPSP_\mathit{tot}(t) = - r_m \, g_\mathit{in} \, P_\mathit{in}^\mathit{tot}(t) \cdot [E_L - E_\mathit{in}]
\end{eqnarray}$

(Slide-28)

# Poisson statistics

An important aspect of our simulation is the statistics of presynaptic events.  Consistent with experimental observations, we assume that presynaptic events exhibit Poisson statistics (i.e., independence and constant hazard).

To generate Poisson events with rate $\nu_\mathit{syn}$ we obtain inter-spike-intervals either by log-transforming samples from a uniform random distribution

$\begin{eqnarray}
t_\mathit{isi}^\mathit{ex} = - \frac{1}{\nu_\mathit{ex}} \log u ,\qquad\qquad t_\mathit{isi}^\mathit{in} = - \frac{1}{\nu_\mathit{in}} \log u ,\qquad\qquad u \in [0,1]
\end{eqnarray}$

or by sampling them directly from an exponential distribution (as was done above in the code of Slide-12 "Compare different densities").

Importantly, combining (adding) series of Poisson events produces another series of Poisson events. For example, when we superimpose two series of Poisson events with rates $\nu_1$ and $\nu_2$, respectively, the series of combined events also exhibits Poisson statistics, with rate $\nu_\mathit{both} = \nu_1 + \nu_2 $.

**Cautionary note:** this is a nice property of Poisson statistics and **does not generally hold** for events with other kinds of statistics (linear hazard, Cauchy, etc.)

Accordingly, if we want to combine $N_\mathit{ex}=800$ presynaptic spike trains, each with a Poisson rate of $\nu_{ex}=10\,Hz$, we can simply generate Poisson events with rate $\nu_\mathit{comb} = N_\mathit{ex} \cdot \nu_\mathit{ex} = 8 \, kHz$.

Similarly, if we want to combine $N_\mathit{in}=200$ presynaptic spike trains with a Poisson rate $\nu_{in}=10\,Hz$, we can simply generate Poisson events with rate $\nu_\mathit{comb} = N_\mathit{in} \cdot \nu_\mathit{in} = 2 \, kHz$.

(Slide-29)

```python
# @title Total synaptic conductance {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# generate Poisson events
def PoissonEvents( Tend, Tmargin, nu ):

    # number of spikes
    Nevent = round( nu * (Tend + Tmargin) );

    # inter-spike-intervals
    t_isi  = - np.log( np.random.rand(Nevent) ) / nu;

    # spike train
    tspike_i = np.cumsum(t_isi);

    # truncate
    ix = (tspike_i <= Tend+Tmargin);
    tspike_i = tspike_i[ix] - Tmargin/2;

    return tspike_i


# obtain total conductance
def TotalConductance( ti, tspike_k, tau_syn ):

    # allocate conductance
    Psyni = np.zeros_like(ti);

    # loop over time-points and extract list of recent events
    for i in range(len(ti)):
        ix_recent = (ti[i]-10*tau_syn<=tspike_k) & (tspike_k<=ti[i]);
        if len(ix_recent)>0:
            alpha_k = ( ti[i] - tspike_k[ ix_recent ] ) / tau_syn;
            Psyn_k  = alpha_k * np.exp(1 - alpha_k);
            Psyni[i] = np.sum( Psyn_k );

    return Psyni


# Define an interactive plot
def plot_total_conductance( Nex = 800, Nin = 200 ):
    """(Slide-30)"""

    rmgex = 0.05;
    rmgin = 0.5;

    tauex = 1;
    tauin = 2;

    nuex  = 0.1;   # 0.1 per ms corresponds to 100 per s
    nuin  = 0.1;

    Eex  =  0.0;
    Ein  = -80.0;

    EL   = -70;

    # time axis
    T         = 500;
    Tmargin   = 50;
    dt        = 0.1;                     # time step, in ms (0.1 ms)
    Ni        = round(T/dt) + 1;         # number of time steps
    ti        = np.linspace(0, T, Ni);   # time vector, with successive time steps from 0 to T

    # obtain spike trains
    tspike_ex = PoissonEvents( T, Tmargin, nuex * Nex );
    tspike_in = PoissonEvents( T, Tmargin, nuin * Nin );

    # obtain total conductance
    Pexi      = TotalConductance( ti, tspike_ex, tauex );
    Pini      = TotalConductance( ti, tspike_in, tauin );

    # obtain EPSP and IPSP
    EPSPi = -rmgex * Pexi * [EL - Eex];
    IPSPi = -rmgin * Pini * [EL - Ein];

    # combined PSP
    PSPi  = EPSPi + IPSPi;

    # mean and variance of PSPi
    muPSPi  = np.mean(PSPi);
    varPSPi = np.var(PSPi);

    str1      = 'Combined PSP'
    str2      = ', Mean= ';
    str3      = f'{muPSPi:.2f}';
    str4      = ', STD = ';
    str5      = f'{np.sqrt(varPSPi):.2f}';

    # Create a figure and a grid of subplots (2 rows, 1 columns)
    fig, axs = plt.subplots(3, 1, figsize=(8, 6))

    # First subplot
    plt.subplot(3, 1, 1)  # (nrows, ncols, plot_index)
    plt.plot(ti, EPSPi, '-', color='blue');
    plt.title('Excitatory PSP', fontsize=18)
    #plt.xlabel('time [ms]', fontsize=14)
    plt.ylabel('EPSP', fontsize=14)

    # Second subplot
    plt.subplot(3, 1, 2)  # (nrows, ncols, plot_index)
    plt.plot(ti, IPSPi, '-', color='red');
    plt.title('Inhibitory PSP', fontsize=18)
    #plt.xlabel('time [ms]', fontsize=14)
    plt.ylabel('IPSP', fontsize=14)

    # Third subplot
    plt.subplot(3, 1, 3)  # (nrows, ncols, plot_index)
    plt.plot(ti, EPSPi + IPSPi, '-', color='black');
    plt.plot((0,T), (muPSPi,muPSPi), ':', color='black');
    plt.plot((0,T), (muPSPi+np.sqrt(varPSPi),muPSPi+np.sqrt(varPSPi)), ':', color='red');
    plt.plot((0,T), (muPSPi-np.sqrt(varPSPi),muPSPi-np.sqrt(varPSPi)), ':', color='red');
    plt.title(str1+str2+str3+str4+str5, fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)
    plt.ylabel('PSP', fontsize=14)

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_total_conductance, Nex = (1, 1200, 1), Nin=(1, 400, 1))

```

# Independently controlling mean and variance

By adjusting the **difference** between $N_{ex}$ and $N_{in}$, we can control the mean level of the combined PSP.

By adjusting the **sum** of $N_{ex}$ and $N_{in}$, we can control the variance of the combined PSP.

Thus, balanced excitation and inhibition offer independent control over both mean and variance of synaptic inputs.

This may be one reason why cortical projections between two locations typically involve both excitatory and inhibitory synapses.

```python
 # @title Membrane potential for balanced inputs {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# iteratively integrate membrane potential
def IterativeIntegration( ti, EL, Vth, Vreset, Veffi, taueffi, spikeflag ):

    dt    = ti[1] - ti[0];

    Vi    = np.zeros_like(ti);
    Vi[0] = EL;

    tspike_k = np.zeros_like(ti);

    kspike = 0;

    for i in range(0,len(ti)-1):

        Vi[i+1] = Veffi[i] + (Vi[i] - Veffi[i]) * np.exp(-dt/taueffi[i]);

        if (Vi[i+1]>Vth) & (spikeflag==1):
            Vi[i+1] = Vreset;
            kspike += 1;
            tspike_k[kspike-1] = ti[i+1];

    # truncate spike list
    tspike_k = tspike_k[0:kspike];

    return Vi, tspike_k


# Define an interactive plot
def plot_membrane_potential_for_balanced_inputs( Nex = 800, Nin = 200, Tend=1000 ):
    """(Slide-31)"""

    rmgex  = 0.125;
    rmgin  = 0.5;

    tauex  = 1;
    tauin  = 2;

    taum   = 100;

    nuex   = 0.01;   # 0.1 per ms corresponds to 100 per s
    nuin   = 0.01;

    Eex    =  0.0;
    Ein    = -80.0;

    EL     = -70;

    Vth    = -54;
    Vreset = -60;   # makes for prettier graphics than -80

    # time axis
    #Tend      = 500;
    Tmargin   = Tend / 10;
    dt        = 0.1;                     # time step, in ms (0.1 ms)
    Ni        = round(Tend/dt) + 1;         # number of time steps
    ti        = np.linspace(0, Tend, Ni);   # time vector, with successive time steps from 0 to T

    # obtain spike trains
    tspike_ex = PoissonEvents( Tend, Tmargin, nuex * Nex );
    tspike_in = PoissonEvents( Tend, Tmargin, nuin * Nin );

    # obtain total conductance
    Pexi      = TotalConductance( ti, tspike_ex, tauex );
    Pini      = TotalConductance( ti, tspike_in, tauin );

    # time-constant and equilibrium potential
    taueffi   = taum / (1 + rmgex*Pexi + rmgin*Pini);
    Veffi     = (EL + rmgex*Pexi*Eex + rmgin*Pini*Ein) / (1 + rmgex*Pexi + rmgin*Pini);

    # obtain membrane potentials with and without spikes
    spikeflag = 1;
    (Vspikei, tspike_k)   = IterativeIntegration( ti, EL, Vth, Vreset, Veffi, taueffi, spikeflag );

    # obtain inter-spike-intervals
    t_isi     = np.diff(tspike_k);

    if len(t_isi)>10:
        CV        = np.std(t_isi) / np.mean(t_isi);
        Rate      = 1000 * len(tspike_k) / Tend;
    else:
        CV        = 0;
        Rate      = 0;

    str1      = 'With spikes, '
    str2      = 'Rate = ';
    str3      = f'{Rate:.2f}';
    str4      = ', CV = ';
    str5      = f'{CV:.2f}';


    spikeflag = 0;
    (Vnospikei, dummy_k) = IterativeIntegration( ti, EL, Vth, Vreset, Veffi, taueffi, spikeflag );

    # Create a figure and a grid of subplots (2 rows, 1 columns)
    fig, axs = plt.subplots(2, 1, figsize=(8, 6))

    # First subplot
    plt.subplot(2, 1, 1)  # (nrows, ncols, plot_index)
    for k in range(len(tspike_k)):
        plt.plot((tspike_k[k], tspike_k[k]), (Vreset,-40), '-', color='black', linewidth=0.5 );
    plt.plot(ti, Vspikei, '-', color='red');
    plt.ylim([Vreset, -40]);
    plt.title(str1 + str2 + str3 + str4 + str5, fontsize=18)
    #plt.xlabel('time [ms]', fontsize=14)
    plt.ylabel('Vm [ms]', fontsize=14)

    # Second subplot
    plt.subplot(2, 1, 2)  # (nrows, ncols, plot_index)
    plt.plot((0,Tend), (Vth, Vth), ':', color = 'black', linewidth=0.5 )
    plt.plot(ti, Vnospikei, '-', color='blue');
    plt.ylim([Vreset, -40]);
    plt.title('Without spikes', fontsize=18)
    plt.xlabel('time [ms]', fontsize=14)
    plt.ylabel('Vm [ms]', fontsize=14)


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_membrane_potential_for_balanced_inputs, Nex = (500, 1200, 50), Nin=(50, 400, 50), Tend=(500,5000,500))

```

# Slow and irregular regime

In the above simulation, the strength of excitatory and inhibitory synapses is $g_{ex}\,r_m = 0.125$ and $g_{in}\,r_m=0.5$, respectively, with $V_{reset} = -60\,mV$.

When the number or excitatory inputs is too low ($N_\mathit{ex} < 600$), the membrane potential does not reach threshold and there are no spikes.

As the number of excitatory inputs increases ($600 \leq N_\mathit{ex} \leq 800$), the membrane potential reaches threshold more frequently, and the firing rate rises gradually from $0\,Hz$ to $25 \, Hz$.

Importantly, the timing of spikes is driven by random fluctuations, so that the spiking is irregular and the coefficient of variation is high, with $C_V \approx 0.8-0.9$.

**Slow and irregular firing** is the physiological operating regime of cortical neurons.


(Slide-31)

# Fast and regular regime

As the number of excitatory inputs increases even further ($800 < N_\mathit{ex} $), the membrane potential is almost constantly driven to threshold, and the firing rate rises steeply to $80 \, Hz$ and more.

In this regime, the timing of spikes is driven more by reset recovery than by input fluctuations, so that the spiking is more regular and the coefficient of variation is low, approaching $C_V \approx 0.5$.

**Fast and regular firing** is a pathological regime of cortical neurons (i.e., epileptic activity).

# Mean and variance of membrane potential


**Fast, regular firing mode:**

In the absence or spikes, mean membrane potential is **above threshold**.  Spike timing is governed **by recovery** of membrane potential from reset.  This results in more **regular** inter-spike intervals.


**Slow, irregular firing mode:**

In the absence of spikes, mean membrane potential is **below threshold**.  Spike timing is governed by **fluctuations** of membrane potential over and above mean.  This results in more **irregular** inter-spike intervals.

# Theory of membrane potential fluctuations (advanced)

To gain a deeper understanding, we investigate a situation that is analytically tractable.

Specifically, we consider synapses with exponential postsynaptic potentials (PSP)

$\begin{eqnarray}
u(t) = w \, P(t) \qquad\qquad w = r_m \, g_s \, [V-E_s], \qquad P(t) = \exp\left(-\frac{t}{\tau_s} \right)
\end{eqnarray}$

where $w$ is synaptic weight and $\tau_s$ is the synaptic time-constant.  


The sign of $w$ determines whether the synapse is excitatory (EPSP) or inhibitory (IPSP).

**What is the collective effect of many of such synapses?  How do they change the membrane potential?**


To answer this question, we assume that synaptic events are independent and equally likely to occur at any time (Poisson assumption).

(Slide-32)

# Mean and variance of PSP (without proof!)

For Poisson PSP events with rate $\nu$, weight $w$, and time-constant $\tau_s$, the expected contribution to the **mean** of the membrane potential is

$\begin{eqnarray}
\mu_V = w\, \nu \, \tau_s
\end{eqnarray}$

whereas the expected contribution to the **variance** is

$\begin{eqnarray}
\sigma^2_V  = \frac{1}{2} w^2\, \nu \, \tau_s
\end{eqnarray}$

Note that the **mean** contribution depends on $w$, but the **variance** contribution on $w^2$.

(Slide-33)

# Balanced excitation and inhibition

For balanced excitation (EPSPs) and inhibition (IPSPs), with identical rates $\nu_+ = \nu_- = \nu$ and equal but opposite weights $w_+ = w$ and $ w_- = -w$, the expected contributions are

$\begin{eqnarray}
\mu_V = w_+\, \nu_+ \, \tau_s + w_-\, \nu_- \, \tau_s = w \, \nu \, \tau_s - w \, \nu \, \tau_s = 0
\end{eqnarray}$

$\begin{eqnarray}
\sigma^2_V =  \frac{1}{2} w_+^2\, \nu_+ \, \tau_s+  \frac{1}{2} w_-^2\, \nu_- \, \tau_s= w^2\, \nu \, \tau_s
\end{eqnarray}$


**Balanced excitation and inhibition cancel in the mean, but add in the variance!**

(Slide-34)

# Fluctuation-driven firing activity

- The average of the membrane potential is set by the **difference** between excitatory and inhibitory input rates

$\begin{eqnarray}
\mu_V = w_+ \, \nu_+ \, \tau_+ - w_{-} \, \nu_{-} \, \tau_{-}
\end{eqnarray}$

- The variance of the membrane potential is set by the **sum** of excitatory and inhibitory input rates

$\begin{eqnarray}
\sigma^2_V = \frac{1}{2} \, w_+^2 \, \nu_+ \, \tau_+ + \frac{1}{2} \,  w_{-}^2 \, \nu_{-} \, \tau_{-}
\end{eqnarray}$

- The **difference** between excitatory and inhibitory input controls the firing mode (slow \& irregular or fast \& regular).

- In the slow \& irregular mode, the **sum** of excitation and inhibition determines **input variance**, which in turn controls output spiking.

- If firing was controlled by **input variance**, this would explain why cortical projections typically combine **excitatory and inhibitory** synapses.

(Slide-35)

# Summary operating regime of cortical neurons

- Cortical neurons receive thousands or tens of thousands of inputs.  Typically, excitatory and inhibitory inputs are approximately balanced.

- LIF neuron models reveal the functional implications of balanced excitation and inhibition.

- There physiological operating mode is $\left\langle V_m \right\rangle < V_\mathit{th}$, with slow, irregular firing ( $C_v\approx 1$ ).

-  A pathological mode is $\left\langle V_m \right\rangle \approx V_\mathit{th}$, with fast, regular firing ( $C_v\approx 0.5$ ).

- The operating regime is controlled by the  **difference** of inputs (excitatory and inhibitory).

- The rate of output firing may be controlled by **input variance**, making use of both excitatory and inhibitory synapses.

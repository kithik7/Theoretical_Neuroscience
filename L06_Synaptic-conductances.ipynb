**Abstract:**

After some general remarks, we model the opening and closing of synaptic conductances with different approximate formulas. Comparing the characteristics of typical mammalian synapses -- **GABA, AMPA and NMDA** -- we find suprisingly large differences. Next we analyze the effects of adding synaptic conductances to LIF neurons. The two main effects are  triggering of **postsynaptic potentials (EPSP or IPSP)** and modulation of the **membrane time-constant**.

As an advanced topic, we investigate the effects of **short-term depression and facilitation (STD and STF)**. To quantify the average effect over a stochastic ensemble of synaptic time-courses, we introduce **Poisson spikes** and learn to compute **expectation values**.

# Ionotropic and metabotropic receptors

If the transmitter binds to receptors in the postsynaptic membrane that are also channel proteins, the synaptic conductance opens and closes rapidly and we speak of *ionotropic receptors*.

If the transmitter binds to receptors that control ion channels through G-proteins, the synaptic conductance opens and closes more slowly and we speak of *metabotropic receptors*.

# **2. Time-course of opening and closing**

The details of synaptic cell biology and morphology are irrelevant for our purposes. We are exclusively concerned with the time-course of the opening and the closing of synaptic conductances, because this is what in the end influences the firing activity the postsynaptic neuron.


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec06-ionotropic.png" width="500">

Postsynaptic receptors and channel proteins create a 'transmitter-gated' conductance in the postsynaptic membrane.  Let's describe this time-varying synaptic conductance $g_s(t)$ as a product of two factors

$\begin{eqnarray}
g_s(t) = \bar g_s \, P_s(t), \qquad\qquad 0 \leq P_s(t) \leq 1
\end{eqnarray}$

where maximal conductance $\bar g_s$ reflects the type and number of channel proteins and open fraction $P_s(t)$ describes time-course of opening and closing of channels.

(Slide-8)

# Relaxation dynamics



We expect the opening and closing of synaptic channel proteins to be governed by rates $\alpha$ and $\beta$, which in turn will depend on the concentration of transmitter molecules in the synaptic cleft (i.e., the conductance is 'transmitter-gated').

To simplify the argument, let's just consider two extreme transmitter levels -- *high* when the synapse is active and *low* when it is inactive. For the resulting extreme rates, we can assume

$\begin{eqnarray}
\alpha_\mathit{high} \gg \beta_\mathit{high} \qquad\qquad \mathrm{and} \qquad\qquad \alpha_\mathit{low} \ll \beta_\mathit{low}
\end{eqnarray}$

Given these assumptions, we expect an **exponential relaxation dynamics** for both opening and closing:

$\begin{eqnarray}
\left.\frac{\partial P_s}{\partial t}\right|_\mathit{high} = \alpha_\mathit{high} \left[ 1 - P_s\right] - \beta_\mathit{high} P_s &\approx& \alpha_\mathit{high} \left[1  -  P_s\right]\qquad\Rightarrow\qquad P_s^{ss}=1
\\
\\
\left.\frac{\partial P_s}{\partial t}\right|_\mathit{low} = \alpha_\mathit{low} \left[ 1 - P_s\right] - \beta_\mathit{low} P_s &\approx&  - \beta_\mathit{low} P_s\qquad\Rightarrow\qquad P_s^{ss}=0
\end{eqnarray}$

Why is this a **relaxation dynamics**?  Because the change $\partial P_s/\partial t$ vanishes at a steady-state $P_s^{ss}$ and grows linearly with the distance from this steady-state.

(Slide-9)



# Approximate time-course of opening and closing

Different functions may be used to approximate the time-course of synaptic conductances. The choice of function depends on the application.

Often, a measured time-course is well fit by a difference between exponentials

$\begin{eqnarray}
P_s(t) = B \, \left[\exp(-t/\tau_1) - \exp(-t/\tau_2) \right], \qquad t \geq 0, \qquad \tau_1>\tau_2, \qquad \int_0^\infty P_s(t) \, dt = B \, (\tau_2-\tau_1)
\end{eqnarray}$

where the characteristic *fall* time is $\tau_\mathit{fall}=\tau_1$ and the characteristic *rise* time is $\tau_\mathit{rise} = \tau_1\tau_2/(\tau_1-\tau_2)$.

The normalization factor $\begin{eqnarray}
B = \left[\left(\frac{\tau_2}{\tau_1}\right)^\frac{\tau_2}{\tau_1-\tau_2} - \left(\frac{\tau_2}{\tau_1}\right)^\frac{\tau_1}{\tau_1-\tau_2} \right]^{-1}
\end{eqnarray}$ ensures that $P(t_\mathit{peak}) = 1$


The peak conductance is reached at time
$\begin{eqnarray}
t_\mathit{peak} = \frac{\tau_1\,\tau_2}{\tau_1-\tau_2} \, \ln\frac{\tau_1}{\tau_2}
\end{eqnarray}$.

D&A Fig. 5-14

(Slide-10)



# Alpha function

Another, simpler approximate function is the *alpha function*. This is what we will use in our simulations.

$\begin{eqnarray}
P_s(t) = \frac{t}{\tau_s} \exp\left(1- \frac{t}{\tau_s} \right), \qquad t \geq 0,  \qquad \int_0^\infty P_s(t) \, dt = e \tau_s
\end{eqnarray}$

where $\tau_s$ is both *rise* and *fall* time and peak conductance  is reached at  $t_\mathit{peak}=\tau_s$.   

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec06_Fig02a.png" width="300">

(Slide-11)

# Exponential decay

Even simpler is an exponential decay:

$\begin{eqnarray}
P_s(t) = \exp\left(- \frac{t}{\tau_s} \right), \qquad t \geq 0, \qquad \int_0^\infty P_s(t) \, dt = \left[-\tau_s \exp\left(- \frac{t}{\tau_s}\right)\right]_0^\infty = \tau_s
\end{eqnarray}$

where the rise is instantaneous and the characteristic *fall* time is  $\tau_\mathit{fall} = \tau_s$

This approximation is appropriate when the rise time can be neglected.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec06_Fig02c.png" width="300">

(Slide-12)

# Synaptic conductance, current and total charge

The synaptic conductance, current, and total charge are related as follows:

$\begin{eqnarray}
g_s(t) = \bar g_s \, P_s(t) \qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\textrm{conductance}
\end{eqnarray}$

$\begin{eqnarray}
I_s(t) = \bar g_s \, P_s(t) \, \left(V_{m} - E_s \right) \qquad\qquad\qquad\qquad\qquad\qquad\textrm{current}
\end{eqnarray}$

$\begin{eqnarray}
Q_s = \int_0^\infty \, I_s(t) \, dt =  \bar g_s \, \left(V_{m} - E_s \right)\int_0^\infty \, P_s(t) \, dt \qquad\qquad\textrm{charge}
\end{eqnarray}$

where $\bar g_s$ is the maximal conductance. The expression for total charge assumes that membrane potential $V_m$ remains constant while the conductance opens and closes.

(Slide-13)

# Total charge of exponential decay

For an exponentially decaying synapse, we have

$\begin{eqnarray}
\int_0^\infty \, P_s(t) \, dt = \int_0^\infty \, e^{-{t}/{\tau_s}} \, dt = \left[ -\tau_s \, e^{-{t}/{\tau_s}}\right]_0^\infty =  \tau_s
\end{eqnarray}$

so that the total charge becomes

$\begin{eqnarray}
Q = \int I_{s} \, dt &=&  \bar g_s   \,\left[ \int_0^\infty \, P_s(t)  \, dt \right] \, \left(V_{ave} - E_{syn} \right) =
\\
&=& \bar g_s  \, \tau_s \, \left(V_{ave} - E_{syn} \right)
\end{eqnarray}$

Thus, the total charge approximately equals *conductance* $\times$ *fall time* $\times$ *driving force*.

(Slide-14)


# Summary of time-course

- The effect of a synapse depends on both strength and speed.

- The speed of ionotropic synapses is governed by the kinetics of transmitter binding.

- Different functions may be used to approximate the opening and closing of synaptic conductance.

- The total synaptic charge is proportional to the maximum conductance and  closing time.

(Slide-15)

# Inhibitory synapses GABA$_A$ and GABA$_B$

Inhibitory synapses open conductances for ions with *hyperpolarizing* reversal potentials, such as $\mathit{K}^+$ or $\mathit{Cl}^-$, which make the membrane potential **more negative**.

The transmitter is often $\gamma-$aminobutyric acid or **GABA** (an amino acid).

# GABA$_A$ and GABA$_B$ receptors

GABA$_A$ receptors are ionotropic and open a fast $\mathit{Cl}^-$ channel, whereas
GABA$_B$ receptors are metabotropic and open a longer-lasting $\mathit{K}^+$ channel.   

In deep cerebellar nuclei, synapses with GABA receptors may be modeled as follows:
$\begin{eqnarray}
I_\mathit{GABA}(t) = g_{syn}(t) \, ( V - E_\mathit{GABA} ) \qquad\qquad E_\mathit{GABA} \approx -75\, mV
\end{eqnarray}$

**Purkinje cells (GABA$_A$)**

$\begin{eqnarray}
g_{syn}(t) = \bar g_{syn} \, e^{-t/\tau}, \qquad\qquad\qquad \bar g_{syn} = 40\,pS, \quad \tau = 5\, ms
\end{eqnarray}$


**Cerebellar granule cells (GABA$_A$ and GABA$_B$)**

$\begin{eqnarray}
g_{syn}(t) = \bar g_1 \, e^{-{t}/{\tau_1}} +  \bar g_2 \, e^{-{t}/{\tau_2}}\qquad\qquad g_1 = 30\,pS, \quad \tau_1 = 5\, ms
\\
\hskip 2.1in g_2 = 10\,pS, \quad \tau_2 = 50\, ms
\end{eqnarray}$

(Slide-18)

Left: Purkinje cell with $40 pS$ of GABA$_A$ receptors

Right: Granule cell with $30 pS$ of GABA$_A$ and $10 pS$ of GABA$_B$ receptors

At a membrane potential of $-65\, mV$, the driving force is $V_m - E_\mathit{GABA} = 10 mV$ and the total synaptic charge is

$\begin{eqnarray}
Q_\mathit{Purkinje} = 40\,pS \cdot 5 \, ms \cdot 10 \, mV = 2 \, aC,
\qquad
 Q_\mathit{Granule} = 1.5 \, aC + 5 \, aC
\end{eqnarray}$

# Excitatory synapses AMPA and NMDA

Excitatory synapses open conductances for ions with *depolarizing* reversal potentials, such as $\mathit{Na}^+$ or $\mathit{Ca}^{+2}$, which make the membrane potential **more positive**.

The transmitter is often **glutamate** (another amino acid).

The receptors are ionotropic with important differences between  types, which are commonly distinguished by the synthetic agonists that they bind specifically, such as **AMPA** ($\alpha$-amino-3-hydroxy-5-methyl-4-isoxazolepropionic acid) or **NMDA** (N-methyl-D-aspartate).

(Slide-20)

# AMPA receptors

Synapses with AMPA receptors in deep cerebellar nuclei may be modeled as follows:

$\begin{eqnarray}
I_\mathit{AMPA} = g_\mathit{AMPA}(t) \, ( V - E_\mathit{AMPA} ) \qquad\qquad E_\mathit{AMPA} \approx 0\, mV
\end{eqnarray}$

$\begin{eqnarray}
g_\mathit{AMPA}(t) = \bar g_\mathit{AMPA} \, B \, \left[ e^{-t/\tau_1} -  e^{-t/\tau_2} \right],\quad\tau_\mathit{rise} = \tau_1\tau_2/(\tau_1-\tau_2), \quad\tau_\mathit{fall}= \tau_1
\end{eqnarray}$

$\begin{eqnarray}
\tau_1 = 1.5\,ms,
\qquad
\tau_2 = 0.09\,ms,
\qquad
\bar g_\mathit{AMPA} = 720 \, pS,
\qquad
B = 1.273
\end{eqnarray}$


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec06-Fig-AMPA.png" width="300">

At a membrane potential of $-65\, mV$, the driving force is $V_m - E_\mathit{AMPA} = 65 mV$ and the total synaptic charge is

$\begin{eqnarray}
Q_\mathit{AMPA} = 720\,pS \cdot 1.273 \cdot (1.5 ms - 0.09 ms) \, \cdot 65 \, mV = 84 \, aC
\end{eqnarray}$

(Slide-21)

# NMDA receptors

NMDA receptors are permeable to several ions  $\mathit{Na}^+$, $\mathit{K}^+$, and $\mathit{Ca}^{2+}$, with an effective reversal potential $E_{syn}^{exc} \approx 0\, mV$.   

Importantly, the conductance depends on both **glutamate levels** and **postsynaptic membrane potential**.  The latter dependence involves extracellular $\left[\mathit{Mg}^{2+}\right]$. When the postsynaptic membrane is at the resting potential, channel pores are blocked by extracellular $\left[\mathit{Mg}^{2+}\right]$.  When the postsynaptic membrane is depolarized, the blockage clears.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec06-BCP-6-21.png" width="300">

Bear, Connors, Paradiso, Fig. 6-21

(Slide-22)

The combined dependence on time $t$, postsynaptic potential $V_{post}$ and extracellular $\mathit{Mg}^{2+}$ concentration can be modeled as follows:

$\begin{eqnarray}
I_\mathit{NMDA}(t) = g_\mathit{NMDA}(t) \, ( V - E_\mathit{NMDA} ) \qquad\qquad E_\mathit{NMDA} \approx 0\, mV
\end{eqnarray}$

$\begin{eqnarray}
g_\mathit{NMDA}(t) = \bar g_\mathit{NMDA} \, B \, \left[ e^{-t/\tau_{decay}} -  e^{-t/\tau_{rise}} \right] \, G_\mathit{NMDA}(V_{post})
\end{eqnarray}$

The factor $G_\mathit{NMDA}$ is a Michaelis-Menten term describing the dependence of $\mathit{Mg}^{2+}$ binding on the postsynaptic membrane potential.

$\begin{eqnarray}
G_\mathit{NMDA}(V_{post}) = \frac{\beta}{\beta + e^{-\alpha\, V_{post}} \, \left[ \mathit{Mg}^{2+}\right]_{out}}
\end{eqnarray}$

$\begin{eqnarray}
\tau_{rise} = 3\,ms,
\quad
\tau_{decay} = 40\,ms,
\quad
\bar g_\mathit{NMDA} = 1200 \, pS,
\quad
B = 1.358
\qquad
\\
\qquad\qquad
\alpha = 0.062\, mV^{-1} ,
\qquad
\beta = 3.57 \, mM,
\qquad
\left[ \mathit{Mg}^{2+}\right]_{out} = 1.2 \, mM
\qquad
\end{eqnarray}$

(Slide-23)

Left: Time-course of synaptic conductance for different postsynaptic membrane potentials.

Right: Dependence of $G_\mathit{NMDA}$ on postsynaptic membrane potential (for physiological $\mathit{Mg}^{2+}$ levels).


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec06-Fig-NMDA.png" width="500">


At a membrane potential of $-65\, mV$, the driving force is $V_m - E_\mathit{NMDA} = 65 mV$ and the total synaptic charge is

$\begin{eqnarray}
Q_\mathit{NMDA} = 1200 \,pS \cdot 1.358 \cdot ( 40 \, ms - 3 \, ms ) \cdot 65 \, mV = 3900 \, aC
\end{eqnarray}$

$\begin{eqnarray}
G_\mathit{NMDA}(-65\, mV) \approx 0.05, \qquad\qquad G_\mathit{NMDA}(-45\,mV) \approx 0.15
\end{eqnarray}$

# Importance of voltage gating

The fact that NMDA receptor channels require both pre- and postsynaptic activity (depolarization) is of supreme importance.  It means that NMDA receptors can act as *coincidence detectors* for pre- and postsynaptic activity.   

This functionality underlies *conditional interactions*, that is, the effect of one activity is conditional on another activity.

The voltage-dependence of NMDA channels is also thought to play a role in synaptic modification and plasticity ("Hebbian learning").

(Slide-25)

# Summary GABA, AMPA, and NMDA


We have introduced simple models for the most common synapse types.

- Inhibitory GABA$_A$ synapses are small and medium fast: $Q=2-5\, aC$, $\tau_{close}=5\, ms$.

- Inhibitory GABA$_B$ synapses also comprise a slow component: $\tau_{close}=50\, ms$.

- Excitatory AMPA synapses are larger and very fast: $Q=80\, aC$,$\tau_{close} = 1.5\,ms$.

- Excitatory NMDA synapses are even larger and slow: $Q=200-600\, aC$, $\tau_{close} = 50\,ms$.

- They are significantly permeable to $Ca^{2+}$ ions.

- Their effectiveness grows with the postsynaptic potential, providing a mechanism for "coincidence detection".

(Slide-26)

# **4. LIF neurons with synapses**

Recall the dynamic equation for the membrane potential $V(t)$ of a LIF neuron

$\begin{eqnarray}
 \tau_m \, \frac{dV(t)}{dt} = -V(t) + E_L  + r_m \, I_e(t)
\end{eqnarray}$

with resting potential $E_L$, leak resistance $r_m=1/g_L$, membrane time constant $\tau_m$, and threshold and reset potentials $V_\mathit{th}$ and $V_\mathit{reset}$, where $I_e(t)$ stands for an (inward) electrode current.

$\begin{eqnarray}
\tau_m=20\,ms, \quad r_m=0.33\, M\Omega, \quad E_L = -70\,mV, \quad V_\mathit{th}= -54\, mV, \quad V_\mathit{reset}=-80\,mV
\end{eqnarray}$

(Slide-27)



# Replace electrode with synapse

Replacing the (inward) electrode current $+I_e(t)$ with an (outward) time-varying synaptic current $-I_s(t)$ gives

$\begin{eqnarray}
 \tau_m \, \frac{dV(t)}{dt} &=& -V(t) + E_L -r_m \, I_s(t) =
 \\
 \\
 &=& -V(t) + E_L -r_m \, g_s \, P_k(t) \, \left[V(t) - E_s\right]
\end{eqnarray}$

with

$\begin{eqnarray}
V_\mathit{PSP}(t) = -\underbrace{r_m \, g_s}_\mathit{strength} \quad \underbrace{P(t)}_\mathit{time-course} \quad \left[\underbrace{V- E_s}_\mathit{driving-force} \right]
\end{eqnarray}$

Note that synaptic strength is determined by the **ratio** of membrane and synaptic conductance:
$\begin{eqnarray}
r_m \, g_s = \frac{g_s}{g_L}
\end{eqnarray}$

(Slide-28)


# Multiple synapses

With multiple synapses of the same kind, we simply add the individual conductances

$\begin{eqnarray}
 \tau_m \, \frac{dV(t)}{dt} = -V(t) + E_L -r_m \, g_s \, \sum_k P_k(t) \, [V(t) - E_s]
\end{eqnarray}$

with

$\begin{eqnarray}
-\underbrace{r_m \, g_s}_{strength} \quad \underbrace{\sum_k P(t)}_{time-course} \quad \left[\underbrace{V-E_s}_{driving-force} \right]
\end{eqnarray}$

(Slide-29)

# Post-synaptic currents and potentials

In the context of the LIF model, we define the postsynaptic (outward) current in a post-synaptic neuron as

$\begin{eqnarray}
I_\mathit{PSC} = -g_s \, \sum_k P_k(t) \, [V(t) - E_s] \qquad\qquad \textrm{post-synaptic  current}
\end{eqnarray}$

and the postsynaptic potential as the contribution to the dynamic equation of a post-synaptic neuron as

$\begin{eqnarray}
V_\mathit{PSP} = -r_m \, g_s \, \sum_k P_k(t) \, [V(t) - E_s] \qquad\qquad \textrm{post-synaptic  potential}
\end{eqnarray}$

Note that **"postynaptic potential"** takes different meanings in different contexts.  In a biological neuron, a postsynaptic potential is sometimes measured in the dendrite and differs from the soma potential. In a single-compartment LIF model, such differences do not exist.

(Slide-30)

# Synaptic effects

To assess the effect of synapses, we rearrange the dynamic equation into the standard form:

$\begin{eqnarray}
\qquad\qquad\qquad\qquad\tau_m^\mathit{eff} \, \frac{dV}{dt} = -V + V_\infty^\mathit{eff}
\end{eqnarray}$

$\begin{eqnarray}
V_\infty^\mathit{eff} =  \frac{E_L + r_m \, g_s \,  \sum_k \, P_k(t) \, E_s }{1 + r_m \, g_s \, \sum_k \, P_k(t) },
\qquad
r_m^\mathit{eff} =  \frac{r_m}{1 + r_m \, g_s \, \sum_k \, P_k(t)},
\qquad
\tau_m^\mathit{eff} = \frac{\tau_m}{1 + r_m \, g_s \, \sum_k \, P_k(t)}
\end{eqnarray}$

Just like any other conductances, synaptic conductance changes the effective equilibrium potential $V_\infty^\mathit{eff}$, the effective membrane resistance $r_m^\mathit{eff}$, and the effective membrane time-constant $\tau_m^\mathit{eff}$.

(Slide-31)

# Synaptic transmission in four logical steps


- Presynaptic spike $t_i$ transiently releases transmitter, thus opening and closing postsynaptic conductance $P_s(t)$, for example with a time-course
$\begin{eqnarray}
\qquad\qquad\qquad P_s(t) = \frac{t-t_i}{\tau_s} \, \exp\left( -\frac{t-t_i}{\tau_s}\right), \qquad\qquad t\geq t_i
\end{eqnarray}$

- Postsynaptic conductance contributes a new term $V_\mathit{PSP}(t)$ in the dynamic equation of the membrane potential:

$\begin{eqnarray}
\qquad\qquad\qquad V_\mathit{PSP}(t) = -r_m \, g_s \, P_s(t) \, \left[V_m(t) - E_s \right]
\end{eqnarray}$

- This new term changes both equilibrium potential $V_\infty(t)$ and time-constant $\tau_m(t)$

$\begin{eqnarray}
\qquad\qquad\qquad V_{\infty}(t) = \frac{E_L + r_m \, g_s \, P_s(t) \, E_s }{1+ r_m \, g_s \, P_s(t) \, E_s}, \qquad \tau^\mathit{eff}(t) = \frac{\tau_m }{1+ r_m \, g_s \, P_s(t) \, E_s}
\end{eqnarray}$

- The membrane potential $V_m(t)$ relaxes towards the new equilibrium potential.

(Slide-32)

# Simulation of excitatory synapses

We model a single LIF neuron being stimulated via excitatory synapses triggered by presynaptic spikes $i$:

$\begin{eqnarray}
V_\mathit{EPSP}(t) = - r_m \, g_s \, \sum_i \, P_i(t) \, [V(t) - E_s]
\end{eqnarray}$

$\begin{eqnarray}
P_{i}(t) =  
\left\{
\begin{array}{cc}
\frac{t-t_i}{\tau_{s}} \, \exp\left(1- \frac{t-t_i}{\tau_{s}} \right) & \mathrm{if} \,\, t\geq t_i \\
0 &  t < t_i \\
\end{array}
\right.
\end{eqnarray}$

$\begin{eqnarray}
E_{s} = 0\, mV \qquad r_m\,g_s=0.25 \qquad \tau_s = 10\,ms
\end{eqnarray}$

The remaining parameters of the LIF model are

$\begin{eqnarray}
r_m=0.33\,M\Omega, \quad\tau_m=20\,ms, \quad E_L = -70\,mV, \quad V_\mathit{th}= -54\, mV, \quad V_\mathit{reset}=-80\,mV
\end{eqnarray}$

We give ourselves the possibility of changing the synapses size -- $r_m \, g_s$ -- a constant electrode current -- $r_m I_e$ -- and the presynaptic firing rate -- $\nu_\mathit{presynaptic}$.

Presynaptic spikes are irregular and generated randomly with **Poisson statistics**. This is explained in the next section.

(Slide-33)

```python
# @title Excitatory synapses onto LIF neuron {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Excitatory synapses onto a LIF neuron
# Define auxiliary functions

def Poisson(Tend, rate):
    N = 2 * round(Tend * rate / 1000);
    dti = -np.log(np.random.rand(N)) / rate
    ti_spike = 1000 * np.cumsum(dti)
    ix       = (ti_spike <= Tend);
    ti_spike = ti_spike[ix];
    return ti_spike

# Define an interactive plot
def plot_excitatory_synapses(gsrm = 0.25, rmIe=10, rate = 20):
    """ Slide-34 """

    #r_m*g_s  = 0.25                            # [1] synaptic strength
    #r_m*I_e = 10                               # [mV] constant electrode current
    #rate = 20                        # [Hz] presynaptic firing rate

    # parameters
    c_m = 10;      # nF/mm^2
    g_L = 3;       # uS/mm^2  (10e-6 S / mm)
    r_m = 1 / g_L;

    g_s = gsrm / r_m;  # deduce g_s
    I_e = rmIe / r_m;  # deduce I_e

    E_L = -70;     # mV
    Vth = -54;     # mV
    Vreset = -80;    # mV
    E_s  = 0;        # mV

    tau_m = 20;     # ms
    tau_s = 10;     # ms

    dt  = 0.02;                  # time step, in ms (0.02 ms)
    T   = 100;                   # total time, in ms (15 ms)
    Ni  = round(T/dt) + 1;       # number of time steps
    ti  = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # allocate vectors for variables to be plotted (membrane potential, steady-state potential, activation variables, currents)

    Vmi     = np.zeros_like(ti)
    Pti     = np.zeros_like(ti)
    EPSPi   = np.zeros_like(ti)
    Vinftyi = np.zeros_like(ti)
    tau_mi  = np.zeros_like(ti)

    ti_dummy = np.zeros_like(ti);

    # make presynaptic spike train

    ti_pre     = Poisson(T, rate)

    # assign initial values where necessary

    Vmi[0]     = E_L                     # membrane potential

    # iterate over successive time steps, starting at resting potential (reversal
    # potential of leak)

    npost = 0;

    for i in range( len(ti) ):

        # current contributions

        ix_spike   = (ti_pre <= ti[i]);
        delt_spike = ti[i] - ti_pre[ix_spike];
        Pt_spike   = delt_spike * np.exp( -delt_spike/tau_s );
        Pti[i]     = np.sum(Pt_spike);

        EPSPi[i]   = -r_m * g_s * Pti[i] * (Vmi[i] - E_s);

        Vinftyi[i] = (E_L + r_m * g_s * Pti[i] * E_s + r_m * I_e) / (1 + r_m * g_s * Pti[i]);

        tau_mi[i]  = tau_m / (1 + r_m * g_s * Pti[i]);

        if i < len(ti)-1:
             # update state variables at time i+1
            Vmi[i + 1] = Vinftyi[i] + (Vmi[i] - Vinftyi[i]) * np.exp(-dt / tau_mi[i])

            if Vmi[i + 1] > Vth:
                Vmi[i + 1] = Vreset;
                npost             = npost + 1;
                ti_dummy[npost-1] = ti[i];


    # truncate list of postsynaptic spikes
    ti_post = ti_dummy[0:npost];

    # Create a figure and a grid of subplots (2 rows, 1 columns)
    fig, axs = plt.subplots(2, 1, figsize=(8, 6))

    # First subplot
    axs[0].plot(ti, Vmi, 'r', ti, Vinftyi, 'r--', ti, EPSPi, 'b');
    for n in range(npost):
        axs[0].plot((ti_post[n],ti_post[n]), (Vreset,0), 'k:');
    axs[0].set_title('Postsynaptic potentials & spikes', fontsize=18);
    axs[0].set_xlabel('time [ms]', fontsize=14);
    axs[0].set_ylabel('potential [mV]', fontsize=14);

    # Second subplot
    axs[1].plot(ti, r_m * g_s * Pti, 'b');
    for i in range(len(ti_pre)):
        axs[1].plot((ti_pre[i],ti_pre[i]), (0,1), 'k:');
    axs[1].set_title('Presynaptic spikes & synaptic conductance', fontsize=18);
    axs[1].set_xlabel('time [ms]', fontsize=14);
    axs[1].set_ylabel('conductance', fontsize=14);


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_excitatory_synapses, gsrm=(0.25, 0.25, 1.0), rmIe=(0, 100, 10), rate=(0, 50, 5))

```

# **5. Short-term depression and facilitation (advanced)**

Synapses show both short-term *depression* and short-term *faciliation* (at different time-scales and in different proportions).  In the first case the amplitude decreases during a rapid series post-synaptic potentials, in the second case it increases.

# Formalization

We wish to model how potential synaptic strength can change with prior use.

State variable $P_\mathit{rel}(t)$ represents the **potential** number of vesicles released at time $t$, or **potential** synaptic strength, in case of a presynaptic spike at time $t$.

We assume that $P_\mathit{rel}(t)$ increments (STF) or decrements (STD) with each use, but relaxes to a default value $P_\mathit{default}$ (low for STF, high for STD) as long as the synapse remains unused.

**Between synaptic events (relaxation):**

$\begin{eqnarray}
\tau_P \, \frac{dP_\mathit{rel}(t)}{dt} = - P_\mathit{rel}(t) + P_\mathit{default},\qquad
\\
\qquad
P_\mathit{rel}(t) = P_\mathit{default} + \left[ P_\mathit{initial} - P_\mathit{default} \right] \, \exp\left(-\frac{t}{\tau_P} \right)
\end{eqnarray}$

**With each synaptic event (decrement or increment):**

$\begin{eqnarray}
P_{rel} \rightarrow f_D\, P_\mathit{rel}(t), \qquad\qquad\qquad 0\leq f_D\leq 1 \qquad{\rm depression}\\
\end{eqnarray}$

$\begin{eqnarray}
P_{rel} \rightarrow P_\mathit{rel}+ f_F \, (1-P_\mathit{rel}), \qquad 0\leq f_F\leq 1 \qquad{\rm facilitation}\\
\end{eqnarray}$

where factors $f_D$ (or $f_F$) describe the fractional decrease (or increase) towards zero (or unity) of $P_\mathit{rel}$ with each synaptic use.

(Slide-41)

```python
# @title Short-term depression or facilitation {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Short-term depression or facilitation of a synapse
# Define auxiliary functions

def Poisson( ti, dt, rate_per_s ):
    rate_per_ms = rate_per_s / 1000;
    hazard_per_bin = rate_per_ms * dt;
    rho_i    = np.random.rand(len(ti))
    ti_spike = np.zeros_like(ti);
    ix       = (rho_i <= hazard_per_bin);
    ti_spike[ix] = 1;
    return ti_spike

def Alphafunc(ti, t_spike, tau_s):
    di       = ( ti - t_spike ) / tau_s;
    ai       = di * np.exp(1 - di)
    ai[ di<0 ] = 0;
    return ai

# Define an interactive plot
def plot_short_term_plasticity(fD=0.75, fF = 0.25, rate = 20):
    """ Slide-42 """

    #flag  = 0.25                    # [0,1] depression or facilitation, respectively
    #fD    = 0.9                     # [1] decrement factor
    #fF    = 0.9                     # [1] increment factor
    #rate  = 20                      # [Herz] spike rate

    # parameters
    tau_s   = 10;      # ms
    tau_PSP = 100;     # ms

    PDzero  = 0.6;
    PDinfty = 1.0;

    PFzero  = 0.4;
    PFinfty = 0.1;

    dt  = 0.02;                  # time step, in ms (0.1 ms)
    T   = 200;                   # total time, in ms (100 ms)
    Ni  = round(T/dt) + 1;       # number of time steps
    ti  = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # obtain presynaptic spikes

    ti_spike = Poisson(ti, dt, rate);

    Nspike     = round( np.sum(ti_spike) );
    Ntimepoint = len(ti);

    # allocate vectors for variables to be plotted

    PDrel_i  = np.zeros_like(ti)
    PSPD_i   = np.zeros_like(ti)
    PSPD_in  = np.zeros((len(ti),Nspike))

    PFrel_i  = np.zeros_like(ti)
    PSPF_i   = np.zeros_like(ti)
    PSPF_in  = np.zeros((len(ti), Nspike))

    # assign initial values where necessary

    PDrel_i[0] = PDzero;
    PFrel_i[0] = PFzero;

    # iterate over successive time steps, starting at resting potential (reversal
    # potential of leak)

    n =  0;

    for i in range( len(ti)-1 ):

        # spike or no spike
        if ti_spike[i] == 1:
            PDrel_i[i+1] = fD * PDrel_i[i];
            PFrel_i[i+1] = PFrel_i[i] + fF * (1 - PFrel_i[i]);

            PSPD_i = PSPD_i + PDrel_i[i] * Alphafunc(ti, ti[i], tau_s);
            PSPF_i = PSPF_i + PFrel_i[i] * Alphafunc(ti, ti[i], tau_s);

            PSPD_in[:,n] = PDrel_i[i] * Alphafunc(ti, ti[i], tau_s);
            PSPF_in[:,n] = PFrel_i[i] * Alphafunc(ti, ti[i], tau_s);

            n = n + 1;
        else:
            PDrel_i[i+1] = PDinfty + ( PDrel_i[i] - PDinfty) * np.exp( -dt / tau_PSP );
            PFrel_i[i+1] = PFinfty + ( PFrel_i[i] - PFinfty) * np.exp( -dt / tau_PSP );


    # Create a figure and a grid of subplots (1 rows, 1 columns)
    fig, axs = plt.subplots(1, 2, figsize=(10, 4))

    # Depression plot
    # Facilitation plot

    axs[0].plot(ti, PDrel_i, 'r:', ti, PSPD_i, 'k'  );
    for n in range(Nspike):
        axs[0].plot(ti, PSPD_in[:,n], 'k:');
    axs[0].plot(ti, PDrel_i, 'r:', ti, PSPD_i, 'k'  );
    axs[0].set_title('PSP and release probability with STD', fontsize=18);
    axs[0].set_xlabel('time [ms]', fontsize=14);
    axs[0].set_ylabel('fraction', fontsize=14);

    axs[1].plot(ti, PFrel_i, 'r:', ti, PSPF_i, 'k'  );
    for n in range(Nspike):
        axs[1].plot(ti, PSPF_in[:,n], 'k:');
    axs[1].set_title('PSP and release probability with STF', fontsize=18);
    axs[1].set_xlabel('time [ms]', fontsize=14);
    #axs[1].set_ylabel('fraction');


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_short_term_plasticity, fD=(0.05, 0.95, 0.05), fF = (0.05, 0.95, 0.05), rate = (5, 50, 5) )
```

# How to handle stochasticity?

Quantifying the overall effect of STD/STF is not straightforward, because of the irregular timing of presynaptic spikes.

If two spikes are separated by time $t_\mathit{isi}$ and release fraction is changed by the preceeding spike from $P_\mathit{initial}$ to $P_\mathit{post}$ and recovers to $P_\mathit{recover}$ until the succeeding spike, these values are related by

$\begin{eqnarray}
P_\mathit{recover} = P_\mathit{default} + \left[P_\mathit{initial} - P_\mathit{default} \right] \, \exp\left(-\frac{t_\mathit{isi}}{\tau_P} \right) \qquad\qquad
\\
\qquad P_\mathit{recover} = f_D \,  P_\mathit{initial} \qquad \mathrm{or} \qquad P_\mathit{recover} = P_\mathit{initial} + f_F (1- P_\mathit{initial})
\end{eqnarray}$

However, presynaptic events occur irregularly, so that $t_\mathit{isi}$, $P_\mathit{initial}$, and $P_\mathit{recover}$ all take different values every time.

(Slide-43)



# Poisson spike train statistics

If presynaptic spikes are independent from each other -- with the same probability of $p^\mathit{spike}_{\Delta t}$ for every small interval $\Delta t$ -- we speak of **Poisson spikes** or **Poisson statistics**.

A consequence of independence is that  the probability of interspike intervals $t_{isi}$ decreases geometrically with duration:

$\begin{eqnarray}
P\left( t_{isi} \right) = \left(1-p^\mathit{spike}_{\Delta t}\right)^{t_{isi}/\Delta t}
\end{eqnarray}$

because the no-spike probability of $1 - p^\mathit{spike}_{\Delta t}$ is also the same for every small interval $\Delta t$.

In the limit of $\Delta t \to 0$, this becomes an exponential distribution
$\begin{eqnarray}
P\left( t_{isi} \right)
= r \, e^{-r \, t_{isi}}, \qquad\qquad \int_0^\infty P\left( t_{isi} \right) \, dt_{isi} = 1
\end{eqnarray}$

where $r$ is the rate (in Hz) of Poisson spikes.

$\begin{eqnarray}
%S(t_{isi}) = 1 - \int_0^{t_{isi}} P\left( t' \right) \, dt' = 1 - %\left[ - e^{-r t'}\right]_0^{t_{isi}} = e^{-r t_{isi}}
%\\
%u = e^{-r t_{isi}} \qquad \Leftrightarrow \qquad t_{isi} = -\frac{1}{r} \log u
\end{eqnarray}$

(Slide-44)


# Probability-weighted average

Returning to our problem of handling stochasticity, we now see a way to average over irregular spike trains.

The **probability-weighted average recovery** during the random the intervals $t_\mathit{isi}$ between Poisson spikes with rate $r$ can be computed as

$\begin{eqnarray}
\left\langle{\exp\left(-\frac{t_\mathit{isi}}{\tau_P}\right)}\right\rangle &=& \int_0^\infty \, \exp\left(-\frac{t_\mathit{isi}} {\tau_P}\right) \, P\left(t_\mathit{isi}\right) \, dt_\mathit{isi} =
\\
\\
&=& r \, \int_0^\infty \,  \exp\left(-t_\mathit{isi} \, r -\frac{t_\mathit{isi}}{\tau_P} \right) \, dt_\mathit{isi} = \frac{r\, \tau_P}{1+r\, \tau_P}
\end{eqnarray}$

where the angle brackets $\left\langle \right\rangle$ denote the probability-weighted average, which is also called 'expected value' or 'expectation value'.

(Slide-45)

# Average release probability

We seek the average release fraction $\left\langle P_\mathit{rel}\right\rangle$ after recovery over an average interval $\left\langle t_\mathit{isi}\right\rangle$.

Specifically, we are interested in the steady-state condition where, on average, the release fraction will be the same before and after covery:
$\left\langle P_\mathit{rel}\right\rangle = \left\langle P_\mathit{initial}\right\rangle$

Starting from $\left\langle P_\mathit{initial}\right\rangle$, the exponential recovery of depression
$\begin{eqnarray}
P_\mathit{rel}(t_{isi}) = P_\mathit{default} + \left[f_D \,  \langle P_\mathit{initial}\rangle - P_\mathit{default} \right] \, \exp\left(-\frac{t_\mathit{isi}}{\tau_P} \right)
\end{eqnarray}$

$\ldots$ after probability-weighted averaging becomes  $\ldots$

$\begin{eqnarray}
\langle{P_\mathit{rel}}\rangle = P_\mathit{default} + \left[f_D \,  \langle{P_\mathit{initial}}\rangle - P_\mathit{default} \right] \, \frac{r\, \tau_P}{1+r\, \tau_P}
\end{eqnarray}$

$\ldots$ which with $\left\langle P_\mathit{rel}\right\rangle = \left\langle P_\mathit{initial}\right\rangle$ rearranges to

$\begin{eqnarray}
%\langle{P_\mathit{rel}}\rangle (1+r\, \tau_P - f_D r\, \tau_P)= P_\infty(1+r\, \tau_P)  - P_\infty r \, \tau_P
\end{eqnarray}$
$\begin{eqnarray}
\langle{P_\mathit{rel}}\rangle = \frac{P_\mathit{default}}{1 + r \, (1-f_D) \, \tau_P}\qquad\qquad
\end{eqnarray}$

For facilitation, the exponential recovery is

$\begin{eqnarray}
P_\mathit{rel}(t_{isi}) = P_\mathit{default} + \left[\langle P_\mathit{initial}\rangle + f_F \, \left(1 - \langle P_\mathit{initial}\rangle\right) - P_\mathit{default} \right] \, \exp\left(-\frac{t_\mathit{isi}}{\tau_P} \right)
\end{eqnarray}$

$\ldots$ which, after probability-weighted averaging, becomes  $\ldots$

$\begin{eqnarray}
\langle{P_\mathit{rel}}\rangle = P_\mathit{default} + \left[\langle P_\mathit{initial}\rangle + f_F \, \left(1 - \langle P_\mathit{initial}\rangle\right) - P_\mathit{default} \right] \, \frac{r\, \tau_P}{1+r\, \tau_P}
\end{eqnarray}$

$\ldots$ which with $\left\langle P_\mathit{rel}\right\rangle = \left\langle P_\mathit{initial}\right\rangle$rearranges to

$\begin{eqnarray}
%\langle{P_\mathit{rel}}\rangle (1+r\, \tau_P - r\, \tau_P + f_F r\, \tau_P)%= P_\mathit{default}(1+r\, \tau_P)  - P_\mathit{default} r \, \tau_P + f_F r\, \tau_P
\end{eqnarray}$
$\begin{eqnarray}
\langle{P_\mathit{rel}}\rangle =
\frac{P_\mathit{default} + f_F r \, \tau_P}{1 + f_F r \, \tau_P}\qquad\qquad\qquad
\end{eqnarray}$

(Slide-46)

# Average transmission

The full postsynaptic effect depends on both synaptic strength (release probability) and presynaptic activity (firing rate). The average release per unit time (**average transmission**), is therefore the number of spikes per unit time (firing rate $r$) times the average release per spike:

$\begin{eqnarray}
\left\langle T_{rel}\right\rangle = r \, \left\langle P_{rel} \right\rangle =
\frac{r \, P_\infty + f_F r^2 \, \tau_P}{1 + f_F r \, \tau_P} \qquad \mathrm{or} \qquad =  \frac{r\,P_0}{1 + r \, (1-f_D) \, \tau_P}
\end{eqnarray}$

Unsurprisingly, *facilitation amplifies* the effect of rising firing rates, whereas *depression attenuates* this effect.

# Transient effect of short-term depression (STD)

The transient effect is to **transmit differences** in firing rates.   If the firing rate doubles from $r$ to $2r$, the transmission $T_{rel}$ initially doubles, too:

$\begin{eqnarray}
\left\langle T_{old}\right\rangle = r \, \left\langle P_{rel}\right\rangle
\qquad\rightarrow\qquad
\left\langle T_{new}\right\rangle = 2r \, \left\langle P_{rel}\right\rangle = 2 \, \left\langle T_{old}\right\rangle
\end{eqnarray}$

# Sustained effects of short-term depression (STD)

The sustained effect is to **attenuate differences** in firing rates.   If a sustained firing rate $r$ doubles to $2r$, the sustained transmission increases by a less than $2$:

$\begin{eqnarray}
\left\langle T_{old}\right\rangle = \frac{P_0 \, r}{1 + r \, (1-f_D) \, \tau_P}
\qquad\rightarrow\qquad
\left\langle T_{new}\right\rangle = \frac{2 \, P_0 \, r }{1 + 2 \, r \, (1-f_D) \, \tau_P}
\end{eqnarray}$

$\begin{eqnarray}
\frac{\left\langle T_{new}\right\rangle}{\left\langle T_{old}\right\rangle} = \frac{2 + 2 r \, (1-f_D) \, \tau_P}{2 + 2 \, r \, (1-f_D) \, \tau_P - 1}= 1 + \frac{1}{1 + 2 \, r \, (1-f_D) \, \tau_P} \leq 2
\end{eqnarray}$

```python
# @title STD larger simulation {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Short-term depression larger simulation
# Define auxiliary functions

def Poisson( Tend, rate ):
    t_isi     = np.random.exponential(1/rate, round( 2 * Tend * rate) );
    #print( 1/np.mean(t_isi) );
    ti        = np.cumsum(t_isi);
    ix        = (ti <= Tend);
    ti_spike  = ti[ix];
    return ti_spike


# Define an interactive plot
def plot_short_term_depression_large(fDA=0.6, fDB=0.8, Nrepeat=50):
    """ Slide-50 """

    #fD    = 0.9                     # [1] decrement factor

    #Nrepeat = 10                    # number of simulations

    # parameters
    tau_PSP = 0.5;     # s

    Pinfty  = 1.0;

    # configure presynaptic spikes
    T        = [0.2, 0.3, 0.5, 0.2];
    R        = [25,  100,  10,  40];
    CT       = np.cumsum(T);

    # time axis
    Tend     = CT[3];
    dt       = 0.001;       # time step, in s (1 ms)
    ti       = np.linspace(0, Tend, round(Tend/dt) + 1);
    Ni       = len(ti);

    # initial release probability
    PrelA_ini = Pinfty / (1 + (1-fDA) * R[0] * tau_PSP);
    PrelB_ini = Pinfty / (1 + (1-fDB) * R[0] * tau_PSP);
    #print( (PrelA_ini, PrelB_ini) )

    # firing rate, above which Prel saturates
    RhalfA     = 1 / ((1-fDA)*tau_PSP);
    RhalfB     = 1 / ((1-fDB)*tau_PSP);
    #print( (RhalfA,RhalfB) )

    # allocate vectors for variables to be plotted
    PrelA_in = np.zeros((Ni,Nrepeat));
    PrelB_in = np.zeros((Ni,Nrepeat));

    # assign initial values
    PrelA_in[0,:] = PrelA_ini;
    PrelB_in[0,:] = PrelB_ini;

    # Iterate over repetions
    for n in range(Nrepeat):

       # generate presynaptic spikes
       tk_0  = Poisson( T[0], R[0] );
       tk_1  = Poisson( T[1], R[1] ) + CT[0];
       tk_2  = Poisson( T[2], R[2] ) + CT[1];
       tk_3  = Poisson( T[3], R[3] ) + CT[2];

       tk_spike = np.concatenate((tk_0, tk_1, tk_2, tk_3));
       Nspike   = len(tk_spike);

       i0 = 0;
       # iterate over spikes
       for k in range( Nspike ):  # stops at Nspike-1
           # bin of next spike
           i1 = round( tk_spike[k] / dt - 0.5 );

           # fill bins until next spike
           for i in range(i0,i1-1):   # starts with filling bin i0+1
                                      # stops after filling bin i1-1
               PrelA_in[i+1,n] = Pinfty + ( PrelA_in[i,n] - Pinfty) * np.exp( -dt / tau_PSP );
               PrelB_in[i+1,n] = Pinfty + ( PrelB_in[i,n] - Pinfty) * np.exp( -dt / tau_PSP );

           #decrement and fill bin i1
           PrelA_in[i1,n] = fDA * PrelA_in[i1-1,n];
           PrelB_in[i1,n] = fDB * PrelB_in[i1-1,n];

           # next spike becomes previous spike
           i0 = i1;

       # fill any remaining bins after last spike
       i1 = Ni-1;
       for i in range(i0,i1): # starts with filling bin i0+1
                              # stops after filling bin i1
           PrelA_in[i+1,n] = Pinfty + ( PrelA_in[i,n] - Pinfty) * np.exp( -dt / tau_PSP );
           PrelB_in[i+1,n] = Pinfty + ( PrelB_in[i,n] - Pinfty) * np.exp( -dt / tau_PSP );

    # average over repeated simulations
    PrelA_i  = np.mean(PrelA_in, axis=1);
    PrelB_i  = np.mean(PrelB_in, axis=1);

    # seperate input ranges
    ix0      =                  (ti <= CT[0] );
    ix1      = (ti >= CT[0] ) & (ti <= CT[1] );
    ix2      = (ti >= CT[1] ) & (ti <= CT[2] );
    ix3      = (ti >= CT[2] ) & (ti <= CT[3] );

    # concatenate results from different input ranges
    ti_i    = np.concatenate((ti[ix0], ti[ix1], ti[ix2], ti[ix3]));

    PrelA0_i = PrelA_i[ix0];
    PrelA1_i = PrelA_i[ix1];
    PrelA2_i = PrelA_i[ix2];
    PrelA3_i = PrelA_i[ix3];

    TrelA0_i = R[0] * PrelA_i[ix0];
    TrelA1_i = R[1] * PrelA_i[ix1];
    TrelA2_i = R[2] * PrelA_i[ix2];
    TrelA3_i = R[3] * PrelA_i[ix3];

    PrelB0_i = PrelB_i[ix0];
    PrelB1_i = PrelB_i[ix1];
    PrelB2_i = PrelB_i[ix2];
    PrelB3_i = PrelB_i[ix3];

    TrelB0_i = R[0] * PrelB_i[ix0];
    TrelB1_i = R[1] * PrelB_i[ix1];
    TrelB2_i = R[2] * PrelB_i[ix2];
    TrelB3_i = R[3] * PrelB_i[ix3];

    TrelA_i  = np.concatenate((TrelA0_i, TrelA1_i, TrelA2_i, TrelA3_i));
    TrelB_i  = np.concatenate((TrelB0_i, TrelB1_i, TrelB2_i, TrelB3_i));

    # Create a figure and a grid of subplots (1 rows, 1 columns)
    fig, axs = plt.subplots(1, 1, figsize=(10, 4))

    # Depression plot
    axs.plot( ti_i, TrelB_i, 'r', ti_i, TrelA_i, 'b' );
    axs.set_title('Average transmission for fDA=0.6 (blue) and fDB=0.8 (red)',fontsize=20);
    axs.set_xlabel('time [s]', fontsize=14);
    axs.set_ylabel('T_rel', fontsize=14);

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_short_term_depression_large, fDA=(0.05, 0.95, 0.05), fDB=(0.05, 0.95, 0.05), Nrepeat = (1, 500, 1) )
```

# Summary of short-term plasticity

- Short-term plasticity temporarily changes the probability of presynaptic transmitter release.

- It presumably serves multiple functions, for example, stabilizing network activity.

- Short-term depression transmits transient changes, but attenuates sustained changes, in presynaptic activity.

- In quantifying the effects of short-term plasticity, we have picked up some useful new tricks:

i) **Poisson spike trains** with exponentially distributed inter-spike-intervals $t_{isi}$.

ii) Probability-weighted averaging an exponential decay to define **expectation value** $\left\langle P_{rel}\right\rangle$.

iii) Formulating a **self-consistency condition** that can be solved for $\left\langle P_{rel}\right\rangle$.

iv) Brute-force **numerical simulation**.

# **General summary**


- Synaptic transmission is too complex for detailed physical modeling.

- Descriptive formulas capture amplitude and time-course of post-synaptic conductance.

- GABA, AMPA, and NMDA synapses differ in strength and speed.

- NMDA synapses depend on presynaptic transmitter release and on postsynaptic potential (``coincidence detection'').

- Synaptic conductances affect equilibrium potential, membrane resistance, and membrane time-constant.

- Short-term depression transmits transient changes in presynaptic firing, but attenuates sustained changes.

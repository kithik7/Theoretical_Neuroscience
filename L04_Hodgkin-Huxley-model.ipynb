# **1. Action potential and squid giant axon**

Action potentials are fast and stereotypical excursions of the membrane potentail. A **rapid depolarization** to positive potentials ($\sim 0.5 \, ms$) is followed by an almost equally **rapid repolarization** ($\sim 0.5 \, ms$) to negativ potentials and a longer period of **after-hyperpolarization** ($\sim 4 \, ms$).

# Noise-driven transitions

Transitions between open and closed states are typically noise-driven and nearly instantaneous. "Noise-driven" means that they occur spontanteously (unprompted) and with constant rates (measured in Hz) of opening and closing, which however are influenced by variables such as temperature and membrane voltage.

# Population of channel proteins

Consider a population of channel proteins, with $[A_{open}]$ channels open and $[A_{closed}]$ channels closed. The total conductance ($\mu S$ to $mS$) is the number of open channels times conductance per channel ($20pS$ to $50pS$). Thus, the total conductance can vary almost continuously between zero and its maximal value.

<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec04_Fig05c.png" width="250">

(Slide-10)



# Reaction kinetics at steady-state

* Transitions of channel proteins are statistically independent of each other.

* The population rate of *opening* is the indivdual rate $\alpha$ , times the number of *closed* channels (i.e., which are available for opening).

$\begin{eqnarray}
\qquad\qquad\frac{d}{dt} \left[ A_{open} \right] = \alpha \left[ A_{closed} \right]
\end{eqnarray}$

* A similar relation holds for the rate of *closing*.

$\begin{eqnarray}
\qquad\qquad\frac{d}{dt} \left[ A_{closed} \right] = \beta \left[ A_{open} \right]
\end{eqnarray}$

* At steady-state, the population rates of *closing* and *opening* are the same:

$\begin{eqnarray}
\qquad\qquad\frac{d}{dt} \left[ A_{open} \right]  = \frac{d}{dt} \left[ A_{closed} \right] = \alpha \left[ A_{closed} \right] = \beta \left[ A_{open} \right] \qquad \Rightarrow \qquad \frac{\left[ A_{open} \right]}{\left[ A_{closed} \right]} =\frac{\alpha}{\beta}
\end{eqnarray}$

In other words, any imbalance in rates results in a proportional imbalance in concentrations.

(Slide-11)

# Reaction kinetics outside steady-state

Outside the steady-state, the population rate for the fraction of *open* channels $x$ is

$\begin{eqnarray}
\qquad\qquad \frac{d x(t)}{dt} &=& \alpha \, (1-x) - \beta \, x
\\
\frac{d x(t)}{dt} &=& \alpha  - (\alpha + \beta) \, x
\\
\qquad\qquad\frac{1}{\alpha+\beta} \, \frac{d x(t)}{dt} &=& - x + \frac{\alpha}{\alpha + \beta}
\end{eqnarray}$


<img src="https://raw.githubusercontent.com/jochen-braun/Theoretical-Neuroscience-I/main/Lec04_Fig11.png" width="400">

The dynamic equation for $x$ reveals an **exponential relaxation** with steady-state $x_\infty$ (equilibrium fraction) and time-constant $\tau_x$ (relaxation time):

$\begin{eqnarray}
\qquad\qquad x_\infty = \frac{\alpha}{\alpha + \beta} \qquad\qquad \tau_x = \frac{1}{\alpha+\beta}
\end{eqnarray}$

(Slide-12)

# Summary

* Individual channel proteins transition spontaneously and independently between *open* and *closed* states.

* Individual transition rates are given by *reaction rates*.

* Population transition rates also depend on the fraction of  channel proteins that are available for the transition.

* A dynamic equation governs changes in the fraction of open channels and, therefore, in the population conductance.

# Activation energy and Boltzmann factor
Moving a net charge $q$ over a potential difference $V$ either requires or releases energy $E = q \cdot  V$.  This energy constitutes an **activation energy** for the conformation change, which can either favour or disfavour this change.

A negative $E$ *energetically favours* the change, while a positive $E$ *energetically disfavours* the change.

In reaction kinetics, activation energies change the rate of thermally driven reactions by a **Boltzmann factor**. Accordingly, we expect a combined voltage and temperature dependence of the opening and closing rates of channel proteins, as follows

$\begin{eqnarray}
\alpha \propto \exp\left(-\frac{E_\alpha}{kT}\right), \qquad\qquad \beta \propto \exp\left(-\frac{E_\beta}{kT}\right)
\end{eqnarray}$

where $E_\alpha = q \cdot V$ and $E_\beta = q \cdot (-V)$ are the activation energies of opening and closing, respectively, $T$ is temperature and $k$ is the Boltzmann constant. Note that opening and closing involve opposite potential differences.

(Slide-15)


# A concrete example

Let's consider a concrete example for the voltage-dependence of opening and closing rates:

$\begin{eqnarray}
\alpha(V) = A_\alpha \exp\left(-B_\alpha \, V \right)  \qquad\qquad\qquad   \beta(V) = A_\beta \exp\left(- B_\beta \, V \right)
\end{eqnarray}$

$\begin{eqnarray}
A_\alpha = 1.22/ms \qquad\qquad\qquad & A_\beta = 0.056/ms
\\
B_\alpha = - 0.04 / mV \qquad\qquad\qquad & B_\beta = 0.0125 / mV
\end{eqnarray}$

where $A_\alpha$ and $A_\beta$ parametrize the kinetic rates and $B_\alpha$ and $B_\beta$ the voltage-dependence. Temperature and Boltzmann constant are omitted for simplicity.

(Slide-16)

```python
# @title Voltage dependent conductance {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Concrete example of voltage-dependent conductance

# parameters

Aalpha = 1.22
Abeta  = 0.056
Balpha = -0.04
Bbeta  = 0.0125

# voltage range
Vi  = np.linspace(-100, 0, 100); # voltage vector, from -100 mV to 0 mV

Alphai = Aalpha * np.exp(-Balpha * Vi)
Betai  = Abeta  * np.exp(-Bbeta  * Vi)

# Create a figure and a grid of subplots (2 rows, 1 columns)
fig, axs = plt.subplots(1, 2, figsize=(6, 3))

# First subplot
axs[0].plot(Vi, Alphai, 'r');
axs[0].set_title('Opening');
axs[0].set_xlabel('potential [mV]');
axs[0].set_ylabel('rate [1/ms]');


# Second subplot
axs[1].plot(Vi, Betai, 'b');
axs[1].set_title('Closing');
axs[1].set_xlabel('potential [mV]');
axs[1].set_ylabel('rate [1/ms]');

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

# Exponential becomes sigmoidal

Recall that the fraction of open channels $x$ relaxes exponentially, with a time-constant and steady-state determined by opening and closing rates:

$\begin{eqnarray}
\tau_x \frac{dx}{dt} = -x + x_\infty\qquad\qquad
\\
\qquad\qquad\tau_x = \frac{1}{\alpha+\beta},\qquad\qquad x_\infty = \frac{\alpha}{\alpha+\beta},
\end{eqnarray}$

An *exponential* voltage-dependence of opening and closing rates, $\alpha$ and $\beta$, implies a *sigmoidal* voltage-dependence of time-constant and steady-state:

$\begin{eqnarray}
\qquad\qquad\tau_x(V) = \frac{1}{\alpha(V)+\beta(V)} = \frac{1}{A_\alpha \exp\left[ -B_\alpha V \right] + A_\beta \exp\left[ -B_\beta V \right]}
\end{eqnarray}$
$\begin{eqnarray}
\qquad\qquad x_\infty(V) = \frac{\alpha(V)}{\alpha(V)+\beta(V)} = \frac{1}{1 + \frac{A_\beta}{A_\alpha} \exp\left[ (B_\alpha-B_\beta) V \right]}
\end{eqnarray}$

We can now understand how membrane voltage alters the exponential relaxation of conductance (fraction of open channels)!

(Slide-18)

```python
# @title Exponential relaxation of conductance {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# See how the conductance (fraction of open channels) relaxes for different choices of membrane potential!

# Define an interactive plot
def plot_conductance_relaxation(X_0=0.1, V_m=-65):
    """(Slide-19)"""

    #X_0     = 0.1                                 # [1] initial fraction of open channels
    #V_m = 0                                       # [mV] membrane potential

    # parameters

    Aalpha = 1.22
    Abeta  = 0.056
    Balpha = -0.04
    Bbeta  = 0.0125

    # voltage range
    Vi  = np.linspace(-100, 0, 100); # voltage vector, from -100 mV to 0 mV

    Alphai = Aalpha * np.exp(-Balpha * Vi)
    Betai  = Abeta  * np.exp(-Bbeta  * Vi)

    Taui_x   = 1 / (Alphai + Betai)
    Xi_infty = Alphai / (Alphai + Betai)

    # rates for chosen membrane potential
    Alpha0 = Aalpha * np.exp(-Balpha * V_m)
    Beta0  = Abeta  * np.exp(-Bbeta  * V_m)

    # time-constant and steady-state for chosen membrane potential
    Tau0_x   = 1 / (Alpha0 + Beta0)
    X0_infty = Alpha0 / (Alpha0 + Beta0)

    # time range
    tk  = np.linspace(0, 5, 100); # time vector, from 0 to 5 ms

    # time-course of exponential relaxation
    Xk = X0_infty + (X_0 - X0_infty) * np.exp( -tk / Tau0_x );

    # Create a figure and a grid of subplots (2 row, 3 columns)
    fig, axs = plt.subplots(2, 3, figsize=(9, 6))

    # First subplot
    axs[0,0].plot(Vi, Alphai, 'r')
    axs[0,0].plot(V_m, Alpha0, 'ro')
    axs[0,0].set_ylim(0,1.2)
    axs[0,0].set_xlabel('potential [mV]')
    axs[0,0].set_ylabel('rate [1/ms]')
    axs[0,0].set_title('Opening');

    # Second subplot
    axs[1,0].plot(Vi, Betai, 'b')
    axs[1,0].plot(V_m, Beta0, 'bo')
    axs[1,0].set_xlabel('potential [mV]')
    axs[1,0].set_ylabel('rate [1/ms]')
    axs[1,0].set_ylim(0,1.2)
    axs[1,0].set_title('Closing')

    # Fourth subplot
    axs[0,1].plot(Vi, Taui_x, 'm')
    axs[0,1].plot(V_m, Tau0_x, 'mo')
    axs[0,1].set_title('Time constant');
    axs[0,1].set_xlabel('potential [mV]')
    axs[0,1].set_ylabel('time [ms]')

    # Fifth subplot
    axs[1,1].plot(Vi, Xi_infty, 'm')
    axs[1,1].plot([-100,0], [X0_infty,X0_infty], 'k--')
    axs[1,1].plot(V_m, X0_infty, 'mo')
    axs[1,1].set_xlabel('potential [mV]')
    axs[1,1].set_ylabel('open fraction')
    axs[1,1].set_ylim(0,1)
    axs[1,1].set_title('Steady-state')

    # Sixth subplot
    axs[1,2].plot(tk, Xk, 'k')
    axs[1,2].plot([0,5], [X0_infty,X0_infty], 'k--')
    axs[1,2].plot([0,5], [X_0,X_0], 'k:')
    axs[1,2].set_xlabel('time [ms]')
    axs[1,2].set_ylabel('open fraction')
    axs[1,2].set_ylim(0,1)
    axs[1,2].set_title('Time-course')


    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_conductance_relaxation, X_0=(0, 1, 0.01), V_m=(-100, 0, 1))
```

* The reaction kinetics of channel proteins with charged groups depends on membrane voltage.

* Reaction rates are expected to depend *exponentially* on membrane voltage.

* The time-constant and steady-state are expected to depend *sigmoidally* on membrane voltage.

* Concrete example: depending on the membrane potential, the conductance (open fraction) relaxes either slowly to a low steady-state, or quickly to a high steady-state.

# **4. Active $\mathit{Na}^+$ and $\mathit{K}^+$ conductances**

* HH explained the action potential with active $\mathit{Na}^+$ and $\mathit{K}^+$ conductances.

* "Active" means that conductance changes with the membrane potential, so that self-reinforcing feedback loops become possible.

* HH postulated three **activation variables** ranging between 0 and 1: $\mathit{Na}^+$-activation $m$, $\mathit{Na}^+$-**in**activation $h$, and $\mathit{K}^+$-activation $n$.

* From a contemporary perspective, $m$, $h$ and $n$ correspond to open fractions of **channel subunits** (not channels!). HH knew nothing about channels and subunits!

* Activation variables relax with time-constants and steady-states that depend sigmoidally on membrane voltage.

* Opening and closing rates of $m$, $h$ and $n$ depend exponentially on membrane voltage.

* Hodgkin and Huxley determined these dependencies experimentally for the squid giant axon.

(Slide-21)

# Active $\mathit{Na}^+$ conductance

$\mathit{Na}^+$ conductance depends on both **activation** and **inactivation** variables, $m(V,t)$ and $h(V,t)$, respectively, with $m\in [0,1]$, $h\in [0,1]$.

$\begin{eqnarray}
\qquad\qquad g_{\mathit{Na}^+}(V,t) = \bar g_{\mathit{Na}^+} \, m^3 \, h  
\end{eqnarray}$

Conductance is proportional to the *third power* of $m$ and to the *first power* of $h$. Here, $\bar g_{\mathit{Na}^+}$ denotes the maximal available conductance.

During the AP, a *rapid increase* of $m$ activates the $\mathit{Na}^+$-conductance and *depolarises* the neuron.  Subsequently, a *slow decrease* of $h$ inactivates the $\mathit{Na}^+$-conductance and helps *re-polarise* the neuron.

(Slide-22)

# Active $\mathit{K}^+$ conductance

$\mathit{K}^+$ conductance depends on **activation variable** $n(V,t)$, with $n\in [0,1]$.

$\begin{eqnarray}
\qquad\qquad g_{\mathit{K}^+}(V,t) = \bar g_{\mathit{K}^+} \, n^4  
\end{eqnarray}$

Conductance is proportional to the *fourth power* of $n$, where $\bar g_{\mathit{K}^+}$ denotes the maximal available conductance.

During the AP, a *slow increase* of $n$ activates the $\mathit{K}^+$-conductance and first *re-polarises* and then *hyper-polarizes* the neuron.  Subsequently, the *slow decrease* of $n$ inactivates the $\mathit{K}^+$-conductance and restores the resting potential.

# Four dynamic equations

Summarizing the above, the four dynamic equations of the HHM are

$\begin{eqnarray}
\tau_m \, \frac{dV}{dt} = I_e - i_m \qquad\qquad\qquad\qquad \frac{dn}{dt} = -\frac{1}{\tau_n} \, \left[ n - n_\infty\right]
\end{eqnarray}$

$\begin{eqnarray}
\frac{dm}{dt} = -\frac{1}{\tau_m} \, \left[ m - m_\infty\right]
\qquad\qquad\qquad\quad
\frac{dh}{dt} = -\frac{1}{\tau_h} \, \left[ h - h_\infty\right]
\\
\\
\end{eqnarray}$

The membrane current has three components, sodium, potassium, and leak:

$\begin{eqnarray}

\\
\end{eqnarray}$

# Voltage dependence

Finally, we need to specify the voltage-dependence of the time-constants and steady-states of activation variables.

\begin{eqnarray}
m_\infty(V) = \frac{\alpha_m(V)}{\alpha_m(V) + \beta_m(V)}
\qquad\qquad
\tau_m(V) = \frac{1}{\alpha_m(V) + \beta_m(V)}
\\
h_\infty(V) = \frac{\alpha_h(V)}{\alpha_h(V) + \beta_h(V)}
\qquad\qquad
\tau_h(V) = \frac{1}{\alpha_h(V) + \beta_h(V)}
\\
n_\infty(V) = \frac{\alpha_n(V)}{\alpha_n(V) + \beta_n(V)}
\qquad\qquad
\tau_n(V) = \frac{1}{\alpha_n(V) + \beta_n(V)}
\end{eqnarray}

```python
# @title Hodgkin-Huxley model of the action potential {"vertical-output":true,"display-mode":"form"}
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Hodgkin-Huxley model of the action potential in the squid giant axon

# Define auxiliary functions

def HH_alpha_M( V ):
    #    alpha = 0.1 * (V+40) / (1 - exp( -0.1*(V + 40) ) );
    va = 0.1 * ( V + 40 );
    alpha = va / ( 1 -  np.exp( -va ) );  # CAUTION: this fails when V equals exactly -40 mV !!!!
    return alpha

def HH_beta_M( V ):
    #    beta  = 4 * exp( -0.0556*(V+65) );
    vb = 0.0556 * ( V + 65 );
    beta = 4 *  np.exp( -vb );
    return beta

def HH_tau( alpha , beta ):
    tau = 1 / (alpha + beta );
    return tau

def HH_ss( alpha , beta ):
    ss = alpha / (alpha + beta );
    return ss

def HH_alpha_H( V ):
    # alpha = 0.07 * exp( -0.05 * (V + 65) )
    va = 0.05 * ( V + 65 );
    alpha = 0.07 *  np.exp( -va );
    return alpha

def HH_beta_H( V ):
    #    beta  = 1 / ( 1 + exp(-0.1*(V + 35) ) )
    vb = 0.1 * ( V + 35 );
    beta = 1 / ( 1 + np.exp( -vb ) );
    return beta

def HH_alpha_N( V ):
    # alpha = 0.01 * (V+55) / (1 - exp(-0.1 * (V+55) ) )
    va = 0.1 * ( V + 55 );                      # CAUTION: this fails when V equals exactly -55 mV !!!!
    alpha = 0.1 * va / ( 1 - np.exp( -va ) );
    return alpha

def HH_beta_N( V ):
    #   beta  = 0.125 * exp( -0.0125*(V+65) );
    vb = 0.0125 * ( V + 65 );
    beta = 0.125 * np.exp( -vb );
    return beta

# Define an interactive plot
def plot_squid_action_potential(i_e=-30, V0_m=-65):
    """(Slide-45)"""
    #i_e  = -30                                 # [nA] constant electrode current
    #V0_m = -65                                 # [mV] initial value of membrane potential

    # parameters
    cm = 10;      # nF/mm^2
    gL = 3;       # uS/mm^2  (10e-6 S / mm)
    gNa = 1200;   # uS/mm^2
    gK = 360;     # uS/mm^2
    EL = -54.402; # mV
    ENa = 50;     # mV
    EK = -77;     # mV

    dt  = 0.1;                   # time step, in ms (0.1 ms)
    T   = 15;                    # total time, in ms (15 ms)
    Ni  = round(T/dt) + 1;       # number of time steps
    ti  = np.linspace(0, T, Ni); # time vector, with successive time steps from 0 to T

    # allocate vectors for varaibles to be plotted (membrane potential, steady-state potential, activation variables, currents)

    Vmi = np.zeros_like(ti)
    Vinftyi = np.zeros_like(ti)

    Mi   = np.zeros_like(ti)
    Hi   = np.zeros_like(ti)
    Ni   = np.zeros_like(ti)

    iLi  = np.zeros_like(ti)
    iNai = np.zeros_like(ti)
    iKi  = np.zeros_like(ti)


    # assign initial values where necessary

    Vmi[0]     = V0_m                     # membrane potential

    alphaMi = HH_alpha_M( Vmi[0] )        # initial rates [1/ms]
    betaMi  = HH_beta_M( Vmi[0] )
    alphaHi = HH_alpha_H( Vmi[0] )
    betaHi  = HH_beta_H( Vmi[0] )
    alphaNi = HH_alpha_N( Vmi[0] )
    betaNi  = HH_beta_N( Vmi[0] )

    ssMi    = HH_ss( alphaMi , betaMi )   # initial steady-states [fractions]
    ssHi    = HH_ss( alphaHi , betaHi )
    ssNi    = HH_ss( alphaMi , betaMi )

    Mi[0]   = ssMi                        # initial activation variables [fractions]
    Hi[0]   = ssHi
    Ni[0]   = ssNi


    # iterate over successive time steps, starting at resting potential (reversal
    # potential of leak)

    for i in range( len(ti) ):

        gNai       = gNa * Hi[i] * Mi[i] ** 3         # conductances [uS/mm^2]
        gKi        = gK * Ni[i] ** 4

        DFLi       = -Vmi[i] + EL                     # driving forces [mV]
        DFNai      = -Vmi[i] + ENa
        DFKi       = -Vmi[i] + EK

        iLi[i]     = gL      * DFLi                   # currents [nA/mm^2 =  mV * uS/mm^2]
        iNai[i]    = gNai    * DFNai
        iKi[i]     = gKi     * DFKi

        # current steady-states and time-constants       [mV]

        Vinftyi[i] = (gL*EL+gNai*ENa+gKi*EK+i_e)/(gL+gNai+gKi);
        taueffi    = cm/(gL+gNai+gKi);                # [ms = (nF/mm^2) / (uS/mm^2)]

        alphaMi    = HH_alpha_M( Vmi[i] )             # rates
        betaMi     = HH_beta_M( Vmi[i] )
        alphaHi    = HH_alpha_H( Vmi[i] )
        betaHi     = HH_beta_H( Vmi[i] )
        alphaNi    = HH_alpha_N( Vmi[i] )
        betaNi     = HH_beta_N( Vmi[i] )

        tauMi   = HH_tau( alphaMi , betaMi )
        ssMi    = HH_ss( alphaMi , betaMi )
        tauHi   = HH_tau( alphaHi , betaHi )
        ssHi    = HH_ss( alphaHi , betaHi )
        tauNi   = HH_tau( alphaNi , betaNi )
        ssNi    = HH_ss( alphaMi , betaMi )

        if i < len(ti)-1:
             # update state variables at time i+1
            Vmi[i + 1] = Vinftyi[i] + (Vmi[i] - Vinftyi[i]) * np.exp(-dt / taueffi)

            Mi[i + 1] = ssMi + (Mi[i] - ssMi) * np.exp(-dt / tauMi)

            Hi[i + 1] = ssHi + (Hi[i] - ssHi) * np.exp(-dt / tauHi)

            Ni[i + 1] = ssNi + (Ni[i] - ssNi) * np.exp(-dt / tauNi)


    # Create a figure and a grid of subplots (1 row, 3 columns)
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))

    # First subplot
    axs[0].plot(ti, Vmi, 'k', ti, Vinftyi, 'k--');
    axs[0].set_title('Potentials');
    axs[0].set_xlabel('time [ms]');
    axs[0].set_ylabel('potential [mV]');

    # Second subplot
    axs[1].plot(ti, Mi, 'r', ti, Hi, 'r--', ti, Ni, 'b');
    axs[1].set_title('Activation');
    axs[1].set_xlabel('time [ms]');
    axs[1].set_ylabel('fraction');

    # Third subplot
    axs[2].plot(ti, iNai/1000, 'r', ti, iKi/1000, 'b', ti, iLi/1000, 'g', ti, (iNai+iKi+iLi)/1000, 'k');
    axs[2].set_title('Currents');
    axs[2].set_xlabel('time [ms]');
    axs[2].set_ylabel('current [uA]');

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plots
    plt.show()

# Creating sliders for concentrations and charge
interact(plot_squid_action_potential, i_e=(-50, 50, 1), V0_m=(-100, 0, 1))

```

# Summary of HHM

* Four state variables interact in the HHM of the squid action potential.

* When a negative membrane potential is rising, this is reinforced by rising activation $m$, which allows $Na^+$ influx.

* With some delay, this depolarization is **slowed** by **in**activation $h$, which closes down this influx.

* With a similar delay, the depolarization is **counteracted** by rising activation $n$, which allows $K^+$ outflux.

* This outflux hyperpolarizes the membrane potential, until activation $n$ has fallen and **in**activation $h$ have risen again.


This beautifully balanced biophysical mechanism continues to be an active topic of research:
